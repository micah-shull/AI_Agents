````python
def build_roi_section(roi_ledger: List[Dict[str, Any]]) -> str:
    """Marketing ROI this period: total spend, estimated value, net ROI %, one-line verdict."""
    if not roi_ledger:
        return "## Marketing ROI\n\nNo ROI ledger data in this period.\n\n"
    total_llm = sum(e.get("llm_cost") or 0 for e in roi_ledger)
    total_human = sum(e.get("human_review_cost") or 0 for e in roi_ledger)
    total_media = sum(e.get("media_spend") or 0 for e in roi_ledger)
    total_value = sum(e.get("estimated_value") or 0 for e in roi_ledger)
    total_net = sum(e.get("net_roi") or 0 for e in roi_ledger)
    total_cost = total_llm + total_human + total_media
    roi_pct = (total_net / total_cost * 100) if total_cost else 0.0
    if total_net > 0:
        roi_status = "positive"
        verdict = "**ROI: Positive** â€” Estimated value exceeds cost this period."
    elif total_net < 0:
        roi_status = "negative"
        verdict = "**ROI: Negative** â€” Cost exceeds estimated value; review underperforming campaigns."
    else:
        roi_status = "neutral"
        verdict = "**ROI: Neutral** â€” Estimated value matches cost."
    lines = [
        "## Marketing ROI",
        "",
        verdict,
        "",
        "| Metric | Amount |",
        "|--------|--------|",
        f"| Total cost (LLM + human review + media) | ${total_cost:,.2f} |",
        f"| Estimated value | ${total_value:,.2f} |",
        f"| Net ROI | ${total_net:,.2f} |",
        f"| ROI % | {roi_pct:+.1f}% |",
        "",
    ]
    return "\n".join(lines)


def build_targets_vs_actuals(
    budget_actions: List[Dict[str, Any]],
    segment_rollups: List[Dict[str, Any]],
    open_risks: List[Dict[str, Any]],
    target_approval_rate_min: float,
    target_segments_on_target_min: float,
    target_open_risks_max: int,
    target_pending_reallocations_max: int,
) -> str:
    """Actual vs target: approval rate, segments on target, open risks, pending reallocations."""
    # Approval rate
    n_budget = len(budget_actions)
    n_approved = sum(1 for b in budget_actions if b.get("approved_by_human"))
    approval_rate = (n_approved / n_budget) if n_budget else 0.0
    approval_ok = approval_rate >= target_approval_rate_min

    # Segments on target = not "stop"
    n_segments = len(segment_rollups)
    on_target = sum(1 for r in segment_rollups if (r.get("recommendation") or "").lower() != "stop")
    segment_ratio = (on_target / n_segments) if n_segments else 1.0
    segments_ok = segment_ratio >= target_segments_on_target_min

    # Open risks
    n_risks = len(open_risks)
    risks_ok = n_risks <= target_open_risks_max

    # Pending reallocations
    pending = [b for b in budget_actions if (b.get("status") or "").lower() == "proposed"]
    n_pending = len(pending)
    pending_ok = n_pending <= target_pending_reallocations_max

    lines = [
        "## Targets vs actuals",
        "",
        "| Metric | Actual | Target | Verdict |",
        "|--------|--------|--------|---------|",
    ]
    lines.append(
        f"| Budget approval rate | {approval_rate:.0%} ({n_approved}/{n_budget}) | â‰¥ {target_approval_rate_min:.0%} | **{'On target' if approval_ok else 'Below target'}** |"
    )
    lines.append(
        f"| Segments on target | {on_target}/{n_segments} ({segment_ratio:.0%}) | â‰¥ {target_segments_on_target_min:.0%} | **{'On target' if segments_ok else 'Below target'}** |"
    )
    lines.append(
        f"| Open risk signals | {n_risks} | â‰¤ {target_open_risks_max} | **{'On target' if risks_ok else 'Above target â€” attention needed'}** |"
    )
    lines.append(
        f"| Pending reallocations | {n_pending} | â‰¤ {target_pending_reallocations_max} | **{'On target' if pending_ok else 'Above target â€” attention needed'}** |"
    )
    lines.append("")
    return "\n".join(lines)


def build_so_what_line(
    open_risks: List[Dict[str, Any]],
    segment_rollups: List[Dict[str, Any]],
    budget_actions: List[Dict[str, Any]],
) -> str:
    """One sentence under the headline: what matters and what to do."""
    n_risks = len(open_risks)
    stop_rollups = [r for r in segment_rollups if (r.get("recommendation") or "").lower() == "stop"]
    n_stops = len(stop_rollups)
    pending = [b for b in budget_actions if (b.get("status") or "").lower() == "proposed"]
    n_pending = len(pending)
    if n_risks == 0 and n_stops == 0 and n_pending == 0:
        return "No open risks, no segment stops, and no pending reallocations â€” address any opportunities below."
    parts = []
    if n_risks:
        parts.append(f"{n_risks} open risk{'s' if n_risks != 1 else ''} require review")
    if n_pending:
        parts.append(f"{n_pending} reallocation{'s' if n_pending != 1 else ''} pending")
    if n_stops:
        parts.append(f"{n_stops} segment{'s' if n_stops != 1 else ''} recommended stop")
    return " â€” ".join(parts) + " â€” address before next cycle."
````
