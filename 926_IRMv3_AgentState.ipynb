# Agent State

## Architecture Review & Explanation

The **Integration & Risk Management Orchestrator (IRMv3)** is designed to monitor the operational health of an AI ecosystem and convert system telemetry into **clear, decision-ready intelligence for leadership**.

Instead of performing tasks like most AI agents, this orchestrator acts as a **portfolio-level monitoring system** that continuously evaluates:

* integration reliability
* dependency exposure
* governance responsiveness
* automation value delivery

These signals are then combined into a **portfolio health score** and an **executive report** that highlights emerging risks.

This design makes the orchestrator function more like an **AI operations command center** than a traditional automation tool.

---

# Core Architecture Concept

IRMv3 follows a **structured state-driven orchestration model**.

The system works by moving information through a shared state object that captures:

1. **Data inputs** from telemetry files
2. **Derived insights** produced by scoring logic
3. **Executive outputs** generated at the end of the workflow

Every stage of the pipeline reads from and writes back to the same state.

This approach makes the entire decision flow **transparent and auditable**, which is essential when systems are used to monitor operational risk.

---

# The State Object: The System’s Memory

The `IRMv3State` object acts as the **central memory of the orchestrator**.

It stores:

* loaded telemetry data
* computed risk signals
* calculated scores
* rollups and triggers
* final reports

Because everything is captured in one structure, the orchestrator always has a **complete picture of the system’s operational condition**.

### Why this matters operationally

Many AI systems produce outputs without preserving the reasoning that led to them.

Here, every score and decision can be traced to:

* specific integrations
* workflow execution logs
* remediation records
* historical performance data

That traceability is essential for **operational governance**.

For executives, this means:

**the system can explain its conclusions.**

# Lookup Structures for Fast Analysis

The state also builds several **lookup maps**:

```
agents_lookup
workflows_by_agent
workflows_by_integration
```

These are small but important architectural decisions.

Instead of repeatedly scanning raw lists, the system can immediately answer questions like:

* Which workflows belong to this agent?
* Which workflows rely on this integration?
* Which systems create dependency concentration?

This improves both **performance and clarity** in the scoring logic.

---

# The Scoring Model

IRMv3 evaluates four core dimensions of AI operations.

### 1. Integration Stability

Measures:

* API reliability
* system latency
* integration health logs

This answers:

**Are the systems connecting reliably?**

---

### 2. Dependency Exposure

Measures:

* concentration of integrations
* shared system dependencies
* incident propagation risk

This answers:

**Could a single failure disrupt multiple AI systems?**

---

### 3. Governance Responsiveness

Measures:

* unresolved risk signals
* overdue remediation actions
* governance review history

This answers:

**Are problems being fixed quickly enough?**

---

### 4. Automation Value

Measures:

* expected vs actual automation performance
* workflow throughput
* operational cost metrics

This answers:

**Is automation delivering the value we expected?**

---

# Portfolio Health Score

These signals are combined into a single **portfolio score**.

```
portfolio_score =
  0.30 Integration Stability
  0.25 Dependency Exposure
  0.20 Governance Responsiveness
  0.25 Automation Value
```

This weighted scoring model allows leadership to quickly understand:

**Is the AI ecosystem healthy overall?**

---

# Portfolio Status Classification

IRMv3 converts scores into **clear operating states**:

```
HEALTHY
AT_RISK
CRITICAL
```

Rules determine these states using configurable thresholds.

Examples:

* Any score below **60** → Critical condition
* Portfolio score below **75** → At risk
* Two weak subscores → At risk

These rules create **deterministic risk classification**.

This is important because leadership needs systems that behave **predictably and consistently**.

---

# Executive Triggers

The orchestrator also generates **trigger signals** when certain conditions occur.

Examples include:

* critical integration reliability
* excessive open risks
* overdue remediation actions
* automation value leakage

Triggers are not vague alerts.

They are **explicit, rule-driven escalation signals**.

This design supports **clear operational ownership**.

````python
# ============================================================================
# Integration & Risk Management Orchestrator v3 (IRMv3)
# ============================================================================
# Dashboard/portfolio agent: integration health, dependency exposure,
# governance responsiveness, automation value → portfolio health signal.

class IRMv3State(TypedDict, total=False):
    """State for Integration & Risk Management Orchestrator v3."""

    # Input
    data_dir: Optional[str]
    reports_dir: Optional[str]

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Raw data (loaded from agents/data)
    agents: List[Dict[str, Any]]
    workflows: List[Dict[str, Any]]
    system_integrations: List[Dict[str, Any]]
    dependency_graph: List[Dict[str, Any]]
    mission_runs: List[Dict[str, Any]]
    task_execution_logs: List[Dict[str, Any]]
    integration_health_logs: List[Dict[str, Any]]
    dependency_incidents: List[Dict[str, Any]]
    risk_signals: List[Dict[str, Any]]
    remediation_actions: List[Dict[str, Any]]
    historical_snapshots: List[Dict[str, Any]]
    governance_reviews: List[Dict[str, Any]]
    expected_vs_actual_value: List[Dict[str, Any]]
    kpis_cost: List[Dict[str, Any]]

    # Lookups
    agents_lookup: Dict[str, Dict[str, Any]]
    workflows_by_agent: Dict[str, List[Dict[str, Any]]]
    workflows_by_integration: Dict[str, List[str]]  # integration_id -> workflow_ids

    # Scores (0–100 each)
    integration_stability_score: float
    dependency_exposure_score: float
    governance_responsiveness_score: float
    automation_value_score: float
    portfolio_score: float
    portfolio_status: str  # HEALTHY | AT_RISK | CRITICAL

    # Rollups and triggers
    score_rollup: Dict[str, Any]
    executive_triggers: List[Dict[str, Any]]
    data_snapshot_loaded_at: Optional[str]
    validation_warnings: List[str]
    loader_record_counts: Dict[str, int]

    # Output
    mission_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class IRMv3Config:
    """Configuration for IRMv3: paths, score weights, thresholds."""

    reports_dir: str = "agents/output/irm_v3_reports"
    data_dir: str = "agents/data"

    # File names (under data_dir)
    agents_file: str = "agents.json"
    workflows_file: str = "workflows.json"
    system_integrations_file: str = "system_integrations.json"
    dependency_graph_file: str = "dependency_graph.json"
    mission_runs_file: str = "mission_runs.json"
    task_execution_logs_file: str = "task_execution_logs.json"
    integration_health_logs_file: str = "integration_health_logs.json"
    dependency_incidents_file: str = "dependency_incidents.json"
    risk_signals_file: str = "risk_signals.json"
    remediation_actions_file: str = "remediation_actions.json"
    historical_snapshots_file: str = "historical_snapshots.json"
    governance_reviews_file: str = "governance_reviews.json"
    expected_vs_actual_value_file: str = "expected_vs_actual_value.json"
    kpis_cost_file: str = "kpis_cost.json"

    # Portfolio status thresholds
    critical_threshold: float = 60.0   # any score < 60 → CRITICAL
    at_risk_portfolio_score: float = 75.0  # portfolio_score < 75 → AT_RISK
    at_risk_two_scores_below: float = 70.0  # two scores < 70 → AT_RISK

    # Executive triggers
    integration_stability_critical: float = 60.0
    dependency_exposure_critical: float = 60.0
    governance_responsiveness_critical: float = 60.0
    automation_value_critical: float = 60.0
    open_risks_critical: int = 8
    overdue_remediations_critical: int = 3

    # Score weights (for portfolio_score)
    weight_integration: float = 0.30
    weight_dependency: float = 0.25
    weight_governance: float = 0.20
    weight_automation: float = 0.25
````
