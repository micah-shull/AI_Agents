# üî∑ 2Ô∏è‚É£ Root Cause Attribution Layer

### Purpose:

Distinguish demand failure vs operational leakage.

### From stock_availability.csv:

We derive:

* `stockout_overlap_weeks`
* `percent_gap_due_to_stockout`

* `demand_drop_flag`

Logic example:

If revenue declined AND stockout occurred same week ‚Üí operational-driven
If revenue declined AND no stockout ‚Üí demand-driven
If zero spend + no stockout ‚Üí behavioral churn

* üî¥ Demand
