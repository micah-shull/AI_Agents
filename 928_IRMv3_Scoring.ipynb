# IRMv3 Scoring System

## Architecture Review & Operational Explanation

The scoring module is responsible for converting raw telemetry into **clear signals about the health of the AI ecosystem**.

Rather than relying on AI interpretation or heuristics, the orchestrator evaluates the system using **deterministic operational metrics**.

Each score represents a critical dimension of AI operations:

| Score                     | What it measures                 |
| ------------------------- | -------------------------------- |
| Integration Stability     | Are systems connecting reliably? |
| Dependency Exposure       | How fragile is the architecture? |
| Governance Responsiveness | Are problems being fixed?        |
| Automation Value          | Is automation delivering ROI?    |

These signals are then combined into a **portfolio health score**, which produces the overall system status.

This approach ensures that the orchestrator produces **consistent and explainable evaluations of AI operations**.

---

# Integration Stability Score

This score measures **whether the underlying systems supporting AI automation are reliable**.

It evaluates four operational signals:

| Signal        | Meaning                                  |
| ------------- | ---------------------------------------- |
| Uptime        | Are integrations available when needed?  |
| Error Rate    | Are requests failing?                    |
| Latency       | Are integrations slowing down workflows? |
| Schema Errors | Are data contracts breaking?             |

The score uses a weighted model:

```
Uptime          40%
Error Rate      30%
Latency         20%
Schema Errors   10%
```

This weighting makes sense operationally.

Availability and errors represent the **most disruptive issues**, while latency and schema mismatches represent **performance degradation**.

### Why leaders care

When integrations degrade, automation quietly breaks.

Workflows slow down.
Manual work reappears.
Customer response times increase.

This score gives leadership a **direct signal of infrastructure reliability**.

---

# Dependency Exposure Score

This score evaluates **architectural fragility**.

The orchestrator analyzes the dependency graph to determine:

* how many workflows depend on each integration
* how many **critical workflows** depend on it
* how often dependency failures occur
* whether failover systems exist

The scoring dimensions are:

```
Dependency concentration   35%
Critical workflow exposure 30%
Incident history           25%
Failover availability      10%
```

This design allows the orchestrator to estimate something extremely valuable:

**blast radius risk**

Example insight:

> If this API fails, how much of our AI operation stops working?

That is exactly the type of risk question most AI deployments **do not currently answer**.

---

# Governance Responsiveness Score

Many AI monitoring systems detect problems.

Few measure **whether organizations actually fix them**.

This score evaluates how effectively the organization responds to operational issues.

It considers:

| Metric                      | Meaning                    |
| --------------------------- | -------------------------- |
| Remediation completion rate | Are issues being resolved? |
| Resolution time             | How long fixes take        |
| Overdue remediation         | Missed deadlines           |

The formula penalizes overdue actions and slow responses.

That makes the score reflect something critical:

**organizational discipline around AI operations**.

### Why this matters

A system may detect problems perfectly.

But if no one fixes them, the system still fails.

This score measures **operational accountability**.

---

# Automation Value Score

This score evaluates whether AI systems are delivering the value they promised.

It looks at four signals:

```
ROI            40%
Cost control   25%
Throughput     20%
Manual labor   15%
```

The score answers questions like:

* Are workflows meeting expected throughput?
* Are costs exceeding projections?
* Are humans stepping back into automated processes?

If automation slows down or requires manual intervention, the score declines.

This is critical because **automation value erosion often happens gradually and silently**.

---

# Portfolio Score

The portfolio score combines the four signals.

```
Integration Stability 30%
Dependency Exposure   25%
Governance Response   20%
Automation Value      25%
```

This weighted score gives leadership a **single indicator of AI operational health**.

Instead of monitoring dozens of dashboards, executives can quickly see whether the AI ecosystem is:

* improving
* stable
* deteriorating

---

# Portfolio Status Classification

The orchestrator converts the numeric score into **clear operating states**.

```
HEALTHY
AT_RISK
CRITICAL
```

Rules are explicit:

CRITICAL if:

* any score < 60

AT_RISK if:

* two scores < 70
* portfolio score < 75

Otherwise:

HEALTHY

These thresholds provide **predictable operational signals**.

There is no ambiguity in how the system interprets risk.

````python
"""
IRMv3 score calculations. All scores 0–100; higher = healthier.
Formulas aligned with IRMv3_score_calculation_formulas.md.
"""

from typing import Any, Dict, List, Optional


# ---------- Integration Stability Score ----------

def compute_integration_stability_score(
    integration_health_logs: List[Dict[str, Any]],
    weights: Optional[Dict[str, float]] = None,
) -> float:
    """
    Are systems connecting reliably?
    Inputs: integration_health_logs (uptime_percent, latency_ms, error_rate_percent, schema_mismatch_events).
    Weights: uptime 0.40, error_rate 0.30, latency 0.20, schema 0.10.
    """
    w = weights or {"uptime": 0.40, "error_rate": 0.30, "latency": 0.20, "schema": 0.10}
    if not integration_health_logs:
        return 85.0  # no data: assume acceptable

    # Use latest per integration (by timestamp if present), then average across integrations
    by_int: Dict[str, Dict[str, Any]] = {}
    for row in integration_health_logs:
        iid = row.get("integration_id", "")
        ts = row.get("timestamp", "")
        if iid not in by_int or (ts and (by_int[iid].get("timestamp") or "") < ts):
            by_int[iid] = row

    scores: List[float] = []
    for row in by_int.values():
        uptime = float(row.get("uptime_percent", row.get("uptime", 99)))
        uptime_score = min(100, max(0, uptime))

        error_rate = float(row.get("error_rate_percent", row.get("error_rate", 0)) or 0)
        error_rate_score = max(0, 100 - error_rate)

        latency_ms = float(row.get("latency_ms", row.get("latency_p95_ms", 300)) or 300)
        latency_score = max(0, 100 - (latency_ms / 300) * 20)

        schema_events = int(row.get("schema_mismatch_events", 0) or 0)
        schema_score = max(0, 100 - schema_events * 10)

        comp = (
            w["uptime"] * uptime_score
            + w["error_rate"] * error_rate_score
            + w["latency"] * latency_score
            + w["schema"] * schema_score
        )
        scores.append(comp)

    return round(sum(scores) / len(scores), 1) if scores else 85.0


# ---------- Dependency Exposure Score ----------

def _count_workflows_per_integration(
    dependency_graph: List[Dict[str, Any]],
    workflows_by_agent: Dict[str, List[Dict[str, Any]]],
) -> Dict[str, int]:
    """integration_id -> count of workflows that depend on it."""
    agent_to_integrations: Dict[str, List[str]] = {}
    for edge in dependency_graph:
        src = edge.get("source_component")
        tgt = edge.get("target_component")
        if src and tgt:
            agent_to_integrations.setdefault(src, []).append(tgt)
    out: Dict[str, int] = {}
    for agent_id, int_ids in agent_to_integrations.items():
        n_wf = len(workflows_by_agent.get(agent_id, []))
        for iid in int_ids:
            out[iid] = out.get(iid, 0) + n_wf
    return out


def _count_critical_workflows_per_integration(
    dependency_graph: List[Dict[str, Any]],
    workflows_by_agent: Dict[str, List[Dict[str, Any]]],
) -> Dict[str, int]:
    """integration_id -> count of critical workflows that depend on it."""
    agent_to_integrations: Dict[str, List[str]] = {}
    for edge in dependency_graph:
        src = edge.get("source_component")
        tgt = edge.get("target_component")
        if src and tgt:
            agent_to_integrations.setdefault(src, []).append(tgt)
    out: Dict[str, int] = {}
    for agent_id, int_ids in agent_to_integrations.items():
        wfs = workflows_by_agent.get(agent_id, [])
        critical = sum(1 for w in wfs if (w.get("criticality") or "").lower() == "high")
        for iid in int_ids:
            out[iid] = out.get(iid, 0) + critical
    return out


def compute_dependency_exposure_score(
    dependency_graph: List[Dict[str, Any]],
    dependency_incidents: List[Dict[str, Any]],
    workflows_by_agent: Dict[str, List[Dict[str, Any]]],
    system_integrations: List[Dict[str, Any]],
    weights: Optional[Dict[str, float]] = None,
) -> float:
    """
    How fragile is the architecture? (Higher = less exposed.)
    Weights: concentration 0.35, criticality 0.30, incident 0.25, failover 0.10.
    """
    w = weights or {"concentration": 0.35, "criticality": 0.30, "incident": 0.25, "failover": 0.10}
    if not dependency_graph and not system_integrations:
        return 80.0

    # Per-integration component scores, then average
    wf_per_int = _count_workflows_per_integration(dependency_graph, workflows_by_agent)
    crit_per_int = _count_critical_workflows_per_integration(dependency_graph, workflows_by_agent)
    incident_count_per_int: Dict[str, int] = {}
    for inc in dependency_incidents:
        comp = inc.get("dependency_component", "")
        if comp:
            incident_count_per_int[comp] = incident_count_per_int.get(comp, 0) + 1

    fallback_by_int: Dict[str, bool] = {}
    for s in system_integrations:
        iid = s.get("integration_id", "")
        fallback_by_int[iid] = bool(s.get("fallback_available", False))
    for edge in dependency_graph:
        tgt = edge.get("target_component", "")
        if tgt and tgt not in fallback_by_int:
            fallback_by_int[tgt] = bool(edge.get("failover_available", False))

    integration_ids = set(wf_per_int) | set(fallback_by_int) | set(s.get("integration_id", "") for s in system_integrations if s.get("integration_id"))
    if not integration_ids:
        return 80.0

    scores: List[float] = []
    for iid in integration_ids:
        dependent_wf = wf_per_int.get(iid, 0)
        concentration_score = max(0, 100 - dependent_wf * 5)

        critical_wf = crit_per_int.get(iid, 0)
        criticality_score = max(0, 100 - critical_wf * 10)

        incidents = incident_count_per_int.get(iid, 0)
        incident_score = max(0, 100 - incidents * 8)

        failover_score = 100 if fallback_by_int.get(iid, False) else 60

        comp = (
            w["concentration"] * concentration_score
            + w["criticality"] * criticality_score
            + w["incident"] * incident_score
            + w["failover"] * failover_score
        )
        scores.append(comp)

    return round(sum(scores) / len(scores), 1) if scores else 80.0


# ---------- Governance Responsiveness Score ----------

def compute_governance_responsiveness_score(
    remediation_actions: List[Dict[str, Any]],
    risk_signals: List[Dict[str, Any]],
    weights: Optional[Dict[str, float]] = None,
) -> float:
    """
    Are problems being fixed? Completion, resolution time, overdue/repeat penalties.
    """
    w = weights or {"completion": 0.50, "resolution": 0.30}
    if not remediation_actions:
        return 75.0

    total = len(remediation_actions)
    resolved = sum(1 for r in remediation_actions if (r.get("status") or "").lower() == "resolved")
    completion_score = (resolved / total * 100) if total else 100

    resolution_days = [
        r.get("days_open") for r in remediation_actions
        if r.get("resolution_date") is not None and isinstance(r.get("days_open"), (int, float))
    ]
    avg_days = sum(resolution_days) / len(resolution_days) if resolution_days else 0
    resolution_score = max(0, 100 - avg_days * 2)

    overdue = sum(1 for r in remediation_actions if r.get("missed_deadline_flag") is True)
    overdue_penalty = overdue * 5
    # Repeat incidents: same risk_id appearing again would need risk_signals linkage; simplify to 0 for now
    repeat_penalty = 0

    score = w["completion"] * completion_score + w["resolution"] * resolution_score - overdue_penalty - repeat_penalty
    return round(max(0, min(100, score)), 1)


# ---------- Automation Value Score ----------

def compute_automation_value_score(
    task_execution_logs: List[Dict[str, Any]],
    workflows: List[Dict[str, Any]],
    expected_vs_actual_value: Optional[List[Dict[str, Any]]] = None,
    kpis_cost: Optional[List[Dict[str, Any]]] = None,
    weights: Optional[Dict[str, float]] = None,
) -> float:
    """
    Are AI systems delivering value? ROI, cost, throughput, manual labor.
    When expected_vs_actual_value / kpis_cost missing, derive from task_execution_logs + workflows.
    """
    w = weights or {"roi": 0.40, "cost": 0.25, "throughput": 0.20, "labor": 0.15}

    if expected_vs_actual_value and len(expected_vs_actual_value) > 0:
        roi_scores = []
        for row in expected_vs_actual_value:
            exp = float(row.get("expected_roi", 1) or 1)
            actual = float(row.get("actual_roi", 0) or 0)
            ratio = actual / exp if exp else 0
            roi_scores.append(min(100, ratio * 100))
        roi_score = sum(roi_scores) / len(roi_scores) if roi_scores else 80
        cost_overrun_pct = 0
        for row in expected_vs_actual_value:
            co = float(row.get("cost_overrun", 0) or row.get("cost_overrun_percentage", 0) or 0)
            cost_overrun_pct = max(cost_overrun_pct, co)
        cost_score = max(0, 100 - cost_overrun_pct)
    else:
        roi_score = 80.0
        cost_score = 85.0

    # Throughput from task_execution_logs vs workflows expected_throughput_per_day
    workflow_expected: Dict[str, float] = {w["workflow_id"]: float(w.get("expected_throughput_per_day", 100) or 100) for w in workflows}
    if task_execution_logs and workflow_expected:
        throughput_ratios = []
        for log in task_execution_logs:
            wf_id = log.get("workflow_id", "")
            exp = workflow_expected.get(wf_id, 100)
            actual = float(log.get("throughput", 0) or 0)
            if exp > 0:
                throughput_ratios.append(min(100, (actual / exp) * 100))
        throughput_score = sum(throughput_ratios) / len(throughput_ratios) if throughput_ratios else 80
    else:
        throughput_score = 80.0

    # Manual labor: human_intervention in task_execution_logs
    if task_execution_logs:
        with_intervention = sum(1 for t in task_execution_logs if t.get("human_intervention") is True)
        total = len(task_execution_logs)
        # Lower manual % = higher score
        labor_pct = (with_intervention / total * 100) if total else 0
        labor_score = max(0, 100 - labor_pct)
    else:
        labor_score = 85.0

    score = (
        w["roi"] * roi_score
        + w["cost"] * cost_score
        + w["throughput"] * throughput_score
        + w["labor"] * labor_score
    )
    return round(max(0, min(100, score)), 1)


# ---------- Portfolio Score & Status ----------

def compute_portfolio_score(
    integration_stability: float,
    dependency_exposure: float,
    governance_responsiveness: float,
    automation_value: float,
    weights: Optional[Dict[str, float]] = None,
) -> float:
    """Weighted combination of the four scores."""
    w = weights or {"integration": 0.30, "dependency": 0.25, "governance": 0.20, "automation": 0.25}
    return round(
        w["integration"] * integration_stability
        + w["dependency"] * dependency_exposure
        + w["governance"] * governance_responsiveness
        + w["automation"] * automation_value,
        1,
    )


def compute_portfolio_status(
    integration_stability: float,
    dependency_exposure: float,
    governance_responsiveness: float,
    automation_value: float,
    portfolio_score: float,
    critical_threshold: float = 60.0,
    at_risk_portfolio: float = 75.0,
    at_risk_two_below: float = 70.0,
) -> str:
    """
    HEALTHY / AT_RISK / CRITICAL.
    - Any score < critical_threshold → CRITICAL
    - Else two scores < at_risk_two_below → AT_RISK
    - Else portfolio_score < at_risk_portfolio → AT_RISK
    - Else HEALTHY
    """
    scores = [integration_stability, dependency_exposure, governance_responsiveness, automation_value]
    if any(s < critical_threshold for s in scores):
        return "CRITICAL"
    below_70 = sum(1 for s in scores if s < at_risk_two_below)
    if below_70 >= 2:
        return "AT_RISK"
    if portfolio_score < at_risk_portfolio:
        return "AT_RISK"
    return "HEALTHY"
````
