## üå± **From Tier 1 ‚Üí Tier 2: Why This Dataset Matters**

* simple routines
* basic product relationships
* essential steps (cleanser ‚Üí moisturizer ‚Üí SPF)

This dataset adds:

* **active ingredients** (Vitamin C, Niacinamide, Retinol, AHA/BHA, etc.)
* **target skin concerns** (acne, aging, fine lines, hyperpigmentation)
* **upgrade paths** from Tier 1 to Tier 2 products
* **routine step mapping** (cleanse ‚Üí prep ‚Üí treat ‚Üí moisturize ‚Üí protect)
* **more diverse customer personas**
* **ingredient ‚Üí concern alignment**
* **advanced cross-sell logic**
* **stronger RFM + churn risk patterns**

## ü§ñ **What the Tier 2 Orchestrator Will Learn**

Using this dataset, the Tier 2 agent will be capable of:

### **1. Ingredient-based reasoning**

Matching Niacinamide to acne concerns, Vitamin C to dullness or dark spots, Retinol to aging, etc.

Recognizing when a customer is ready to move from Tier 1 to Tier 2 products.

### **3. Concern-based cross-sells**

Recommending treatment serums, masks, or advanced moisturizers depending on skin goals.

### **4. Routine-step adjacency logic**

Ensuring recommendations follow skincare order:
cleanser ‚Üí toner ‚Üí serum ‚Üí moisturizer ‚Üí SPF.

### **5. Better revenue optimization**

Using RFM, price sensitivity, and CLV to prioritize the highest-value opportunities.

### **6. Higher-quality personalization**

Taking into account:

* skin sensitivity
* product actives
* routine completeness
* ingredient conflicts
* customer tier preference

# data validation script

```python
#!/usr/bin/env python3
"""
Validation script for Tier 2 customer and product data
"""
import json
from datetime import datetime, timedelta
from collections import defaultdict

# Load data
with open('data_tier_2/product_catalog.json') as f:
    products = json.load(f)

with open('data_tier_2/customers.json') as f:
    customers = json.load(f)

# Get product IDs and create lookup
product_ids = {p['product_id'] for p in products}
product_lookup = {p['product_id']: p for p in products}

print("=" * 70)
print("TIER 2 DATA VALIDATION REPORT")
print("=" * 70)
print(f"\nüì¶ Products loaded: {len(products)}")
print(f"üë• Customers loaded: {len(customers)}\n")

errors = []
warnings = []
info = []

# Validate products first
print("üîç Validating Product Catalog...")
for product in products:
    pid = product['product_id']

    # Check tier
    if product.get('tier') != 2:
        errors.append(f"Product {pid}: Should be tier 2, got {product.get('tier')}")

    # Check price range
    price = product.get('price', 0)
    if not (25.0 <= price <= 59.99):
        errors.append(f"Product {pid}: Price ${price:.2f} out of range ($25-$60)")

    # Check required fields
    required_fields = ['routine_step', 'active_ingredients', 'target_concerns', 'upgrades_from', 'recommended_cross_sells']
    for field in required_fields:
        if field not in product:
            errors.append(f"Product {pid}: Missing required field '{field}'")

    # Check routine_step
    if 'routine_step' in product:
        rs = product['routine_step']
        if not (1 <= rs <= 5):
            errors.append(f"Product {pid}: routine_step {rs} should be 1-5")

    # Check cross-sells
    if 'recommended_cross_sells' in product:
        cross_sells = product['recommended_cross_sells']
        if not (3 <= len(cross_sells) <= 5):
            errors.append(f"Product {pid}: Should have 3-5 cross-sells, got {len(cross_sells)}")
        if pid in cross_sells:
            errors.append(f"Product {pid}: Self-references in cross-sells")
        for cs_id in cross_sells:
            if cs_id not in product_ids:
                errors.append(f"Product {pid}: Cross-sell {cs_id} does not exist")

print(f"‚úÖ Product validation complete: {len(errors)} errors found\n")

# Validate customers
print("üîç Validating Customer Data...")
today = datetime.now()

for customer in customers:
    cid = customer['customer_id']

    # Check products_owned
    for product in customer.get('products_owned', []):
        pid = product['product_id']
        if pid not in product_ids:
            errors.append(f"Customer {cid}: Missing product {pid} in products_owned")
        else:
            # Check if product matches customer's skin concerns
            prod = product_lookup[pid]
            customer_concerns = set(customer.get('skin_concerns', []))
            product_concerns = set(prod.get('target_concerns', []))

            # Allow some flexibility - at least one concern should match or be general
            if not customer_concerns.intersection(product_concerns) and 'general' not in product_concerns:
                warnings.append(f"Customer {cid}: Product {pid} concerns {product_concerns} don't match customer concerns {customer_concerns}")

    # Check purchase_history
    purchase_history = customer.get('purchase_history', [])
    for purchase in purchase_history:
        pid = purchase['product_id']
        if pid not in product_ids:
            errors.append(f"Customer {cid}: Missing product {pid} in purchase_history")

    # Validate RFM consistency
    if purchase_history:
        try:
            dates = [datetime.strptime(p['date'], '%Y-%m-%d') for p in purchase_history]
            most_recent = max(dates)
            recency_calc = (today - most_recent).days
            recency_actual = customer.get('rfm', {}).get('recency_days', 0)

            if abs(recency_calc - recency_actual) > 5:
                errors.append(f"Customer {cid}: recency_days mismatch (calc: {recency_calc}, actual: {recency_actual}, most recent: {most_recent.date()})")

            cutoff = today - timedelta(days=90)
            frequency_calc = sum(1 for d in dates if d >= cutoff)
            frequency_actual = customer.get('rfm', {}).get('frequency_90d', 0)
            if frequency_calc != frequency_actual:
                errors.append(f"Customer {cid}: frequency_90d mismatch (calc: {frequency_calc}, actual: {frequency_actual})")

            monetary_calc = sum(p['amount'] for p, d in zip(purchase_history, dates) if d >= cutoff)
            monetary_actual = customer.get('rfm', {}).get('monetary_90d', 0)
            if abs(monetary_calc - monetary_actual) > 0.01:
                errors.append(f"Customer {cid}: monetary_90d mismatch (calc: ${monetary_calc:.2f}, actual: ${monetary_actual:.2f})")
        except (ValueError, KeyError) as e:
            errors.append(f"Customer {cid}: Error validating RFM - {e}")

    # Check churn_risk format
    churn = customer.get('churn_risk', 0)
    if not (0.0 <= churn <= 1.0):
        errors.append(f"Customer {cid}: churn_risk should be 0.0-1.0, got {churn}")

    # Check lifetime_value consistency
    total_history = sum(p['amount'] for p in purchase_history)
    ltv = customer.get('lifetime_value', 0)
    if abs(total_history - ltv) > 1.0:
        warnings.append(f"Customer {cid}: lifetime_value (${ltv:.2f}) differs from purchase_history sum (${total_history:.2f})")

    # Check churn_risk vs recency alignment
    recency = customer.get('rfm', {}).get('recency_days', 0)
    if recency < 30 and churn > 0.3:
        warnings.append(f"Customer {cid}: Low recency ({recency}d) but high churn_risk ({churn:.2f}) - may need review")
    elif recency > 60 and churn < 0.3:
        warnings.append(f"Customer {cid}: High recency ({recency}d) but low churn_risk ({churn:.2f}) - may need review")

    # Check routine completeness
    owned_categories = set(customer.get('categories', []))
    essential_categories = {'cleanser', 'moisturizer', 'spf'}
    missing_essentials = essential_categories - owned_categories
    if missing_essentials:
        info.append(f"Customer {cid}: Missing essential categories: {missing_essentials} (routine gap opportunity)")

# Print results
print("=" * 70)
if errors:
    print(f"‚ùå ERRORS FOUND: {len(errors)}")
    print("-" * 70)
    for error in errors[:20]:  # Show first 20
        print(f"  ‚Ä¢ {error}")
    if len(errors) > 20:
        print(f"  ... and {len(errors) - 20} more errors")
else:
    print("‚úÖ No critical errors found!")

print()
if warnings:
    print(f"‚ö†Ô∏è  WARNINGS: {len(warnings)}")
    print("-" * 70)
    for warning in warnings[:15]:  # Show first 15
        print(f"  ‚Ä¢ {warning}")
    if len(warnings) > 15:
        print(f"  ... and {len(warnings) - 15} more warnings")
else:
    print("‚úÖ No warnings!")

print()
if info:
    print(f"‚ÑπÔ∏è  INFO: {len(info)} routine gap opportunities found")
    print("-" * 70)
    for i in info[:10]:  # Show first 10
        print(f"  ‚Ä¢ {i}")
    if len(info) > 10:
        print(f"  ... and {len(info) - 10} more opportunities")

print()
print("=" * 70)
print("SUMMARY")
print("=" * 70)
print(f"Total Errors:   {len(errors)}")
print(f"Total Warnings: {len(warnings)}")
print(f"Opportunities:  {len(info)}")
print()

if len(errors) == 0:
    print("üéâ Data validation PASSED! Ready for orchestrator development.")
else:
    print("‚ö†Ô∏è  Please fix errors before proceeding.")

```
