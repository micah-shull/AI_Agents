````python
"""Integration & Risk Management Orchestrator - LangGraph Workflow"""

from langgraph.graph import StateGraph, END
from config import (
    IntegrationRiskManagementOrchestratorState,
    IntegrationRiskManagementOrchestratorConfig
)
from agents.integration_risk_orchestrator.nodes import (
    goal_node,
    planning_node,
    data_loading_node,
    integration_health_node,
    risk_analysis_node,
    workflow_kpi_analysis_node,
    value_analysis_node,
    prioritization_node,
    report_generation_node
)


def create_integration_risk_orchestrator(config: IntegrationRiskManagementOrchestratorConfig = None):
    """Create and return the Integration & Risk Management Orchestrator workflow"""

    if config is None:
        config = IntegrationRiskManagementOrchestratorConfig()

    # Create workflow graph
    workflow = StateGraph(IntegrationRiskManagementOrchestratorState)

    # Add all nodes
    workflow.add_node("goal", goal_node)
    workflow.add_node("planning", planning_node)
    workflow.add_node("data_loading", lambda state: data_loading_node(state, config))
    workflow.add_node("integration_health", lambda state: integration_health_node(state, config))
    workflow.add_node("risk_analysis", lambda state: risk_analysis_node(state, config))
    workflow.add_node("workflow_kpi_analysis", lambda state: workflow_kpi_analysis_node(state, config))
    workflow.add_node("value_analysis", lambda state: value_analysis_node(state, config))
    workflow.add_node("prioritization", lambda state: prioritization_node(state, config))
    workflow.add_node("report_generation", lambda state: report_generation_node(state, config))

    # Set entry point
    workflow.set_entry_point("goal")

    # Linear flow
    workflow.add_edge("goal", "planning")
    workflow.add_edge("planning", "data_loading")

    # Parallel analysis nodes (all depend on data_loading)
    workflow.add_edge("data_loading", "integration_health")
    workflow.add_edge("data_loading", "risk_analysis")
    workflow.add_edge("data_loading", "workflow_kpi_analysis")
    workflow.add_edge("data_loading", "value_analysis")

    # Prioritization depends on all analysis nodes completing
    # LangGraph will wait for all incoming edges before executing prioritization
    workflow.add_edge("integration_health", "prioritization")
    workflow.add_edge("risk_analysis", "prioritization")
    workflow.add_edge("workflow_kpi_analysis", "prioritization")
    workflow.add_edge("value_analysis", "prioritization")

    # Generate report after prioritization
    workflow.add_edge("prioritization", "report_generation")
    workflow.add_edge("report_generation", END)

    return workflow.compile()


def run_orchestrator(
    agent_id: str = None,
    config: IntegrationRiskManagementOrchestratorConfig = None
) -> IntegrationRiskManagementOrchestratorState:
    """Run the orchestrator with given parameters"""

    if config is None:
        config = IntegrationRiskManagementOrchestratorConfig()

    orchestrator = create_integration_risk_orchestrator(config)

    initial_state: IntegrationRiskManagementOrchestratorState = {
        "agent_id": agent_id,
        "errors": []
    }

    result = orchestrator.invoke(initial_state)
    return result
````
