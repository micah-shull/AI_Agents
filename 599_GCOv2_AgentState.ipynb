````python
# ----------------------------------------------------------------------------
# Governance & Compliance Orchestrator Agent â€” V2 (MVPv2)
# ----------------------------------------------------------------------------
# V2 adds: multiple agents, historical runs, evolving risk, bias/drift trends,
# human overrides, regulatory categories, root-cause tracing, executive
# triggers, portfolio-level governance. From "auditor bot" â†’ "enterprise AI
# risk command center."

class GovernanceComplianceOrchestratorStateV2(TypedDict, total=False):
    """State for Governance & Compliance Orchestrator Agent (v2 â€” enterprise risk command center)"""

    # Input fields
    agent_name: Optional[str]              # Specific agent to analyze (None = analyze all)
    time_window_days: Optional[int]        # Time window for analysis (None = use default)

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Data Ingestion (V1 + expanded)
    agent_action_logs: List[Dict[str, Any]]   # All agent events (batches + day_02 with cost_usd, latency_ms, environment, etc.)
    policy_rules: List[Dict[str, Any]]
    bias_signals: List[Dict[str, Any]]
    drift_signals: List[Dict[str, Any]]

    # V2: Human escalation / overrides
    policy_enforcement_events: List[Dict[str, Any]]  # override_reason, approver_role, time_to_resolution, remediation_action
    governance_cases: List[Dict[str, Any]]          # case_id, triggered_by, risk_score, remediation_steps, financial_exposure_usd

    # V2: Historical runs / snapshots (trend analysis)
    bias_signals_history: List[Dict[str, Any]]       # Per-run bias snapshots (run_id, trend_direction, regulatory_frameworks)
    drift_signals_history: List[Dict[str, Any]]     # Per-run drift snapshots
    governance_portfolio_summary: List[Dict[str, Any]]  # Per-run: overall_risk_score, risk_trend, open_cases, executive_summary

    # Lookups
    policy_lookup: Dict[str, Dict[str, Any]]
    events_lookup: Dict[str, Dict[str, Any]]
    cases_lookup: Dict[str, Dict[str, Any]]          # case_id -> case

    # Policy Evaluation
    policy_evaluations: List[Dict[str, Any]]
    compliance_events: List[Dict[str, Any]]

    # Risk Assessment (V1 + V2 evolving risk)
    risk_scores: Dict[str, Any]                     # agent_scores + overall_risk_score
    risk_trends: Dict[str, Any]                     # V2: trend_direction, delta vs prior run, early_warning_flags

    # V2: Portfolio-level governance
    portfolio_rollup: Dict[str, Any]                # high_risk_agents, open_cases, regulatory_exposure_usd, top_root_causes
    executive_triggers: List[Dict[str, Any]]        # Conditions that warrant exec attention (risk spike, critical case, etc.)
    root_cause_summary: List[Dict[str, Any]]        # Top root causes with agent/case linkage

    # Prioritization
    prioritized_issues: List[Dict[str, Any]]

    # Summary
    summary: Dict[str, Any]

    # Output
    audit_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class GovernanceComplianceOrchestratorConfigV2:
    """Configuration for Governance & Compliance Orchestrator Agent (v2)"""

    llm_model: str = os.getenv("LLM_MODEL", "gpt-4o-mini")
    temperature: float = 0.3

    data_dir: str = "agents/data"
    agent_logs_files: List[str] = field(default_factory=lambda: [
        "agent_action_logs_batch_1.json",
        "agent_action_logs_batch_2.json",
        "agent_action_logs_batch_3.json",
        "agent_action_logs_day_02.json",
    ])
    policy_rules_file: str = "policy_rules.json"
    bias_signals_file: str = "bias_signals.json"
    drift_signals_file: str = "drift_and_degradation_signals.json"
    policy_enforcement_events_file: str = "policy_enforcement_events.json"
    governance_cases_file: str = "governance_cases.json"
    bias_signals_history_file: str = "bias_signals_history.json"
    drift_signals_history_file: str = "drift_signals_history.json"
    governance_portfolio_summary_file: str = "governance_portfolio_summary.json"

    reports_dir: str = "output/governance_compliance_reports"
    default_time_window_days: int = 30

    severity_weights: Dict[str, float] = field(default_factory=lambda: {
        "critical": 1.0,
        "high": 0.75,
        "medium": 0.50,
        "low": 0.25
    })
    bias_delta_threshold: float = 0.20
    bias_risk_levels: Dict[str, float] = field(default_factory=lambda: {
        "critical": 0.50,
        "high": 0.30,
        "medium": 0.20,
        "low": 0.10
    })
    drift_threshold_multiplier: float = 1.2
    priority_scoring_weights: Dict[str, float] = field(default_factory=lambda: {
        "severity": 0.40,
        "urgency": 0.30,
        "impact": 0.20,
        "frequency": 0.10
    })

    # V2: Executive trigger thresholds
    executive_risk_threshold: float = 75.0       # Overall risk score above this â†’ executive trigger
    executive_open_cases_threshold: int = 2     # Open cases >= this â†’ executive trigger
    executive_regulatory_exposure_threshold_usd: float = 500_000.0

    enable_prioritization: bool = True
    enable_reporting: bool = True
    enable_llm_explanations: bool = False
    llm_explanation_max_events: int = 5
````
