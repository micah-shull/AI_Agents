```python
"""
GCO v2 orchestrator nodes: goal, plan, load_data, policy_eval, risk, trend, portfolio, prioritize, report.
"""

import time
from typing import Any, Dict, List

from config import (
    GovernanceComplianceOrchestratorConfigV2,
    GovernanceComplianceOrchestratorStateV2,
)
from toolshed.prioritization.scoring import calculate_priority_score, normalize_severity

from agents.governance_compliance_orchestrator_v2.orchestrator.utilities.data_loading import (
    load_all_gco_v2_data,
)
from agents.governance_compliance_orchestrator_v2.orchestrator.utilities.policy_evaluation import (
    evaluate_events_against_policies,
)
from agents.governance_compliance_orchestrator_v2.orchestrator.utilities.risk_scoring import (
    compute_risk_scores,
)
from agents.governance_compliance_orchestrator_v2.orchestrator.utilities.trend_analysis import (
    compute_risk_trends,
)
from agents.governance_compliance_orchestrator_v2.orchestrator.utilities.portfolio_rollup import (
    build_portfolio_rollup,
    compute_executive_triggers,
    build_root_cause_summary,
)
from agents.governance_compliance_orchestrator_v2.orchestrator.utilities.reporting import (
    generate_audit_report_v2,
)


def goal_node(state: GovernanceComplianceOrchestratorStateV2) -> Dict[str, Any]:
    """Set fixed goal for GCO v2 run."""
    return {
        "goal": {
            "objective": "Governance & Compliance portfolio analysis (v2)",
            "scope": "all agents",
            "outputs": ["audit_report", "portfolio_rollup", "executive_triggers"],
        }
    }


def plan_node(state: GovernanceComplianceOrchestratorStateV2) -> Dict[str, Any]:
    """Set execution plan (template-based)."""
    return {
        "plan": [
            {"step": "load_data", "description": "Load all v2 data sources"},
            {"step": "policy_eval", "description": "Evaluate events against policies"},
            {"step": "risk", "description": "Compute risk scores"},
            {"step": "trend", "description": "Compute risk trends from history"},
            {"step": "portfolio", "description": "Build portfolio rollup and executive triggers"},
            {"step": "prioritize", "description": "Prioritize compliance issues"},
            {"step": "report", "description": "Generate audit report"},
        ]
    }


def data_loading_node(
    state: GovernanceComplianceOrchestratorStateV2,
    config: GovernanceComplianceOrchestratorConfigV2,
    project_root: str,
) -> Dict[str, Any]:
    """Load all GCO v2 data and build lookups."""
    start = time.perf_counter()
    errors = list(state.get("errors") or [])

    try:
        data = load_all_gco_v2_data(
            data_dir=config.data_dir,
            agent_logs_files=config.agent_logs_files,
            policy_rules_file=config.policy_rules_file,
            bias_signals_file=config.bias_signals_file,
            drift_signals_file=config.drift_signals_file,
            policy_enforcement_events_file=config.policy_enforcement_events_file,
            governance_cases_file=config.governance_cases_file,
            bias_signals_history_file=config.bias_signals_history_file,
            drift_signals_history_file=config.drift_signals_history_file,
            governance_portfolio_summary_file=config.governance_portfolio_summary_file,
            project_root=project_root,
        )
    except Exception as e:
        errors.append(f"data_loading: {e}")
        return {"errors": errors}

    agent_action_logs = data.get("agent_action_logs") or []
    policy_rules = data.get("policy_rules") or []
    if isinstance(policy_rules, list) and len(policy_rules) and isinstance(policy_rules[0], dict):
        policy_lookup = {p.get("policy_id", ""): p for p in policy_rules if p.get("policy_id")}
    else:
        policy_lookup = {}
    events_lookup = {e.get("event_id", ""): e for e in agent_action_logs if e.get("event_id")}
    governance_cases = data.get("governance_cases") or []
    cases_lookup = {c.get("case_id", ""): c for c in governance_cases if isinstance(c, dict) and c.get("case_id")}

    bias_signals = data.get("bias_signals") or []
    drift_signals = data.get("drift_signals") or []

    return {
        "agent_action_logs": agent_action_logs,
        "policy_rules": policy_rules,
        "policy_lookup": policy_lookup,
        "events_lookup": events_lookup,
        "cases_lookup": cases_lookup,
        "bias_signals": bias_signals,
        "drift_signals": drift_signals,
        "policy_enforcement_events": data.get("policy_enforcement_events") or [],
        "governance_cases": governance_cases,
        "bias_signals_history": data.get("bias_signals_history") or [],
        "drift_signals_history": data.get("drift_signals_history") or [],
        "governance_portfolio_summary": data.get("governance_portfolio_summary") or [],
        "errors": errors,
        "processing_time": state.get("processing_time", 0) + (time.perf_counter() - start),
    }


def policy_eval_node(
    state: GovernanceComplianceOrchestratorStateV2,
) -> Dict[str, Any]:
    """Evaluate agent events against policy rules; produce policy_evaluations and compliance_events."""
    start = time.perf_counter()
    logs = state.get("agent_action_logs") or []
    policy_rules = state.get("policy_rules") or []
    policy_evaluations, compliance_events = evaluate_events_against_policies(logs, policy_rules)
    # Enrich compliance_events with agent_name from events_lookup
    events_lookup = state.get("events_lookup") or {}
    for ev in compliance_events:
        ev["agent_name"] = (events_lookup.get(ev.get("event_id") or "") or {}).get("agent_name")

    return {
        "policy_evaluations": policy_evaluations,
        "compliance_events": compliance_events,
        "processing_time": (state.get("processing_time") or 0) + (time.perf_counter() - start),
    }


def risk_node(
    state: GovernanceComplianceOrchestratorStateV2,
    config: GovernanceComplianceOrchestratorConfigV2,
) -> Dict[str, Any]:
    """Compute risk_scores (agent_scores + overall_risk_score) using events_lookup."""
    start = time.perf_counter()
    risk_scores = compute_risk_scores(
        compliance_events=state.get("compliance_events") or [],
        bias_signals=state.get("bias_signals") or [],
        drift_signals=state.get("drift_signals") or [],
        severity_weights=config.severity_weights,
        bias_risk_levels=config.bias_risk_levels,
        events_lookup=state.get("events_lookup"),
    )
    return {
        "risk_scores": risk_scores,
        "processing_time": (state.get("processing_time") or 0) + (time.perf_counter() - start),
    }


def trend_node(state: GovernanceComplianceOrchestratorStateV2) -> Dict[str, Any]:
    """Compute risk_trends from portfolio summary and bias/drift history."""
    start = time.perf_counter()
    risk_trends = compute_risk_trends(
        governance_portfolio_summary=state.get("governance_portfolio_summary") or [],
        bias_signals_history=state.get("bias_signals_history") or [],
        drift_signals_history=state.get("drift_signals_history") or [],
    )
    return {
        "risk_trends": risk_trends,
        "processing_time": (state.get("processing_time") or 0) + (time.perf_counter() - start),
    }


def portfolio_node(
    state: GovernanceComplianceOrchestratorStateV2,
    config: GovernanceComplianceOrchestratorConfigV2,
) -> Dict[str, Any]:
    """Build portfolio_rollup, executive_triggers, root_cause_summary."""
    start = time.perf_counter()
    portfolio_rollup = build_portfolio_rollup(
        governance_cases=state.get("governance_cases") or [],
        governance_portfolio_summary=state.get("governance_portfolio_summary") or [],
        risk_scores=state.get("risk_scores") or {},
    )
    executive_triggers = compute_executive_triggers(
        portfolio_rollup=portfolio_rollup,
        risk_scores=state.get("risk_scores") or {},
        executive_risk_threshold=config.executive_risk_threshold,
        executive_open_cases_threshold=config.executive_open_cases_threshold,
        executive_regulatory_exposure_threshold_usd=config.executive_regulatory_exposure_threshold_usd,
    )
    root_cause_summary = build_root_cause_summary(
        governance_cases=state.get("governance_cases") or [],
        portfolio_rollup=portfolio_rollup,
    )
    return {
        "portfolio_rollup": portfolio_rollup,
        "executive_triggers": executive_triggers,
        "root_cause_summary": root_cause_summary,
        "processing_time": (state.get("processing_time") or 0) + (time.perf_counter() - start),
    }


def prioritize_node(
    state: GovernanceComplianceOrchestratorStateV2,
    config: GovernanceComplianceOrchestratorConfigV2,
) -> Dict[str, Any]:
    """Prioritize compliance issues using toolshed weights; produce prioritized_issues."""
    start = time.perf_counter()
    compliance_events = state.get("compliance_events") or []
    events_lookup = state.get("events_lookup") or {}
    weights = config.priority_scoring_weights

    issues: List[Dict[str, Any]] = []
    for ev in compliance_events:
        sev = (ev.get("severity") or "medium").lower()
        priority_score = (
            weights.get("severity", 0.4) * normalize_severity(sev) / 100.0 * 100
            + weights.get("urgency", 0.3) * (50.0 if sev in ("high", "critical") else 25.0) / 100
            + weights.get("impact", 0.2) * 50.0 / 100
            + weights.get("frequency", 0.1) * 30.0 / 100
        )
        issues.append({
            "compliance_event_id": ev.get("compliance_event_id"),
            "event_id": ev.get("event_id"),
            "agent_name": ev.get("agent_name") or (events_lookup.get(ev.get("event_id") or "") or {}).get("agent_name"),
            "severity": sev,
            "priority_score": round(priority_score, 1),
            "risk_type": ev.get("risk_type"),
            "policy_id": ev.get("policy_id"),
        })

    prioritized_issues = sorted(issues, key=lambda x: (-(x.get("priority_score") or 0), x.get("severity", "")))

    return {
        "prioritized_issues": prioritized_issues,
        "summary": {
            "total_events_analyzed": len(state.get("agent_action_logs") or []),
            "total_violations": len(compliance_events),
            "high_severity_count": sum(1 for e in compliance_events if (e.get("severity") or "").lower() in ("high", "critical")),
            "bias_signals_count": len(state.get("bias_signals") or []),
            "drift_signals_count": len(state.get("drift_signals") or []),
            "agents_affected": list({e.get("agent_name") for e in compliance_events if e.get("agent_name")}),
        },
        "processing_time": (state.get("processing_time") or 0) + (time.perf_counter() - start),
    }


def report_node(
    state: GovernanceComplianceOrchestratorStateV2,
    config: GovernanceComplianceOrchestratorConfigV2,
    project_root: str,
) -> Dict[str, Any]:
    """Generate audit report and optionally write to file."""
    start = time.perf_counter()
    audit_report = generate_audit_report_v2(
        summary=state.get("summary") or {},
        risk_scores=state.get("risk_scores") or {},
        risk_trends=state.get("risk_trends") or {},
        portfolio_rollup=state.get("portfolio_rollup") or {},
        executive_triggers=state.get("executive_triggers") or [],
        root_cause_summary=state.get("root_cause_summary") or [],
        prioritized_issues=state.get("prioritized_issues") or [],
        compliance_events=state.get("compliance_events") or [],
        bias_signals=state.get("bias_signals") or [],
        drift_signals=state.get("drift_signals") or [],
        governance_cases=state.get("governance_cases") or [],
        policy_enforcement_events=state.get("policy_enforcement_events") or [],
    )

    report_file_path = None
    try:
        import os
        out_dir = os.path.join(project_root, config.reports_dir)
        os.makedirs(out_dir, exist_ok=True)
        from datetime import datetime
        fname = f"gco_audit_v2_{datetime.utcnow().strftime('%Y%m%d_%H%M%S')}.md"
        path = os.path.join(out_dir, fname)
        with open(path, "w", encoding="utf-8") as f:
            f.write(audit_report)
        report_file_path = path
    except Exception as e:
        (state.get("errors") or []).append(f"report_write: {e}")

    return {
        "audit_report": audit_report,
        "report_file_path": report_file_path,
        "processing_time": (state.get("processing_time") or 0) + (time.perf_counter() - start),
    }
```
