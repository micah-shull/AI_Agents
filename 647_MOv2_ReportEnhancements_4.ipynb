````python
def build_experiment_snapshot_section(
    experiments: List[Dict[str, Any]],
    performance_metrics: List[Dict[str, Any]],
    experiments_lookup: Dict[str, Dict[str, Any]],
    metrics_by_experiment: Dict[str, List[Dict[str, Any]]],
) -> str:
    """Short experiments section: # running, # completed, average lift (variant vs control)."""
    if not experiments:
        return "## Experiments\n\nNo experiments in this period.\n\n"
    running = [e for e in experiments if (e.get("status") or "").lower() == "running"]
    completed = [e for e in experiments if (e.get("status") or "").lower() == "completed"]
    lifts: List[float] = []
    for exp in experiments:
        exp_id = exp.get("experiment_id", "")
        metrics = metrics_by_experiment.get(exp_id, [])
        if len(metrics) < 2:
            continue
        metric_name = exp.get("metric", "")
        control_asset = exp.get("control_asset", "")
        variant_asset = exp.get("variant_asset", "")
        if not metric_name or not control_asset or not variant_asset:
            continue
        by_asset: Dict[str, List[float]] = defaultdict(list)
        for m in metrics:
            aid = m.get("asset_id")
            val = m.get(metric_name)
            if aid is not None and val is not None:
                by_asset[aid].append(float(val))
        c_vals = by_asset.get(control_asset, [])
        v_vals = by_asset.get(variant_asset, [])
        if not c_vals or not v_vals:
            continue
        c_avg = sum(c_vals) / len(c_vals)
        v_avg = sum(v_vals) / len(v_vals)
        if c_avg != 0:
            lift_pct = (v_avg - c_avg) / c_avg * 100
            lifts.append(lift_pct)
    avg_lift_str = f"{sum(lifts) / len(lifts):+.1f}%" if lifts else "â€”"
    lines = [
        "## Experiments",
        "",
        f"**Running:** {len(running)} | **Completed:** {len(completed)} | **Average lift (variant vs control):** {avg_lift_str}",
        "",
    ]
    return "\n".join(lines)


def build_portfolio_momentum_line(segment_rollups: List[Dict[str, Any]]) -> str:
    """One line: Segment momentum â€” N improving, M declining, K stable (from trend_vs_prior_window)."""
    if not segment_rollups:
        return ""
    improving = sum(1 for r in segment_rollups if (r.get("trend_vs_prior_window") or 0) > 0)
    declining = sum(1 for r in segment_rollups if (r.get("trend_vs_prior_window") or 0) < 0)
    stable = len(segment_rollups) - improving - declining
    return f"**Segment momentum:** {improving} improving, {declining} declining, {stable} stable."
````
