````python
"""Executive markdown report for WDO v3."""

from __future__ import annotations

from typing import Any, Dict, List


def _fmt_pct(x: float, decimals: int = 0) -> str:
    return f"{x * 100:.{decimals}f}%" if abs(x) <= 1.0 else f"{x:.{decimals}f}%"


def _verdict_line(metrics: Dict[str, Any]) -> str:
    if metrics.get("status") != "ok":
        return "**Verdict:** Insufficient snapshot history to compute organization trajectory; load at least two periods per department."
    traj = metrics.get("trajectory_status", "unknown")
    th = metrics.get("transformation_health", "unknown")
    st = metrics.get("workforce_stability_signal", "unknown")
    gap = metrics.get("adaptation_gap_index")
    gap_s = f"{gap:+.3f}" if isinstance(gap, (int, float)) else "n/a"
    return (
        f"**Verdict:** Workforce trajectory is **{traj.replace('_', ' ')}**; "
        f"transformation health **{th.replace('_', ' ')}**; "
        f"stability signal **{st.replace('_', ' ')}** "
        f"(adaptation gap index {gap_s}: automation pressure minus reskilling velocity)."
    )


def build_wdo_v3_executive_report_md(
    metrics: Dict[str, Any],
    department_analyses: List[Dict[str, Any]],
    scenarios: List[Dict[str, Any]],
    recommendations: List[str],
    data_counts: Dict[str, int],
) -> str:
    lines: List[str] = [
        "# Workforce Development Orchestrator v3 — Executive Brief",
        "",
        _verdict_line(metrics),
        "",
        "## Data traceability",
        "",
        f"Rows loaded: {', '.join(f'{k}={v}' for k, v in sorted(data_counts.items()))}.",
        "",
    ]

    if metrics.get("status") == "ok":
        d = metrics.get("deltas") or {}
        cur = metrics.get("current_period") or {}
        lines.extend(
            [
                "## Headline metrics (org-wide, headcount-weighted latest vs prior period)",
                "",
                "| Metric | Current | Δ |",
                "|--------|---------|---|",
                f"| Readiness score | {cur.get('readiness_score', 0):.1f} | {d.get('readiness_score_delta', 0):+.1f} |",
                f"| Skill gap index | {cur.get('skill_gap_index', 0):.2f} | {d.get('skill_gap_index_delta', 0):+.2f} |",
                f"| Automation exposure | {cur.get('avg_automation_exposure', 0):.2f} | {d.get('avg_automation_exposure_delta', 0):+.2f} |",
                f"| Reskilling progress (share) | {_fmt_pct(float(cur.get('reskilling_progress_pct', 0)), 1)} | "
                f"{_fmt_pct(float(d.get('reskilling_progress_pct_delta', 0)), 1)} (pts) |",
                f"| Attrition risk index | {cur.get('attrition_risk_index', 0):.2f} | {d.get('attrition_risk_index_delta', 0):+.2f} |",
                "",
            ]
        )

        roi = metrics.get("training_roi") or {}
        cov = metrics.get("retraining_coverage_weighted_mean")
        disp = metrics.get("displacement_risk_weighted_mean")
        lines.extend(
            [
                "## Investment signals",
                "",
                f"- **Training ROI (risk-adjusted mean, uplift % × completion prob / $):** "
                f"{roi.get('roi_risk_adjusted_uplift_pct_per_usd_mean') or 0:.6f}",
                f"- **Risk-adjusted uplift per $1k spend (mean):** "
                f"{roi.get('risk_adjusted_uplift_pct_per_1000_usd_mean') or 0:.2f} percentage points",
                f"- **Retraining coverage (headcount-weighted):** "
                f"{_fmt_pct(float(cov), 0) if cov is not None else 'n/a'}",
                f"- **Displacement risk (headcount-weighted):** "
                f"{_fmt_pct(float(disp), 0) if disp is not None else 'n/a'}",
                "",
            ]
        )

    warn = metrics.get("validation_warnings") or []
    if warn:
        lines.extend(["## Validation notes", "", *[f"- {w}" for w in warn], ""])

    lines.extend(["## Department trajectory", ""])
    for row in department_analyses:
        did = row.get("department_id")
        if not row.get("trend_available"):
            lines.append(f"- **{did}:** insufficient history for trend.")
            continue
        lines.append(
            f"- **{did}:** readiness {row.get('readiness_score', 0):.0f} "
            f"({row.get('readiness_delta', 0):+.1f}), "
            f"gaps {row.get('skill_gap_index', 0):.2f} ({row.get('skill_gap_delta', 0):+.2f}), "
            f"trajectory *{row.get('trajectory')}*, "
            f"transformation *{row.get('transformation_health')}*."
        )
    lines.append("")

    drivers = metrics.get("drivers") or {}
    lines.append("## Driver narrative (from snapshot driver library)")
    lines.append("")
    if "by_department" in drivers:
        for did, pack in sorted(drivers["by_department"].items()):
            pos = ", ".join(pack.get("top_positive_drivers") or []) or "—"
            neg = ", ".join(pack.get("top_negative_drivers") or []) or "—"
            lines.append(f"- **{did}:** positives: {pos}; negatives: {neg}.")
    else:
        pos = ", ".join(drivers.get("top_positive_drivers") or []) or "—"
        neg = ", ".join(drivers.get("top_negative_drivers") or []) or "—"
        lines.append(f"- Positives: {pos}")
        lines.append(f"- Negatives: {neg}")
    lines.append("")

    lines.append("## Scenario library (stubs for planning)")
    lines.append("")
    for sc in scenarios:
        lines.append(
            f"- **{sc.get('scenario_id')}:** {sc.get('description', '')} "
            f"(horizon {sc.get('time_horizon_months')} mo, budget ${sc.get('training_budget_usd', 0):,})."
        )
    lines.append("")

    lines.append("## Recommended next moves")
    lines.append("")
    for i, r in enumerate(recommendations, 1):
        lines.append(f"{i}. {r}")
    lines.append("")

    lines.append("---")
    lines.append("*Deterministic metrics; trajectory compares latest vs prior department snapshots (headcount-weighted at org level).*")
    return "\n".join(lines)
````
