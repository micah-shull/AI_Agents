# Purpose

This document defines the **data generation standard** for the Sales Enablement Orchestrator v3.

The goal is not to create large datasets.

The goal is:

> **Generate minimal, structured, time-aware data that enables deterministic trend analysis, risk velocity detection, and executive-level insight**

This dataset must support:

* deal trajectory analysis
* pipeline trend evaluation
* rep performance monitoring
* risk velocity detection

---

# 🧭 1. Time Standard (Global)

All v3 datasets must follow a consistent time structure.

```python
time_granularity = "monthly"
period_count = 6
period_range = ["2025-06", "2025-11"]
```

# 📊 3. Dataset Specifications

---

## 3.1 deal_history.json (REQUIRED)

### Purpose:

Enable:

* probability trend
* stage progression
* deal stagnation
* forecast drift

---

### Scope:

* 10–12 deals (not all deals)
* include:

  * high-value deals
  * active deals
  * 1–2 won/lost deals for contrast

---

### Structure:

```json
{
  "deal_id": "D-001",
  "monthly_history": [
    {
      "period": "2025-06",
      "stage": "Discovery",
      "probability": 0.35,
      "deal_value_usd": 75000
    }
  ]
}
```

---

### Required Fields:

* period
* stage
* probability
* deal_value_usd

---

### Data Rules:

* stage progression must be logical
* probability must:

  * increase for healthy deals
  * decrease for at-risk deals
* no random jumps > ±10% unless modeling instability

---

### Required Patterns:

Include at least:

* 📈 improving deal
* 📉 deteriorating deal
* ➖ stalled deal

---

---

## 3.2 rep_performance_history.json

### Purpose:

Enable:

* rep performance trend

### Scope:

Each rep must represent:

* stable performer

* volatile performer

---

### Required Patterns:

Include:

* 📈 growth period

# ⚠️ 4. Data Consistency Rules

All datasets must follow:

---

## 4.1 Logical Consistency

* win_rate ↓ → weighted_pipeline should ↓
* pipeline ↓ → deal_count or deal_value should reflect it
* probability trends must align with stage movement

---

## 4.2 Controlled Variability

* introduce noise ±2–5%
* avoid random spikes unless intentional

---

## 4.3 Realistic Behavior

* deals rarely jump stages backward
* high-value deals move slower
* negotiation stages often stall

---

# 📈 5. v3 Trend Requirements

Each dataset must support:

* ≥ 3 historical periods (we use 6)
* trend direction calculation
* volatility detection
* delta computation

---

```python
risk_delta = current_risk - previous_risk
```

And classification:

* increasing
* stable
* decreasing

# 🚨 7. Executive Trigger Support

* high-value deal deterioration
* deal stagnation
* declining rep performance
* pipeline contraction
* falling win rate

* [ ] no logical inconsistencies
