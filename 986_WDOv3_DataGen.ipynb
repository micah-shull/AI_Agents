# 📘 **WDO v3 — Data Generation Plan**

### *Minimal, Deterministic, Trend-Ready Workforce Data Design*

---

## **Purpose**

This document defines how data is generated for the Workforce Development Orchestrator v3.

The goal is not to simulate reality in full detail.

The goal is to:

* support **deterministic metrics**
* enable **trend analysis**
* preserve **causal explainability**
* avoid **data bloat**

---

## 🧭 **Design Philosophy**

### 1. **Metrics First**

All data must exist to support a defined metric.

If a field does not map to a metric → it should not exist.

---

### 2. **Snapshots Over Logs**

We do not track:

* individual events
* task-level time series
* daily changes

We track:

> **pre-aggregated snapshots at meaningful intervals**

---

### 3. **Minimal Time Depth**

Use:

* **2–3 time points per entity**

This is sufficient to compute:

* deltas
* trends
* trajectory classification

---

### 4. **Deterministic Data**

All values must be:

* explainable
* reproducible
* logically consistent

---

### 5. **Narrative-Ready**

Data must support executive storytelling:

> “What changed, why it changed, and what to do next”

---

# 🏗 **Core Data Structure**

The WDO v3 dataset is composed of **three layers**:

---


---

## 🧠 Snapshot Design Rules

### 1. Time Points

Use:

* Q1
* Q2
* (Optional) Q3

👉 Do NOT exceed 3–4 snapshots

---

### 2. Controlled Variation

Each metric must:

* change meaningfully
* reflect logical relationships

---

### 3. Required Patterns

Each department should demonstrate **at least one pattern**:

| Pattern   | Example                        |
| --------- | ------------------------------ |
| Improving | readiness ↑, skill gaps ↓      |
| At Risk   | automation ↑, reskilling ↓     |
| Mixed     | some improving, some worsening |

---

## 📊 Example Pattern (GOOD)

```json id="f7v5b0"
Q1:
  readiness: 63
  skill_gap: 0.42
  reskilling: 0.31

Q2:
  readiness: 71
  skill_gap: 0.35
  reskilling: 0.48
```

👉 This supports:

* positive trajectory
* causal interpretation

---

---

## **Layer 3 — Driver Attribution (NEW, MINIMAL)**

### Dataset: `workforce_snapshot_drivers.json`

---

### Purpose

Explain:

> **Why metrics changed**

---

### Structure

```json id="m4z2lp"
{
  "run_id": "string",
  "department_id": "string",
  "top_positive_drivers": [string],
  "top_negative_drivers": [string]
}
```

---

### Driver Categories (Standardized)

Use ONLY these:

#### Positive Drivers

* increased_learning_completion
* reduced_skill_gaps
* improved_role_alignment
* effective_training_investment

#### Negative Drivers

* rising_automation_exposure
* insufficient_reskilling
* widening_skill_gaps
* attrition_pressure

---

### Design Rules

* Max **2–3 drivers per category**
* Must align with metric deltas
* Must be deterministic (not random)

---

# ⚙️ **Data Generation Rules**

---

## Rule 1 — Logical Consistency

If:

* reskilling ↑
  → skill gaps should ↓

If not:

* must be explained via drivers

---

## Rule 2 — Bounded Changes

Avoid unrealistic jumps:

* readiness: ±5–10 max per period
* skill gap: ±0.05–0.10
* reskilling: ±0.10–0.20

---

## Rule 3 — Cross-Metric Relationships

Maintain relationships:

| Metric     | Dependency                  |
| ---------- | --------------------------- |
| readiness  | driven by gaps + reskilling |
| skill gaps | impacted by training        |
| attrition  | rises with instability      |

---

## Rule 4 — Department Differentiation

Each department must behave differently:

* Sales → higher automation pressure
* HR → higher fairness/regulatory sensitivity
* Operations → balanced
* Analytics → lower automation risk

---

---

---

Ensure:

* trends make sense

---

## Step 5 — Assign Drivers

Map:

* metric deltas → driver labels

* ❌ excessive snapshots
