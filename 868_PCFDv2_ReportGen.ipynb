#Executive Report Generation

### Product–Customer Fit Discovery Orchestrator v2 (PCFDv2)

The discovery engine identifies signals across customers, products, features, and trends.

However, raw insights alone are not enough for business decision-makers.

Executives need information that is:

* concise
* structured
* traceable

The purpose of the **report generation module** is to transform the orchestrator’s analytical output into a **clear executive briefing**.

This component converts structured discovery signals into a **decision-ready report** that leadership teams can quickly interpret.


````python
"""
Executive-grade PCFD v2 report: verdict, so-what, one-view, discoveries, traceability.
"""

from typing import Any, Dict, List, Optional


def _table_rows(headers: List[str], rows: List[Dict[str, Any]], keys: List[str]) -> str:
    """Render a markdown table from headers and row dicts with given keys."""
    line = "| " + " | ".join(headers) + " |"
    sep = "| " + " | ".join(["---"] * len(headers)) + " |"
    out = [line, sep]
    for r in rows:
        out.append("| " + " | ".join(str(r.get(k, "")) for k in keys) + " |")
    return "\n".join(out)


def build_pcfd_report(
    goal: Dict[str, Any],
    loader_counts: Dict[str, int],
    data_snapshot_loaded_at: Optional[str],
    validation_warnings: List[str],
    customer_intel: Dict[str, Any],
    product_intel: Dict[str, Any],
    feature_intel: Dict[str, Any],
    trend_intel: Dict[str, Any],
    strategic_signals: Dict[str, Any],
) -> str:
    """Build full markdown report: one-view verdict, so-what, discoveries, traceability."""
    # Traceability
    counts_str = ", ".join(f"{k}: {v}" for k, v in sorted(loader_counts.items()) if v is not None)
    trace = f"**Data:** {counts_str}"
    if data_snapshot_loaded_at:
        trace += f" · Snapshot: {data_snapshot_loaded_at}"
    if validation_warnings:
        trace += "\n\n**Validation:** " + "; ".join(validation_warnings)

    # One-view verdict
    n_bundles = len(strategic_signals.get("bundle_opportunities", []))
    n_emerging = len(trend_intel.get("emerging_products", []))
    n_declining = len(trend_intel.get("declining_products", []))
    n_high_margin = len(product_intel.get("high_margin_products", []))
    verdict = "Product–Customer Fit Discovery complete."
    so_what_parts = []
    if n_bundles:
        so_what_parts.append(f"{n_bundles} bundle opportunity(ies)")
    if n_emerging:
        so_what_parts.append(f"{n_emerging} emerging product(s)")
    if n_declining:
        so_what_parts.append(f"{n_declining} declining product(s) to watch")
    if n_high_margin:
        so_what_parts.append(f"{n_high_margin} high-margin product(s)")
    so_what = " · ".join(so_what_parts) if so_what_parts else "No strong signals above thresholds; review thresholds or data."

    sections = [
        "# Product–Customer Fit Discovery — Executive Summary",
        "",
        "## One view",
        "",
        f"{verdict} {so_what}",
        "",
        "---",
        "",
        "## Traceability",
        "",
        trace,
        "",
        "---",
        "",
        "## Customer intelligence",
        "",
        "High-value segments (by revenue):",
        "",
    ]

    segs = customer_intel.get("high_value_segments", [])
    if segs:
        sections.append(_table_rows(
            ["Segment", "Customers", "Total revenue", "Revenue %", "Avg CLV", "Avg retention risk"],
            segs,
            ["segment", "customer_count", "total_revenue", "revenue_pct", "avg_clv", "avg_retention_risk"],
        ))
    else:
        sections.append("*No segment breakdown (missing or empty customer_metrics).*")
    sections.extend(["", "---", "", "## Product intelligence", "", "High-margin products:", ""])

    prods = product_intel.get("high_margin_products", [])
    if prods:
        sections.append(_table_rows(
            ["Product ID", "Product type", "Profit margin %", "Usage count"],
            prods,
            ["product_id", "product_type", "profit_margin_pct", "usage_count"],
        ))
    else:
        sections.append("*No high-margin products above threshold.*")
    sections.extend(["", "---", "", "## Feature intelligence", "", "Feature demand surges:", ""])

    surges = feature_intel.get("feature_demand_surges", [])
    if surges:
        sections.append(_table_rows(["Feature", "Usage count", "Requests"], surges, ["feature", "usage_count", "requests"]))
    else:
        sections.append("*No feature demand surges above threshold.*")
    sections.extend(["", "---", "", "## Trend intelligence", "", "Emerging products:", ""])

    emerging = trend_intel.get("emerging_products", [])
    declining = trend_intel.get("declining_products", [])
    fast_seg = trend_intel.get("fast_expanding_segments", [])
    if emerging:
        sections.append(_table_rows(["Product ID", "Growth %"], emerging, ["product_id", "growth_pct"]))
    else:
        sections.append("*None.*")
    sections.extend(["", "Declining products:", ""])
    if declining:
        sections.append(_table_rows(["Product ID", "Growth %"], declining, ["product_id", "growth_pct"]))
    else:
        sections.append("*None.*")
    sections.extend(["", "Fast-expanding segments:", ""])
    if fast_seg:
        sections.append(_table_rows(["Segment", "Growth %"], fast_seg, ["segment", "growth_pct"]))
    else:
        sections.append("*None.*")
    sections.extend(["", "---", "", "## Strategic signals", "", "Bundle / cross-sell opportunities:", ""])

    bundles = strategic_signals.get("bundle_opportunities", [])
    if bundles:
        sections.append(_table_rows(
            ["Bundle", "Product A", "Product B", "Segment", "Customers", "Avg revenue/cust", "Score"],
            bundles,
            ["bundle_id", "product_a", "product_b", "segment", "observed_customers", "avg_revenue_per_customer", "opportunity_score"],
        ))
    else:
        sections.append("*No bundle signals above threshold.*")
    sections.extend(["", "---", "", "## Next steps", ""])
    next_steps = []
    if n_bundles:
        next_steps.append("Prioritize top bundle opportunities for packaging or cross-sell campaigns.")
    if n_emerging:
        next_steps.append("Double down on emerging products (invest in capacity and marketing).")
    if n_declining:
        next_steps.append("Review declining products for repositioning or sunset decisions.")
    if n_high_margin:
        next_steps.append("Push high-margin products in high-value segments.")
    if not next_steps:
        next_steps.append("Review discovery thresholds and data coverage; re-run after data refresh.")
    for i, step in enumerate(next_steps, 1):
        sections.append(f"{i}. {step}")
    sections.append("")

    return "\n".join(sections)
````
