# Intervention Planning Utilities for Customer Journey Orchestrator

````python
"""
Intervention Planning Utilities for Customer Journey Orchestrator

Utilities for generating intervention recommendations based on risk scores,
signals, and customer journey state.
"""

from typing import Dict, Any, List, Optional
from datetime import datetime


def generate_intervention_recommendation(
    customer_id: str,
    customer_data: Dict[str, Any],
    risk_score: Dict[str, Any],
    aggregated_signals: Optional[Dict[str, Any]],
    journey_evaluation: Optional[Dict[str, Any]],
    signals: List[Dict[str, Any]],
    confidence_threshold: float,
    high_value_threshold: float
) -> Optional[Dict[str, Any]]:
    """
    Generate intervention recommendation for a customer.

    Args:
        customer_id: Customer identifier
        customer_data: Customer data dictionary
        risk_score: Risk score dictionary
        aggregated_signals: Aggregated signals for this customer
        journey_evaluation: Journey evaluation for this customer
        signals: List of signals for this customer
        confidence_threshold: Minimum confidence to recommend intervention
        high_value_threshold: Account value threshold for high-value escalation

    Returns:
        Intervention recommendation dictionary or None if no intervention needed
    """
    overall_risk = risk_score.get("overall_risk_score", 0.0)
    risk_tier = risk_score.get("risk_tier", "low")
    urgency = risk_score.get("urgency", "low")

    # Calculate confidence based on risk score and signal strength
    confidence = overall_risk

    # Boost confidence if multiple signals
    if aggregated_signals:
        signal_count = aggregated_signals.get("total_signals", 0)
        if signal_count >= 2:
            confidence = min(1.0, confidence + 0.1)
        if signal_count >= 3:
            confidence = min(1.0, confidence + 0.1)

    # Only recommend if confidence meets threshold
    if confidence < confidence_threshold:
        return None

    # Determine recommended action based on journey stage and signals
    journey_stage = journey_evaluation.get("current_stage", "unknown") if journey_evaluation else "unknown"
    signal_types = [s.get("signal_type") for s in signals] if signals else []

    recommended_action = determine_intervention_action(
        journey_stage,
        signal_types,
        risk_tier,
        urgency
    )

    # Determine if human approval is required
    account_value = customer_data.get("account_value", 0)
    requires_human_approval = (
        risk_tier == "high" or
        urgency == "high" or
        account_value >= high_value_threshold or
        confidence >= 0.75
    )

    # Get signal IDs that triggered this intervention
    triggered_by_signals = [s.get("signal_id") for s in signals if s.get("signal_id")]

    return {
        "customer_id": customer_id,
        "journey_stage": journey_stage,
        "recommended_action": recommended_action,
        "confidence": round(confidence, 2),
        "requires_human_approval": requires_human_approval,
        "triggered_by_signals": triggered_by_signals,
        "risk_score": overall_risk,
        "customer_value": account_value,
        "priority_score": calculate_priority_score(risk_score, account_value),
        "evaluation_latency_ms": 0,  # Will be set by node
        "generated_at": datetime.now().strftime("%Y-%m-%d")
    }


def determine_intervention_action(
    journey_stage: str,
    signal_types: List[str],
    risk_tier: str,
    urgency: str
) -> str:
    """
    Determine the recommended intervention action based on context.

    Args:
        journey_stage: Current journey stage
        signal_types: List of signal types detected
        risk_tier: Risk tier (low/medium/high)
        urgency: Urgency level (low/medium/high)

    Returns:
        Recommended action string
    """
    # Stage-specific actions
    if journey_stage == "onboarding":
        if "failed_onboarding_step" in signal_types:
            return "onboarding_specialist_call"
        elif "low_product_usage" in signal_types:
            return "onboarding_nudge_email"
        elif "support_ticket_spike" in signal_types or "negative_sentiment" in signal_types:
            return "proactive_outreach"
        else:
            return "onboarding_checkin"

    elif journey_stage == "engagement":
        if "usage_drop" in signal_types:
            return "usage_review_checkin"
        elif "inactivity_warning" in signal_types:
            return "reengagement_campaign"
        elif "usage_recovery" in signal_types:
            return "monitor_only"
        else:
            return "engagement_checkin"

    elif journey_stage == "support":
        if "repeat_support_tickets" in signal_types or "support_satisfaction_low" in signal_types:
            return "support_manager_followup"
        elif "negative_sentiment" in signal_types:
            return "escalated_support_review"
        else:
            return "support_followup"

    elif journey_stage == "retention":
        if "renewal_intent" in signal_types:
            return "renewal_health_check"
        else:
            return "retention_checkin"

    # Default action
    return "general_checkin"


def calculate_priority_score(
    risk_score: Dict[str, Any],
    account_value: float
) -> float:
    """
    Calculate priority score for intervention (0-100).

    Higher score = higher priority.
    Combines risk and customer value.

    Args:
        risk_score: Risk score dictionary
        account_value: Customer account value

    Returns:
        Priority score (0-100)
    """
    overall_risk = risk_score.get("overall_risk_score", 0.0)
    urgency = risk_score.get("urgency", "low")

    # Urgency multiplier
    urgency_multipliers = {
        "high": 1.2,
        "medium": 1.0,
        "low": 0.8
    }
    urgency_mult = urgency_multipliers.get(urgency, 1.0)

    # Value component (normalize to 0-1, max around 200k)
    value_component = min(1.0, account_value / 200000.0)

    # Priority = (risk * 0.6 + value * 0.4) * urgency_mult * 100
    priority = (
        overall_risk * 0.6 +
        value_component * 0.4
    ) * urgency_mult * 100

    return round(min(100.0, max(0.0, priority)), 1)


def generate_all_interventions(
    customers_lookup: Dict[str, Dict[str, Any]],
    risk_scores: List[Dict[str, Any]],
    aggregated_signals: List[Dict[str, Any]],
    journey_evaluations: List[Dict[str, Any]],
    signals_lookup: Dict[str, List[Dict[str, Any]]],
    confidence_threshold: float,
    high_value_threshold: float
) -> List[Dict[str, Any]]:
    """
    Generate intervention recommendations for all customers.

    Args:
        customers_lookup: Lookup dictionary for customer data
        risk_scores: List of risk score dictionaries
        aggregated_signals: List of aggregated signal dictionaries
        journey_evaluations: List of journey evaluation dictionaries
        signals_lookup: Lookup dictionary mapping customer_id -> list of signals
        confidence_threshold: Minimum confidence to recommend intervention
        high_value_threshold: Account value threshold for high-value escalation

    Returns:
        List of intervention recommendation dictionaries
    """
    # Build lookups for fast access
    risk_scores_lookup = {rs["customer_id"]: rs for rs in risk_scores}
    aggregated_signals_lookup = {agg["customer_id"]: agg for agg in aggregated_signals}
    evaluations_lookup = {eval["customer_id"]: eval for eval in journey_evaluations}

    interventions = []

    for customer_id, customer_data in customers_lookup.items():
        risk_score = risk_scores_lookup.get(customer_id)
        aggregated_signal = aggregated_signals_lookup.get(customer_id)
        journey_evaluation = evaluations_lookup.get(customer_id)
        signals = signals_lookup.get(customer_id, [])

        if not risk_score:
            continue

        intervention = generate_intervention_recommendation(
            customer_id,
            customer_data,
            risk_score,
            aggregated_signal,
            journey_evaluation,
            signals,
            confidence_threshold,
            high_value_threshold
        )

        if intervention:
            interventions.append(intervention)

    return interventions

````

# Nodes

````python
def intervention_planning_node(
    state: CustomerJourneyOrchestratorState,
    config: CustomerJourneyOrchestratorConfig
) -> Dict[str, Any]:
    """
    Intervention Planning Node: Generate intervention recommendations.

    Generates intervention recommendations based on risk scores, signals,
    and journey evaluations.
    """
    errors = state.get("errors", [])
    customers_lookup = state.get("customers_lookup", {})
    risk_scores = state.get("risk_scores", [])
    aggregated_signals = state.get("aggregated_signals", [])
    journey_evaluations = state.get("journey_evaluations", [])
    signals_lookup = state.get("signals_lookup", {})

    if not customers_lookup or not risk_scores:
        return {
            "errors": errors + ["intervention_planning_node: customers_lookup and risk_scores are required"]
        }

    try:
        start_time = time.time()

        recommended_interventions = generate_all_interventions(
            customers_lookup,
            risk_scores,
            aggregated_signals,
            journey_evaluations,
            signals_lookup,
            config.intervention_confidence_threshold,
            config.high_value_customer_threshold
        )

        # Add intervention IDs and evaluation latency
        evaluation_latency_ms = int((time.time() - start_time) * 1000)
        for i, intervention in enumerate(recommended_interventions, 1):
            if "intervention_id" not in intervention:
                intervention["intervention_id"] = f"I{i:03d}"
            intervention["evaluation_latency_ms"] = evaluation_latency_ms

        return {
            "recommended_interventions": recommended_interventions,
            "errors": errors
        }
    except Exception as e:
        return {
            "errors": errors + [f"intervention_planning_node: {str(e)}"]
        }


````
