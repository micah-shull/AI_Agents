# Evaluation-as-a-Service (EaaS) Orchestrator v3

## Why This Agent Exists

As AI systems become more deeply embedded in business operations, a critical question emerges:

> **How do we know an AI system is actually safe, reliable, and ready for real customers?**

Many AI deployments today rely on informal testing, ad-hoc prompt experimentation, or occasional benchmark runs. These approaches may show whether a model can perform well in isolation, but they rarely answer the question executives care about most:

**Can this system operate reliably in the real world — and will we know if something starts to go wrong?**

The **Evaluation-as-a-Service (EaaS) Orchestrator** was designed to answer that question.

Instead of evaluating AI systems once and hoping for the best, EaaS turns evaluation into a **continuous, structured, and auditable process.**

It simulates real customer journeys, measures how an orchestrator responds, tracks performance over time, and surfaces risk signals before they become operational failures.

The result is an evaluation system that behaves much more like **production monitoring** than a one-time benchmark.

---

# What This Agent Actually Does

At a practical level, the EaaS Orchestrator performs four core functions.

### 1. Simulates Real Customer Scenarios

The system evaluates an orchestration agent against realistic customer situations such as:

* delivery delays
* lost packages
* warehouse issues
* “item not received” claims
* churn-risk support cases

Each scenario includes the expected classification, response path, and outcome.

This allows the evaluator to measure not just whether the system responds, but whether it responds **correctly and responsibly**.

---

### 2. Verifies the Orchestrator’s Decision Logic

The evaluator checks three critical elements of each response:

**Issue Classification**

Did the system correctly identify the problem?

**Resolution Path**

Did it call the correct specialist agents?

**Final Outcome**

Did it produce the correct result for the customer?

These checks ensure that the orchestrator is not simply producing plausible responses — it is **following the correct operational workflow.**

---

### 3. Measures Operational Risk

Accuracy alone is not enough.

The evaluator also tracks signals that matter in real operations:

* high-risk failures
* human review rates
* latency
* policy violations
* scenario severity

Failures involving high-risk situations — such as lost packages or churn-risk customers — receive greater weight than minor issues.

This ensures that the evaluation reflects **business impact**, not just statistical performance.

---

### 4. Tracks System Stability Over Time

One of the most valuable features of EaaS v3 is its ability to evaluate **system stability across releases.**

Instead of looking at one run in isolation, the evaluator tracks:

* performance trends
* regressions
* improvements
* risk signals across runs

For example, it can detect when a seemingly small rule change causes a drop in performance for high-value customers.

This allows teams to catch problems **before customers experience them.**

Most AI agents being developed today focus on **capability**.

But very few systems are built to **evaluate and monitor those agents themselves.**

This creates a system that can be **audited and trusted.**

### Scenario-Based Evaluation

# Purpose

This document defines the **official scoring framework** used by the Evaluation-as-a-Service (EaaS) Orchestrator v3.

The evaluator measures how reliably a customer-support orchestration system:

1. **classifies issues**
2. **routes to correct specialist agents**
3. **produces the correct final outcome**
4. **operates safely and efficiently**
5. **maintains stability across releases**

The goal of the evaluator is not simply to produce accuracy metrics.

The goal is to determine:

> **Is the orchestrator safe, reliable, and production-ready for real customer interactions?**

The evaluator therefore measures:

* functional correctness
* operational risk
* performance
* historical stability

---

# Evaluation Architecture

The evaluator measures system performance at **two levels**:

### Scenario Level

Individual customer journeys are evaluated.

Example:

```
Customer reports lost package
Expected outcome → refund issued
```

The evaluator measures whether the orchestrator:

* classified the issue correctly
* called the correct agents
* produced the correct outcome

---

### Run Level

A **run** represents a full evaluation pass across all scenarios.

Example:

```
RUN_2026_01_24
```

Run-level metrics summarize:

* overall system accuracy
* risk signals
* latency
* stability trends

---

# Scoring Layers

The evaluation score is composed of **four layers**.

```
Functional Accuracy
Risk & Safety
Operational Performance
Historical Stability
```

Each layer produces metrics used in the final score.

---

# Layer 1 — Functional Accuracy

Measures whether the orchestrator performs the correct actions.

This is the **most important layer**.

### Issue Classification Accuracy

```
issue_classification_accuracy =
correct_issue_predictions / total_scenarios
```

Example:

```
Correct classifications: 9
Total scenarios: 10

Accuracy = 0.90
```

---

### Resolution Path Accuracy

Measures whether the correct **specialist agents** were invoked.

Example:

Expected path:

```
shipping_update_agent → apology_message_agent
```

Predicted path:

```
shipping_update_agent → apology_message_agent
```

If the paths match, the scenario passes.

Formula:

```
resolution_path_accuracy =
correct_resolution_paths / total_scenarios
```

---

### Outcome Accuracy

Measures whether the orchestrator produced the correct final outcome.

Example:

Expected:

```
issue_refund_and_notify_customer
```

Predicted:

```
issue_refund_and_notify_customer
```

Formula:

```
outcome_accuracy =
correct_outcomes / total_scenarios
```

Outcome accuracy is considered the **most critical functional metric**.

---

# Layer 2 — Risk & Safety

Even if the orchestrator is accurate, it may still introduce risk.

This layer measures safety signals.

---

### High-Risk Failures

A failure is considered **high risk** if it involves:

* lost package mishandling
* incorrect refund decisions
* unresolved churn-risk scenarios
* escalation failures

Metric:

```
high_risk_failures
```

Thresholds:

| Failures | Risk Level |
| -------- | ---------- |
| 0        | Healthy    |
| 1-2      | Moderate   |
| 3+       | Critical   |

---

### Human Review Rate

Measures how often the system cannot resolve issues autonomously.

```
human_review_rate =
human_review_cases / total_scenarios
```

High review rates indicate:

* insufficient automation
* rule conflicts
* uncertain classifications

---

### Hallucination Rate (optional)

Measures cases where the orchestrator produces unsupported conclusions.

```
hallucination_rate =
hallucination_cases / total_scenarios
```

This metric is expected to be **near zero** for deterministic orchestrators.

---

# Layer 3 — Operational Performance

Measures how efficiently the orchestrator operates.

---

### Average Latency

```
avg_latency_ms
```

Target:

```
< 900 ms
```

---

### P95 Latency

```
p95_latency_ms
```

Target:

```
< 1300 ms
```

Latency matters because customer support systems must respond quickly.

---

# Layer 4 — Historical Stability (v3 Feature)

EaaS v3 introduces **trend analysis**.

The evaluator tracks performance across multiple runs.

---

### Pass Rate Trend

```
pass_rate =
scenarios_passed / total_scenarios
```

Tracked across historical runs.

Example:

```
RUN_2025_12_20 → 0.75
RUN_2025_12_27 → 0.82
RUN_2026_01_03 → 0.90
```

Trend analysis detects:

* steady improvement
* regressions
* instability.

---

### Regression Detection

A regression is flagged when:

```
current_pass_rate < previous_pass_rate - 0.05
```

Example:

```
Previous run: 0.90
Current run: 0.80
Regression flag triggered
```

---

### Improvement Detection

An improvement is flagged when:

```
current_pass_rate > previous_pass_rate + 0.05
```

---

### Risk Velocity

Measures how quickly risk signals are changing.

Example signals:

* increase in high-risk failures
* increase in human review rate
* latency degradation

Risk velocity helps detect emerging system instability.

---

# Final Evaluation Score

The evaluator produces a **single headline score**.

This score summarizes overall system health.

Formula:

```
evaluation_score =
0.40 * outcome_accuracy
+0.30 * resolution_path_accuracy
+0.20 * issue_classification_accuracy
+0.10 * (1 - human_review_rate)
```

Example:

```
Outcome accuracy = 0.92
Resolution path accuracy = 0.91
Issue classification accuracy = 0.94
Human review rate = 0.15
```

Evaluation score:

```
0.92
```

---

# Risk Flags

Certain conditions trigger **executive risk alerts**.

### Critical Risk

Triggered if:

```
high_risk_failures >= 3
```

or

```
evaluation_score < 0.75
```

---

### Regression Alert

Triggered if:

```
pass_rate drops > 5% from previous run
```

---

### Operational Alert

Triggered if:

```
avg_latency_ms > 1000
```

Each evaluation run produces:

These metrics feed the **EaaS executive report**.

## Recommended structure

```python
# agents/eaas_v3/scoring.py

from __future__ import annotations

from dataclasses import dataclass
from typing import Any, Dict, List, Optional


@dataclass
class ScenarioScore:
    scenario_id: str
    run_id: str

    issue_classification_correct: bool
    resolution_path_correct: bool
    outcome_correct: bool

    latency_ms: float
    human_review_required: bool
    hallucination_flag: bool = False
    policy_violation_flag: bool = False

    severity_weight: float = 1.0
    business_risk: str = "low"


@dataclass
class RunMetrics:
    run_id: str

    total_scenarios: int
    scenarios_passed: int

    overall_pass_rate: float
    issue_classification_accuracy: float
    resolution_path_accuracy: float
    outcome_accuracy: float

    high_risk_failures: int
    human_review_rate: float

    avg_latency_ms: float
    p95_latency_ms: float

    hallucination_rate: float
    policy_violation_rate: float

    weighted_failure_rate: float
    evaluation_score: float

    regression_flags: List[str]
    improvement_flags: List[str]
    trigger_flags: List[str]
```

## Core helper functions

These are the building blocks.

### 1. Path comparison

For MVP, I recommend exact ordered match.

```python
from typing import Sequence


def paths_match(expected_path: Sequence[str], predicted_path: Sequence[str]) -> bool:
    """Return True when the predicted resolution path exactly matches expected."""
    return list(expected_path) == list(predicted_path)
```

Later, you could add partial credit, but exact match is better for v3 MVP.

---

### 2. Percentile helper

```python
import math


def percentile(values: List[float], pct: float) -> float:
    """Compute a simple percentile from a list of numeric values."""
    if not values:
        return 0.0

    sorted_vals = sorted(values)
    k = (len(sorted_vals) - 1) * pct
    floor_idx = math.floor(k)
    ceil_idx = math.ceil(k)

    if floor_idx == ceil_idx:
        return float(sorted_vals[int(k)])

    floor_val = sorted_vals[floor_idx]
    ceil_val = sorted_vals[ceil_idx]
    return float(floor_val + (ceil_val - floor_val) * (k - floor_idx))
```

## Scenario-level scoring

This is where each scenario gets evaluated.

```python
def score_scenario(
    *,
    run_id: str,
    scenario_id: str,
    expected_issue_type: str,
    predicted_issue_type: str,
    expected_resolution_path: List[str],
    predicted_resolution_path: List[str],
    expected_outcome: str,
    predicted_outcome: str,
    latency_ms: float,
    human_review_required: bool,
    severity_weight: float = 1.0,
    business_risk: str = "low",
    hallucination_flag: bool = False,
    policy_violation_flag: bool = False,
) -> ScenarioScore:
    """Score one evaluation scenario."""

    issue_classification_correct = predicted_issue_type == expected_issue_type
    resolution_path_correct = paths_match(expected_resolution_path, predicted_resolution_path)
    outcome_correct = predicted_outcome == expected_outcome

    return ScenarioScore(
        scenario_id=scenario_id,
        run_id=run_id,
        issue_classification_correct=issue_classification_correct,
        resolution_path_correct=resolution_path_correct,
        outcome_correct=outcome_correct,
        latency_ms=latency_ms,
        human_review_required=human_review_required,
        hallucination_flag=hallucination_flag,
        policy_violation_flag=policy_violation_flag,
        severity_weight=severity_weight,
        business_risk=business_risk,
    )
```

## High-risk failure logic

This is important because your evaluator is not just measuring correctness. It is measuring **operational risk**.

```python
def is_high_risk_failure(score: ScenarioScore) -> bool:
    """
    Return True when a failure occurs on a high-risk scenario.
    Conservative logic for MVP.
    """
    scenario_failed = not (
        score.issue_classification_correct
        and score.resolution_path_correct
        and score.outcome_correct
    )

    if not scenario_failed:
        return False

    return score.business_risk == "high" or score.severity_weight >= 4.0
```

That gives you a clean, deterministic bridge between scenario severity metadata and risk reporting.

## Run-level rollup

This is the heart of the system.

```python
def calculate_weighted_failure_rate(scores: List[ScenarioScore]) -> float:
    """Weighted failure rate using scenario severity weights."""
    if not scores:
        return 0.0

    total_weight = sum(s.severity_weight for s in scores)
    if total_weight == 0:
        return 0.0

    failed_weight = 0.0
    for s in scores:
        scenario_passed = (
            s.issue_classification_correct
            and s.resolution_path_correct
            and s.outcome_correct
        )
        if not scenario_passed:
            failed_weight += s.severity_weight

    return failed_weight / total_weight
```

```python
def calculate_evaluation_score(
    *,
    outcome_accuracy: float,
    resolution_path_accuracy: float,
    issue_classification_accuracy: float,
    human_review_rate: float,
) -> float:
    """
    Official v3 headline score from SCORING_EAASv3.md.
    """
    score = (
        0.40 * outcome_accuracy
        + 0.30 * resolution_path_accuracy
        + 0.20 * issue_classification_accuracy
        + 0.10 * (1.0 - human_review_rate)
    )
    return round(score, 4)
```

```python
def rollup_run_metrics(run_id: str, scenario_scores: List[ScenarioScore]) -> RunMetrics:
    """Aggregate scenario-level scores into run-level metrics."""
    total = len(scenario_scores)

    if total == 0:
        return RunMetrics(
            run_id=run_id,
            total_scenarios=0,
            scenarios_passed=0,
            overall_pass_rate=0.0,
            issue_classification_accuracy=0.0,
            resolution_path_accuracy=0.0,
            outcome_accuracy=0.0,
            high_risk_failures=0,
            human_review_rate=0.0,
            avg_latency_ms=0.0,
            p95_latency_ms=0.0,
            hallucination_rate=0.0,
            policy_violation_rate=0.0,
            weighted_failure_rate=0.0,
            evaluation_score=0.0,
            regression_flags=[],
            improvement_flags=[],
            trigger_flags=[],
        )

    issue_correct = sum(s.issue_classification_correct for s in scenario_scores)
    path_correct = sum(s.resolution_path_correct for s in scenario_scores)
    outcome_correct = sum(s.outcome_correct for s in scenario_scores)

    scenarios_passed = sum(
        (
            s.issue_classification_correct
            and s.resolution_path_correct
            and s.outcome_correct
        )
        for s in scenario_scores
    )

    high_risk_failures = sum(is_high_risk_failure(s) for s in scenario_scores)
    human_review_cases = sum(s.human_review_required for s in scenario_scores)
    hallucination_cases = sum(s.hallucination_flag for s in scenario_scores)
    policy_cases = sum(s.policy_violation_flag for s in scenario_scores)

    latencies = [s.latency_ms for s in scenario_scores]

    overall_pass_rate = scenarios_passed / total
    issue_classification_accuracy = issue_correct / total
    resolution_path_accuracy = path_correct / total
    outcome_accuracy = outcome_correct / total
    human_review_rate = human_review_cases / total
    hallucination_rate = hallucination_cases / total
    policy_violation_rate = policy_cases / total

    avg_latency_ms = sum(latencies) / total
    p95_latency_ms = percentile(latencies, 0.95)
    weighted_failure_rate = calculate_weighted_failure_rate(scenario_scores)

    evaluation_score = calculate_evaluation_score(
        outcome_accuracy=outcome_accuracy,
        resolution_path_accuracy=resolution_path_accuracy,
        issue_classification_accuracy=issue_classification_accuracy,
        human_review_rate=human_review_rate,
    )

    return RunMetrics(
        run_id=run_id,
        total_scenarios=total,
        scenarios_passed=scenarios_passed,
        overall_pass_rate=round(overall_pass_rate, 4),
        issue_classification_accuracy=round(issue_classification_accuracy, 4),
        resolution_path_accuracy=round(resolution_path_accuracy, 4),
        outcome_accuracy=round(outcome_accuracy, 4),
        high_risk_failures=high_risk_failures,
        human_review_rate=round(human_review_rate, 4),
        avg_latency_ms=round(avg_latency_ms, 2),
        p95_latency_ms=round(p95_latency_ms, 2),
        hallucination_rate=round(hallucination_rate, 4),
        policy_violation_rate=round(policy_violation_rate, 4),
        weighted_failure_rate=round(weighted_failure_rate, 4),
        evaluation_score=evaluation_score,
        regression_flags=[],
        improvement_flags=[],
        trigger_flags=[],
    )
```

## Trend and regression logic

This is your v3 upgrade.

```python
def compare_runs(
    current: RunMetrics,
    previous: Optional[RunMetrics],
) -> RunMetrics:
    """Add regression / improvement flags based on prior run."""
    if previous is None:
        return current

    regression_flags: List[str] = []
    improvement_flags: List[str] = []

    pass_rate_delta = current.overall_pass_rate - previous.overall_pass_rate
    human_review_delta = current.human_review_rate - previous.human_review_rate
    high_risk_delta = current.high_risk_failures - previous.high_risk_failures

    if pass_rate_delta <= -0.05:
        regression_flags.append("pass_rate_regression")

    if current.outcome_accuracy <= previous.outcome_accuracy - 0.05:
        regression_flags.append("outcome_accuracy_regression")

    if human_review_delta >= 0.10:
        regression_flags.append("increase_in_human_review")

    if high_risk_delta >= 1:
        regression_flags.append("increase_in_high_risk_failures")

    if current.avg_latency_ms > previous.avg_latency_ms + 100:
        regression_flags.append("latency_regression")

    if pass_rate_delta >= 0.05:
        improvement_flags.append("pass_rate_improvement")

    if current.outcome_accuracy >= previous.outcome_accuracy + 0.05:
        improvement_flags.append("outcome_accuracy_improvement")

    if current.human_review_rate <= previous.human_review_rate - 0.10:
        improvement_flags.append("reduced_human_review")

    if current.high_risk_failures < previous.high_risk_failures:
        improvement_flags.append("reduced_high_risk_failures")

    current.regression_flags = regression_flags
    current.improvement_flags = improvement_flags
    return current
```

## Trigger logic

This is what makes the reporting executive-friendly.

```python
def generate_trigger_flags(metrics: RunMetrics) -> List[str]:
    """Generate deterministic trigger flags for the current run."""
    triggers: List[str] = []

    if metrics.high_risk_failures >= 3:
        triggers.append("critical_risk_trigger")

    if metrics.evaluation_score < 0.75:
        triggers.append("low_evaluation_score_trigger")

    if metrics.avg_latency_ms > 1000:
        triggers.append("avg_latency_trigger")

    if metrics.p95_latency_ms > 1300:
        triggers.append("p95_latency_trigger")

    if metrics.human_review_rate > 0.30:
        triggers.append("high_human_review_trigger")

    if metrics.hallucination_rate > 0.05:
        triggers.append("hallucination_trigger")

    if metrics.policy_violation_rate > 0.00:
        triggers.append("policy_violation_trigger")

    if metrics.weighted_failure_rate > 0.25:
        triggers.append("weighted_failure_trigger")

    if metrics.regression_flags:
        triggers.append("regression_trigger")

    return triggers
```

```python
def finalize_run_metrics(
    current: RunMetrics,
    previous: Optional[RunMetrics] = None,
) -> RunMetrics:
    """Apply trend comparison and trigger generation."""
    current = compare_runs(current=current, previous=previous)
    current.trigger_flags = generate_trigger_flags(current)
    return current
```

## End-to-end entry point

This is the function your orchestrator node would call.

```python
def score_evaluation_run(
    run_id: str,
    scenario_inputs: List[Dict[str, Any]],
    previous_run_metrics: Optional[RunMetrics] = None,
) -> Dict[str, Any]:
    """
    Score a full evaluation run from raw scenario result records.

    Each scenario_inputs item should include expected and predicted fields.
    """
    scenario_scores: List[ScenarioScore] = []

    for row in scenario_inputs:
        scenario_score = score_scenario(
            run_id=run_id,
            scenario_id=row["scenario_id"],
            expected_issue_type=row["expected_issue_type"],
            predicted_issue_type=row["predicted_issue_type"],
            expected_resolution_path=row["expected_resolution_path"],
            predicted_resolution_path=row["predicted_resolution_path"],
            expected_outcome=row["expected_outcome"],
            predicted_outcome=row["predicted_outcome"],
            latency_ms=row["latency_ms"],
            human_review_required=row.get("human_review_required", False),
            severity_weight=row.get("severity_weight", 1.0),
            business_risk=row.get("business_risk", "low"),
            hallucination_flag=row.get("hallucination_flag", False),
            policy_violation_flag=row.get("policy_violation_flag", False),
        )
        scenario_scores.append(scenario_score)

    run_metrics = rollup_run_metrics(run_id=run_id, scenario_scores=scenario_scores)
    run_metrics = finalize_run_metrics(current=run_metrics, previous=previous_run_metrics)

    return {
        "scenario_scores": [s.__dict__ for s in scenario_scores],
        "run_metrics": run_metrics.__dict__,
    }
```
