## üèóÔ∏è Compliance Sentinel Agent - LangGraph Node Scaffold

### **Core Architecture Overview**
The agent follows a **"Watchdog Layer"** pattern that wraps around existing AI workflows, providing compliance monitoring, evaluation, and escalation capabilities.

---

### **1. Entry & Input Processing Nodes**

#### **`input_receiver`**
- **Purpose**: Entry point for content to be audited
- **Inputs**: Raw content from other agents (marketing copy, contracts, proposals, etc.)
- **Processing**:
  - Content validation and sanitization
  - Metadata extraction (source agent, timestamp, content type)
  - Content preprocessing (text cleaning, structure detection)
- **Outputs**: Structured content object with metadata

#### **`content_classifier`**
- **Purpose**: Determine content type and applicable compliance domains
- **Inputs**: Processed content from `input_receiver`
- **Processing**:
  - Content type classification (marketing, legal, HR, financial, etc.)
  - Risk level assessment (high/medium/low)
  - Applicable regulation identification (GDPR, HIPAA, SOX, etc.)
- **Outputs**: Content classification with risk profile

---

### **2. Knowledge Retrieval Nodes**

#### **`policy_retriever`**
- **Purpose**: Fetch relevant compliance policies and regulations
- **Inputs**: Content classification from `content_classifier`
- **Processing**:
  - Vector similarity search in compliance knowledge base
  - Multi-domain policy retrieval (internal policies + regulations)
  - Context-aware policy selection
- **Outputs**: Relevant policy documents and regulations

#### **`rule_engine`**
- **Purpose**: Apply specific compliance rulesets
- **Inputs**: Content + retrieved policies
- **Processing**:
  - Rule matching and application
  - Cross-reference with industry standards
  - Historical precedent checking
- **Outputs**: Applicable rules and compliance criteria

---

### **3. Evaluation & Analysis Nodes**

#### **`compliance_evaluator`**
- **Purpose**: Core LLM-based compliance assessment
- **Inputs**: Content + policies + rules
- **Processing**:
  - Structured compliance checking using LLM
  - Violation detection and severity assessment
  - Risk scoring and confidence levels
- **Outputs**: Compliance verdict (pass/warning/violation) with rationale

#### **`risk_assessor`**
- **Purpose**: Calculate risk impact and business implications
- **Inputs**: Compliance verdict + content context
- **Processing**:
  - Financial risk calculation (potential fines, lawsuits)
  - Reputational risk assessment
  - Regulatory exposure analysis
- **Outputs**: Risk score and business impact assessment

---

### **4. Decision & Routing Nodes**

#### **`decision_router`**
- **Purpose**: Route content based on compliance verdict
- **Inputs**: Compliance evaluation + risk assessment
- **Processing**:
  - Decision logic for pass/warning/escalation
  - Threshold-based routing
  - Priority determination
- **Outputs**: Routing decision and next steps

#### **`escalation_manager`**
- **Purpose**: Handle flagged content and human escalation
- **Inputs**: Violations requiring human review
- **Processing**:
  - Escalation path determination
  - Context preparation for human reviewers
  - Notification scheduling
- **Outputs**: Escalation package with full context

---

### **5. Action & Response Nodes**

#### **`approval_handler`**
- **Purpose**: Process human approvals/overrides
- **Inputs**: Human decisions on flagged content
- **Processing**:
  - Approval validation and logging
  - Override reason capture
  - Learning from human feedback
- **Outputs**: Final approval status and reasoning

#### **`notification_sender`**
- **Purpose**: Send alerts and notifications
- **Inputs**: Escalation packages and alerts
- **Processing**:
  - Multi-channel notification (Slack, Teams, email)
  - Message formatting and personalization
  - Delivery confirmation tracking
- **Outputs**: Notification status and delivery confirmations

---

### **6. Audit & Logging Nodes**

#### **`audit_logger`**
- **Purpose**: Comprehensive audit trail creation
- **Inputs**: All decision points and actions
- **Processing**:
  - Immutable log entry creation
  - Evidence chain preservation
  - Regulatory compliance formatting
- **Outputs**: Structured audit logs

#### **`report_generator`**
- **Purpose**: Create compliance reports and dashboards
- **Inputs**: Audit logs and compliance data
- **Processing**:
  - Trend analysis and reporting
  - Executive summary generation
  - Regulatory submission formatting
- **Outputs**: Compliance reports and analytics

---

### **7. Learning & Improvement Nodes**

#### **`feedback_processor`**
- **Purpose**: Learn from human decisions and improve accuracy
- **Inputs**: Human overrides and feedback
- **Processing**:
  - Pattern recognition in human decisions
  - Model fine-tuning data preparation
  - Rule refinement suggestions
- **Outputs**: Learning insights and improvement recommendations

#### **`model_updater`**
- **Purpose**: Update compliance models based on feedback
- **Inputs**: Learning insights from feedback processor
- **Processing**:
  - Model retraining triggers
  - Rule set updates
  - Performance monitoring
- **Outputs**: Updated models and rules

## Key Conditional Routing Functions:

```python
def route_compliance_decision(state: ComplianceState) -> str:
    """Route based on compliance evaluation result"""
    verdict = state.compliance_verdict
    
    if verdict == "pass":
        return "pass"
    elif verdict == "warning":
        return "warning"
    elif verdict == "violation":
        return "violation"
    else:
        return "error"

def route_after_approval(state: ComplianceState) -> str:
    """Route based on human approval decision"""
    approval_status = state.human_approval_status
    
    if approval_status == "approved":
        return "approved"
    elif approval_status == "rejected":
        return "rejected"
    elif approval_status == "requires_changes":
        return "requires_changes"
    else:
        return "escalate_further"

def route_for_learning(state: ComplianceState) -> str:
    """Route for learning and improvement based on feedback availability"""
    if state.has_human_feedback:
        return "process_feedback"
    elif state.model_update_needed:
        return "update_models"
    else:
        return "generate_report"
```

````python
"""
Compliance Sentinel Agent - MVP Implementation

A LangGraph-based agent that monitors AI-generated content for regulatory compliance.
This MVP focuses on marketing compliance using a simplified 6-node workflow.
"""

import os
import json
import logging
from typing import Dict, List, Any, TypedDict
from datetime import datetime

from langgraph.graph import StateGraph, END
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage
from dotenv import load_dotenv

# Load environment variables
load_dotenv("API_KEYS.env")

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('logs/compliance_audit.log'),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)

# State Schema
class ComplianceState(TypedDict):
    # Goal and criteria
    compliance_goal: str
    success_criteria: Dict[str, Any]

    # Content processing
    original_content: str
    content_metadata: Dict[str, Any]

    # Policy and rules
    retrieved_policies: List[Dict[str, Any]]

    # Evaluation results
    compliance_verdict: str  # "pass", "warning", "violation"
    compliance_rationale: str
    risk_score: float

    # Audit
    audit_log: List[Dict[str, Any]]

# Initialize LLM
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.1,
    max_tokens=2000,
    api_key=os.getenv("OPENAI_API_KEY")
)

def set_compliance_goal_and_criteria(state: ComplianceState) -> ComplianceState:
    """Set the compliance goal and success criteria for this evaluation"""
    logger.info("üéØ Setting compliance goal and criteria")

    goal = "Ensure all AI-generated marketing content meets regulatory and company policy requirements"
    criteria = {
        "ftc_compliance": "No false or misleading claims",
        "gdpr_compliance": "Proper consent and data handling",
        "brand_compliance": "Adheres to company brand guidelines",
        "disclaimer_requirements": "Appropriate disclaimers present"
    }

    state["compliance_goal"] = goal
    state["success_criteria"] = criteria

    logger.info(f"Goal set: {goal}")
    return state

def receive_and_validate_content(state: ComplianceState) -> ComplianceState:
    """Receive and validate the content to be checked"""
    logger.info("üì• Receiving and validating content")

    # For MVP, we'll expect content to be passed in the state
    if "original_content" not in state or not state["original_content"]:
        raise ValueError("No content provided for compliance checking")

    content = state["original_content"]

    # Basic validation
    if len(content) < 10:
        raise ValueError("Content too short for meaningful compliance check")

    # Extract metadata
    metadata = {
        "content_length": len(content),
        "timestamp": datetime.now().isoformat(),
        "content_type": "marketing",  # For MVP, assume marketing
        "source": "ai_generated"
    }

    state["content_metadata"] = metadata

    logger.info(f"Content validated: {len(content)} characters")
    return state

def retrieve_relevant_policies(state: ComplianceState) -> ComplianceState:
    """Retrieve relevant compliance policies based on content analysis"""
    logger.info("üîç Retrieving relevant compliance policies")

    content = state["original_content"].lower()
    policies = []

    # Load compliance policies from file
    try:
        with open("docs/Marketing Compliance Policy v1.0.txt", "r") as f:
            policy_content = f.read()
    except FileNotFoundError:
        logger.error("Compliance policy file not found")
        return state

    # Simple keyword-based policy retrieval for MVP
    policy_sections = policy_content.split("MCP-")

    for i, section in enumerate(policy_sections[1:], 1):
        policy_id = f"MCP-{i:02d}"
        policy_text = f"MCP-{i:02d}" + section

        # Check if this policy is relevant to the content
        is_relevant = False

        if "guarantee" in content and "MCP-01" in policy_text:
            is_relevant = True
        elif "unsubscribe" in content and "MCP-05" in policy_text:
            is_relevant = True
        elif "gdpr" in content or "consent" in content and "MCP-08" in policy_text:
            is_relevant = True
        elif "disclaimer" in content and "MCP-03" in policy_text:
            is_relevant = True
        elif "social" in content and "MCP-12" in policy_text:
            is_relevant = True

        if is_relevant:
            policies.append({
                "id": policy_id,
                "content": policy_text,
                "relevance_reason": f"Content contains keywords related to {policy_id}"
            })

    # If no specific policies found, include general marketing policies
    if not policies:
        policies.append({
            "id": "MCP-01",
            "content": policy_sections[1] if len(policy_sections) > 1 else "",
            "relevance_reason": "General marketing compliance check"
        })

    state["retrieved_policies"] = policies
    logger.info(f"Retrieved {len(policies)} relevant policies")

    return state

def evaluate_compliance_status(state: ComplianceState) -> ComplianceState:
    """Evaluate compliance status using LLM"""
    logger.info("‚öñÔ∏è Evaluating compliance status")

    content = state["original_content"]
    policies = state["retrieved_policies"]

    # Prepare policy context
    policy_context = "\n\n".join([p["content"] for p in policies])

    # Create compliance evaluation prompt
    system_prompt = f"""You are a compliance expert evaluating marketing content against company policies and regulations.

COMPLIANCE POLICIES:
{policy_context}

EVALUATION CRITERIA:
- FTC Compliance: No false or misleading claims
- GDPR Compliance: Proper consent and data handling
- Brand Compliance: Adheres to company guidelines
- Disclaimer Requirements: Appropriate disclaimers present

Respond with a JSON object containing:
{{
    "verdict": "pass|warning|violation",
    "rationale": "Detailed explanation of compliance status",
    "risk_score": 0.0-1.0,
    "specific_issues": ["list of specific compliance issues found"],
    "recommendations": ["list of recommended actions"]
}}"""

    human_prompt = f"""Please evaluate this marketing content for compliance:

CONTENT TO EVALUATE:
{content}

Provide your evaluation in the specified JSON format."""

    try:
        messages = [
            SystemMessage(content=system_prompt),
            HumanMessage(content=human_prompt)
        ]

        response = llm.invoke(messages)
        result = json.loads(response.content)

        state["compliance_verdict"] = result["verdict"]
        state["compliance_rationale"] = result["rationale"]
        state["risk_score"] = result["risk_score"]

        logger.info(f"Compliance verdict: {result['verdict']} (risk: {result['risk_score']})")

    except Exception as e:
        logger.error(f"Error in compliance evaluation: {e}")
        state["compliance_verdict"] = "error"
        state["compliance_rationale"] = f"Evaluation error: {str(e)}"
        state["risk_score"] = 1.0

    return state

def route_based_on_compliance(state: ComplianceState) -> str:
    """Route based on compliance evaluation result"""
    verdict = state["compliance_verdict"]

    if verdict == "pass":
        return "pass"
    elif verdict == "warning":
        return "warning"
    elif verdict == "violation":
        return "violation"
    else:
        return "pass"  # Default to pass for MVP

def create_audit_trail(state: ComplianceState) -> ComplianceState:
    """Create comprehensive audit trail"""
    logger.info("üìã Creating audit trail")

    audit_entry = {
        "timestamp": datetime.now().isoformat(),
        "content_preview": state["original_content"][:100] + "...",
        "compliance_verdict": state["compliance_verdict"],
        "compliance_rationale": state["compliance_rationale"],
        "risk_score": state["risk_score"],
        "policies_checked": [p["id"] for p in state["retrieved_policies"]],
        "content_metadata": state["content_metadata"]
    }

    # Initialize audit log if not exists
    if "audit_log" not in state:
        state["audit_log"] = []

    state["audit_log"].append(audit_entry)

    # Log to file
    logger.info(f"Audit entry created: {audit_entry['compliance_verdict']}")

    return state

def create_compliance_sentinel_agent():
    """Create the Compliance Sentinel Agent workflow - MVP Version"""
    logger.info("üèóÔ∏è Building Compliance Sentinel Agent Workflow (MVP)...")

    # Create the workflow
    workflow = StateGraph(ComplianceState)

    # Add nodes (core compliance flow only)
    workflow.add_node("set_compliance_goal", set_compliance_goal_and_criteria)
    workflow.add_node("input_receiver", receive_and_validate_content)
    workflow.add_node("policy_retriever", retrieve_relevant_policies)
    workflow.add_node("compliance_evaluator", evaluate_compliance_status)
    workflow.add_node("audit_logger", create_audit_trail)

    # Add edges (linear flow with conditional routing)
    workflow.add_edge("set_compliance_goal", "input_receiver")
    workflow.add_edge("input_receiver", "policy_retriever")
    workflow.add_edge("policy_retriever", "compliance_evaluator")

    # Conditional routing based on compliance decision
    workflow.add_conditional_edges(
        "compliance_evaluator",
        route_based_on_compliance,
        {
            "pass": "audit_logger",
            "warning": "audit_logger",
            "violation": "audit_logger"
        }
    )

    # All paths lead to audit logging
    workflow.add_edge("audit_logger", END)

    # Set entry point
    workflow.set_entry_point("set_compliance_goal")

    # Compile the workflow
    app = workflow.compile()

    logger.info("‚úÖ Compliance Sentinel Agent workflow compiled successfully")
    return app

# Test function
def test_compliance_agent():
    """Test the compliance agent with sample content"""
    logger.info("üß™ Testing Compliance Sentinel Agent")

    app = create_compliance_sentinel_agent()

    # Test with sample content
    test_content = "Get 100% guaranteed ROI with our AI system! No risk, maximum returns!"

    initial_state = {
        "original_content": test_content
    }

    try:
        result = app.invoke(initial_state)
        logger.info("‚úÖ Test completed successfully")
        logger.info(f"Final verdict: {result['compliance_verdict']}")
        logger.info(f"Rationale: {result['compliance_rationale']}")
        return result
    except Exception as e:
        logger.error(f"‚ùå Test failed: {e}")
        return None

if __name__ == "__main__":
    # Run test
    test_compliance_agent()
````
