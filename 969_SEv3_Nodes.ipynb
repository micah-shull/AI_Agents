````python
from __future__ import annotations

from datetime import datetime, timezone
from pathlib import Path
from typing import Any, Dict, List

from config import SalesEnablementOrchestratorV3Config, SalesEnablementOrchestratorV3State
from toolshed.sales import save_report

from .utilities.data_loading import load_sev3_inputs
from .utilities.enablement import build_enablement_snapshot_md
from .utilities.reporting import build_sev3_report_md
from .utilities.temporal_metrics import compute_temporal_metrics


def goal_node(state: SalesEnablementOrchestratorV3State) -> Dict[str, Any]:
    """Define the analysis objective (rule-based v3 execution)."""
    view_mode = state.get("view_mode") or "portfolio"
    goal = {
        "objective": "Temporal revenue intelligence: trajectory, risk velocity, forecast stability, and execution visibility.",
        "view_mode": view_mode,
    }
    return {
        "goal": goal,
        "errors": state.get("errors", []),
    }


def planning_node(state: SalesEnablementOrchestratorV3State) -> Dict[str, Any]:
    """Create a deterministic, linear execution plan."""
    plan: List[Dict[str, Any]] = [
        {
            "step": 1,
            "name": "data_loading",
            "description": "Load snapshot + temporal histories from agents/data.",
            "outputs": ["data_bundle", "data_counts", "reference_datetime_utc"],
        },
        {
            "step": 2,
            "name": "temporal_metrics",
            "description": "Compute deal/rep/pipeline trajectory metrics and trigger evolution.",
            "outputs": ["temporal_metrics"],
        },
        {
            "step": 3,
            "name": "enablement_snapshot",
            "description": "Build current enablement actions (lead + deal + rep nudges).",
            "outputs": ["enablement_report_md"],
        },
        {
            "step": 4,
            "name": "report_generation",
            "description": "Assemble v3 executive temporal intelligence + enablement actions.",
            "outputs": ["report_file_path"],
        },
    ]
    return {"plan": plan, "errors": state.get("errors", [])}


def make_data_loading_node(config: SalesEnablementOrchestratorV3Config, project_root: str):
    def _node(state: SalesEnablementOrchestratorV3State) -> Dict[str, Any]:
        start = datetime.now(timezone.utc)
        errors = state.get("errors", [])
        validation_warnings: List[str] = state.get("validation_warnings", [])

        try:
            bundle = load_sev3_inputs(project_root=project_root, data_dir=config.data_dir)
        except Exception as e:
            return {"errors": errors + [f"data_loading_node failed: {e}"]}

        deals = bundle.get("deals") or []
        reps = bundle.get("sales_reps") or []
        interactions = bundle.get("interactions") or []
        deal_history = bundle.get("deal_history") or []

        # Basic integrity checks to avoid silent partial outputs.
        deal_history_deal_ids = {d.get("deal_id") for d in deal_history if d.get("deal_id")}
        deals_deal_ids = {d.get("deal_id") for d in deals if d.get("deal_id")}
        missing_in_current = sorted(list(deal_history_deal_ids - deals_deal_ids))
        if missing_in_current:
            validation_warnings.append(
                f"deal_history includes deals not present in deals.json (skipped): {', '.join(missing_in_current[:10])}"
            )

        missing_temporal = []
        for k in ["pipeline_snapshots", "rep_performance_history"]:
            if not bundle.get(k):
                missing_temporal.append(k)
        if missing_temporal:
            validation_warnings.append(f"missing temporal datasets: {', '.join(missing_temporal)}")

        ref_dt = bundle.get("reference_datetime_utc")
        if not ref_dt:
            validation_warnings.append("reference_datetime_utc not determined; using fallback period end.")

        data_counts = {
            "deals_count": len(deals),
            "reps_count": len(reps),
            "interactions_count": len(interactions),
            "deal_history_deals_count": len(deal_history),
        }

        processing_time = (datetime.now(timezone.utc) - start).total_seconds()
        return {
            "data_bundle": bundle,
            "data_counts": data_counts,
            "reference_datetime_utc": ref_dt,
            "validation_warnings": validation_warnings,
            "processing_time": (state.get("processing_time") or 0.0) + processing_time,
            "errors": errors,
        }

    return _node


def make_temporal_metrics_node(config: SalesEnablementOrchestratorV3Config):
    def _node(state: SalesEnablementOrchestratorV3State) -> Dict[str, Any]:
        errors = state.get("errors", [])
        bundle = state.get("data_bundle") or {}
        thresholds = bundle.get("thresholds") or {}
        stage_actions = bundle.get("stage_actions") or []

        try:
            temporal = compute_temporal_metrics(
                deals=bundle.get("deals") or [],
                deal_history=bundle.get("deal_history") or [],
                interactions=bundle.get("interactions") or [],
                signals_lookup=bundle.get("signals_lookup") or {},
                stage_actions=stage_actions,
                thresholds=thresholds,
                pipeline_snapshots=bundle.get("pipeline_snapshots") or [],
                sales_reps=bundle.get("sales_reps") or [],
                rep_performance_history=bundle.get("rep_performance_history") or [],
            )
        except Exception as e:
            return {"errors": errors + [f"temporal_metrics_node failed: {e}"]}

        return {"temporal_metrics": temporal, "errors": errors}

    return _node


def make_enablement_snapshot_node(config: SalesEnablementOrchestratorV3Config):
    def _node(state: SalesEnablementOrchestratorV3State) -> Dict[str, Any]:
        errors = state.get("errors", [])
        bundle = state.get("data_bundle") or {}
        ref_raw = state.get("reference_datetime_utc") or bundle.get("reference_datetime_utc")

        try:
            reference_dt = (
                datetime.fromisoformat(ref_raw) if ref_raw else datetime.now(timezone.utc)
            )
            if reference_dt.tzinfo is None:
                reference_dt = reference_dt.replace(tzinfo=timezone.utc)
            else:
                reference_dt = reference_dt.astimezone(timezone.utc)
        except Exception:
            reference_dt = datetime.now(timezone.utc)

        try:
            enablement = build_enablement_snapshot_md(
                data=bundle,
                reference_datetime_utc=reference_dt,
                reports_dir=config.reports_dir,
            )
        except Exception as e:
            return {"errors": errors + [f"enablement_snapshot_node failed: {e}"]}

        return {
            "enablement_report_md": enablement.get("enablement_report_md") or "",
            "errors": errors,
        }

    return _node


def make_report_generation_node(config: SalesEnablementOrchestratorV3Config):
    def _node(state: SalesEnablementOrchestratorV3State) -> Dict[str, Any]:
        errors = state.get("errors", [])
        temporal = state.get("temporal_metrics") or {}
        enablement_report_md = state.get("enablement_report_md") or ""

        try:
            md = build_sev3_report_md(
                data_counts=state.get("data_counts") or {},
                temporal_metrics=temporal,
                enablement_report_md=enablement_report_md,
            )
        except Exception as e:
            return {"errors": errors + [f"report_generation_node failed building markdown: {e}"]}

        # Deterministic filename: include current period when available.
        period = (
            (temporal.get("pipeline_metrics") or {}).get("current_period")
            or (state.get("reference_datetime_utc") or "")[:10]
            or "run"
        )
        filename = f"sales_enablement_orchestrator_v3_{period}.md"

        try:
            out_path = save_report(md, Path(config.reports_dir), filename=filename)
        except Exception as e:
            return {"errors": errors + [f"report_generation_node save failed: {e}"]}

        return {
            "report_file_path": out_path,
            "errors": errors,
        }

    return _node

````
