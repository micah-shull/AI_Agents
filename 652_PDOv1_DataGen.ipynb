## ðŸ”§ Recommended Enhancements (MVP-Critical)

### 1. **Consistent Timestamping Across All Files**

**Issue:** Timestamps are inconsistent or missing, which limits time-series analysis and KPI calculations.

**Recommendation:**
- Add `created_at` (ISO 8601 format: `"2026-01-10T14:30:00Z"`) to ALL files
- Add `updated_at` where records can change (documents, outcomes)
- Add `started_at` and `completed_at` to `workflow_stages.json` (not just duration)
- Add `reviewed_at` to `review_events.json`
- Add `checked_at` to `compliance_checks.json`
- Add `tracked_at` to `cost_tracking.json`
- Add `submitted_at` or `completed_at` to `outcomes.json`

**Why:** Enables bottleneck detection, time-to-completion analysis, and proper before/after comparisons.

---

### 2. **Cross-File Linking for Workflow Analysis**

**Issue:** Missing foreign keys limit ability to trace document lifecycle and calculate end-to-end metrics.

**Recommendation:**
- Add `version_id` to `review_events.json` (which version was reviewed?)
- Add `stage_id` to `cost_tracking.json` (optional: break down costs by stage)
- Add `version_id` to `compliance_checks.json` (which version was checked?)
- Consider adding `parent_stage_id` to `workflow_stages.json` if stages have dependencies

**Why:** Enables "revision 3 failed compliance check X, triggering review Y" analysis.

---

### 3. **Workflow Sequence Tracking**

**Issue:** `workflow_stages.json` doesn't show stage order, making bottleneck analysis difficult.

**Recommendation:**
- Add `stage_order` or `sequence_number` field to `workflow_stages.json`
- Example: `"stage_order": 1, 2, 3...`

**Why:** Enables "stage 3 is always the bottleneck" insights.

---

### 4. **Error/Failure Context for Debugging**

**Issue:** Failed stages and compliance checks lack error details, limiting root cause analysis.

**Recommendation:**
- Add `error_message` or `failure_reason` to `workflow_stages.json` when `status: "failed"`
- Add `error_message` or `details` to `compliance_checks.json` when `status: "failed"`
- Example: `"error_message": "Template validation failed: missing required section 'executive_summary'"`

**Why:** Critical for MVP debugging and proving the system catches issues early.

---

### 5. **Cost Granularity (Optional but Valuable)**

**Issue:** `cost_tracking.json` aggregates all LLM costs, making it hard to identify expensive stages.

**Recommendation (MVP-optional):**
- Consider adding `cost_by_stage` array or separate `stage_costs.json` file
- OR keep aggregated but add `stage_breakdown` field:
  ```json
  {

    "llm_cost_usd": 2.40,
    "stage_breakdown": {
      "structure_planning": 0.30,
      "content_generation": 1.80,
      "compliance_check": 0.30
    }
  }
  ```

**Why:** Helps identify optimization opportunities (e.g., "content_generation is 75% of LLM cost").
