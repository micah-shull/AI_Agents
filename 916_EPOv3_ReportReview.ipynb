# Refinement 1 — Risk Narrative Capitalization

Right now:

```
Data Quality Issues are currently the dominant portfolio risk
```

Technically this should be sentence case:

```
Data quality issues are currently the dominant portfolio risk
```

Because it's no longer a proper title.

So the section would read:

```
Primary risk driver: data_quality_issue
Data quality issues are currently the dominant portfolio risk, driven by 2 critical events and 2 blocked experiments.
```

Small detail, but it reads more naturally.

---

# Refinement 2 — “Executive Attention Required” Tone

Your summary sentence says:

```
Executive attention required — see Next steps.
```

But the verdict is **Watch**, not **Attention**.

To keep tone aligned with verdict levels, you could say:

```
Leadership review recommended — see Next steps.
```

or

```
Monitoring required — see Next steps.
```

This keeps the language consistent with the verdict scale:

```
On target
Watch
Attention
Escalate
```

Not required, but it makes the tone slightly more consistent.
