````python
def load_all_raa_data(
    data_dir: str,
    project_root: str,
    contracts_file: str = "sample_contracts.json",
    invoices_file: str = "sample_invoices.csv",
    usage_file: str = "usage_data.json",
    discount_approvals_file: str = "discount_approvals.json",
    recovery_log_file: str = "recovery_log.json",
) -> Dict[str, Any]:
    """
    Load all RAA data from data_dir (resolved relative to project_root).
    Returns dict with: contracts, invoices, usage_records, discount_approvals, recovery_log, data_snapshot_loaded_at.
    """
    base = Path(project_root) / data_dir
    data_snapshot_loaded_at = datetime.now(timezone.utc).isoformat()

    warnings: List[str] = []
    paths = {
        "contracts": base / contracts_file,
        "invoices": base / invoices_file,
        "usage": base / usage_file,
        "discount_approvals": base / discount_approvals_file,
        "recovery_log": base / recovery_log_file,
    }

    contracts = _read_json(paths["contracts"])
    invoices = _read_invoices_csv(paths["invoices"])
    usage_records = _read_json(paths["usage"])
    discount_approvals = _read_json(paths["discount_approvals"])
    recovery_log = _read_json(paths["recovery_log"])

    if not contracts:
        warnings.append("Contracts file missing or empty")
    if not invoices:
        warnings.append("Invoices file missing or empty")
    if not usage_records:
        warnings.append("Usage file missing or empty")

    data_counts = {
        "contracts": len(contracts),
        "invoices": len(invoices),
        "usage_records": len(usage_records),
        "discount_approvals": len(discount_approvals),
        "recovery_log": len(recovery_log),
    }
    files_loaded = {k: str(v) for k, v in paths.items()}

    return {
        "contracts": contracts,
        "invoices": invoices,
        "usage_records": usage_records,
        "discount_approvals": discount_approvals,
        "recovery_log": recovery_log,
        "data_snapshot_loaded_at": data_snapshot_loaded_at,
        "validation_warnings": warnings,
        "data_counts": data_counts,
        "files_loaded": files_loaded,
    }
````
