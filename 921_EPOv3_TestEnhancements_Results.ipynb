# Planned Testing Enhancements — EPOv3

The EPOv3 testing framework already provides strong coverage across utilities, nodes, and full-graph integration. The planned enhancements focus on **increasing robustness, protecting report semantics, and strengthening state contract guarantees** without significantly increasing test complexity.

These improvements expand validation around three areas:

1. **Report narrative correctness**
2. **Node output integrity**
3. **End-to-end orchestration metadata propagation**

The goal is to ensure the system remains **deterministic, traceable, and structurally stable** as the orchestrator evolves.

---

# 1. Utility Test Enhancements

### Risk Narrative Validation

The report now includes an interpreted narrative explaining the dominant portfolio risk. A new test ensures this explanation appears when appropriate.

Example scenario:

```text
Data quality issues are currently the dominant portfolio risk.
```

This protects the narrative logic responsible for translating raw risk data into executive-level interpretation.

---

### Portfolio Trend Sentence

The report also includes a trend interpretation such as:

```text
Trend: Portfolio improving as risk declined and learning velocity increased.
```

A new test validates that the trend interpretation always appears when trend signals are present, preventing silent regressions in the reporting logic.

---

### Snapshot Format Validation

A lightweight snapshot-style test will validate the presence of core report sections:

```text
Experimentation Portfolio — Executive Report
## One view
## Risk & governance
```

This protects the **report contract** so future refactoring cannot accidentally break the executive report format.

---

# 2. Node Test Enhancements

### Run ID Format Validation

Goal node tests will verify the run identifier structure:

```text
epo_v3_<timestamp>
```

This ensures that orchestration runs remain uniquely traceable and prevents accidental changes to run ID formatting.

---

### Health Score Type Safety

The portfolio node will now explicitly validate that the computed health score is an integer:

```python
assert isinstance(portfolio_health_score, int)
```

This protects against unintended type changes (e.g., `None`, strings, floats) that could break reporting or analytics pipelines.

---

# 3. Integration Test Enhancements

### Run Metadata Verification

Integration tests will confirm that the orchestration run metadata is correctly propagated through the graph:

```python
assert result.get("run_id")
assert result.get("run_started_at")
```

This ensures that traceability metadata generated in the goal node reaches the final output state.

---

### Health Score Presence in Report

Integration tests will also verify that the portfolio health score appears in the generated report:

```python
assert "health score" in report_content
```

Because the health score is the primary portfolio metric, this check ensures the executive report always contains this key signal.
