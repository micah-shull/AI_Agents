````python
"""
Build CFO-grade Revenue Assurance report: one-view, rollup metrics, issue table, next steps.
"""
from typing import Any, Dict, List, Optional


def _so_what_line(rollup: Dict[str, Any], triggers: List[Dict[str, Any]]) -> str:
    """One sentence: revenue at risk, recovered, open issues, and whether attention is required."""
    rar = rollup.get("revenue_at_risk") or 0
    recovered = rollup.get("revenue_recovered") or 0
    open_count = rollup.get("open_issues_count") or 0
    if triggers:
        return f"Revenue at risk ${rar:,.0f}; {open_count} open issue(s); executive attention required â€” see next steps below."
    return f"Revenue at risk ${rar:,.0f}; ${recovered:,.0f} recovered to date; {open_count} open issue(s) to address."


def _next_steps_from_triggers_and_issues(
    triggers: List[Dict[str, Any]],
    top_issues: List[Dict[str, Any]],
    max_steps: int = 5,
) -> List[str]:
    """Build 2â€“5 concrete next steps from triggers and top issues (critical-first)."""
    steps: List[str] = []
    for t in triggers[:2]:
        steps.append(t.get("message", ""))
    # Add issue-driven steps: assign highest-impact to owners
    by_owner: Dict[str, List[Dict[str, Any]]] = {}
    for i in top_issues[:10]:
        owner = i.get("recommended_owner") or "Finance"
        by_owner.setdefault(owner, []).append(i)
    for owner, items in by_owner.items():
        if len(steps) >= max_steps:
            break
        customers = list(dict.fromkeys([x.get("customer_id", "") for x in items if x.get("customer_id")]))[:5]
        cust_str = ", ".join(customers) if customers else "affected accounts"
        steps.append(f"Assign {cust_str} ({owner}): {len(items)} issue(s); require status within 5 business days.")
    if not steps:
        steps.append("Continue monitoring; no executive action required this run.")
    return steps[:max_steps]


def build_revenue_report(
    rollup: Dict[str, Any],
    executive_triggers: List[Dict[str, Any]],
    top_issues: List[Dict[str, Any]],
    data_snapshot_loaded_at: Optional[str] = None,
    validation_warnings: Optional[List[str]] = None,
) -> str:
    """
    Generate CFO-grade markdown report: headline, blockquote if triggers, rollup metrics,
    top issues table, next steps (built from triggers + top_issues), data snapshot note.
    """
    next_steps = _next_steps_from_triggers_and_issues(executive_triggers, top_issues)
    lines: List[str] = []
    lines.append("# Revenue Assurance â€” Executive Report")
    lines.append("")
    lines.append("**Revenue Assurance Orchestrator v2** â€” Continuous revenue integrity and recovery.")
    lines.append("")

    if executive_triggers:
        lines.append("> **Executive attention required.** Triggers exceeded; see next steps below.")
        lines.append("")
    lines.append("## One view")
    lines.append("")
    so_what = _so_what_line(rollup, executive_triggers)
    lines.append(f"**So what:** {so_what}")
    lines.append("")
    lines.append("| Metric | Value |")
    lines.append("|--------|-------|")
    lines.append(f"| Revenue recovered (to date) | ${rollup.get('revenue_recovered', 0):,.2f} |")
    lines.append(f"| Revenue at risk (this run) | ${rollup.get('revenue_at_risk', 0):,.2f} |")
    lines.append(f"| Upsell pipeline (identified) | ${rollup.get('upsell_pipeline_value', 0):,.2f} |")
    lines.append(f"| Recovery rate | {rollup.get('recovery_rate', 0):.1f}% |")
    lines.append(f"| Open issues | {rollup.get('open_issues_count', 0)} |")
    lines.append("")
    lines.append("## Next steps")
    lines.append("")
    for step in next_steps:
        lines.append(f"- {step}")
    lines.append("")
    lines.append("## Top issues by revenue impact")
    lines.append("")
    if not top_issues:
        lines.append("No open issues identified this run.")
    else:
        lines.append("| Customer | Type | Revenue impact | Severity | Owner |")
        lines.append("|----------|------|----------------|----------|-------|")
        for i in top_issues:
            cust = i.get("customer_id", "")
            typ = i.get("issue_type", "")
            impact = i.get("revenue_impact") or 0
            sev = i.get("severity", "")
            owner = i.get("recommended_owner", "")
            lines.append(f"| {cust} | {typ} | ${impact:,.2f} | {sev} | {owner} |")
        lines.append("")
    lines.append("## Methodology")
    lines.append("")
    lines.append("Reconciliation: contracts vs invoices (pricing, discount, calculation, billing cycle); usage vs contract (overage); discount vs approvals. Issues are quantified, prioritized by revenue impact, and assigned to Finance, Sales Ops, or Account Manager.")
    if data_snapshot_loaded_at:
        lines.append("")
        lines.append(f"*Data snapshot: {data_snapshot_loaded_at} (UTC)*")
    if validation_warnings:
        lines.append("")
        lines.append("**Validation warnings:**")
        for w in validation_warnings:
            lines.append(f"- {w}")
    lines.append("")
    return "\n".join(lines)
````
