# Portfolio & Risk Rollup Layer

This code implements the **signal extraction layer** of the Experimentation Portfolio Orchestrator.

Its role is to convert raw telemetry into **portfolio intelligence**.

The orchestrator does this in three stages:

1. **Portfolio Rollup** — summarizes experimentation health
2. **Risk Rollup** — aggregates governance and operational risk
3. **Executive Triggers** — detects when leadership attention is required

These steps transform operational telemetry into **decision-ready signals**.

Instead of drowning leaders in dashboards or statistical detail, the system produces **clear indicators of experimentation program health**.

---

# 1. Snapshot Ordering

```python
def _parse_snapshot_date(s: str) -> date:
```

This helper function converts snapshot timestamps into date objects so snapshots can be sorted chronologically.

Experimentation telemetry is stored as a **series of portfolio snapshots**, each representing the state of the experimentation program at a point in time.

By converting timestamps into date objects, the system can reliably determine:

* the **latest portfolio state**
* the **previous state**
* how the system has changed over time

This is a small but important detail.

Without deterministic ordering, portfolio analysis becomes unreliable.

# 2. Portfolio Rollup

```python
def build_portfolio_rollup(...)
```

This function creates a **summary of the experimentation program**.

It answers questions such as:

* How many experiments are running?
* How many are at risk?
* Are we learning quickly enough?
* Is the experimentation portfolio improving or deteriorating?

## Handling Missing Snapshots

If no snapshots exist, the system returns a **safe default state**.

Example:

```python
"portfolio_status": "unknown"
```

This ensures the orchestrator **never produces misleading intelligence** when data is missing.

Instead of fabricating insights, the system acknowledges uncertainty.

This is a critical trust-building behavior.

# 3. Snapshot Selection

Snapshots are sorted by date:

```python

```

The most recent snapshot becomes the **authoritative portfolio state**.

The second snapshot (if available) becomes the **baseline for trend comparison**.

This design enables the orchestrator to detect **directional change**, which is far more useful than static metrics.

````python
"""Portfolio and risk rollups for EPOv3."""

from datetime import date
from typing import Any, Dict, List


def _parse_snapshot_date(s: str) -> date:
    """Parse snapshot_date (YYYY-MM-DD) to date for ordering."""
    return date.fromisoformat(s) if s else date.min


def build_portfolio_rollup(
    snapshots: List[Dict[str, Any]],
) -> Dict[str, Any]:
    """
    Build portfolio rollup from snapshots. Uses chronologically latest snapshot;
    adds deltas vs prior snapshot when available.
    """
    if not snapshots:
        return {
            "latest_snapshot": None,
            "prior_snapshot": None,
            "active_experiments": 0,
            "experiments_at_risk": 0,
            "high_risk_experiments": 0,
            "portfolio_status": "unknown",
            "learning_velocity_last_30d": 0,
            "scale_decisions_ytd": 0,
            "iterate_decisions_ytd": 0,
            "stop_decisions_ytd": 0,
            "estimated_annual_value_in_flight_usd": 0,
            "avg_primary_metric_lift_pct": 0.0,
            "delta_experiments_at_risk": None,
            "delta_learning_velocity": None,
            "delta_portfolio_status": None,
        }

    sorted_snapshots = sorted(
        snapshots,
        key=lambda x: _parse_snapshot_date(x.get("snapshot_date") or ""),
        reverse=True,
    )
    latest = sorted_snapshots[0]
    prior = sorted_snapshots[1] if len(sorted_snapshots) > 1 else None

    rollup: Dict[str, Any] = {
        "latest_snapshot": latest,
        "prior_snapshot": prior,
        "snapshot_date": latest.get("snapshot_date"),
        "active_experiments": latest.get("active_experiments", 0),
        "completed_experiments": latest.get("completed_experiments", 0),
        "planned_experiments": latest.get("planned_experiments", 0),
        "experiments_at_risk": latest.get("experiments_at_risk", 0),
        "high_risk_experiments": latest.get("high_risk_experiments", 0),
        "portfolio_status": latest.get("portfolio_status", "unknown"),
        "learning_velocity_last_30d": latest.get("learning_velocity_last_30d", 0),
        "scale_decisions_ytd": latest.get("scale_decisions_ytd", 0),
        "iterate_decisions_ytd": latest.get("iterate_decisions_ytd", 0),
        "stop_decisions_ytd": latest.get("stop_decisions_ytd", 0),
        "estimated_annual_value_in_flight_usd": latest.get("estimated_annual_value_in_flight_usd", 0),
        "avg_primary_metric_lift_pct": latest.get("avg_primary_metric_lift_pct", 0.0),
        "human_reviews_last_30d": latest.get("human_reviews_last_30d", 0),
        "top_risk_type": latest.get("top_risk_type", "none"),
    }

    if prior:
        rollup["delta_experiments_at_risk"] = (
            latest.get("experiments_at_risk", 0) - prior.get("experiments_at_risk", 0)
        )
        rollup["delta_learning_velocity"] = (
            latest.get("learning_velocity_last_30d", 0)
            - prior.get("learning_velocity_last_30d", 0)
        )
        rollup["delta_portfolio_status"] = (
            latest.get("portfolio_status") != prior.get("portfolio_status")
        )
    else:
        rollup["delta_experiments_at_risk"] = None
        rollup["delta_learning_velocity"] = None
        rollup["delta_portfolio_status"] = None

    return rollup


def build_risk_rollup(risk_events: List[Dict[str, Any]]) -> Dict[str, Any]:
    """Build risk rollup: by severity, by risk_type, open count."""
    open_statuses = {"open", "investigating", "blocked"}
    by_severity: Dict[str, int] = {}
    by_risk_type: Dict[str, int] = {}
    open_count = 0
    critical_count = 0

    for e in risk_events:
        sev = e.get("severity") or "unknown"
        by_severity[sev] = by_severity.get(sev, 0) + 1
        if sev == "critical":
            critical_count += 1
        rt = e.get("risk_type") or "unknown"
        by_risk_type[rt] = by_risk_type.get(rt, 0) + 1
        if (e.get("status") or "").lower() in open_statuses:
            open_count += 1

    return {
        "total_events": len(risk_events),
        "open_count": open_count,
        "critical_count": critical_count,
        "by_severity": by_severity,
        "by_risk_type": by_risk_type,
        "events": risk_events,
    }


def build_executive_triggers(
    portfolio_rollup: Dict[str, Any],
    risk_rollup: Dict[str, Any],
    *,
    experiments_at_risk_critical: int = 2,
    high_risk_experiments_critical: int = 1,
    open_risk_events_critical: int = 5,
    learning_velocity_min_target: int = 2,
) -> List[Dict[str, Any]]:
    """Derive executive triggers from rollups and config thresholds."""
    triggers: List[Dict[str, Any]] = []
    at_risk = portfolio_rollup.get("experiments_at_risk", 0)
    high_risk = portfolio_rollup.get("high_risk_experiments", 0)
    open_events = risk_rollup.get("open_count", 0)
    learning_vel = portfolio_rollup.get("learning_velocity_last_30d", 0)

    if at_risk >= experiments_at_risk_critical:
        triggers.append({
            "trigger_type": "experiments_at_risk",
            "message": f"Experiments at risk ({at_risk}) at or above critical threshold ({experiments_at_risk_critical}).",
            "value": at_risk,
            "threshold": experiments_at_risk_critical,
        })
    if high_risk >= high_risk_experiments_critical:
        triggers.append({
            "trigger_type": "high_risk_experiments",
            "message": f"High-risk experiments ({high_risk}) at or above critical threshold ({high_risk_experiments_critical}).",
            "value": high_risk,
            "threshold": high_risk_experiments_critical,
        })
    if open_events >= open_risk_events_critical:
        triggers.append({
            "trigger_type": "open_risk_events",
            "message": f"Open risk events ({open_events}) at or above critical threshold ({open_risk_events_critical}).",
            "value": open_events,
            "threshold": open_risk_events_critical,
        })
    if learning_velocity_min_target and learning_vel < learning_velocity_min_target:
        triggers.append({
            "trigger_type": "learning_velocity_below_target",
            "message": f"Learning velocity last 30d ({learning_vel}) below target ({learning_velocity_min_target}).",
            "value": learning_vel,
            "threshold": learning_velocity_min_target,
        })
    return triggers
````
