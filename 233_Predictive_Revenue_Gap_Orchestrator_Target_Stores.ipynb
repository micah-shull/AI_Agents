````python
"""LLM enhancement utilities for Predictive Revenue Gap Orchestrator

Following ORCHESTRATOR_AGENTS_GUIDE_3.md Phase 8 pattern:
- LLM as enhancement layer (not requirement)
- Graceful fallback to rule-based
- Cost-controlled (only enhance top N)
- Configurable on/off
"""

from typing import Dict, List, Any, Optional
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate


def generate_llm_rationale(
    gap: Dict[str, Any],
    customer_data: Dict[str, Any],
    baseline: Dict[str, Any],
    opportunity: Dict[str, Any],
    config: Any
) -> Optional[str]:
    """
    Generate personalized rationale using LLM.
    Falls back to rule-based if LLM fails.

    Args:
        gap: Revenue gap dictionary
        customer_data: Customer demographic data
        baseline: Revenue baseline metrics
        opportunity: Recovery opportunity dictionary
        config: Configuration object with LLM settings

    Returns:
        Personalized rationale string, or None if LLM fails
    """
    try:
        # Build context for LLM
        customer_id = gap.get("customer_id", "Unknown")
        gap_type = gap.get("gap_type", "unknown").replace("_", " ")
        gap_amount = abs(gap.get("gap_amount", 0.0))
        severity = gap.get("severity", "unknown")

        customer_age = customer_data.get("age", "Unknown")
        loyalty_status = "loyalty member" if customer_data.get("loyalty_member") else "non-member"
        household_size = customer_data.get("household_size", "Unknown")

        baseline_avg = baseline.get("baseline_weeks_avg", 0.0)
        recent_avg = baseline.get("recent_weeks_avg", 0.0)
        trend = baseline.get("revenue_trend", "unknown")

        opportunity_type = opportunity.get("opportunity_type", "unknown").replace("_", " ")
        potential_recovery = opportunity.get("potential_revenue", 0.0)

        # Create prompt
        prompt = ChatPromptTemplate.from_messages([
            ("system", """You are a caring customer success specialist for a retail business.
Your goal is to create personalized, empathetic rationales for revenue gaps that help
customers understand their situation and feel valued. Be concise (2-3 sentences), warm,
and focus on the customer's value to the business."""),
            ("human", """Create a personalized rationale for this revenue gap:

Customer: {customer_id} (Age: {age}, {loyalty_status}, Household: {household_size})
Gap Type: {gap_type}
Severity: {severity}
Gap Amount: \${gap_amount:.2f}
Baseline Average: \${baseline_avg:.2f}/week
Recent Average: \${recent_avg:.2f}/week
Trend: {trend}
Opportunity Type: {opportunity_type}
Potential Recovery: \${potential_recovery:.2f}

Write a brief, personalized rationale (2-3 sentences) that:
1. Acknowledges the customer's value
2. Explains the situation clearly
3. Shows empathy and understanding
4. Focuses on positive next steps

Keep it professional but warm, and avoid being pushy or salesy.""")
        ])

        # Initialize LLM
        llm = ChatOpenAI(
            model=config.llm_model,
            temperature=config.temperature
        )

        # Generate rationale
        chain = prompt | llm
        response = chain.invoke({
            "customer_id": customer_id,
            "age": customer_age,
            "loyalty_status": loyalty_status,
            "household_size": household_size,
            "gap_type": gap_type,
            "severity": severity,
            "gap_amount": gap_amount,
            "baseline_avg": baseline_avg,
            "recent_avg": recent_avg,
            "trend": trend,
            "opportunity_type": opportunity_type,
            "potential_recovery": potential_recovery
        })

        rationale = response.content.strip()

        # Validate rationale (basic check)
        if rationale and len(rationale) > 20:  # Minimum length check
            return rationale
        else:
            return None

    except Exception as e:
        print(f"LLM rationale generation failed: {e}, using rule-based rationale")
        return None


def generate_llm_opportunity_description(
    gap: Dict[str, Any],
    customer_data: Dict[str, Any],
    opportunity: Dict[str, Any],
    config: Any
) -> Optional[str]:
    """
    Generate personalized opportunity description using LLM.
    Falls back to rule-based if LLM fails.

    Args:
        gap: Revenue gap dictionary
        customer_data: Customer demographic data
        opportunity: Recovery opportunity dictionary
        config: Configuration object with LLM settings

    Returns:
        Personalized opportunity description, or None if LLM fails
    """
    try:
        customer_id = gap.get("customer_id", "Unknown")
        opportunity_type = opportunity.get("opportunity_type", "unknown").replace("_", " ")
        recommended_actions = opportunity.get("recommended_actions", [])
        potential_revenue = opportunity.get("potential_revenue", 0.0)

        loyalty_status = "loyalty member" if customer_data.get("loyalty_member") else "non-member"

        # Create prompt
        prompt = ChatPromptTemplate.from_messages([
            ("system", """You are a customer success specialist. Create brief, actionable
opportunity descriptions that explain what actions can be taken to recover revenue.
Be specific and customer-focused."""),
            ("human", """Create a brief opportunity description for this customer:

Customer: {customer_id} ({loyalty_status})
Opportunity Type: {opportunity_type}
Potential Recovery: \${potential_revenue:.2f}
Recommended Actions: {actions}

Write 1-2 sentences describing the opportunity in a way that's clear and actionable
for a customer success team member.""")
        ])

        # Initialize LLM
        llm = ChatOpenAI(
            model=config.llm_model,
            temperature=config.temperature
        )

        # Generate description
        chain = prompt | llm
        response = chain.invoke({
            "customer_id": customer_id,
            "loyalty_status": loyalty_status,
            "opportunity_type": opportunity_type,
            "potential_revenue": potential_revenue,
            "actions": ", ".join(recommended_actions)
        })

        description = response.content.strip()

        if description and len(description) > 15:
            return description
        else:
            return None

    except Exception as e:
        print(f"LLM opportunity description generation failed: {e}")
        return None


def enhance_gaps_with_llm(
    top_gaps: List[Dict[str, Any]],
    all_customers: List[Dict[str, Any]],
    customer_revenue_baseline: Dict[str, Dict[str, Any]],
    revenue_recovery_opportunities: List[Dict[str, Any]],
    config: Any,
    max_gaps: int = 10
) -> List[Dict[str, Any]]:
    """
    Enhance top gaps with LLM-generated content.

    Only enhances top N gaps (cost control).
    Falls back to rule-based if LLM fails.

    Args:
        top_gaps: Top priority gaps to enhance
        all_customers: All customer records
        customer_revenue_baseline: Baseline metrics per customer
        revenue_recovery_opportunities: Recovery opportunities
        config: Configuration object
        max_gaps: Maximum number of gaps to enhance (cost control)

    Returns:
        List of enhanced gaps
    """
    if not config.enable_llm_recommendations:
        return top_gaps

    # Create lookup dictionaries
    customer_lookup = {c["customer_id"]: c for c in all_customers}
    baseline_lookup = customer_revenue_baseline
    opportunity_lookup = {o["customer_id"]: o for o in revenue_recovery_opportunities}

    enhanced_gaps = []

    # Only enhance top N gaps
    gaps_to_enhance = top_gaps[:max_gaps]

    for gap in top_gaps:
        customer_id = gap.get("customer_id")
        customer_data = customer_lookup.get(customer_id, {})
        baseline = baseline_lookup.get(customer_id, {})
        opportunity = opportunity_lookup.get(customer_id, {})

        # Enhance if in top N
        if gap in gaps_to_enhance:
            try:
                # Generate LLM rationale
                llm_rationale = generate_llm_rationale(
                    gap, customer_data, baseline, opportunity, config
                )

                # Generate LLM opportunity description
                llm_description = generate_llm_opportunity_description(
                    gap, customer_data, opportunity, config
                )

                # Add LLM enhancements to gap
                enhanced_gap = gap.copy()

                if llm_rationale:
                    enhanced_gap["llm_rationale"] = llm_rationale
                    enhanced_gap["rationale_source"] = "llm"
                else:
                    enhanced_gap["rationale_source"] = "rule_based"

                if llm_description:
                    enhanced_gap["llm_opportunity_description"] = llm_description
                    if "opportunity" in enhanced_gap:
                        enhanced_gap["opportunity"]["llm_description"] = llm_description

            except Exception as e:
                # Fallback: continue with rule-based
                print(f"LLM enhancement failed for customer {customer_id}: {e}")
                enhanced_gap = gap.copy()
                enhanced_gap["rationale_source"] = "rule_based"
        else:
            # Not in top N, keep rule-based
            enhanced_gap = gap.copy()
            enhanced_gap["rationale_source"] = "rule_based"

        enhanced_gaps.append(enhanced_gap)

    return enhanced_gaps

````
