````python
"""
Revenue Assurance Orchestrator v2 â€” LangGraph workflow.
Linear: goal â†’ planning â†’ data_loading â†’ reconciliation â†’ detection â†’ prioritization â†’ report â†’ END.
"""
from pathlib import Path

from langgraph.graph import StateGraph, END

from config import RAAv2OrchestratorConfig, RAAv2OrchestratorState
from agents.raa_v2.orchestrator.nodes import (
    goal_node,
    planning_node,
    make_data_loading_node,
    reconciliation_node,
    detection_node,
    make_prioritization_node,
    make_report_node,
)


def _project_root() -> Path:
    """Resolve project root (orchestrator.py is agents/raa_v2/orchestrator/orchestrator.py)."""
    return Path(__file__).resolve().parent.parent.parent.parent


def create_raa_v2_orchestrator(config: RAAv2OrchestratorConfig | None = None):
    """Build and compile the RAA v2 graph. Uses config for closure nodes."""
    if config is None:
        config = RAAv2OrchestratorConfig()
    project_root = str(_project_root())
    data_loading_node = make_data_loading_node(config, project_root)
    prioritization_node = make_prioritization_node(config)
    report_node = make_report_node(config, project_root)

    workflow = StateGraph(RAAv2OrchestratorState)
    workflow.add_node("goal", goal_node)
    workflow.add_node("planning", planning_node)
    workflow.add_node("data_loading", data_loading_node)
    workflow.add_node("reconciliation", reconciliation_node)
    workflow.add_node("detection", detection_node)
    workflow.add_node("prioritization", prioritization_node)
    workflow.add_node("report_generation", report_node)

    workflow.set_entry_point("goal")
    workflow.add_edge("goal", "planning")
    workflow.add_edge("planning", "data_loading")
    workflow.add_edge("data_loading", "reconciliation")
    workflow.add_edge("reconciliation", "detection")
    workflow.add_edge("detection", "prioritization")
    workflow.add_edge("prioritization", "report_generation")
    workflow.add_edge("report_generation", END)

    return workflow.compile()
````
