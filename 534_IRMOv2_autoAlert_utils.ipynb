````python
"""Automated alerting utilities for Integration & Risk Management Orchestrator"""

from typing import Dict, List, Any, Optional
from datetime import datetime


def create_alert(
    alert_type: str,
    severity: str,
    title: str,
    description: str,
    agent_id: Optional[str] = None,
    system_id: Optional[str] = None,
    workflow_id: Optional[str] = None,
    priority_score: Optional[float] = None,
    metadata: Optional[Dict[str, Any]] = None
) -> Dict[str, Any]:
    """Create an alert"""
    return {
        "alert_id": f"alert_{datetime.now().strftime('%Y%m%d_%H%M%S')}_{alert_type}",
        "alert_type": alert_type,
        "severity": severity,
        "title": title,
        "description": description,
        "agent_id": agent_id,
        "system_id": system_id,
        "workflow_id": workflow_id,
        "priority_score": priority_score or 0.0,
        "created_at": datetime.now().isoformat(),
        "metadata": metadata or {}
    }


def generate_critical_issue_alerts(
    prioritized_issues: List[Dict[str, Any]],
    alert_threshold: float = 30.0
) -> List[Dict[str, Any]]:
    """Generate alerts for critical issues"""
    alerts = []

    for issue in prioritized_issues:
        priority_score = issue.get("priority_score", 0.0)
        severity = issue.get("severity", "medium")
        issue_type = issue.get("type", "unknown")

        # Alert on high priority or critical severity issues
        if priority_score >= alert_threshold or severity in ["critical", "high"]:
            alert = create_alert(
                alert_type=issue_type,
                severity=severity,
                title=f"{issue_type.title()} Issue: {issue.get('description', 'N/A')[:60]}",
                description=issue.get("description", "N/A"),
                agent_id=issue.get("agent_id"),
                system_id=issue.get("system_id"),
                workflow_id=issue.get("workflow_id"),
                priority_score=priority_score,
                metadata={
                    "issue_id": issue.get("issue_id"),
                    "recommended_action": issue.get("recommended_action"),
                    "recommendations": issue.get("recommendations", [])
                }
            )
            alerts.append(alert)

    return sorted(alerts, key=lambda x: x.get("priority_score", 0.0), reverse=True)


def generate_value_leakage_alerts(
    value_leakage_analysis: List[Dict[str, Any]],
    threshold: float = 50.0
) -> List[Dict[str, Any]]:
    """Generate alerts for value leakage issues"""
    alerts = []

    for analysis in value_leakage_analysis:
        leakage_score = analysis.get("value_leakage_score", 0.0)
        agent_id = analysis.get("agent_id")

        if leakage_score >= threshold:
            roi_gap = analysis.get("roi_gap", 0.0)
            roi_gap_percent = analysis.get("roi_gap_percent", 0.0)

            alert = create_alert(
                alert_type="value_leakage",
                severity="critical" if leakage_score >= 80.0 else "high",
                title=f"Value Leakage Alert: {agent_id}",
                description=f"Agent {agent_id} has value leakage score of {leakage_score:.1f} (ROI gap: {roi_gap_percent:.1f}%)",
                agent_id=agent_id,
                priority_score=leakage_score,
                metadata={
                    "leakage_score": leakage_score,
                    "roi_gap": roi_gap,
                    "roi_gap_percent": roi_gap_percent,
                    "cost_overrun": analysis.get("cost_overrun", 0.0),
                    "recommendations": analysis.get("recommendations", [])
                }
            )
            alerts.append(alert)

    return alerts


def generate_action_item_alerts(
    action_items_analysis: Dict[str, Any]
) -> List[Dict[str, Any]]:
    """Generate alerts for overdue action items"""
    alerts = []

    overdue_items = action_items_analysis.get("overdue_items", [])

    for item in overdue_items:
        agent_id = item.get("agent_id")
        days_open = item.get("days_open", 0)
        action_item = item.get("action_item", "N/A")

        alert = create_alert(
            alert_type="action_item_overdue",
            severity="critical" if days_open > 30 else "high",
            title=f"Overdue Action Item: {agent_id}",
            description=f"Action item for {agent_id} is {days_open} days overdue: {action_item[:80]}",
            agent_id=agent_id,
            priority_score=min(100.0, days_open * 2.0),  # Higher score for longer overdue
            metadata={
                "action_id": item.get("action_id"),
                "review_id": item.get("review_id"),
                "reviewer": item.get("reviewer"),
                "days_open": days_open,
                "action_item": action_item
            }
        )
        alerts.append(alert)

    return alerts


def generate_integration_health_alerts(
    integration_health: List[Dict[str, Any]]
) -> List[Dict[str, Any]]:
    """Generate alerts for degraded/critical integrations"""
    alerts = []

    for health in integration_health:
        health_status = health.get("health_status", "healthy")
        system_id = health.get("system_id")
        overall_score = health.get("overall_score", 100.0)

        if health_status in ["degraded", "critical"]:
            affected_agents = health.get("affected_agents", [])

            alert = create_alert(
                alert_type="integration_health",
                severity=health_status,
                title=f"Integration Health Alert: {system_id}",
                description=f"Integration {system_id} is {health_status} (score: {overall_score:.1f}). Affects: {', '.join(affected_agents)}",
                system_id=system_id,
                priority_score=100.0 - overall_score,
                metadata={
                    "health_status": health_status,
                    "overall_score": overall_score,
                    "affected_agents": affected_agents,
                    "issues": health.get("issues", [])
                }
            )
            alerts.append(alert)

    return alerts


def generate_all_alerts(
    prioritized_issues: List[Dict[str, Any]],
    value_leakage_analysis: List[Dict[str, Any]],
    action_items_analysis: Dict[str, Any],
    integration_health: List[Dict[str, Any]],
    alert_threshold: float = 30.0
) -> Dict[str, Any]:
    """Generate all alerts"""
    alerts = []

    # Critical issue alerts
    alerts.extend(generate_critical_issue_alerts(prioritized_issues, alert_threshold))

    # Value leakage alerts
    alerts.extend(generate_value_leakage_alerts(value_leakage_analysis))

    # Action item alerts
    alerts.extend(generate_action_item_alerts(action_items_analysis))

    # Integration health alerts
    alerts.extend(generate_integration_health_alerts(integration_health))

    # Sort by priority score
    alerts = sorted(alerts, key=lambda x: x.get("priority_score", 0.0), reverse=True)

    # Summary
    critical_count = len([a for a in alerts if a.get("severity") == "critical"])
    high_count = len([a for a in alerts if a.get("severity") == "high"])

    return {
        "total_alerts": len(alerts),
        "critical_alerts": critical_count,
        "high_alerts": high_count,
        "alerts": alerts
    }
````
