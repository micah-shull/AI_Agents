# 📘 **WDO v3 — Metric Framework**

### *Deterministic Workforce Transformation Intelligence*

---

## **Purpose**

This document defines the **official metric framework** for the Workforce Development Orchestrator v3.

The purpose of this framework is to ensure that workforce transformation is:

* **measurable**
* **traceable over time**
* **explainable to leadership**
* **actionable at the executive level**

The system does not aim to produce static workforce analytics.

The system answers:

> **“Is the workforce becoming more prepared or more at risk over time — and why?”**

---

## 🧭 **Design Principles**

The metric framework follows five core principles:

### 1. **Trend Over Snapshot**

Metrics must support **change over time**, not just current state.

---

### 2. **Deterministic Over Black Box**

All metrics must be:

* reproducible
* explainable
* rule-based

---

### 3. **Minimal but Sufficient**

The system uses:

* **pre-aggregated snapshots**
* **delta-based calculations**

Avoid:

* raw logs
* unnecessary granularity
* bloated time series

---

### 4. **Executive Interpretability**

Every metric must translate into:

* risk
* readiness
* velocity
* stability
* ROI

---

### 5. **Causal Framing (Not Just Measurement)**

Metrics must support answering:

> **“What changed — and what caused it?”**

---

# 🏗 **Metric Architecture**

The framework is structured into **three layers**:

---

## **Layer 1 — Core Workforce Metrics (State)**

These describe the current condition of the workforce.

### 1. Workforce Readiness Score

**Definition:**
Composite measure of workforce preparedness for future work.

**Source:**
`workforce_snapshots.json`

**Interpretation:**

* High → workforce is prepared
* Low → workforce is at risk

---

### 2. Skill Gap Index

**Definition:**
Normalized measure of skill deficiencies relative to role requirements.

**Source:**
Derived from:

* skill gaps dataset

**Interpretation:**

* High → significant capability gaps
* Low → workforce aligned with future needs

---

### 3. Automation Exposure

**Definition:**
Average exposure of tasks to automation risk.

**Source:**
Derived from:

* task automation risk scores

**Interpretation:**

* High → roles at risk of disruption
* Low → roles resilient to automation

---

### 4. Reskilling Progress

**Definition:**
Percentage of workforce actively progressing through reskilling.

**Source:**
`workforce_snapshots.json`

**Interpretation:**

* High → workforce adapting
* Low → transformation lagging

---

### 5. Attrition Risk Index

**Definition:**
Likelihood of workforce loss during transformation.

**Source:**
`workforce_snapshots.json`

**Interpretation:**

* High → instability risk
* Low → workforce stable

---

---

## **Layer 2 — Trend Metrics (Change Over Time)**

These are the **core of v3**.

They measure **direction and velocity**.

---

### 6. Workforce Readiness Delta

```text
readiness_delta =
    current_readiness - previous_readiness
```

**Interpretation:**

* Positive → improving workforce
* Negative → deteriorating workforce

---

### 7. Reskilling Velocity ⭐

```text
reskilling_velocity =
    current_reskilling_progress - previous_reskilling_progress
```

**Interpretation:**

* High → rapid adaptation
* Low → stagnation

---

### 8. Skill Gap Trend

```text
skill_gap_delta =
    current_skill_gap_index - previous_skill_gap_index
```

**Interpretation:**

* Negative → gaps closing (good)
* Positive → gaps widening (risk)

---

### 9. Automation Pressure Trend

```text
automation_delta =
    current_automation_exposure - previous_automation_exposure
```

**Interpretation:**

* Positive → increasing disruption pressure
* Negative → stabilization

---

### 10. Workforce Stability Trend

```text
attrition_delta =
    current_attrition_risk - previous_attrition_risk
```

**Interpretation:**

* Positive → rising instability
* Negative → improving retention outlook

---

---

## **Layer 3 — Efficiency & Investment Metrics**

These evaluate whether interventions are effective.

---

### 11. Training ROI Signal ⭐

```text
roi_signal =
    expected_productivity_uplift / cost_usd
```

**Source:**

* training investments

**Interpretation:**

* High → efficient investment
* Low → poor return

---

### 12. Retraining Coverage Ratio

**Definition:**
Percentage of at-risk employees receiving training.

**Source:**

* workforce controls

**Interpretation:**

* High → proactive mitigation
* Low → exposure risk

---

# 🧠 **Derived Executive Signals**

Metrics are translated into **decision-ready signals**.

---

## 1. Workforce Trajectory Status

| Condition                    | Status        |
| ---------------------------- | ------------- |
| readiness ↑ AND skill gaps ↓ | **Improving** |
| mixed signals                | **Stable**    |
| readiness ↓ AND skill gaps ↑ | **At Risk**   |

---

## 2. Transformation Health

| Condition                         | Status                  |
| --------------------------------- | ----------------------- |
| reskilling ↑ offsets automation ↑ | **Healthy Adaptation**  |
| automation ↑ exceeds reskilling   | **Transformation Risk** |

---

## 3. Workforce Stability Signal

| Condition        | Status               |
| ---------------- | -------------------- |
| attrition risk ↓ | **Stable Workforce** |
| attrition risk ↑ | **Retention Risk**   |

---

---

# ⚙️ **Deterministic Classification Rules**

Example:

```python
if readiness_delta > 0 and skill_gap_delta < 0:
    trajectory = "improving"
elif readiness_delta < 0 and skill_gap_delta > 0:
    trajectory = "at_risk"
else:
    trajectory = "stable"
```

# 🧾 **Metric Constraints**

To maintain system quality:

* Use **max 2–3 snapshots per entity**
* Avoid raw event-level data
* Prefer **aggregated indices**
* All metrics must:

  * support delta calculation
  * map to executive meaning
