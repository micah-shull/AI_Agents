````python
"""Prioritization utilities for Workforce Development Orchestrator

Following the pattern: Utilities implement, nodes orchestrate.
Uses toolshed.prioritization for reusable prioritization logic.
"""

from typing import Dict, List, Any
from toolshed.prioritization import (
    calculate_priority_score,
    normalize_severity
)
from config import WorkforceDevelopmentOrchestratorConfig


def prioritize_skill_gaps(
    skill_gap_analysis: List[Dict[str, Any]],
    config: WorkforceDevelopmentOrchestratorConfig
) -> List[Dict[str, Any]]:
    """Prioritize skill gaps using toolshed prioritization"""
    prioritized = []

    for gap in skill_gap_analysis:
        # Use existing priority_score if available, otherwise calculate
        if "priority_score" not in gap:
            # Calculate using toolshed
            priority_score = calculate_priority_score(
                issue=gap,
                weights={
                    "priority_score": 0.4,  # Use existing priority_score if present
                    "urgency_score": 0.3,
                    "automation_risk": 0.3
                },
                field_mappers={
                    "priority_score": lambda i, c: normalize_severity(i.get("priority", "medium")),
                    "urgency_score": lambda i, c: i.get("urgency_score", 0.5) * 100.0,
                    "automation_risk": lambda i, c: (
                        100.0 if i.get("automation_risk_context") == "high"
                        else 50.0 if i.get("automation_risk_context") == "medium"
                        else 25.0
                    )
                }
            )
            gap["priority_score"] = priority_score

        prioritized.append(gap)

    # Sort by priority_score (descending)
    prioritized.sort(key=lambda x: x.get("priority_score", 0.0), reverse=True)

    return prioritized


def prioritize_learning_paths(
    learning_path_recommendations: List[Dict[str, Any]],
    config: WorkforceDevelopmentOrchestratorConfig
) -> List[Dict[str, Any]]:
    """Prioritize learning path recommendations"""
    prioritized = []

    for rec in learning_path_recommendations:
        # Calculate priority score using toolshed
        priority_score = calculate_priority_score(
            issue=rec,
            weights={
                "match_score": 0.4,
                "priority": 0.3,
                "prerequisites_met": 0.2,
                "urgency": 0.1
            },
            field_mappers={
                "match_score": lambda i, c: i.get("match_score", 0.0) * 100.0,
                "priority": lambda i, c: normalize_severity(i.get("priority", "medium")),
                "prerequisites_met": lambda i, c: 100.0 if i.get("prerequisites_met") else 50.0,
                "urgency": lambda i, c: 75.0  # Default urgency
            }
        )
        rec["priority_score"] = priority_score
        prioritized.append(rec)

    # Sort by priority_score (descending)
    prioritized.sort(key=lambda x: x.get("priority_score", 0.0), reverse=True)

    return prioritized


def prioritize_role_evolutions(
    role_evolution_recommendations: List[Dict[str, Any]],
    config: WorkforceDevelopmentOrchestratorConfig
) -> List[Dict[str, Any]]:
    """Prioritize role evolution recommendations"""
    prioritized = []

    for rec in role_evolution_recommendations:
        # Use existing readiness_score as base, enhance with toolshed
        readiness_score = rec.get("readiness_score", 0.0) * 100.0

        priority_score = calculate_priority_score(
            issue=rec,
            weights={
                "readiness_score": 0.5,
                "implementation_priority": 0.3,
                "affected_employees": 0.2
            },
            field_mappers={
                "readiness_score": lambda i, c: i.get("readiness_score", 0.0) * 100.0,
                "implementation_priority": lambda i, c: normalize_severity(
                    i.get("implementation_priority", "medium")
                ),
                "affected_employees": lambda i, c: min(
                    100.0,
                    len(i.get("affected_employees", [])) * 10.0
                )
            }
        )
        rec["priority_score"] = priority_score
        prioritized.append(rec)

    # Sort by priority_score (descending)
    prioritized.sort(key=lambda x: x.get("priority_score", 0.0), reverse=True)

    return prioritized


def calculate_workforce_summary(
    employees: List[Dict[str, Any]],
    roles: List[Dict[str, Any]],
    tasks: List[Dict[str, Any]],
    skill_gap_analysis: List[Dict[str, Any]],
    learning_path_recommendations: List[Dict[str, Any]],
    role_evolution_recommendations: List[Dict[str, Any]],
    automation_risk_analysis: List[Dict[str, Any]]
) -> Dict[str, Any]:
    """Calculate workforce summary metrics"""
    # Count employees at risk (high automation risk)
    employees_at_risk = 0
    for risk in automation_risk_analysis:
        if risk.get("risk_level") == "high":
            employees_at_risk += len(risk.get("affected_employees", []))

    # Count high priority gaps
    high_priority_gaps = [
        g for g in skill_gap_analysis
        if g.get("priority") == "high" or g.get("priority_score", 0.0) > 70.0
    ]

    # Calculate overall workforce readiness score
    # Based on: automation risk, skill gaps, learning path coverage
    total_employees = len(employees)
    employees_with_gaps = len(set(g.get("employee_id") for g in skill_gap_analysis))
    employees_with_recommendations = len(set(
        r.get("employee_id") for r in learning_path_recommendations
    ))

    # Simple readiness calculation
    if total_employees > 0:
        gap_coverage = (employees_with_gaps / total_employees) * 100.0
        recommendation_coverage = (employees_with_recommendations / total_employees) * 100.0
        readiness_score = (gap_coverage + recommendation_coverage) / 2.0
    else:
        readiness_score = 0.0

    summary = {
        "total_employees": total_employees,
        "total_roles": len(roles),
        "total_tasks": len(tasks),
        "employees_at_risk": employees_at_risk,
        "total_skill_gaps": len(skill_gap_analysis),
        "high_priority_gaps": len(high_priority_gaps),
        "total_learning_recommendations": len(learning_path_recommendations),
        "roles_requiring_evolution": len(role_evolution_recommendations),
        "overall_workforce_readiness_score": round(readiness_score, 2)
    }

    return summary

````
