````python
"""
Build executive-style recommendations report: one-view, traceability, top opportunities, trend.
"""
from typing import Any, Dict, List, Optional


def build_recommendations_report(
    customer_id: str,
    customer_data: Dict[str, Any],
    top_opportunities: List[Dict[str, Any]],
    opportunity_summary: Dict[str, Any],
    offer_performance_rollup: Optional[Dict[str, Any]] = None,
    customers_count: Optional[int] = None,
    products_count: Optional[int] = None,
    offers_count: Optional[int] = None,
    responses_count: Optional[int] = None,
    data_snapshot_loaded_at: Optional[str] = None,
    validation_warnings: Optional[List[str]] = None,
) -> str:
    """Build markdown report with one-view, traceability, top opportunities table, trend section."""
    lines = []
    name = customer_data.get("name") or customer_id
    tier = customer_data.get("loyalty_tier", "—")
    summary = opportunity_summary or {}
    total_cross = summary.get("total_cross_sell_opportunities", 0)
    total_upsell = summary.get("total_upsell_opportunities", 0)
    total_rev = summary.get("total_potential_revenue", 0)
    high_val = summary.get("high_value_opportunities", 0)

    # Title and one-view
    lines.append("# Cross-Sell & Upsell Recommendations")
    lines.append("")
    lines.append("## One-view")
    lines.append("")
    verdict = "Strong opportunities" if (total_cross + total_upsell) > 0 else "No opportunities identified"
    lines.append(f"- **Verdict:** {verdict}")
    lines.append(f"- **Customer:** {name} ({customer_id}) · Tier: {tier}")
    lines.append(f"- **Opportunities:** {total_cross} cross-sell, {total_upsell} upsell · High-value: {high_val}")
    lines.append(f"- **Potential revenue (this set):** ${total_rev:,.2f}")
    lines.append("")

    # Traceability
    lines.append("## Traceability")
    lines.append("")
    trace_parts = []
    if data_snapshot_loaded_at:
        trace_parts.append(f"Data snapshot: {data_snapshot_loaded_at}")
    if customers_count is not None:
        trace_parts.append(f"Customers: {customers_count}")
    if products_count is not None:
        trace_parts.append(f"Products: {products_count}")
    if offers_count is not None:
        trace_parts.append(f"Offers: {offers_count}")
    if responses_count is not None:
        trace_parts.append(f"Responses: {responses_count}")
    if trace_parts:
        lines.append(" · ".join(trace_parts))
    else:
        lines.append("(No traceability data)")
    lines.append("")

    # Data quality (when loader reported validation warnings)
    if validation_warnings:
        lines.append("## Data quality")
        lines.append("")
        for w in validation_warnings[:10]:
            lines.append(f"- {w}")
        if len(validation_warnings) > 10:
            lines.append(f"- … and {len(validation_warnings) - 10} more")
        lines.append("")

    # Top opportunities table
    lines.append("## Top opportunities")
    lines.append("")
    if not top_opportunities:
        lines.append("No opportunities to show.")
    else:
        lines.append("| # | Product | Category | Type | Price | Score | Rationale |")
        lines.append("|---|--------|----------|------|-------|-------|-----------|")
        for i, o in enumerate(top_opportunities, 1):
            product_name = (o.get("product_name") or "")[:40]
            category = o.get("category", "")
            rec_type = o.get("recommendation_type", "")
            price = o.get("price", 0)
            score = o.get("raw_score", 0)
            rationale = (o.get("rationale") or "")[:50]
            lines.append(f"| {i} | {product_name} | {category} | {rec_type} | ${price:,.2f} | {score:.1f} | {rationale} |")
    lines.append("")

    # Offer performance / trend
    if offer_performance_rollup:
        lines.append("## Offer performance (trend)")
        lines.append("")
        lines.append(f"- **Sent:** {offer_performance_rollup.get('total_sent', 0)} · "
                    f"**Accepted:** {offer_performance_rollup.get('total_accepted', 0)} · "
                    f"**Acceptance rate:** {offer_performance_rollup.get('acceptance_rate_pct', 0):.1f}%")
        lines.append(f"- **Revenue generated:** ${offer_performance_rollup.get('total_revenue_generated', 0):,.2f} · "
                    f"**$/sent:** ${offer_performance_rollup.get('revenue_per_sent_offer', 0):,.2f} · "
                    f"**$/accepted:** ${offer_performance_rollup.get('revenue_per_accepted_offer', 0):,.2f}")
        lines.append("")
        by_type = offer_performance_rollup.get("by_offer_type") or {}
        if by_type:
            lines.append("### By offer type")
            lines.append("")
            lines.append("| Type | Sent | Accepted | Accept % | Revenue |")
            lines.append("|------|------|----------|----------|--------|")
            for t, v in sorted(by_type.items()):
                sent = v.get("sent", 0)
                acc = v.get("accepted", 0)
                rate = v.get("acceptance_rate_pct", 0)
                rev = v.get("revenue", 0)
                lines.append(f"| {t} | {sent} | {acc} | {rate:.1f}% | ${rev:,.2f} |")
            lines.append("")
        by_channel = offer_performance_rollup.get("by_channel") or {}
        if by_channel:
            lines.append("### By channel")
            lines.append("")
            lines.append("| Channel | Sent | Accepted | Accept % | Revenue |")
            lines.append("|---------|------|----------|----------|--------|")
            for ch, v in sorted(by_channel.items()):
                sent = v.get("sent", 0)
                acc = v.get("accepted", 0)
                rate = v.get("acceptance_rate_pct", 0)
                rev = v.get("revenue", 0)
                lines.append(f"| {ch} | {sent} | {acc} | {rate:.1f}% | ${rev:,.2f} |")
            lines.append("")

    # Summary metrics
    lines.append("## Summary")
    lines.append("")
    lines.append(f"- Routine completeness: {summary.get('routine_completeness_percent', 0):.1f}%")
    lines.append(f"- Replenishment urgency count: {summary.get('replenishment_urgency_count', 0)}")
    lines.append("")

    # Next step
    lines.append("## Next step")
    lines.append("")
    if top_opportunities:
        lines.append("1. Prioritize top opportunities above for outreach.")
        if offer_performance_rollup:
            lines.append("2. Use offer performance by type/channel to tailor messaging and channel mix.")
    else:
        lines.append("No actions this run; review customer segment or data freshness.")
    lines.append("")

    return "\n".join(lines)
````
