````python
"""Portfolio rollup for executive one-view (Third-Party Risk Orchestrator v2)."""

from datetime import datetime, timezone
from typing import Any, Dict, List


def _days_to_renewal(renewal_date_str: str | None) -> int | None:
    if not renewal_date_str:
        return None
    try:
        d = datetime.fromisoformat(renewal_date_str.replace("Z", "+00:00"))
        if d.tzinfo is None:
            d = d.replace(tzinfo=timezone.utc)
        return (d - datetime.now(timezone.utc)).days
    except (ValueError, TypeError):
        return None


def build_portfolio_rollup(
    vendor_risk_scores: List[Dict[str, Any]],
    triggered_rules: List[Dict[str, Any]],
    vendor_contracts_lookup: Dict[str, Dict[str, Any]],
    high_risk_score_threshold: float,
    critical_risk_score_threshold: float,
    renewal_window_days: int,
) -> Dict[str, Any]:
    """Build one-view rollup for executive report."""
    high = sum(1 for r in vendor_risk_scores if (r.get("overall_risk_score") or 0) >= high_risk_score_threshold)
    critical = sum(1 for r in vendor_risk_scores if (r.get("overall_risk_score") or 0) >= critical_risk_score_threshold)

    renewals_in_window = 0
    for vid, contract in vendor_contracts_lookup.items():
        days = _days_to_renewal(contract.get("renewal_date"))
        if days is not None and 0 <= days <= renewal_window_days:
            renewals_in_window += 1

    return {
        "total_vendors": len(vendor_risk_scores),
        "high_risk_count": high,
        "critical_risk_count": critical,
        "triggered_count": len(triggered_rules),
        "renewals_in_window": renewals_in_window,
    }
````
