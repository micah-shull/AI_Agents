````python
"""
ROI Calculation Utilities

Calculate ROI, revenue impact, and business value from evaluation results.
"""

from typing import Dict, Any, List


def calculate_revenue_impact(
    evaluation_summary: Dict[str, Any],
    evaluation_scores: List[Dict[str, Any]],
    agent_performance_summary: Dict[str, Any],
    config: Any
) -> Dict[str, float]:
    """
    Calculate revenue impact from evaluation results.

    Revenue impact comes from:
    1. Prevented churn (early detection of critical failures)
    2. CSAT improvements (better agent performance)
    3. Efficiency gains (reduced manual review)

    Args:
        evaluation_summary: Summary metrics
        evaluation_scores: Individual evaluation scores
        agent_performance_summary: Agent performance data
        config: Orchestrator config with revenue settings

    Returns:
        Dictionary with revenue impact breakdown
    """
    if not config.enable_roi_calculation:
        return {
            "prevented_churn_usd": 0.0,
            "csat_improvement_usd": 0.0,
            "efficiency_gains_usd": 0.0,
            "total_revenue_impact_usd": 0.0
        }

    # 1. Prevented Churn Value
    # Critical failures caught early prevent customer churn
    critical_agents = evaluation_summary.get("critical_agents", 0)
    failed_evaluations = evaluation_summary.get("total_evaluations", 0) - evaluation_summary.get("total_passed", 0)

    # Estimate: Each critical agent failure caught prevents ~1 customer churn
    # High-risk failures caught = critical agents * avg customers per agent
    high_risk_failures_prevented = critical_agents * 2  # Conservative estimate
    prevented_churn_value = high_risk_failures_prevented * config.avg_customer_lifetime_value_usd * config.churn_prevention_value_multiplier

    # 2. CSAT Improvement Value
    # Better agent performance improves customer satisfaction
    overall_pass_rate = evaluation_summary.get("overall_pass_rate", 0.0)
    avg_score = evaluation_summary.get("average_score", 0.0)

    # Estimate CSAT improvement based on pass rate and average score
    # Higher pass rate = better CSAT
    csat_improvement_points = (overall_pass_rate * 0.5) + (avg_score * 0.3)  # Normalized to 0-1 scale
    csat_improvement_value = csat_improvement_points * config.csat_improvement_value_per_point * evaluation_summary.get("total_evaluations", 0)

    # 3. Efficiency Gains
    # Reduced manual review time saves costs
    total_evaluations = evaluation_summary.get("total_evaluations", 0)
    # Estimate: Each evaluation that passes reduces manual review by 0.1 hours
    manual_review_hours_saved = total_evaluations * 0.1 * config.efficiency_gain_percent
    efficiency_gains_value = manual_review_hours_saved * config.manual_review_cost_per_hour_usd

    total_revenue_impact = prevented_churn_value + csat_improvement_value + efficiency_gains_value

    return {
        "prevented_churn_usd": round(prevented_churn_value, 2),
        "csat_improvement_usd": round(csat_improvement_value, 2),
        "efficiency_gains_usd": round(efficiency_gains_value, 2),
        "total_revenue_impact_usd": round(total_revenue_impact, 2)
    }


def calculate_roi(
    total_cost_usd: float,
    revenue_impact: Dict[str, float],
    config: Any
) -> Dict[str, Any]:
    """
    Calculate ROI metrics from cost and revenue impact.

    Args:
        total_cost_usd: Total cost of evaluation
        revenue_impact: Revenue impact breakdown
        config: Orchestrator config

    Returns:
        Dictionary with ROI analysis
    """
    if not config.enable_roi_calculation:
        return {
            "net_roi_usd": 0.0,
            "roi_percent": 0.0,
            "roi_ratio": 0.0,
            "roi_status": "neutral"
        }

    total_revenue = revenue_impact.get("total_revenue_impact_usd", 0.0)
    net_roi = total_revenue - total_cost_usd

    # Calculate ROI percentage: (revenue - cost) / cost * 100
    roi_percent = (net_roi / total_cost_usd * 100) if total_cost_usd > 0 else 0.0

    # Calculate ROI ratio: revenue / cost
    roi_ratio = (total_revenue / total_cost_usd) if total_cost_usd > 0 else 0.0

    # Determine ROI status
    if net_roi > 0:
        roi_status = "positive"
    elif net_roi < 0:
        roi_status = "negative"
    else:
        roi_status = "neutral"

    return {
        "net_roi_usd": round(net_roi, 2),
        "roi_percent": round(roi_percent, 2),
        "roi_ratio": round(roi_ratio, 2),
        "roi_status": roi_status
    }


def calculate_business_value_metrics(
    evaluation_summary: Dict[str, Any],
    cost_analysis: Dict[str, Any],
    roi_analysis: Dict[str, Any],
    config: Any
) -> Dict[str, Any]:
    """
    Calculate business value metrics and attribution.

    Args:
        evaluation_summary: Evaluation summary metrics
        cost_analysis: Cost analysis results
        roi_analysis: ROI analysis results
        config: Orchestrator config

    Returns:
        Dictionary with business value metrics
    """
    if not config.enable_roi_calculation:
        return {
            "business_impact_score": 0.0,
            "risk_reduction_percent": 0.0,
            "efficiency_gain_percent": 0.0,
            "quality_improvement_percent": 0.0,
            "value_per_evaluation": 0.0,
            "value_per_dollar_spent": 0.0
        }

    total_evaluations = evaluation_summary.get("total_evaluations", 0)
    total_cost = cost_analysis.get("total_cost_usd", 0.0)
    total_revenue = roi_analysis.get("total_revenue_impact_usd", 0.0)

    # Business Impact Score (0-1): Composite of pass rate, ROI, and risk reduction
    pass_rate = evaluation_summary.get("overall_pass_rate", 0.0)
    roi_score = min(roi_analysis.get("roi_ratio", 0.0) / 10.0, 1.0)  # Normalize ROI ratio
    risk_reduction = min(evaluation_summary.get("critical_agents", 0) / 10.0, 1.0)  # Normalize risk

    business_impact_score = (pass_rate * 0.4) + (roi_score * 0.4) + (risk_reduction * 0.2)

    # Risk Reduction: Percentage of critical issues caught early
    critical_agents = evaluation_summary.get("critical_agents", 0)
    total_agents = evaluation_summary.get("agents_evaluated", 1)
    risk_reduction_percent = (critical_agents / total_agents * 100) if total_agents > 0 else 0.0

    # Efficiency Gain: Based on pass rate (higher pass rate = less manual review)
    efficiency_gain_percent = pass_rate * 100 * config.efficiency_gain_percent

    # Quality Improvement: Based on average score improvement
    avg_score = evaluation_summary.get("average_score", 0.0)
    quality_improvement_percent = (avg_score - 0.5) * 100  # Assume 0.5 is baseline

    # Value metrics
    value_per_evaluation = total_revenue / total_evaluations if total_evaluations > 0 else 0.0
    value_per_dollar_spent = total_revenue / total_cost if total_cost > 0 else 0.0

    return {
        "business_impact_score": round(business_impact_score, 3),
        "risk_reduction_percent": round(risk_reduction_percent, 1),
        "efficiency_gain_percent": round(efficiency_gain_percent, 1),
        "quality_improvement_percent": round(quality_improvement_percent, 1),
        "value_per_evaluation": round(value_per_evaluation, 2),
        "value_per_dollar_spent": round(value_per_dollar_spent, 2)
    }
````
