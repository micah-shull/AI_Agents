````python
# ============================================================================
# Compliance Sentinel v2 â€” Enterprise AI Governance Orchestrator
# ============================================================================
# Monitors AI outputs, compliance findings, financial exposure, escalations,
# remediation, and human reviews. Portfolio rollup + executive triggers;
# board-ready report with one-view, segment table (R/Y/G), next steps.

class ComplianceSentinelV2State(TypedDict, total=False):
    """State for Compliance Sentinel v2 (AI governance command center)"""

    # Input
    data_dir: Optional[str]              # Data directory (default: agents/data_v2)
    reports_dir: Optional[str]          # Reports directory
    view_mode: Optional[str]           # "portfolio" | "system" (default: portfolio)

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Raw data (from loader)
    ai_output_events: List[Dict[str, Any]]
    ai_system_registry: List[Dict[str, Any]]
    compliance_findings: List[Dict[str, Any]]
    remediation_actions: List[Dict[str, Any]]
    escalation_events: List[Dict[str, Any]]
    financial_exposure_models: List[Dict[str, Any]]
    human_review_actions: List[Dict[str, Any]]
    portfolio_snapshots: List[Dict[str, Any]]

    # Lookups (built in data_loading node)
    systems_lookup: Dict[str, Dict[str, Any]]       # system_id -> system
    findings_lookup: Dict[str, Dict[str, Any]]     # finding_id -> finding
    exposure_lookup: Dict[str, Dict[str, Any]]     # finding_id -> exposure model

    # Rollups & triggers
    portfolio_rollup: Dict[str, Any]               # open_cases, total_exposure_usd, by_severity, by_system, etc.
    executive_triggers: List[Dict[str, Any]]       # Fired threshold triggers
    alerts: List[Dict[str, Any]]                   # Alerts created when triggers fire

    # Forensics
    data_snapshot_loaded_at: Optional[str]         # UTC ISO when data was loaded
    validation_warnings: List[str]

    # Output
    compliance_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class ComplianceSentinelV2Config:
    """Configuration for Compliance Sentinel v2"""
    reports_dir: str = "agents/output/compliance_sentinel_v2_reports"

    # Data directory and file names (under data_dir, e.g. agents/data_v2)
    data_dir: str = "agents/data_v2"
    ai_output_events_file: str = "ai_output_events_v2.json"
    ai_system_registry_file: str = "ai_system_registry_v2.json"
    compliance_findings_file: str = "compliance_findings_v2.json"
    remediation_actions_file: str = "emediation_actions_v2.json"  # filename as on disk
    escalation_events_file: str = "escalation_events_v2.json"
    financial_exposure_models_file: str = "financial_exposure_models_v2.json"
    human_review_actions_file: str = "human_review_actions_v2.json"
    portfolio_snapshots_file: str = "portfolio_snapshots_v2.json"

    # Executive trigger thresholds (rollup key -> threshold)
    open_cases_critical: int = 5           # Trigger if open_cases >= this
    total_exposure_usd_critical: float = 10_000_000   # Trigger if total_estimated_exposure_usd >= this
    portfolio_risk_score_critical: float = 75.0       # Trigger if portfolio_risk_score >= this
    critical_findings_critical: int = 2    # Trigger if critical_cases >= this
    executive_escalations_critical: int = 2 # Trigger if executive_escalations >= this
````
