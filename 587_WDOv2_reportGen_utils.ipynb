```python
"""
Report generation for Workforce Development Orchestrator v2.

Builds markdown from state (summary, risk, gaps, recommendations).
"""

from datetime import datetime
from typing import Any, Dict, List


def generate_workforce_report_v2(
    workforce_summary: Dict[str, Any],
    automation_risk_analysis: List[Dict[str, Any]],
    skill_gap_analysis: List[Dict[str, Any]],
    learning_path_recommendations: List[Dict[str, Any]],
    role_evolution_recommendations: List[Dict[str, Any]],
    role_transition_recommendations: List[Dict[str, Any]],
    prioritized_gaps: List[Dict[str, Any]],
    prioritized_recommendations: List[Dict[str, Any]],
    prioritized_evolutions: List[Dict[str, Any]],
) -> str:
    """
    Generate markdown workforce development report from analysis results.
    """
    s = workforce_summary
    report = f"""# Workforce Development Report (v2)

**Generated:** {datetime.now().strftime('%Y-%m-%d %H:%M')}

---

## Executive Summary

| Metric | Value |
|--------|--------|
| Total employees | {s.get('total_employees', 0)} |
| Total roles | {s.get('total_roles', 0)} |
| Employees at risk (high automation exposure) | {s.get('employees_at_risk', 0)} |
| Total skill gaps | {s.get('total_skill_gaps', 0)} |
| High-priority gaps | {s.get('high_priority_gaps', 0)} |
| Learning path recommendations | {s.get('total_learning_recommendations', 0)} |
| Roles requiring evolution | {s.get('roles_requiring_evolution', 0)} |
| **Overall workforce readiness score** | **{s.get('overall_workforce_readiness_score', 0)}** |

---

## Automation Risk by Role

"""
    for a in automation_risk_analysis[:10]:
        report += f"\n### {a.get('role_name', a.get('role_id'))}\n"
        report += f"- **Risk level:** {a.get('risk_level', 'N/A')} (score: {a.get('overall_risk_score', 0)})\n"
        report += f"- High-risk tasks: {len(a.get('high_risk_tasks', []))}\n"
        report += f"- Affected employees: {', '.join(a.get('affected_employees', []))}\n"
        for rec in a.get("recommendations", []):
            report += f"- {rec}\n"

    report += "\n---\n\n## Top Skill Gaps (Prioritized)\n\n"
    for g in prioritized_gaps[:10]:
        report += f"- **{g.get('employee_name', g.get('employee_id'))}** ({g.get('role_name', '')}): "
        report += f"missing {g.get('skill_name', g.get('skill_id'))} â€” priority: {g.get('priority', 'N/A')}, urgency: {g.get('urgency_score', 0)}\n"

    report += "\n---\n\n## Learning Path Recommendations\n\n"
    for r in prioritized_recommendations[:10]:
        report += f"- **{r.get('employee_name', r.get('employee_id'))}** â†’ {r.get('learning_path_name', '')} "
        report += f"(target: {r.get('target_skill', '')}, priority: {r.get('priority', '')})\n"

    report += "\n---\n\n## Role Evolution Recommendations\n\n"
    for e in prioritized_evolutions[:5]:
        report += f"- **{e.get('role_name', '')}** ({e.get('evolution_type', '')}): {e.get('description', '')[:80]}...\n"
        report += f"  Affected: {', '.join(e.get('affected_employees', []))}\n"

    report += "\n---\n\n## Feasible Role Transitions\n\n"
    for t in role_transition_recommendations[:5]:
        report += f"- {t.get('from_role_name', '')} â†’ {t.get('to_role_name', '')} "
        report += f"(probability: {t.get('transition_probability', 0)}, time: {t.get('time_to_transition_weeks', 0)} weeks)\n"

    report += "\n---\n*End of report*\n"
    return report
```
