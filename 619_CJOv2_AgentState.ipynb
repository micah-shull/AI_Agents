````python
# ============================================================================
# Customer Journey Orchestrator v2 (CJO v2)
# ============================================================================

class CJOv2OrchestratorState(TypedDict, total=False):
    """State for Customer Journey Orchestrator v2 â€” portfolio intelligence, playbooks, trajectory, governance."""

    # Input
    customer_id: Optional[str]              # Optional: single customer; None = full portfolio

    # Goal & Planning (MVP: fixed goal, template-based plan)
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Data Ingestion (agents/data)
    customers: List[Dict[str, Any]]
    journey_state_log: List[Dict[str, Any]]
    signals: List[Dict[str, Any]]
    journey_playbooks: List[Dict[str, Any]]
    journey_forecasts: List[Dict[str, Any]]
    interventions: List[Dict[str, Any]]
    outcomes: List[Dict[str, Any]]
    playbook_performance_log: List[Dict[str, Any]]
    portfolio_snapshots: List[Dict[str, Any]]

    # Lookups (for fast access)
    customers_lookup: Dict[str, Dict[str, Any]]
    journey_states_lookup: Dict[str, Dict[str, Any]]
    signals_by_customer: Dict[str, List[Dict[str, Any]]]
    playbooks_lookup: Dict[str, Dict[str, Any]]
    forecasts_lookup: Dict[str, Dict[str, Any]]
    interventions_by_customer: Dict[str, List[Dict[str, Any]]]

    # Signal Fusion (aggregated per customer)
    fused_signals: List[Dict[str, Any]]

    # Trajectory & Risk (from journey_forecasts or lightweight rules)
    trajectory_risk: List[Dict[str, Any]]

    # Intervention Strategy (matched playbooks â†’ recommended interventions)
    recommended_interventions: List[Dict[str, Any]]

    # Governance & HITL
    pending_approvals: List[Dict[str, Any]]
    approval_history: List[Dict[str, Any]]

    # Executive / Portfolio
    portfolio_summary: Dict[str, Any]
    executive_summary_text: Optional[str]

    # Output
    journey_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: Annotated[List[str], operator.add]
    processing_time: Optional[float]


@dataclass
class CJOv2OrchestratorConfig:
    """Configuration for Customer Journey Orchestrator v2."""
    data_dir: str = "agents/data"
    reports_dir: str = "output/cjo_v2_reports"
    auto_approve_for_testing: bool = True
    intervention_confidence_threshold: float = 0.50
    high_value_escalation_threshold: float = 50000.0
````
