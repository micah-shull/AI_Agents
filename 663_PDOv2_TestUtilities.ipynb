```python
"""
PDO V2 utilities tests (run from project root: pytest test_pdo_v2_utilities.py -v).

Tests: build_lookups, portfolio_rollup (anomalies, bottlenecks, reviewer economics,
risk_tier_summary, compute_portfolio_rollup), escalation (triggers, alerts), report.
"""
import pytest
from pathlib import Path

# Project root in path for imports
import sys
_root = Path(__file__).resolve().parent
if str(_root) not in sys.path:
    sys.path.insert(0, str(_root))

from agents.proposal_document_orchestrator_v2.orchestrator.utilities.data_loading import (
    load_all_pdo_v2_data,
    build_lookups,
)
from agents.proposal_document_orchestrator_v2.orchestrator.utilities.portfolio_rollup import (
    compute_stage_anomalies,
    compute_bottleneck_stages,
    compute_reviewer_economics,
    compute_risk_tier_summary,
    compute_portfolio_rollup,
)
from agents.proposal_document_orchestrator_v2.orchestrator.utilities.escalation import (
    compute_executive_triggers,
    create_escalation_alerts,
)
from agents.proposal_document_orchestrator_v2.orchestrator.utilities.report import (
    build_executive_report,
)


# ---- build_lookups ----
def test_build_lookups_empty():
    data = {
        "documents": [],
        "document_versions": [],
        "workflow_stages": [],
        "review_events": [],
        "compliance_checks": [],
        "cost_tracking": [],
        "outcomes": [],
        "reviewer_registry": [],
        "document_risk_profile": [],
        "stage_cost_baselines": [],
    }
    lookups = build_lookups(data)
    assert lookups["documents_lookup"] == {}
    assert lookups["reviewer_registry_lookup"] == {}
    assert lookups["stage_cost_baselines_lookup"] == {}


def test_build_lookups_documents_and_lookups():
    data = {
        "documents": [{"document_id": "D1", "status": "active"}, {"document_id": "D2"}],
        "document_versions": [{"document_id": "D1", "version_id": "V1"}],
        "workflow_stages": [],
        "review_events": [],
        "compliance_checks": [],
        "cost_tracking": [],
        "outcomes": [],
        "reviewer_registry": [{"reviewer_id": "R1", "role": "legal"}],
        "document_risk_profile": [{"document_id": "D1", "risk_tier": "tier_2"}],
        "stage_cost_baselines": [{"stage_name": "content_generation", "expected_minutes": 45}],
    }
    lookups = build_lookups(data)
    assert lookups["documents_lookup"]["D1"]["status"] == "active"
    assert lookups["document_versions_lookup"]["D1"] == [{"document_id": "D1", "version_id": "V1"}]
    assert lookups["reviewer_registry_lookup"]["R1"]["role"] == "legal"
    assert lookups["document_risk_profile_lookup"]["D1"]["risk_tier"] == "tier_2"
    assert lookups["stage_cost_baselines_lookup"]["content_generation"]["expected_minutes"] == 45


# ---- stage anomalies ----
def test_compute_stage_anomalies_empty():
    assert compute_stage_anomalies([], {}) == []


def test_compute_stage_anomalies_over_baseline():
    stages = [
        {
            "stage_id": "S1",
            "document_id": "D1",
            "stage_name": "content_generation",
            "status": "completed",
            "started_at": "2026-01-01T10:00:00Z",
            "completed_at": "2026-01-01T11:00:00Z",  # 60 min
        }
    ]
    baselines = {"content_generation": {"expected_minutes": 20, "variance_threshold_pct": 50}}
    # 60 vs 20 = 200% over -> over 50% threshold
    anomalies = compute_stage_anomalies(stages, baselines)
    assert len(anomalies) == 1
    assert anomalies[0]["stage_name"] == "content_generation"
    assert anomalies[0]["actual"] == 60.0
    assert anomalies[0]["expected"] == 20
    assert anomalies[0]["variance_pct"] > 50


def test_compute_stage_anomalies_skip_failed_and_in_progress():
    stages = [
        {"stage_name": "x", "status": "failed", "started_at": "2026-01-01T10:00:00Z", "completed_at": "2026-01-01T10:30:00Z"},
        {"stage_name": "y", "status": "in_progress", "started_at": "2026-01-01T10:00:00Z", "completed_at": None},
    ]
    baselines = {"x": {"expected_minutes": 10, "variance_threshold_pct": 50}, "y": {"expected_minutes": 10, "variance_threshold_pct": 50}}
    assert compute_stage_anomalies(stages, baselines) == []


# ---- bottlenecks ----
def test_compute_bottleneck_stages_aggregates():
    stages = [
        {"document_id": "D1", "stage_name": "legal_review", "status": "completed", "started_at": "2026-01-01T10:00:00Z", "completed_at": "2026-01-01T10:30:00Z"},
        {"document_id": "D2", "stage_name": "legal_review", "status": "failed", "started_at": "2026-01-01T10:00:00Z", "completed_at": "2026-01-01T10:20:00Z"},
    ]
    baselines = {"legal_review": {"expected_minutes": 20}}
    out = compute_bottleneck_stages(stages, baselines)
    assert len(out) == 1
    assert out[0]["stage_name"] == "legal_review"
    assert out[0]["document_count"] == 2
    assert out[0]["failure_count"] == 1
    assert out[0]["failure_rate"] == 0.5
    assert out[0]["avg_duration_minutes"] is not None


# ---- reviewer economics ----
def test_compute_reviewer_economics():
    events = [
        {"reviewer_id": "R1", "time_spent_minutes": 60, "reviewer_role": "legal"},
        {"reviewer_id": "R1", "time_spent_minutes": 30, "reviewer_role": "legal"},
    ]
    registry = {"R1": {"hourly_cost_usd": 120, "role": "legal"}}
    out = compute_reviewer_economics(events, registry)
    assert out["total_reviewer_cost_usd"] == 180.0  # 90 min * 120/60
    assert out["by_role"]["legal"]["review_count"] == 2
    assert out["total_reviews"] == 2


def test_compute_reviewer_economics_unknown_reviewer():
    events = [{"reviewer_id": "R_UNK", "time_spent_minutes": 60, "reviewer_role": "legal"}]
    out = compute_reviewer_economics(events, {})
    assert out["total_reviewer_cost_usd"] == 0.0
    assert out["by_role"]["legal"]["review_count"] == 1


# ---- risk tier summary ----
def test_compute_risk_tier_summary():
    documents = [{"document_id": "D1"}, {"document_id": "D2"}, {"document_id": "D3"}]
    profile_lookup = {
        "D1": {"risk_tier": "tier_1", "executive_visibility": False},
        "D2": {"risk_tier": "tier_3", "executive_visibility": True},
        "D3": {"risk_tier": "tier_3", "executive_visibility": True},
    }
    out = compute_risk_tier_summary(profile_lookup, documents)
    assert out["total_documents"] == 3
    assert out["by_tier"]["tier_1"] == 1
    assert out["by_tier"]["tier_3"] == 2
    assert out["executive_visibility_count"] == 2
    assert out["tier_3_count"] == 2
    assert out["high_risk_count"] == 2


# ---- portfolio rollup ----
def test_compute_portfolio_rollup_shape():
    documents = [{"document_id": "D1"}]
    profile = {"D1": {"risk_tier": "tier_2", "executive_visibility": True}}
    baselines = {"content_generation": {"expected_minutes": 45, "variance_threshold_pct": 60}}
    rollup = compute_portfolio_rollup(
        documents=documents,
        workflow_stages=[],
        review_events=[],
        document_risk_profile_lookup=profile,
        stage_cost_baselines_lookup=baselines,
        reviewer_registry_lookup={},
    )
    assert "total_documents" in rollup
    assert "high_risk_count" in rollup
    assert "tier_3_count" in rollup
    assert "stage_anomaly_count" in rollup
    assert "executive_visibility_count" in rollup
    assert "total_reviewer_cost_usd" in rollup
    assert "risk_tier_summary" in rollup
    assert "stage_anomalies" in rollup
    assert "bottleneck_stages" in rollup
    assert "reviewer_economics" in rollup


# ---- escalation triggers ----
def test_compute_executive_triggers_none_when_below():
    rollup = {"high_risk_count": 0, "tier_3_count": 0, "stage_anomaly_count": 0, "executive_visibility_count": 0}
    triggers = compute_executive_triggers(
        rollup,
        threshold_high_risk_count=3,
        threshold_stage_anomaly_count=2,
        threshold_executive_visibility_count=5,
        threshold_tier_3_count=2,
    )
    assert triggers == []


def test_compute_executive_triggers_fires_when_above():
    rollup = {"high_risk_count": 5, "tier_3_count": 1, "stage_anomaly_count": 0, "executive_visibility_count": 2}
    triggers = compute_executive_triggers(
        rollup,
        threshold_high_risk_count=3,
        threshold_stage_anomaly_count=2,
        threshold_executive_visibility_count=5,
        threshold_tier_3_count=2,
    )
    assert len(triggers) >= 1
    high_risk = [t for t in triggers if t.get("trigger_type") == "high_risk_count"]
    assert len(high_risk) == 1
    assert high_risk[0]["value"] == 5
    assert high_risk[0]["threshold"] == 3


def test_create_escalation_alerts():
    triggers = [
        {"trigger_type": "high_risk_count", "message": "High-risk count (5) at or above threshold (3)", "value": 5, "threshold": 3},
    ]
    alerts = create_escalation_alerts(triggers)
    assert len(alerts) == 1
    assert alerts[0]["alert_type"] == "pdo_v2_executive_trigger"
    assert "High-risk" in alerts[0]["title"]
    assert alerts[0]["metadata"]["value"] == 5


# ---- report ----
def test_build_executive_report_contains_sections():
    rollup = {
        "total_documents": 2,
        "high_risk_count": 0,
        "executive_visibility_count": 1,
        "stage_anomaly_count": 0,
        "total_reviewer_cost_usd": 100,
        "risk_tier_summary": {"by_tier": {"tier_1": 1, "tier_2": 1}},
        "bottleneck_stages": [{"stage_name": "legal_review", "failure_rate": 0.2}],
        "reviewer_economics": {"total_reviewer_cost_usd": 100, "by_role": {"legal": {"cost_usd": 100, "review_count": 2}}},
        "stage_anomalies": [],
    }
    report = build_executive_report(portfolio_rollup=rollup, executive_triggers=[])
    assert "One-view" in report
    assert "Within guardrails" in report
    assert "So what" in report
    assert "Next steps" in report
    assert "Portfolio by risk tier" in report
    assert "Stage bottlenecks" in report
    assert "Reviewer economics" in report
    assert "tier_1" in report or "tier_2" in report


def test_build_executive_report_with_triggers():
    rollup = {
        "total_documents": 5,
        "high_risk_count": 4,
        "executive_visibility_count": 4,
        "stage_anomaly_count": 0,
        "total_reviewer_cost_usd": 200,
        "risk_tier_summary": {},
        "bottleneck_stages": [],
        "reviewer_economics": {},
        "stage_anomalies": [],
    }
    triggers = [{"trigger_type": "high_risk_count", "message": "High-risk (4) >= 3", "value": 4, "threshold": 3}]
    report = build_executive_report(portfolio_rollup=rollup, executive_triggers=triggers)
    assert "Executive attention required" in report
    assert "Attention required" in report
    assert "Triggers exceeded" in report or "escalation" in report.lower()


# ---- data loading (integration: requires agents/data) ----
@pytest.mark.skipif(
    not (Path(__file__).resolve().parent / "agents" / "data" / "documents.json").exists(),
    reason="agents/data not present",
)
def test_load_all_pdo_v2_data_and_lookups():
    root = Path(__file__).resolve().parent
    data = load_all_pdo_v2_data(data_dir="agents/data", project_root=root)
    assert "documents" in data
    assert "reviewer_registry" in data
    assert "document_risk_profile" in data
    assert "stage_cost_baselines" in data
    assert len(data["documents"]) >= 1
    lookups = build_lookups(data)
    assert len(lookups["documents_lookup"]) == len(data["documents"])
    assert "stage_cost_baselines_lookup" in lookups
```
