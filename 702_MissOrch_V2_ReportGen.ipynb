````python
"""
Mission Control v2 executive report generation.

One-view bar, portfolio health, risk & governance, execution history, HITL metrics,
performance-driven routing summary, and next steps.
"""

from datetime import datetime, timezone
from typing import Any, Dict, List, Optional


def generate_mission_control_v2_report(
    *,
    view_mode: str = "portfolio",
    mission_id: Optional[str] = None,
    missions: Optional[List[Dict[str, Any]]] = None,
    mission_runs: Optional[List[Dict[str, Any]]] = None,
    task_execution_logs: Optional[List[Dict[str, Any]]] = None,
    agent_performance_stats: Optional[List[Dict[str, Any]]] = None,
    mission_risk_events: Optional[List[Dict[str, Any]]] = None,
    risk_rollup: Optional[Dict[str, Any]] = None,
    portfolio_rollup: Optional[Dict[str, Any]] = None,
    executive_triggers: Optional[List[Dict[str, Any]]] = None,
    performance_routing_lookup: Optional[Dict[str, str]] = None,
    agents_lookup: Optional[Dict[str, Dict[str, Any]]] = None,
) -> str:
    """
    Generate the Mission Control v2 markdown report.

    All arguments are optional; missing sections are omitted or show "N/A".
    """
    missions = missions or []
    mission_runs = mission_runs or []
    task_execution_logs = task_execution_logs or []
    agent_performance_stats = agent_performance_stats or []
    mission_risk_events = mission_risk_events or []
    risk_rollup = risk_rollup or {}
    portfolio_rollup = portfolio_rollup or {}
    executive_triggers = executive_triggers or []
    performance_routing_lookup = performance_routing_lookup or {}
    agents_lookup = agents_lookup or {}

    generated = datetime.now(timezone.utc).strftime("%Y-%m-%d %H:%M:%S UTC")

    # ---- Executive one-view ----
    trigger_block = ""
    if executive_triggers:
        trigger_block = "\n> **Executive attention required.** Triggers exceeded; see Next Steps below.\n"
    mar = portfolio_rollup.get("missions_at_risk", 0)
    verdict = "HIGH" if mar >= 2 else "MEDIUM" if mar >= 1 else "LOW"
    headline = f"Portfolio risk: **{verdict}**  |  Missions at risk: {mar}  |  View: {view_mode}"

    report = f"""# Mission Control v2 â€” Executive Report

**Generated:** {generated}
{trigger_block}
## One-View

{headline}

---

## Portfolio Health (Latest Snapshot)

| Metric | Value |
|--------|--------|
| Snapshot date | {portfolio_rollup.get('snapshot_date') or 'N/A'} |
| Active runs | {portfolio_rollup.get('active_runs', 0)} |
| Completed runs | {portfolio_rollup.get('completed_runs', 0)} |
| Missions at risk | {portfolio_rollup.get('missions_at_risk', 0)} |
| Avg KPI improvement % | {portfolio_rollup.get('avg_kpi_improvement_pct', 0)}% |
| Top bottleneck task | {portfolio_rollup.get('top_bottleneck_task_name') or 'N/A'} ({portfolio_rollup.get('top_bottleneck_task_id') or ''}) |
| Escalations (last 24h) | {portfolio_rollup.get('escalations_last_24h', 0)} |
| Human interventions (last 24h) | {portfolio_rollup.get('human_interventions_last_24h', 0)} |

---

## Risk & Governance

**Risk events rollup:** Total {risk_rollup.get('total_count', 0)} (Open: {risk_rollup.get('open_count', 0)}, Closed: {risk_rollup.get('closed_count', 0)})

**By severity:** High {risk_rollup.get('by_severity', {}).get('high', 0)} | Medium {risk_rollup.get('by_severity', {}).get('medium', 0)} | Low {risk_rollup.get('by_severity', {}).get('low', 0)}

**By type:** {_fmt_dict(risk_rollup.get('by_risk_type', {}))}

"""
    if mission_risk_events:
        report += "### Recent risk events\n\n"
        for e in mission_risk_events[:10]:
            desc = (e.get("description") or "")[:80]
            if len((e.get("description") or "")) > 80:
                desc += "..."
            report += f"- **{e.get('event_id', '')}** [{e.get('severity', '')}] {e.get('risk_type', '')}: {desc}\n"
        report += "\n"

    report += "---\n\n## Execution History & Learning\n\n"
    report += f"**Mission runs:** {len(mission_runs)} total.\n\n"
    if mission_runs:
        report += "| Run ID | Mission | Status | Duration (min) | KPIs | Risk flagged |\n"
        report += "|-------|---------|--------|----------------|------|-------------|\n"
        for r in mission_runs[:15]:
            kpis = r.get("kpis") or {}
            kpi_str = ", ".join(f"{k}:{v}" for k, v in list(kpis.items())[:2])
            report += f"| {r.get('run_id', '')} | {r.get('mission_id', '')} | {r.get('status', '')} | {r.get('total_duration_minutes', 0)} | {kpi_str} | {r.get('risk_flagged', False)} |\n"
        report += "\n"

    report += "### Task-level telemetry (sample)\n\n"
    if task_execution_logs:
        report += "| Run | Task | Agent | Duration (min) | Approval required | Human delay (min) | Outcome |\n"
        report += "|-----|------|-------|----------------|-------------------|-------------------|--------|\n"
        for t in task_execution_logs[:15]:
            report += f"| {t.get('run_id', '')} | {t.get('task_id', '')} | {t.get('agent_id', '')} | {t.get('actual_duration_minutes', 0)} | {t.get('human_approval_required', False)} | {t.get('human_delay_minutes', 0)} | {t.get('outcome', '')} |\n"
        report += "\n"
    else:
        report += "No task execution logs.\n\n"

    report += "---\n\n## Agent Performance (routing input)\n\n"
    if agent_performance_stats:
        report += "| Agent ID | Avg task duration (min) | Approval rate | Escalation rate |\n"
        report += "|----------|--------------------------|---------------|------------------|\n"
        for s in agent_performance_stats:
            report += f"| {s.get('agent_id', '')} | {s.get('avg_task_duration_minutes', 0)} | {s.get('approval_rate', 0)} | {s.get('escalation_rate', 0)} |\n"
        report += "\n"
    else:
        report += "No agent performance stats.\n\n"

    if performance_routing_lookup and agents_lookup:
        report += "### Performance-based task routing (recommended agent per task)\n\n"
        for task_id, agent_id in sorted(performance_routing_lookup.items()):
            name = (agents_lookup.get(agent_id) or {}).get("name") or agent_id
            report += f"- **{task_id}** â†’ {agent_id} ({name})\n"
        report += "\n"

    report += "---\n\n## Next Steps\n\n"
    if executive_triggers:
        for t in executive_triggers:
            report += f"- {t.get('message', '')}\n"
        report += "\n"
    else:
        report += "- No executive triggers fired. Continue monitoring portfolio and risk events.\n\n"

    report += "---\n\n*Mission Control v2 â€” outcome-driven, risk-aware, continuously improving.*\n"
    return report


def _fmt_dict(d: Dict[str, Any]) -> str:
    return ", ".join(f"{k}: {v}" for k, v in d.items())
````
