````python
"""Deterministic run normalization, trends, health, and historical triggers."""

from __future__ import annotations

import statistics
from collections import defaultdict
from typing import Any, Dict, List, Optional, Sequence, Tuple

from agents.mission_orchestrator_v3.orchestrator.schema import MissionOrchV3Config


def _mission_name_lookup(business_missions: List[Dict[str, Any]]) -> Dict[str, str]:
    return {str(m["mission_id"]): str(m.get("mission_name", m["mission_id"])) for m in business_missions}


def _task_defs_by_mission(decomposed: List[Dict[str, Any]]) -> Dict[str, Dict[str, Dict[str, Any]]]:
    out: Dict[str, Dict[str, Dict[str, Any]]] = {}
    for block in decomposed:
        mid = str(block["mission_id"])
        out[mid] = {}
        for t in block.get("tasks", []):
            tid = str(t["task_id"])
            out[mid][tid] = t
    return out


def _kpi_targets_for_mission(mission_kpis_row: Dict[str, Any]) -> Dict[str, float]:
    """Map semantic metric name -> target/baseline where discoverable from template keys."""
    kpis = mission_kpis_row.get("kpis") or {}
    resolved: Dict[str, float] = {}
    for k, v in kpis.items():
        if not isinstance(v, (int, float)):
            continue
        if k.startswith("target_"):
            suffix = k.replace("target_", "")
            if suffix.endswith("_days") and "onboarding" in suffix:
                resolved["onboarding_time_days_target"] = float(v)
            elif suffix.endswith("_days") and "pipeline" in suffix:
                resolved["pipeline_days_target"] = float(v)
            elif suffix.endswith("_hours") and "resolution" in suffix:
                resolved["resolution_time_hours_target"] = float(v)
            elif k == "target_onboarding_time_days":
                resolved["onboarding_time_days_target"] = float(v)
            elif k == "target_pipeline_days":
                resolved["pipeline_days_target"] = float(v)
            elif k == "target_resolution_time_hours":
                resolved["resolution_time_hours_target"] = float(v)
        elif k.startswith("baseline_"):
            if k == "baseline_onboarding_time_days":
                resolved["onboarding_time_days_baseline"] = float(v)
            elif k == "baseline_pipeline_days":
                resolved["pipeline_days_baseline"] = float(v)
            elif k == "baseline_resolution_time_hours":
                resolved["resolution_time_hours_baseline"] = float(v)
        elif k == "csat_threshold":
            resolved["csat_target"] = float(v)
    return resolved


def _lookup_target_baseline(
    primary_name: str,
    targets: Dict[str, float],
) -> Tuple[Optional[float], Optional[float]]:
    if primary_name == "onboarding_time_days":
        return targets.get("onboarding_time_days_target"), targets.get("onboarding_time_days_baseline")
    if primary_name == "pipeline_days":
        return targets.get("pipeline_days_target"), targets.get("pipeline_days_baseline")
    if primary_name == "resolution_time_hours":
        return targets.get("resolution_time_hours_target"), targets.get("resolution_time_hours_baseline")
    return None, None


def _better(
    direction: str,
    a: float,
    b: float,
    tolerance_rel: float,
) -> bool:
    """True if a is strictly better than b beyond tolerance."""
    if direction == "lower_is_better":
        if b == 0:
            return a < b
        return a < b * (1 - tolerance_rel)
    if direction == "higher_is_better":
        if b == 0:
            return a > b
        return a > b * (1 + tolerance_rel)
    return False


def _flat(
    direction: str,
    a: float,
    b: float,
    tolerance_rel: float,
) -> bool:
    if a == b:
        return True
    base = max(abs(a), abs(b), 1e-9)
    return abs(a - b) / base <= tolerance_rel


def classify_trend(
    values: Sequence[float],
    direction: str,
    tolerance_rel: float,
    min_runs: int,
) -> str:
    if len(values) < min_runs:
        return "insufficient_data"
    first, prev, latest = values[0], values[-2], values[-1]
    if _flat(direction, latest, prev, tolerance_rel) and _flat(direction, latest, first, tolerance_rel):
        return "flat"
    imp_prev = _better(direction, latest, prev, tolerance_rel)
    imp_first = _better(direction, latest, first, tolerance_rel)
    worse_prev = _better(direction, prev, latest, tolerance_rel)
    worse_first = _better(direction, first, latest, tolerance_rel)
    if imp_prev and imp_first:
        return "improving"
    if worse_prev and worse_first:
        return "worsening"
    return "flat"


def _percent_improvement_from_baseline(
    latest: float,
    baseline: Optional[float],
    direction: str,
) -> Optional[float]:
    if baseline is None or baseline == 0:
        return None
    if direction == "lower_is_better":
        return round(100.0 * (baseline - latest) / baseline, 2)
    return round(100.0 * (latest - baseline) / baseline, 2)


def _remaining_gap_to_target_pct(
    latest: float,
    target: Optional[float],
    direction: str,
) -> Optional[float]:
    if target is None or target == 0:
        return None
    if direction == "lower_is_better":
        if latest <= target:
            return 0.0
        return round(100.0 * (latest - target) / target, 2)
    if latest >= target:
        return 0.0
    return round(100.0 * (target - latest) / target, 2)


def normalize_run_history(
    bundle: Dict[str, Any],
    config: MissionOrchV3Config,
) -> List[Dict[str, Any]]:
    runs: List[Dict[str, Any]] = list(bundle.get("mission_runs") or [])
    task_logs: List[Dict[str, Any]] = list(bundle.get("task_execution_logs") or [])
    decomposed = list(bundle.get("decomposed_mission_tasks") or [])
    mission_kpis = list(bundle.get("mission_kpis") or [])

    task_by_mission = _task_defs_by_mission(decomposed)
    kpi_row_by_mission = {str(r["mission_id"]): r for r in mission_kpis}

    logs_by_run: Dict[str, List[Dict[str, Any]]] = defaultdict(list)
    for row in task_logs:
        logs_by_run[str(row["run_id"])].append(row)

    records: List[Dict[str, Any]] = []
    for run in runs:
        rid = str(run["run_id"])
        mid = str(run["mission_id"])
        spec = config.primary_kpi_by_mission.get(mid)
        if not spec:
            continue
        primary_name, _direction = spec
        kpis = run.get("kpis") or {}
        if primary_name not in kpis:
            continue
        targets = _kpi_targets_for_mission(kpi_row_by_mission.get(mid, {}))
        target_v, baseline_v = _lookup_target_baseline(primary_name, targets)

        tlogs = logs_by_run.get(rid, [])
        approval_delay_total = 0.0
        overrun_total = 0.0
        tasks_completed = 0
        defs = task_by_mission.get(mid, {})
        for tl in tlogs:
            tid = str(tl["task_id"])
            est = float((defs.get(tid) or {}).get("estimated_duration_minutes") or 0)
            actual = float(tl.get("actual_duration_minutes") or 0)
            delay = float(tl.get("human_delay_minutes") or 0)
            if tl.get("human_approval_required"):
                approval_delay_total += delay
            overrun_total += max(0.0, actual - est)
            tasks_completed += 1

        start = str(run.get("start_time") or "")
        run_date = start[:10] if len(start) >= 10 else ""

        records.append(
            {
                "run_id": rid,
                "mission_id": mid,
                "run_date": run_date,
                "status": run.get("status"),
                "total_duration_minutes": float(run.get("total_duration_minutes") or 0),
                "primary_kpi_name": primary_name,
                "primary_kpi_value": float(kpis[primary_name]),
                "target_value": target_v,
                "baseline_value": baseline_v,
                "human_interventions": int(run.get("human_interventions") or 0),
                "escalations": int(run.get("escalations") or 0),
                "risk_flagged": bool(run.get("risk_flagged")),
                "tasks_completed": tasks_completed,
                "approval_delay_minutes_total": round(approval_delay_total, 2),
                "task_overrun_minutes_total": round(overrun_total, 2),
            }
        )

    records.sort(key=lambda r: (r["mission_id"], r["run_date"], r["run_id"]))
    return records


def _rolling_avg(values: Sequence[float], n: int) -> Optional[float]:
    if not values:
        return None
    window = values[-n:]
    return round(sum(window) / len(window), 4)


def _volatility_class(duration_vals: List[float], kpi_vals: List[float]) -> str:
    def cv(xs: List[float]) -> float:
        if len(xs) < 2:
            return 0.0
        m = statistics.mean(xs)
        if m == 0:
            return 0.0
        return statistics.pstdev(xs) / abs(m)

    dcv = cv(duration_vals)
    kcv = cv(kpi_vals)
    score = (dcv + kcv) / 2
    if score < 0.08:
        return "stable"
    if score < 0.15:
        return "improving_stability"
    if score < 0.25:
        return "recovering"
    return "unstable"


def _subscores(
    latest: float,
    target: Optional[float],
    baseline: Optional[float],
    direction: str,
    trend_direction: str,
    avg_interventions: float,
    avg_escalations: float,
    avg_approval_delay: float,
    pct_runs_risk: float,
    risk_severity_weight: float,
    duration_vals: List[float],
    kpi_vals: List[float],
) -> Tuple[float, float, float, float, float]:
    # Outcome 0-100
    outcome = 70.0
    if target is not None and baseline is not None:
        if direction == "lower_is_better":
            span = baseline - target
            if span > 0:
                outcome = 100.0 * max(0.0, min(1.0, (baseline - latest) / span))
        else:
            span = target - baseline
            if span > 0:
                outcome = 100.0 * max(0.0, min(1.0, (latest - baseline) / span))
    outcome = max(0.0, min(100.0, outcome))

    traj_map = {"improving": 92.0, "flat": 74.0, "worsening": 38.0, "insufficient_data": 62.0}
    trajectory = traj_map.get(trend_direction, 65.0)

    friction_raw = avg_interventions * 12 + avg_escalations * 18 + min(40.0, avg_approval_delay * 0.8)
    friction = max(0.0, min(100.0, 100.0 - friction_raw))

    risk_raw = pct_runs_risk * 55.0 + min(45.0, risk_severity_weight * 6.0)
    risk = max(0.0, min(100.0, 100.0 - risk_raw))

    vol = _volatility_class(duration_vals, kpi_vals)
    stab_map = {"stable": 90.0, "improving_stability": 78.0, "recovering": 62.0, "unstable": 40.0}
    stability = stab_map.get(vol, 65.0)

    return outcome, trajectory, friction, risk, stability


def mission_health_band(score: float) -> str:
    if score >= 85:
        return "ON_TRACK"
    if score >= 70:
        return "WATCH"
    return "ATTENTION_REQUIRED"


def compute_mission_summaries(
    bundle: Dict[str, Any],
    run_history: List[Dict[str, Any]],
    config: MissionOrchV3Config,
) -> List[Dict[str, Any]]:
    names = _mission_name_lookup(list(bundle.get("business_missions") or []))
    risks: List[Dict[str, Any]] = list(bundle.get("mission_risk_events") or [])

    by_mission: Dict[str, List[Dict[str, Any]]] = defaultdict(list)
    for r in run_history:
        by_mission[str(r["mission_id"])].append(r)
    for mid in by_mission:
        by_mission[mid].sort(key=lambda x: (x["run_date"], x["run_id"]))

    risk_by_mission_run: Dict[str, List[Dict[str, Any]]] = defaultdict(list)
    for ev in risks:
        key = f'{ev.get("mission_id")}:{ev.get("run_id")}'
        risk_by_mission_run[key].append(ev)

    summaries: List[Dict[str, Any]] = []
    for mid, rows in sorted(by_mission.items()):
        spec = config.primary_kpi_by_mission.get(mid)
        if not spec:
            continue
        primary_name, direction = spec
        values = [float(r["primary_kpi_value"]) for r in rows]
        trend_direction = classify_trend(values, direction, config.trend_tolerance_rel, config.min_runs_for_trend)
        latest = values[-1]
        previous = values[-2] if len(values) >= 2 else latest
        first = values[0]
        target_v = rows[-1].get("target_value")
        baseline_v = rows[-1].get("baseline_value")

        sec_trend = None
        sec_spec = config.secondary_kpi_by_mission.get(mid)
        if sec_spec:
            sec_name, sec_dir = sec_spec
            sec_vals: List[float] = []
            for r in rows:
                raw_run = next((x for x in bundle.get("mission_runs") or [] if str(x["run_id"]) == str(r["run_id"])), {})
                kp = (raw_run.get("kpis") or {}).get(sec_name)
                if kp is not None:
                    sec_vals.append(float(kp))
            if len(sec_vals) >= config.min_runs_for_trend:
                sec_trend = classify_trend(sec_vals, sec_dir, config.trend_tolerance_rel, config.min_runs_for_trend)

        interventions_first = int(rows[0]["human_interventions"])
        interventions_last = int(rows[-1]["human_interventions"])
        escalations_first = int(rows[0]["escalations"])
        escalations_last = int(rows[-1]["escalations"])
        avg_interventions = statistics.mean([float(r["human_interventions"]) for r in rows])
        avg_escalations = statistics.mean([float(r["escalations"]) for r in rows])
        avg_approval_delay = statistics.mean([float(r["approval_delay_minutes_total"]) for r in rows])
        avg_overrun = statistics.mean([float(r["task_overrun_minutes_total"]) for r in rows])
        pct_risk = sum(1 for r in rows if r["risk_flagged"]) / max(1, len(rows))

        duration_vals = [float(r["total_duration_minutes"]) for r in rows]
        kpi_vals = values

        risk_weight = 0.0
        for r in rows:
            for ev in risk_by_mission_run.get(f'{mid}:{r["run_id"]}', []):
                sev = str(ev.get("severity") or "low").lower()
                risk_weight += {"high": 3.0, "medium": 2.0, "low": 1.0}.get(sev, 1.0)

        outcome_s, traj_s, fric_s, risk_s, stab_s = _subscores(
            latest,
            float(target_v) if target_v is not None else None,
            float(baseline_v) if baseline_v is not None else None,
            direction,
            trend_direction,
            avg_interventions,
            avg_escalations,
            avg_approval_delay,
            pct_risk * 100.0,
            risk_weight,
            duration_vals,
            kpi_vals,
        )
        mission_health_score = round(
            0.35 * outcome_s
            + 0.25 * traj_s
            + 0.20 * fric_s
            + 0.10 * risk_s
            + 0.10 * stab_s,
            2,
        )
        health = mission_health_band(mission_health_score)

        imp_base = _percent_improvement_from_baseline(latest, float(baseline_v) if baseline_v is not None else None, direction)
        gap_tgt = _remaining_gap_to_target_pct(latest, float(target_v) if target_v is not None else None, direction)
        roll = _rolling_avg(values, config.rolling_n)

        autonomy_score = round(
            max(
                0.0,
                min(
                    100.0,
                    100.0
                    - avg_interventions * 14.0
                    - avg_escalations * 18.0
                    - min(35.0, avg_approval_delay * 0.9)
                    - pct_risk * 40.0,
                ),
            ),
            0,
        )
        stability_score = round(stab_s, 0)

        summaries.append(
            {
                "mission_id": mid,
                "mission_name": names.get(mid, mid),
                "runs_analyzed": len(rows),
                "primary_kpi_name": primary_name,
                "primary_kpi_direction": direction,
                "latest_value": latest,
                "previous_value": previous,
                "first_value": first,
                "baseline_value": baseline_v,
                "target_value": target_v,
                "trend_direction": trend_direction,
                "secondary_kpi_trend": sec_trend,
                "improvement_from_baseline_pct": imp_base,
                "remaining_gap_to_target_pct": gap_tgt,
                "rolling_avg_last_n": roll,
                "first_run_human_interventions": interventions_first,
                "last_run_human_interventions": interventions_last,
                "first_run_escalations": escalations_first,
                "last_run_escalations": escalations_last,
                "avg_human_interventions": round(avg_interventions, 2),
                "avg_escalations": round(avg_escalations, 2),
                "avg_approval_delay_minutes": round(avg_approval_delay, 2),
                "avg_task_overrun_minutes": round(avg_overrun, 2),
                "pct_runs_risk_flagged": round(pct_risk * 100.0, 2),
                "volatility_class": _volatility_class(duration_vals, kpi_vals),
                "autonomy_score": int(autonomy_score),
                "stability_score": int(stability_score),
                "mission_health_score": mission_health_score,
                "mission_health": health,
                "subscores": {
                    "outcome": round(outcome_s, 1),
                    "trajectory": round(traj_s, 1),
                    "friction": round(fric_s, 1),
                    "risk": round(risk_s, 1),
                    "stability": round(stab_s, 1),
                },
            }
        )
    return summaries


def compute_task_bottlenecks(
    bundle: Dict[str, Any],
    run_history: List[Dict[str, Any]],
    config: MissionOrchV3Config,
) -> List[Dict[str, Any]]:
    task_logs: List[Dict[str, Any]] = list(bundle.get("task_execution_logs") or [])
    decomposed = list(bundle.get("decomposed_mission_tasks") or [])
    risks: List[Dict[str, Any]] = list(bundle.get("mission_risk_events") or [])
    names = _mission_name_lookup(list(bundle.get("business_missions") or []))
    task_by_mission = _task_defs_by_mission(decomposed)

    runs_by_mission: Dict[str, List[str]] = defaultdict(list)
    for r in run_history:
        runs_by_mission[str(r["mission_id"])].append(str(r["run_id"]))

    logs_by_key: Dict[Tuple[str, str, str], List[Dict[str, Any]]] = defaultdict(list)
    for tl in task_logs:
        key = (str(tl["mission_id"]), str(tl["run_id"]), str(tl["task_id"]))
        logs_by_key[key].append(tl)

    risk_counts: Dict[Tuple[str, str], Tuple[int, float]] = defaultdict(lambda: (0, 0.0))
    for ev in risks:
        k = (str(ev.get("mission_id")), str(ev.get("task_id")))
        c, w = risk_counts[k]
        sev = str(ev.get("severity") or "low").lower()
        weight = {"high": 3.0, "medium": 2.0, "low": 1.0}.get(sev, 1.0)
        risk_counts[k] = (c + 1, w + weight)

    summaries: List[Dict[str, Any]] = []
    for mid, run_ids in runs_by_mission.items():
        defs = task_by_mission.get(mid, {})
        for tid, tdef in defs.items():
            samples: List[Dict[str, Any]] = []
            for rid in run_ids:
                row = logs_by_key.get((mid, rid, tid))
                if row:
                    samples.append(row[0])
            if not samples:
                continue
            est = float(tdef.get("estimated_duration_minutes") or 0)
            avg_actual = statistics.mean([float(s.get("actual_duration_minutes") or 0) for s in samples])
            avg_delay = statistics.mean([float(s.get("human_delay_minutes") or 0) for s in samples])
            overrun_freq = sum(
                1 for s in samples if float(s.get("actual_duration_minutes") or 0) > est + 1e-6
            ) / max(1, len(samples))
            rc, rw = risk_counts.get((mid, tid), (0, 0.0))
            score = avg_delay * 2.0 + max(0.0, avg_actual - est) + rw * 3.0
            if avg_delay >= config.approval_delay_elevated_minutes and tdef.get("requires_human_approval"):
                driver = "approval_delay"
            elif avg_actual > est * 1.05:
                driver = "task_overrun"
            elif rc > 0:
                driver = "risk"
            else:
                driver = "mixed"

            top_runs = sorted(run_ids)[-3:]
            top_hits = sum(
                1
                for rid in top_runs
                if logs_by_key.get((mid, rid, tid))
                and float(logs_by_key[(mid, rid, tid)][0].get("human_delay_minutes") or 0)
                >= config.approval_delay_elevated_minutes
            )
            if len(samples) >= 3 and top_hits >= 2:
                bstatus = "persistent"
            elif len(samples) >= 2 and top_hits >= 1:
                bstatus = "recurring"
            else:
                bstatus = "isolated"

            summaries.append(
                {
                    "mission_id": mid,
                    "mission_name": names.get(mid, mid),
                    "task_id": tid,
                    "task_name": str(tdef.get("task") or tid),
                    "runs_observed": len(samples),
                    "estimated_duration_minutes": est,
                    "avg_actual_duration_minutes": round(avg_actual, 2),
                    "avg_human_delay_minutes": round(avg_delay, 2),
                    "overrun_frequency": round(overrun_freq, 2),
                    "risk_event_count": rc,
                    "severity_weighted_risk_score": round(rw, 2),
                    "bottleneck_score": round(score, 2),
                    "bottleneck_status": bstatus,
                    "bottleneck_driver": driver,
                }
            )

    best_by_mission: Dict[str, Dict[str, Any]] = {}
    for s in summaries:
        mid = s["mission_id"]
        cur = best_by_mission.get(mid)
        if cur is None or float(s["bottleneck_score"]) > float(cur["bottleneck_score"]):
            best_by_mission[mid] = s

    for s in summaries:
        s["is_top_bottleneck_for_mission"] = best_by_mission.get(s["mission_id"], {}).get("task_id") == s["task_id"]

    return summaries


def attach_top_bottlenecks(mission_summaries: List[Dict[str, Any]], task_summaries: List[Dict[str, Any]]) -> None:
    top = {s["mission_id"]: s for s in task_summaries if s.get("is_top_bottleneck_for_mission")}
    for m in mission_summaries:
        t = top.get(m["mission_id"])
        if not t:
            m["top_bottleneck_task_id"] = None
            m["top_bottleneck_driver"] = None
        else:
            m["top_bottleneck_task_id"] = t["task_id"]
            m["top_bottleneck_driver"] = t["bottleneck_driver"]


def evaluate_triggers(
    mission_summaries: List[Dict[str, Any]],
    task_summaries: List[Dict[str, Any]],
    config: MissionOrchV3Config,
) -> List[Dict[str, Any]]:
    fired: List[Dict[str, Any]] = []
    for m in mission_summaries:
        mid = m["mission_id"]
        if m.get("trend_direction") == "worsening":
            fired.append({"id": "MISSION_TREND_WORSENING", "mission_id": mid, "detail": m.get("primary_kpi_name")})
        gap = m.get("remaining_gap_to_target_pct")
        if gap is not None and float(gap) > 8.0 and m.get("trend_direction") in ("flat", "worsening"):
            fired.append({"id": "TARGET_GAP_PERSISTING", "mission_id": mid, "detail": f"gap_pct={gap}"})
        latest = float(m["latest_value"])
        prev = float(m["previous_value"])
        first = float(m["first_value"])
        direction = m["primary_kpi_direction"]
        if m["runs_analyzed"] >= 3:
            band = max(abs(first), abs(prev), abs(latest), 1e-9) * config.plateau_tolerance_rel
            if abs(latest - prev) <= band and abs(latest - first) <= band:
                fired.append({"id": "KPI_PLATEAU", "mission_id": mid, "detail": m.get("primary_kpi_name")})

        if m.get("avg_approval_delay_minutes", 0) >= config.approval_delay_elevated_minutes and m.get(
            "trend_direction"
        ) not in ("improving",):
            fired.append({"id": "APPROVAL_DELAY_PERSISTING", "mission_id": mid})

        if m.get("avg_escalations", 0) > 0.4 and float(m.get("avg_escalations", 0)) >= float(
            m.get("avg_human_interventions", 0)
        ):
            fired.append({"id": "ESCALATION_BURDEN_RISING", "mission_id": mid})

        if m.get("pct_runs_risk_flagged", 0) >= 50.0:
            fired.append({"id": "RISK_CLUSTER_FORMING", "mission_id": mid})

        if int(m.get("autonomy_score", 100)) < 55:
            fired.append({"id": "AUTONOMY_REGRESSION", "mission_id": mid})

        if int(m.get("stability_score", 100)) < 60:
            fired.append({"id": "STABILITY_DROPPING", "mission_id": mid})

    for t in task_summaries:
        if not t.get("is_top_bottleneck_for_mission"):
            continue
        if t.get("bottleneck_status") in ("persistent", "recurring"):
            fired.append(
                {
                    "id": "REPEATED_BOTTLENECK",
                    "mission_id": t["mission_id"],
                    "detail": f'{t["task_id"]}:{t.get("bottleneck_driver")}',
                }
            )

    for ev in fired:
        if "severity" not in ev:
            ev["severity"] = {"MISSION_TREND_WORSENING": "high", "AUTONOMY_REGRESSION": "high"}.get(
                ev["id"], "medium"
            )
    return fired


def portfolio_verdict_from_missions(mission_summaries: List[Dict[str, Any]]) -> str:
    if not mission_summaries:
        return "WATCH"
    bands = [m.get("mission_health") for m in mission_summaries]
    if any(b == "ATTENTION_REQUIRED" for b in bands):
        return "ATTENTION_REQUIRED"
    if any(b == "WATCH" for b in bands):
        return "WATCH"
    return "ON_TRACK"
````
