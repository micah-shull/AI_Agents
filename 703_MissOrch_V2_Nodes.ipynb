````python
"""
Mission Control v2 orchestrator nodes.

Nodes are thin: orchestrate and call utilities. Config-dependent nodes use closures.
"""

from typing import Any, Dict, List

from pathlib import Path

from agents.mission_control_v2.orchestrator.utilities import (
    load_all_mission_control_data,
    build_risk_rollup,
    build_executive_triggers,
    build_portfolio_rollup,
    build_performance_routing_lookup,
    generate_mission_control_v2_report,
)


def goal_node(state: Dict[str, Any]) -> Dict[str, Any]:
    """Set goal for Mission Control v2 run (portfolio health, risk, learning)."""
    view_mode = state.get("view_mode") or "portfolio"
    mission_id = state.get("mission_id")
    goal = {
        "objective": "Assess portfolio health, risk & governance, and execution learning for Mission Control v2.",
        "view_mode": view_mode,
        "mission_id": mission_id,
    }
    return {"goal": goal}


def planning_node(state: Dict[str, Any]) -> Dict[str, Any]:
    """Produce a fixed plan: data load â†’ risk â†’ portfolio â†’ report."""
    plan = [
        {"step": "data_loading", "description": "Load all mission, run, risk, and portfolio data from agents/data"},
        {"step": "risk_governance", "description": "Build risk rollup and executive triggers"},
        {"step": "portfolio", "description": "Build portfolio rollup from latest snapshot"},
        {"step": "performance_routing", "description": "Build performance-based task â†’ agent routing"},
        {"step": "report", "description": "Generate executive report and save to file"},
    ]
    return {"plan": plan}


def make_data_loading_node(config: Any, project_root: str):
    """Data loading node with config and project_root (closure)."""
    def _node(state: Dict[str, Any]) -> Dict[str, Any]:
        data_dir = state.get("data_dir") or config.data_dir
        data = load_all_mission_control_data(
            data_dir,
            project_root=project_root,
            business_missions_file=config.business_missions_file,
            decomposed_mission_tasks_file=config.decomposed_mission_tasks_file,
            specialized_agents_file=config.specialized_agents_file,
            agent_capabilities_matrix_file=config.agent_capabilities_matrix_file,
            mission_kpis_file=config.mission_kpis_file,
            mission_runs_file=config.mission_runs_file,
            task_execution_logs_file=config.task_execution_logs_file,
            agent_performance_stats_file=config.agent_performance_stats_file,
            mission_risk_events_file=config.mission_risk_events_file,
            mission_portfolio_snapshot_file=config.mission_portfolio_snapshot_file,
        )
        agents_lookup = {a["agent_id"]: a for a in data["specialized_agents"]}
        capabilities_lookup = {m["task_id"]: m["capable_agents"] for m in data["capabilities_matrix"]}
        return {
            "missions": data["missions"],
            "mission_tasks_by_mission": data["mission_tasks_by_mission"],
            "mission_kpis_by_mission": data["mission_kpis_by_mission"],
            "specialized_agents": data["specialized_agents"],
            "capabilities_matrix": data["capabilities_matrix"],
            "agents_lookup": agents_lookup,
            "capabilities_lookup": capabilities_lookup,
            "mission_runs": data["mission_runs"],
            "task_execution_logs": data["task_execution_logs"],
            "agent_performance_stats": data["agent_performance_stats"],
            "mission_risk_events": data["mission_risk_events"],
            "mission_portfolio_snapshot": data["mission_portfolio_snapshot"],
        }
    return _node


def risk_governance_node(state: Dict[str, Any]) -> Dict[str, Any]:
    """Build risk rollup and executive triggers (uses state only; thresholds from config via closure in orchestrator)."""
    # Thresholds will be injected by orchestrator when building the node
    risk_rollup = build_risk_rollup(state.get("mission_risk_events") or [])
    portfolio_rollup = build_portfolio_rollup(state.get("mission_portfolio_snapshot") or [])
    return {"risk_rollup": risk_rollup, "portfolio_rollup": portfolio_rollup}


def make_risk_governance_node(config: Any):
    """Risk + triggers node with config thresholds (closure)."""
    def _node(state: Dict[str, Any]) -> Dict[str, Any]:
        risk_rollup = build_risk_rollup(state.get("mission_risk_events") or [])
        portfolio_rollup = build_portfolio_rollup(state.get("mission_portfolio_snapshot") or [])
        triggers = build_executive_triggers(
            risk_rollup,
            portfolio_rollup,
            missions_at_risk_critical=config.missions_at_risk_critical,
            escalations_24h_critical=config.escalations_24h_critical,
            risk_events_open_critical=config.risk_events_open_critical,
        )
        return {"risk_rollup": risk_rollup, "portfolio_rollup": portfolio_rollup, "executive_triggers": triggers}
    return _node


def portfolio_node(state: Dict[str, Any]) -> Dict[str, Any]:
    """Build portfolio rollup only (if not already done in risk_governance)."""
    portfolio_rollup = build_portfolio_rollup(state.get("mission_portfolio_snapshot") or [])
    return {"portfolio_rollup": portfolio_rollup}


def performance_routing_node(state: Dict[str, Any]) -> Dict[str, Any]:
    """Build performance-based task â†’ agent lookup."""
    capabilities_lookup = state.get("capabilities_lookup") or {}
    agent_performance_stats = state.get("agent_performance_stats") or []
    performance_routing_lookup = build_performance_routing_lookup(
        capabilities_lookup, agent_performance_stats, prefer_fastest=True
    )
    return {"performance_routing_lookup": performance_routing_lookup}


def make_report_node(config: Any, project_root: str):
    """Generate report and save to file (closure)."""
    def _node(state: Dict[str, Any]) -> Dict[str, Any]:
        report = generate_mission_control_v2_report(
            view_mode=state.get("view_mode") or "portfolio",
            mission_id=state.get("mission_id"),
            missions=state.get("missions"),
            mission_runs=state.get("mission_runs"),
            task_execution_logs=state.get("task_execution_logs"),
            agent_performance_stats=state.get("agent_performance_stats"),
            mission_risk_events=state.get("mission_risk_events"),
            risk_rollup=state.get("risk_rollup"),
            portfolio_rollup=state.get("portfolio_rollup"),
            executive_triggers=state.get("executive_triggers"),
            performance_routing_lookup=state.get("performance_routing_lookup"),
            agents_lookup=state.get("agents_lookup"),
        )
        reports_dir = state.get("reports_dir") or config.reports_dir
        if project_root:
            report_path = Path(project_root) / reports_dir / "mission_control_v2_report.md"
        else:
            report_path = Path(reports_dir) / "mission_control_v2_report.md"
        report_path.parent.mkdir(parents=True, exist_ok=True)
        report_path.write_text(report, encoding="utf-8")
        return {"mission_report": report, "report_file_path": str(report_path)}
    return _node
````
