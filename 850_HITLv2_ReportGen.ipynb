````python
"""Build HITL v2 executive report (one-pager)."""

from typing import Any, Dict, List


def build_rollup(routing_decisions: List[Dict[str, Any]], simulated_reviews: List[Dict[str, Any]]) -> Dict[str, Any]:
    """Compute counts for report: auto_approved, human_review, escalate, and human decision breakdown."""
    auto_approved = sum(1 for d in routing_decisions if d.get("action") == "auto_approve")
    human_review = sum(1 for d in routing_decisions if d.get("action") == "human_review")
    escalate = sum(1 for d in routing_decisions if d.get("action") == "escalate")
    approved = sum(1 for r in simulated_reviews if r.get("human_decision") == "approve")
    modified = sum(1 for r in simulated_reviews if r.get("human_decision") == "modify")
    rejected = sum(1 for r in simulated_reviews if r.get("human_decision") == "reject")
    overridden = sum(1 for r in simulated_reviews if r.get("human_decision") == "override")
    return {
        "auto_approved_count": auto_approved,
        "human_review_count": human_review,
        "escalated_count": escalate,
        "human_approved_count": approved,
        "human_modified_count": modified,
        "human_rejected_count": rejected,
        "human_override_count": overridden,
        "total_tasks": len(routing_decisions),
    }


def build_hitl_report(
    rollup: Dict[str, Any],
    routing_decisions: List[Dict[str, Any]],
    simulated_reviews: List[Dict[str, Any]],
    new_audit_entries: List[Dict[str, Any]],
    new_feedback_entries: List[Dict[str, Any]],
    tasks_count: int,
    agent_outputs_count: int,
    reviewers_count: int,
    rules_count: int,
    errors: List[str],
) -> str:
    """
    Build executive one-pager: verdict, routing summary, rollup, audit trail, feedback signals, next steps.
    """
    lines = []
    lines.append("# Human-in-the-Loop Collaboration — Executive Summary")
    lines.append("")
    # Verdict
    total = rollup.get("total_tasks", 0)
    human_involved = (rollup.get("human_review_count", 0) + rollup.get("escalated_count", 0))
    if total == 0:
        verdict = "N/A"
    elif human_involved == 0:
        verdict = "LOW — All tasks auto-approved; no human oversight required."
    elif rollup.get("human_override_count", 0) + rollup.get("human_rejected_count", 0) > 0:
        verdict = "MEDIUM — Human overrides or rejections present; review feedback events."
    else:
        verdict = "LOW — Routine human review; no overrides."
    lines.append(f"**Governance verdict:** {verdict}")
    lines.append("")
    # Traceability
    lines.append(f"*Data: {tasks_count} tasks, {agent_outputs_count} agent outputs, {reviewers_count} reviewers, {rules_count} routing rules.*")
    if errors:
        lines.append(f"*Warnings: {len(errors)} load/validation issue(s).*")
    lines.append("")
    # Rollup
    lines.append("## Routing summary")
    lines.append("")
    lines.append("| Metric | Count |")
    lines.append("|--------|------|")
    lines.append(f"| Auto-approved | {rollup.get('auto_approved_count', 0)} |")
    lines.append(f"| Human review | {rollup.get('human_review_count', 0)} |")
    lines.append(f"| Escalated | {rollup.get('escalated_count', 0)} |")
    lines.append(f"| **Total tasks** | **{total}** |")
    lines.append("")
    if simulated_reviews:
        lines.append("| Human decision | Count |")
        lines.append("|----------------|------|")
        lines.append(f"| Approved | {rollup.get('human_approved_count', 0)} |")
        lines.append(f"| Modified | {rollup.get('human_modified_count', 0)} |")
        lines.append(f"| Rejected | {rollup.get('human_rejected_count', 0)} |")
        lines.append(f"| Override | {rollup.get('human_override_count', 0)} |")
        lines.append("")
    # Per-task routing (compact table)
    lines.append("## Per-task routing")
    lines.append("")
    lines.append("| Task | Risk | Confidence | Action | Human decision |")
    lines.append("|------|------|------------|--------|----------------|")
    review_by_task = {r["task_id"]: r for r in simulated_reviews}
    for d in routing_decisions:
        task_id = d.get("task_id", "")
        risk = d.get("risk_level", "")
        conf = d.get("confidence_score")
        conf_str = f"{conf:.2f}" if conf is not None else "—"
        action = d.get("action", "")
        r = review_by_task.get(task_id)
        human_dec = r.get("human_decision", "—") if r else "—"
        lines.append(f"| {task_id} | {risk} | {conf_str} | {action} | {human_dec} |")
    lines.append("")
    # Audit trail (this run)
    if new_audit_entries:
        lines.append("## Audit trail (this run)")
        lines.append("")
        lines.append("| Task | Routing | Human involved | Final decision | Source | Latency (s) |")
        lines.append("|------|---------|----------------|----------------|--------|-------------|")
        for e in new_audit_entries:
            lines.append(f"| {e.get('task_id', '')} | {e.get('routing_decision', '')} | {e.get('human_involved', '')} | {e.get('final_decision', '')} | {e.get('decision_source', '')} | {e.get('latency_seconds', '')} |")
        lines.append("")
    # Feedback events (this run)
    if new_feedback_entries:
        lines.append("## Feedback events (this run)")
        lines.append("")
        for e in new_feedback_entries:
            lines.append(f"- **{e.get('task_id', '')}**: {e.get('error_type', '')} — {e.get('notes', '')}")
        lines.append("")
    # Next steps
    lines.append("## Next steps")
    lines.append("")
    if new_feedback_entries:
        lines.append("1. Review feedback events for model/policy calibration.")
    if rollup.get("escalated_count", 0) > 0:
        lines.append("2. Review escalated tasks and assigned roles.")
    lines.append("3. Use audit log for compliance and traceability.")
    lines.append("")
    return "\n".join(lines)
````
