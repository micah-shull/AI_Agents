````python
# ============================================================================
# Revenue Gap Orchestrator v2 (RGO v2)
# ============================================================================
# Revenue Exposure Intelligence: REI, structural risk, root cause attribution,
# store-level stockout, exposure by tier and by cause. Data: agents/data/ (CSVs).
# Specs: agents/RGOv2*.md, RGOv2_5_Data_and_Scope_Revisions.md.


class RGOv2State(TypedDict, total=False):
    """State for Revenue Gap Orchestrator v2 (exposure intelligence)."""

    # Input
    project_root: Optional[str]      # Project root path (for resolving data_dir, reports_dir)
    data_dir: Optional[str]           # Data directory (default from config: agents/data)
    reports_dir: Optional[str]       # Reports directory (default from config)

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Raw data (from loader)
    customers: List[Dict[str, Any]]   # retail_customers rows
    weekly_sales: List[Dict[str, Any]]  # retail_weekly_sales rows (weekly_spend, week_start_date, store_id, etc.)
    stock_availability: List[Dict[str, Any]]  # stock_availability rows (store_id, week_start, on_hand_units)
    data_snapshot_loaded_at: Optional[str]   # UTC ISO when data was loaded
    validation_warnings: List[str]

    # Lookups (built from raw data in nodes)
    customers_lookup: Dict[str, Dict[str, Any]]   # customer_id -> row
    sales_by_customer: Dict[str, List[Dict[str, Any]]]  # customer_id -> list of weekly rows (sorted by week)
    stockout_store_weeks: Any                    # set of (store_id, week_str) with stockout; or dict composite key

    # Derived: per-customer metrics (list of dicts, one per customer with gap or at risk)
    customer_metrics: List[Dict[str, Any]]
    # Each item: customer_id, tier, current_revenue, baseline_revenue, gap_amount, decline_percent,
    #   consecutive_decline_weeks, consecutive_zero_spend_weeks, decline_velocity_percent, volatility_score,
    #   structural_score, structural_tier, stockout_overlap_weeks, percent_gap_due_to_stockout,
    #   root_cause, churn_risk_bucket, rei, (and any needed for report)

    # Portfolio rollup
    portfolio_rollup: Dict[str, Any]
    # total_gap, total_rei, exposure_by_tier, exposure_by_root_cause, top_priority_count, etc.

    # Top N priority interventions (for report)
    top_priority_gaps: List[Dict[str, Any]]

    # Output
    report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class RGOv2Config:
    """Configuration for Revenue Gap Orchestrator v2. All thresholds and weights in one place."""

    # Paths
    data_dir: str = "agents/data"
    reports_dir: str = "output/rgo_v2_reports"
    retail_customers_file: str = "retail_customers.csv"
    retail_weekly_sales_file: str = "retail_weekly_sales.csv"
    stock_availability_file: str = "stock_availability.csv"

    # Baseline & current (RGOv2_5)
    baseline_window_weeks: int = 6
    current_window_weeks: int = 4
    structural_window_weeks: int = 8

    # REI (RGOv2_1) â€” MVP: no recurrence
    churn_weights: Dict[str, float] = field(default_factory=lambda: {
        "low": 1.05, "medium": 1.15, "high": 1.30, "critical": 1.50
    })
    structural_multipliers: Dict[str, float] = field(default_factory=lambda: {
        "none": 1.00, "moderate": 1.10, "high": 1.25, "severe": 1.40
    })
    tier_multipliers: Dict[str, float] = field(default_factory=lambda: {
        "standard": 1.00, "high": 1.25, "strategic": 1.50
    })
    max_exposure_multiplier: float = 2.5

    # Churn risk buckets (map churn score/indicators to bucket for REI)
    churn_risk_high_weeks_zero: int = 2   # >= this many zero-spend weeks -> high/critical churn bucket

    # Structural risk (RGOv2_2): thresholds and points
    structural_decline_weeks: Dict[str, int] = field(default_factory=lambda: {"moderate": 2, "high": 3, "severe": 4})
    structural_zero_weeks: Dict[str, int] = field(default_factory=lambda: {"moderate": 1, "high": 2, "severe": 3})
    structural_velocity_thresholds: Dict[str, float] = field(default_factory=lambda: {"moderate": -0.20, "high": -0.35, "severe": -0.50})
    structural_volatility_thresholds: Dict[str, float] = field(default_factory=lambda: {"moderate": 0.30, "high": 0.45, "severe": 0.60})
    structural_tier_cutoffs: Dict[str, int] = field(default_factory=lambda: {"moderate": 2, "high": 4, "severe": 6})

    # Root cause (RGOv2_3)
    stockout_attribution_threshold: float = 0.40   # min fraction of gap weeks with stockout for operational
    zero_spend_behavioral_threshold: int = 2     # consecutive zero weeks >= this -> behavioral_churn candidate

    # Reporting
    top_n_priority: int = 20
````
