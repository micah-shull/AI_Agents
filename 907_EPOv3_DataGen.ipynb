# Experimentation Portfolio Orchestrator v3

## Full Data Blueprint (MVP Telemetry Architecture)

### Agent Type

**Portfolio Intelligence Agent**

### Domain

Experimentation / A-B testing / product experimentation governance

### Primary Question

> *Is our experimentation program generating reliable business value, and which experiments should scale, iterate, or stop?*

---

# 1. Core Portfolio Registry

### `experiment_portfolio.json`

Purpose
Defines **the portfolio of experiments under management**.

Row count (MVP)
3–4 experiments

Key fields

```
experiment_id
experiment_name
domain
owner
start_date
status
risk_tier
estimated_annual_value_usd

secondary_metrics

```

```
product_growth
```

Example statuses

```

design_review
running
monitoring
completed
blocked
```

Why this matters

This dataset defines **the experimentation surface area**.

Everything else references it.

Allows the agent to determine:

* design quality
* statistical credibility
* alignment between hypothesis and metrics

# 14. Signature Metrics for v3

The orchestrator should compute four core signals.

### Experiment Effectiveness Score

Measures experiment success quality.

Factors

* lift magnitude
* statistical confidence
* guardrail pass rate
* segment consistency

---

### Design Quality Score

Measures experiment rigor.

Factors

* hypothesis clarity
* metric alignment
* minimum effect size
* power assumptions
* guardrail completeness

---

### Risk Exposure Score

Measures operational and compliance risk.

Factors

* risk tier
* risk events
* human dependency
* guardrail failures
* compliance flags

---

### Learning Velocity Score

Measures knowledge generation.

Factors

* reusable learnings
* experiment closures
* cross-domain lessons
