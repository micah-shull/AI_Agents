# IRMv3 Executive Report Generator

## Architecture Review & Operational Explanation

The reporting module transforms the orchestrator’s internal state into a **structured executive briefing**.

Rather than presenting raw metrics or technical logs, it produces a **concise, board-ready report** that answers the essential leadership questions:

* Is the AI ecosystem stable?
* What risks are emerging?
* Are problems being fixed?
* What action is required?

The output is a **Markdown document** that can easily be:

* saved to reports
* rendered in dashboards
* attached to governance documentation
* archived for audit history

This design allows the orchestrator to generate **repeatable executive intelligence snapshots**.

---

# Report Structure

The report follows a deliberate structure designed to mirror how executives consume information.

The document moves from **summary → explanation → operational evidence → action**.

Sections include:

1. Executive header
2. One-view summary
3. Operational traceability
4. Core score breakdown
5. Executive triggers
6. Trend signals
7. Risk visibility
8. Remediation progress
9. Recommended next steps

This progression mirrors how real executive briefings work.

---

# Executive Header

The report opens with a **clear declaration of system status**.

```text
Integration & Risk Management — Executive Operating Intelligence
Portfolio Health: HEALTHY | AT_RISK | CRITICAL
```

This ensures leadership immediately understands the most important signal.

Executives do not want to search through dashboards to determine system status.

The report answers that question **in the first two lines**.

---

# One-View Section

The “One View” table acts as the **executive dashboard summary**.

It presents the key metrics in a compact format:

| Metric                    | Meaning                     |
| ------------------------- | --------------------------- |
| Verdict                   | Overall system condition    |
| Portfolio Score           | Weighted health score       |
| Integration Stability     | Reliability of integrations |
| Dependency Exposure       | Architectural fragility     |
| Governance Responsiveness | Issue resolution discipline |
| Automation Value          | Business ROI from AI        |

This table allows leadership to **quickly interpret the operational state of the AI ecosystem**.

It functions as the equivalent of a **board one-pager**.

---

# “So What” Narrative

Immediately after presenting the metrics, the report provides a short interpretation.

Example outcomes:

* Critical ecosystem instability
* Emerging operational risk
* Stable environment

This section is important because metrics alone do not provide context.

Executives need a brief explanation of **what the numbers mean for operations**.

The message remains short and direct.

---

# The “One Ask”

The report then identifies the **single most important leadership action**.

Examples include:

* addressing triggered operational risks
* prioritizing remediation work
* continuing routine monitoring

This design is powerful because it reduces complexity.

Rather than presenting dozens of recommendations, the report focuses leadership attention on **one clear operational priority**.

---

# Data & Run Traceability

The next section provides **operational transparency**.

It includes:

* the data snapshot timestamp
* record counts for each dataset
* validation warnings

Example signals:

* missing telemetry
* incomplete datasets
* degraded observability

This traceability is critical for governance.

If leadership reviews the report during an incident investigation, they can determine **exactly what data was available at the time of analysis**.

---

# Core Score Breakdown

This section explains the four operational dimensions that produced the portfolio score.

Each score is displayed alongside a **health band classification**:

| Band            | Meaning                         |
| --------------- | ------------------------------- |
| Stable / Strong | Healthy operational state       |
| Acceptable      | Minor issues                    |
| Risk emerging   | Operational concerns            |
| Critical        | Immediate intervention required |

This helps leadership understand **which dimension is driving system risk**.

---

# Executive Triggers

If escalation signals exist, they are surfaced here.

Triggers represent operational conditions such as:

* integration instability
* fragile dependencies
* slow remediation
* automation value erosion

Each trigger includes a **clear explanation**.

This ensures that leadership understands **why the alert occurred**, not just that it occurred.

---

# Trend Visibility

The mission run section displays recent portfolio results.

Example metrics include:

* portfolio score history
* agents at risk
* detected risk flags

This allows leadership to identify **system trajectory**.

Is the ecosystem improving?

Or deteriorating?

Trend information is often more important than a single snapshot.

---

# Open Risks

The report also surfaces currently unresolved risks.

Each risk includes:

* risk ID
* risk type
* affected system
* severity level

This section ensures that **known operational risks remain visible**.

Without this visibility, unresolved issues can easily be forgotten as systems scale.

---

# Remediation Status

The remediation summary shows how effectively issues are being addressed.

Metrics include:

* resolved actions
* work in progress
* overdue fixes

This section is particularly important for governance.

It measures **organizational responsiveness**, not just technical system health.

A system can detect problems perfectly, but if no one fixes them, the operation still fails.

---

# Next Steps

The report concludes with recommended actions tailored to the portfolio status.

Examples include:

Critical state:

* prioritize remediation work
* assign ownership
* validate fixes with a follow-up run

Moderate risk:

* address triggered issues
* monitor remediation progress

Healthy state:

* maintain routine monitoring
* resolve minor risks through normal governance cycles

This section ensures the report ends with **clear operational direction**.

````python
"""Build the IRMv3 executive markdown report."""

from typing import Any, Dict, List, Optional


def build_irm_report(
    portfolio_status: str,
    portfolio_score: float,
    integration_stability: float,
    dependency_exposure: float,
    governance_responsiveness: float,
    automation_value: float,
    score_rollup: Dict[str, Any],
    executive_triggers: List[Dict[str, Any]],
    loader_record_counts: Dict[str, int],
    data_snapshot_loaded_at: Optional[str] = None,
    validation_warnings: Optional[List[str]] = None,
    mission_runs: Optional[List[Dict[str, Any]]] = None,
    risk_signals: Optional[List[Dict[str, Any]]] = None,
    remediation_actions: Optional[List[Dict[str, Any]]] = None,
) -> str:
    """Produce executive markdown report (one-view, scores, triggers, next steps)."""
    lines: List[str] = []

    # Title
    lines.append("# Integration & Risk Management — Executive Operating Intelligence")
    lines.append("")
    lines.append("**Portfolio Health:** " + portfolio_status)
    lines.append("")
    lines.append("---")
    lines.append("")

    # One-view (board one-pager style)
    lines.append("## One View")
    lines.append("")
    lines.append("| Metric | Value |")
    lines.append("|--------|-------|")
    lines.append(f"| **Verdict** | {portfolio_status} |")
    lines.append(f"| **Portfolio Score** | {portfolio_score}/100 |")
    lines.append(f"| Integration Stability | {integration_stability} |")
    lines.append(f"| Dependency Exposure | {dependency_exposure} |")
    lines.append(f"| Governance Responsiveness | {governance_responsiveness} |")
    lines.append(f"| Automation Value | {automation_value} |")
    lines.append("")

    # So-what
    if portfolio_status == "CRITICAL":
        so_what = "At least one core score is in the critical range. Prioritize integration or governance remediation before scaling."
    elif portfolio_status == "AT_RISK":
        so_what = "Portfolio is at risk; address low scores and open remediations to avoid further degradation."
    else:
        so_what = "Ecosystem is stable; maintain monitoring and address any open risks in normal cycle."
    lines.append("**So what:** " + so_what)
    lines.append("")

    # One ask
    if executive_triggers:
        lines.append("**One ask:** Address triggered items below (integration, dependency, governance, or overdue remediations).")
    else:
        lines.append("**One ask:** Continue monitoring; no critical triggers this run.")
    lines.append("")
    lines.append("---")
    lines.append("")

    # Traceability
    lines.append("## Data & Run")
    lines.append("")
    if data_snapshot_loaded_at:
        lines.append(f"- **Data snapshot:** {data_snapshot_loaded_at} (UTC)")
    counts = loader_record_counts or {}
    parts = [f"{k}: {v}" for k, v in sorted(counts.items()) if k != "loader_record_counts"]
    if parts:
        lines.append("- **Record counts:** " + ", ".join(parts))
    if validation_warnings:
        lines.append("- **Warnings:** " + "; ".join(validation_warnings))
    lines.append("")
    lines.append("---")
    lines.append("")

    # Four scores (detail)
    lines.append("## Core Scores (0–100, higher = healthier)")
    lines.append("")
    lines.append("| Score | Value | Band |")
    lines.append("|-------|-------|------|")
    for name, val in [
        ("Integration Stability", integration_stability),
        ("Dependency Exposure", dependency_exposure),
        ("Governance Responsiveness", governance_responsiveness),
        ("Automation Value", automation_value),
    ]:
        lines.append(f"| {name} | {val} | {_band(val)} |")
    lines.append("")
    lines.append("**Portfolio score (weighted):** " + str(portfolio_score))
    lines.append("")
    lines.append("---")
    lines.append("")

    # Executive triggers
    if executive_triggers:
        lines.append("## Executive Triggers")
        lines.append("")
        for t in executive_triggers:
            lines.append(f"- **{t.get('trigger_type', '')}:** {t.get('message', '')}")
        lines.append("")
        lines.append("---")
        lines.append("")

    # Recent mission runs (trend)
    if mission_runs and len(mission_runs) > 0:
        lines.append("## Recent Mission Runs (Trend)")
        lines.append("")
        lines.append("| Run | Portfolio Score | Agents at Risk | Risk Flags |")
        lines.append("|-----|-----------------|----------------|------------|")
        for run in mission_runs[-5:]:
            run_id = run.get("run_id", "")
            score = run.get("portfolio_health_score", "")
            at_risk = run.get("agents_at_risk", "")
            flags = run.get("risk_flags", [])
            flags_str = ", ".join(flags[:3]) if isinstance(flags, list) else str(flags)
            lines.append(f"| {run_id} | {score} | {at_risk} | {flags_str} |")
        lines.append("")
        lines.append("---")
        lines.append("")

    # Open risks summary
    if risk_signals:
        open_risks = [r for r in risk_signals if (r.get("status") or "").lower() == "open"]
        if open_risks:
            lines.append("## Open Risks")
            lines.append("")
            lines.append("| Risk ID | Type | Source | Severity |")
            lines.append("|---------|------|--------|----------|")
            for r in open_risks[:10]:
                lines.append(f"| {r.get('risk_id', '')} | {r.get('risk_type', '')} | {r.get('source_system', '')} | {r.get('severity', '')} |")
            lines.append("")
            lines.append("---")
            lines.append("")

    # Remediation status
    if remediation_actions:
        overdue = [r for r in remediation_actions if r.get("missed_deadline_flag") is True]
        in_progress = [r for r in remediation_actions if (r.get("status") or "").lower() == "in_progress"]
        resolved = [r for r in remediation_actions if (r.get("status") or "").lower() == "resolved"]
        lines.append("## Remediation Summary")
        lines.append("")
        lines.append(f"- Resolved: {len(resolved)} | In progress: {len(in_progress)} | Overdue: {len(overdue)}")
        lines.append("")
        lines.append("---")
        lines.append("")

    # Next steps
    lines.append("## Next Steps")
    lines.append("")
    if portfolio_status == "CRITICAL":
        lines.append("1. Review executive triggers and prioritize integration or governance fixes.")
        lines.append("2. Assign owners for overdue remediations and set short-term targets.")
        lines.append("3. Re-run after remediation to confirm score improvement.")
    elif executive_triggers:
        lines.append("1. Address triggered items; focus on lowest-scoring dimension.")
        lines.append("2. Track remediation due dates to avoid additional overdue items.")
    else:
        lines.append("1. Continue routine monitoring.")
        lines.append("2. Resolve open risks in normal governance cycle.")
    lines.append("")

    return "\n".join(lines)


def _band(score: float) -> str:
    if score >= 90:
        return "Stable / Strong"
    if score >= 75:
        return "Acceptable"
    if score >= 60:
        return "Risk emerging"
    return "Critical"
````
