## Optimization Summary

## How It Works Now

1. **Lazy Loading**: When you request a framework (e.g., `porter_five_forces`), only that framework's template file is loaded
2. **Smart File Mapping**: Uses a direct mapping from framework ID to filename (avoids directory scanning)
3. **Fallback**: If a framework isn't found via the mapping, it falls back to scanning the directory
4. **Caching**: Once loaded, frameworks are cached in memory (singleton pattern)

## Benefits

- Faster startup: No unnecessary file I/O
- Lower memory: Only loads what's needed
- Cleaner logs: Only shows the framework you're actually using
- Backward compatible: Still works if you call `list_frameworks()`

Instead of all 6 frameworks. The Strategic Analysis Agent will only load the 4 frameworks it supports (swot, porter_five_forces, pestel, ansoff_matrix), and the Portfolio Agent will only load its 2 frameworks (bcg_matrix, ge_mckinsey_matrix) when needed.

This reduces unnecessary resource usage while maintaining the same functionality.

````python
"""Framework loader for strategic analysis templates"""

import re
import yaml
from pathlib import Path
from typing import Dict, Any, Optional, List
from dataclasses import dataclass, field

logger = __import__("logging").getLogger(__name__)


@dataclass
class FrameworkConfig:
    """Configuration for a strategic analysis framework"""
    template_id: str
    template_name: str
    template_type: str
    complexity: str
    confidence_required: float
    framework_dimensions: Dict[str, Any]  # Framework-specific dimensions
    analysis_steps: List[str]  # Framework-specific analysis steps
    output_format: Dict[str, Any]  # Output format specification
    description: str
    file_path: Path

    # Metadata from frontmatter
    required_data_sources: List[str] = field(default_factory=list)
    chainable_with: List[str] = field(default_factory=list)
    data_quality_thresholds: Dict[str, Any] = field(default_factory=dict)


class FrameworkLoader:
    """Loads and manages strategic analysis framework templates"""

    def __init__(self, templates_dir: Optional[Path] = None):
        """
        Initialize framework loader.

        Args:
            templates_dir: Directory containing framework template files.
                          Defaults to docs/project/Analysis_Templates/
        """
        if templates_dir is None:
            # Default to project templates directory
            project_root = Path(__file__).parent.parent
            templates_dir = project_root / "docs" / "project" / "Analysis_Templates"

        self.templates_dir = Path(templates_dir)
        self._frameworks: Dict[str, FrameworkConfig] = {}
        self._loaded_all: bool = False  # Track if we've scanned the directory

    def _ensure_framework_loaded(self, framework_id: str) -> bool:
        """
        Lazy-load a specific framework if not already loaded.

        Args:
            framework_id: The framework ID to load

        Returns:
            True if framework was found and loaded, False otherwise
        """
        # If already loaded, return True
        if framework_id in self._frameworks:
            return True

        # Try to load this specific framework
        if not self.templates_dir.exists():
            logger.warning(f"Templates directory not found: {self.templates_dir}")
            return False

        # Map framework IDs to expected file names
        framework_file_map = {
            "swot": "SWOT.md",
            "porter_five_forces": "Porter's_Five_Forces.md",
            "pestel": "PESTEL.md",
            "ansoff_matrix": "Ansoff_Matrix.md",
            "bcg_matrix": "BCG_Matrix.md",
            "ge_mckinsey_matrix": "GEâ€“McKinsey_Matrix.md"
        }

        # Try to find the file
        expected_file = framework_file_map.get(framework_id)
        if expected_file:
            template_file = self.templates_dir / expected_file
            if template_file.exists():
                return self._load_single_framework(template_file)

        # Fallback: scan directory if we haven't already
        if not self._loaded_all:
            self._scan_and_load_all()

        return framework_id in self._frameworks

    def _load_single_framework(self, template_file: Path) -> bool:
        """
        Load a single framework from a file.

        Args:
            template_file: Path to the framework template file

        Returns:
            True if loaded successfully, False otherwise
        """
        try:
            framework = self._load_framework(template_file)
            if framework:
                self._frameworks[framework.template_id] = framework
                logger.info(f"âœ… Loaded framework: {framework.template_name} ({framework.template_id})")
                return True
        except Exception as e:
            logger.error(f"âŒ Failed to load {template_file.name}: {e}")
        return False

    def _scan_and_load_all(self):
        """Scan directory and load all framework templates (for list operations)"""
        if self._loaded_all:
            return

        if not self.templates_dir.exists():
            logger.warning(f"Templates directory not found: {self.templates_dir}")
            self._loaded_all = True
            return

        # Load all .md files except base guidelines
        template_files = [
            f for f in self.templates_dir.glob("*.md")
            if f.name != "_Base_Template_Guidelines.md"
        ]

        for template_file in template_files:
            self._load_single_framework(template_file)

        self._loaded_all = True

    def _load_framework(self, file_path: Path) -> Optional[FrameworkConfig]:
        """Load a single framework template from a markdown file"""
        content = file_path.read_text(encoding='utf-8')

        # Parse YAML frontmatter
        frontmatter_match = re.match(r'^---\n(.*?)\n---\n', content, re.DOTALL)
        if not frontmatter_match:
            logger.warning(f"No frontmatter found in {file_path.name}")
            return None

        frontmatter = yaml.safe_load(frontmatter_match.group(1))

        # Extract framework dimensions from markdown table
        dimensions = self._extract_dimensions(content)

        # Extract analysis steps
        analysis_steps = self._extract_analysis_steps(content)

        # Extract output format
        output_format = self._extract_output_format(content)

        # Extract description
        description = self._extract_description(content)

        return FrameworkConfig(
            template_id=frontmatter.get("template_id", ""),
            template_name=frontmatter.get("template_name", ""),
            template_type=frontmatter.get("template_type", ""),
            complexity=frontmatter.get("complexity", "medium"),
            confidence_required=frontmatter.get("confidence_required", 0.6),
            framework_dimensions=dimensions,
            analysis_steps=analysis_steps,
            output_format=output_format,
            description=description,
            file_path=file_path,
            required_data_sources=frontmatter.get("required_data_sources", []),
            chainable_with=frontmatter.get("chainable_with", []),
            data_quality_thresholds=frontmatter.get("data_quality_thresholds", {})
        )

    def _extract_dimensions(self, content: str) -> Dict[str, Any]:
        """Extract framework dimensions from markdown table"""
        dimensions = {}

        # Look for "Framework Dimensions" section
        dimensions_match = re.search(
            r'## Framework Dimensions\s*\n(.*?)(?=\n##|\n---|\Z)',
            content,
            re.DOTALL
        )

        if dimensions_match:
            table_text = dimensions_match.group(1)
            # Parse markdown table
            lines = [line.strip() for line in table_text.split('\n') if line.strip() and '|' in line]

            if len(lines) > 1:  # Header + at least one row
                # Parse header
                headers = [h.strip() for h in lines[0].split('|')[1:-1]]

                # Parse rows
                for line in lines[2:]:  # Skip header and separator
                    cells = [c.strip() for c in line.split('|')[1:-1]]
                    if len(cells) >= len(headers):
                        dimension_name = cells[0].replace('**', '').strip()
                        dimension_data = {headers[i]: cells[i] for i in range(1, min(len(headers), len(cells)))}
                        dimensions[dimension_name] = dimension_data

        return dimensions

    def _extract_analysis_steps(self, content: str) -> List[str]:
        """Extract analysis steps from numbered list"""
        steps = []

        steps_match = re.search(
            r'## Analysis Steps.*?\n(.*?)(?=\n##|\n---|\Z)',
            content,
            re.DOTALL
        )

        if steps_match:
            steps_text = steps_match.group(1)
            # Extract numbered steps
            step_pattern = r'\d+\.\s*\*\*(.*?)\*\*'
            matches = re.findall(step_pattern, steps_text)
            steps.extend(matches)

        return steps

    def _extract_output_format(self, content: str) -> Dict[str, Any]:
        """Extract output format specification"""
        output_format = {
            "summary_table": None,
            "executive_narrative": True,
            "optional_visuals": []
        }

        format_match = re.search(
            r'## Output Format\s*\n(.*?)(?=\n##|\n---|\Z)',
            content,
            re.DOTALL
        )

        if format_match:
            format_text = format_match.group(1)

            # Extract table structure
            table_match = re.search(r'### (.*?) Summary Table\s*\n(.*?)(?=\n###|\n---|\Z)', format_text, re.DOTALL)
            if table_match:
                table_name = table_match.group(1)
                table_content = table_match.group(2)
                output_format["summary_table"] = {
                    "name": table_name,
                    "structure": table_content
                }

            # Check for optional visuals
            visuals_match = re.search(r'### Optional Visuals\s*\n(.*?)(?=\n##|\Z)', format_text, re.DOTALL)
            if visuals_match:
                visuals_text = visuals_match.group(1)
                # Extract bullet points
                visuals = re.findall(r'-\s*\*\*(.*?)\*\*', visuals_text)
                output_format["optional_visuals"] = visuals

        return output_format

    def _extract_description(self, content: str) -> str:
        """Extract framework description"""
        desc_match = re.search(
            r'## Description\s*\n(.*?)(?=\n##|\n---|\Z)',
            content,
            re.DOTALL
        )

        if desc_match:
            return desc_match.group(1).strip()

        return ""

    def get_framework(self, framework_id: str) -> Optional[FrameworkConfig]:
        """
        Get a framework configuration by ID (lazy-loads if needed).

        Args:
            framework_id: The framework ID to retrieve

        Returns:
            FrameworkConfig if found, None otherwise
        """
        # Lazy-load the framework if not already loaded
        self._ensure_framework_loaded(framework_id)
        return self._frameworks.get(framework_id)

    def list_frameworks(self) -> List[str]:
        """List all available framework IDs (scans directory if needed)"""
        # Scan directory to ensure we have all frameworks
        self._scan_and_load_all()
        return list(self._frameworks.keys())

    def get_framework_names(self) -> Dict[str, str]:
        """Get mapping of framework IDs to names (scans directory if needed)"""
        # Scan directory to ensure we have all frameworks
        self._scan_and_load_all()
        return {fid: f.template_name for fid, f in self._frameworks.items()}


# Global framework loader instance
_framework_loader: Optional[FrameworkLoader] = None


def get_framework_loader() -> FrameworkLoader:
    """Get or create the global framework loader instance"""
    global _framework_loader
    if _framework_loader is None:
        _framework_loader = FrameworkLoader()
    return _framework_loader

````
