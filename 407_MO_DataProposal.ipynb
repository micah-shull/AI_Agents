## Data proposal review

The proposal aligns with the orchestrator requirements. Summary and refinements:

### Strengths

1. Covers the full campaign lifecycle: campaigns ‚Üí segments ‚Üí channels ‚Üí assets ‚Üí experiments ‚Üí metrics ‚Üí decisions ‚Üí ROI
2. Supports experimentation: experiments table with control/variant linking
3. Enables ROI tracking: dedicated ROI ledger with cost breakdown
4. Supports transparency: orchestrator decisions table for auditability
5. Governance-ready: `approved_by_human` and `risk_flags` in creative assets
6. MVP-appropriate: small datasets (10‚Äì20 rows) for learning

### Recommended refinements

Add these fields to support the KPIs and workflows:

#### 1. Performance Metrics ‚Äî add timestamp
```json
{
  "asset_id": "ASSET_005",
  "timestamp": "2026-03-15T10:00:00",  // ADD THIS
  "impressions": 12000,
  ...
}
```
Why: Supports "longitudinal tracking" and time-series analysis for effectiveness KPIs.

#### 2. Experiments ‚Äî add dates
```json
{
  "experiment_id": "EXP_001",
  "campaign_id": "CAMP_001",
  "start_date": "2026-03-10",  // ADD THIS
  "end_date": null,  // ADD THIS (null if running)
  "type": "A/B",
  ...
}
```
Why: Enables "experiment velocity" KPI (tests per campaign / time period).

#### 3. Campaigns ‚Äî add created_at
```json
{
  "campaign_id": "CAMP_001",
  "created_at": "2026-02-25T09:00:00",  // ADD THIS
  "name": "Spring Promo Awareness",
  ...
}
```
Why: Supports "latency from signal detection to action" operational KPI.

#### 4. Creative Assets ‚Äî add created_at
```json
{
  "asset_id": "ASSET_005",
  "created_at": "2026-03-05T14:30:00",  // ADD THIS
  "campaign_id": "CAMP_001",
  ...
}
```
Why: Supports "reduction in time from insight to action" effectiveness KPI.

#### 5. Performance Metrics ‚Äî optional experiment_id link
```json
{
  "asset_id": "ASSET_005",
  "experiment_id": "EXP_001",  // OPTIONAL: makes analysis easier
  "impressions": 12000,
  ...
}
```
Why: Simplifies linking metrics to experiments for analysis (can be inferred via asset_id if omitted).

### Optional enhancements (not required for MVP)

- Segment-level performance: add `segment_id` to performance_metrics if you need segment-level analysis
- Experiment segments: add `segment_id` to experiments if experiments are segment-specific

### Final verdict

The proposal is acceptable for MVP. The refinements above are minimal and support the stated KPIs. Without them, you can still build the orchestrator, but you‚Äôll miss some operational and effectiveness metrics.

Recommendation: Add the timestamp/date fields (#1‚Äì4) for MVP. They‚Äôre small additions that enable the KPIs described in the orchestrator doc.

**Unlocks:**

* Lift over time
* Decay effects

* Experiment velocity KPI
