````python
"""Metrics and decision analysis for GCO v3."""

from __future__ import annotations

from typing import Any, Dict, List, Optional

from .comparison_engine import (
    build_comparison,
    classify_trajectory,
    estimate_time_to_recovery,
    estimate_time_to_threshold,
    project_next_value,
    summarize_comparison,
)


def _avg(values: List[float]) -> float:
    return sum(values) / len(values) if values else 0.0


def _extract_numeric_history(rows: List[Dict[str, Any]], keys: List[str]) -> List[float]:
    history: List[float] = []
    for row in rows:
        for key in keys:
            value = row.get(key)
            if isinstance(value, (int, float)):
                history.append(float(value))
                break
    return history


def compute_risk_overview(bundle: Dict[str, Any]) -> Dict[str, Any]:
    portfolio = bundle.get("governance_portfolio_summary") or {}
    case_rows = bundle.get("governance_cases") or []
    bias_rows = bundle.get("bias_signals") or []
    drift_rows = bundle.get("drift_signals") or []

    current_risk = float(portfolio.get("overall_risk_score", 0.0))
    prior_risk = portfolio.get("prior_overall_risk_score")
    if isinstance(prior_risk, (int, float)):
        prior_risk = float(prior_risk)
    else:
        history_risk = _extract_numeric_history(bundle.get("drift_signals_history", []), ["risk_score", "value"])
        prior_risk = history_risk[-2] if len(history_risk) >= 2 else None

    risk_comparison = build_comparison("portfolio_risk_score", current_risk, prior_risk)

    exposure = 0.0
    for row in case_rows:
        value = row.get("financial_exposure_usd")
        if isinstance(value, (int, float)):
            exposure += float(value)
    prior_exposure = portfolio.get("prior_regulatory_exposure_usd")
    if isinstance(prior_exposure, (int, float)):
        prior_exposure = float(prior_exposure)
    else:
        prior_exposure = None
    exposure_comparison = build_comparison("financial_exposure_usd", exposure, prior_exposure)

    bias_scores = _extract_numeric_history(bias_rows, ["score", "value", "bias_score"])
    drift_scores = _extract_numeric_history(drift_rows, ["score", "value", "drift_score"])
    signal_risk = _avg(bias_scores + drift_scores)

    return {
        "risk_comparison": risk_comparison,
        "exposure_comparison": exposure_comparison,
        "signal_risk_score": signal_risk,
        "open_cases": len(case_rows),
        "comparison_summaries": [
            summarize_comparison(risk_comparison),
            summarize_comparison(exposure_comparison),
        ],
    }


def compute_trajectory_and_projection(bundle: Dict[str, Any], thresholds: Dict[str, float]) -> Dict[str, Any]:
    risk_history = _extract_numeric_history(
        bundle.get("drift_signals_history", []), ["risk_score", "value", "drift_score"]
    )
    if not risk_history:
        risk_history = _extract_numeric_history(bundle.get("bias_signals_history", []), ["risk_score", "value"])

    if len(risk_history) >= 2:
        current = risk_history[-1]
        prior = risk_history[-2]
    else:
        current = 0.0
        prior = None

    trajectory = classify_trajectory(risk_history) if risk_history else "unknown"
    threshold = float(thresholds.get("risk_score_critical", 80.0))
    projected_next = project_next_value(current, prior, future_steps=1)
    projected_two = project_next_value(current, prior, future_steps=2)
    time_to_breach = estimate_time_to_threshold(current, prior, threshold)
    time_to_recovery = estimate_time_to_recovery(current, prior, threshold)

    return {
        "risk_history": risk_history,
        "trajectory": trajectory,
        "current_risk_from_history": current,
        "projected_risk_next_run": projected_next,
        "projected_risk_two_runs": projected_two,
        "time_to_threshold_breach_runs": time_to_breach,
        "time_to_recovery_runs": time_to_recovery,
        "threshold_risk_score_critical": threshold,
    }


def determine_decision(overview: Dict[str, Any], trajectory: Dict[str, Any], thresholds: Dict[str, float]) -> Dict[str, Any]:
    risk = overview["risk_comparison"]
    exposure = overview["exposure_comparison"]
    risk_value = float(risk["value"] or 0.0)
    risk_velocity = risk["velocity"]
    risk_trend = risk["trend_direction"]
    exposure_multiple = exposure.get("multiple")

    high_risk = risk_value >= float(thresholds.get("risk_score_high", 60.0))
    critical_risk = risk_value >= float(thresholds.get("risk_score_critical", 80.0))
    accelerating = risk_velocity == "fast" or risk_trend == "increasing"
    exposure_spike = isinstance(exposure_multiple, (int, float)) and exposure_multiple >= float(
        thresholds.get("exposure_multiple_escalation", 2.0)
    )
    deteriorating = trajectory.get("trajectory") == "deteriorating"

    if critical_risk or (high_risk and accelerating and exposure_spike):
        verdict = "immediate_intervention_required"
        escalation = True
    elif high_risk or deteriorating:
        verdict = "targeted_mitigation_required"
        escalation = False
    else:
        verdict = "continue_monitored_operations"
        escalation = False

    actions: List[str] = []
    if escalation:
        actions.append("Trigger executive governance review within 24 hours.")
        actions.append("Freeze high-risk agent policies until remediation evidence is captured.")
    if high_risk:
        actions.append("Assign case owners for top risk contributors and set 7-day remediation checkpoints.")
    if trajectory.get("trajectory") in {"deteriorating", "volatile"}:
        actions.append("Increase monitoring cadence to every run and require human override audit.")
    if not actions:
        actions.append("Maintain weekly monitoring and validate threshold assumptions next cycle.")

    return {
        "verdict": verdict,
        "executive_escalation_triggered": escalation,
        "recommended_actions": actions[:3],
        "decision_factors": {
            "risk_value": risk_value,
            "risk_velocity": risk_velocity,
            "risk_trend": risk_trend,
            "trajectory": trajectory.get("trajectory"),
            "exposure_multiple": exposure_multiple,
        },
    }
````
