````python
# ============================================================================
# Third-Party Risk Orchestrator V2
# ============================================================================
# Governance-grade vendor risk: contracts, risk events, financial exposure,
# control maturity trending, executive trigger policies. State and config are
# v2-only; original TPRO (v1) remains unchanged elsewhere.
# ============================================================================

class ThirdPartyRiskOrchestratorV2State(TypedDict, total=False):
    """State for Third-Party Risk Orchestrator v2 (governance-grade vendor risk)."""

    # Input
    vendor_ids_filter: Optional[List[str]]  # Optional: limit to specific vendors; None = all

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Loaded data (raw lists from agents/data)
    third_parties: List[Dict[str, Any]]
    vendor_contracts: List[Dict[str, Any]]
    vendor_controls: List[Dict[str, Any]]
    risk_events: List[Dict[str, Any]]
    financial_exposure: List[Dict[str, Any]]
    executive_trigger_rules: List[Dict[str, Any]]
    control_maturity_history: List[Dict[str, Any]]
    risk_assessments: List[Dict[str, Any]]
    assessment_history: List[Dict[str, Any]]
    mitigation_actions: List[Dict[str, Any]]
    human_reviews: List[Dict[str, Any]]
    external_signals: List[Dict[str, Any]]
    vendor_performance: List[Dict[str, Any]]
    risk_domains: List[Dict[str, Any]]
    orchestrator_metrics: List[Dict[str, Any]]

    # Lookups (vendor_id -> entity or list) for fast scoring/trigger evaluation
    third_parties_lookup: Dict[str, Dict[str, Any]]
    vendor_contracts_lookup: Dict[str, Dict[str, Any]]
    vendor_controls_lookup: Dict[str, List[Dict[str, Any]]]
    risk_events_lookup: Dict[str, List[Dict[str, Any]]]
    financial_exposure_lookup: Dict[str, Dict[str, Any]]

    # Per-vendor risk and trigger evaluation (rule-based)
    vendor_risk_scores: List[Dict[str, Any]]   # [{vendor_id, overall_risk_score, factors, trend_direction}, ...]
    triggered_rules: List[Dict[str, Any]]      # Executive trigger rules that fired
    escalation_alerts: List[Dict[str, Any]]    # Alerts for HITL / board

    # Rollup / executive one-view
    portfolio_rollup: Dict[str, Any]   # high_risk_count, critical_triggers_count, renewal_at_risk, etc.

    # Output
    executive_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class ThirdPartyRiskOrchestratorV2Config:
    """Configuration for Third-Party Risk Orchestrator v2."""
    llm_model: str = os.getenv("LLM_MODEL", "gpt-4o-mini")
    temperature: float = 0.3

    data_dir: str = "agents/data"
    reports_dir: str = "agents/output/third_party_risk_orchestrator_v2"

    # Data file names (under data_dir)
    third_parties_file: str = "third_parties.json"
    vendor_contracts_file: str = "vendor_contracts.json"
    vendor_controls_file: str = "vendor_controls.json"
    risk_events_file: str = "risk_events.json"
    financial_exposure_file: str = "financial_exposure.json"
    executive_trigger_rules_file: str = "executive_trigger_rules.json"
    control_maturity_history_file: str = "control_maturity_history.json"
    risk_assessments_file: str = "risk_assessments.json"
    assessment_history_file: str = "assessment_history.json"
    mitigation_actions_file: str = "mitigation_actions.json"
    human_reviews_file: str = "human_reviews.json"
    external_signals_file: str = "external_signals.json"
    vendor_performance_file: str = "vendor_performance.json"
    risk_domains_file: str = "risk_domains.json"
    orchestrator_metrics_file: str = "orchestrator_metrics.json"

    # Risk / escalation thresholds (for report bar and triggers)
    high_risk_score_threshold: float = 65.0
    critical_risk_score_threshold: float = 80.0
    renewal_window_days: int = 90
````
