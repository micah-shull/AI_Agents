# Lessons Learned

## Overview

This document captures what worked, what didn't, and key insights from building the strategic analysis agent system.

## âœ… What Worked Well

### 1. Persona-Based Prompts
**What:** Framework-specific personas in prompts  
**Why it worked:** LLM behavior improved significantly with specific expertise  
**Impact:** 40-50% improvement in output quality  
**Takeaway:** Always use framework-specific personas, not generic ones

### 2. Source Reliability Scoring
**What:** Scoring sources by reliability and adjusting confidence  
**Why it worked:** Better confidence scores, flags low-quality sources  
**Impact:** More accurate confidence ratings  
**Takeaway:** Score sources early, adjust confidence based on reliability

### 3. Refinement Loops
**What:** Quality assessment â†’ Review â†’ Rewrite cycle  
**Why it worked:** Iterative improvement of output  
**Impact:** Reports improved with each iteration  
**Takeaway:** Post-hoc refinement works, but guided generation is better

### 4. Template-Based Analyzers
**What:** Single analyzer that reads framework templates  
**Why it worked:** DRY principle, easy to add frameworks  
**Impact:** One analyzer for multiple frameworks  
**Takeaway:** Template-based approach scales well

### 5. Framework Loader
**What:** Lazy-loading framework templates  
**Why it worked:** Faster startup, only load what's needed  
**Impact:** Better resource usage  
**Takeaway:** Lazy loading is worth it

## âŒ What Didn't Work

### 1. Single Monolithic Agent
**What:** One agent for all frameworks  
**Why it failed:** Too much conditional logic, can't optimize  
**Impact:** Lower quality, harder to maintain  
**Takeaway:** Split into specialized agents

### 2. Generic Personas
**What:** Same persona for all frameworks  
**Why it failed:** Not specific enough, generic output  
**Impact:** Lower quality insights  
**Takeaway:** Framework-specific personas are essential

### 3. Post-Hoc Validation Only
**What:** Check quality after generation  
**Why it failed:** Can't fix issues during generation  
**Impact:** Lower quality, more refinement needed  
**Takeaway:** Need guided generation, not just validation

### 4. Dict-Based State
**What:** Using dicts for all state  
**Why it failed:** Key collisions, no validation, hard to debug  
**Impact:** Bugs, unclear state structure  
**Takeaway:** Use Pydantic models and ContextStore

### 5. No Source Filtering
**What:** All sources in analysis context  
**Why it failed:** Low-quality sources dominate, poor insights  
**Impact:** Lower quality, generic insights  
**Takeaway:** Filter sources before analysis

### 6. No Competitive Benchmarking
**What:** Missing competitive context  
**Why it failed:** Insights too generic, no comparisons  
**Impact:** Less strategic value  
**Takeaway:** Benchmarking is essential for competitive analysis

### 7. Single-Pass Analysis
**What:** One LLM call tries to do everything  
**Why it failed:** Can't enforce depth, implications, benchmarks  
**Impact:** Generic insights, missing strategic implications  
**Takeaway:** Multi-stage analysis is needed

### 8. Qualitative Impact Only
**What:** Just "High/Medium/Low" labels  
**Why it failed:** No justification, feels arbitrary  
**Impact:** Less rigorous, harder to prioritize  
**Takeaway:** Need quantitative scoring with rubric

## ðŸŽ¯ Key Insights

### Insight 1: Guided Generation > Post-Hoc Validation
**Finding:** Checking quality after generation is too late. Need to guide during generation.

**Evidence:**
- Persona improvements helped, but not enough
- Generic insights still appeared
- Strategic implications often missing

**Solution:**
- Multi-stage analysis
- Depth validation during generation
- Implication enforcement
- Benchmark enrichment

### Insight 2: Source Quality Matters Most
**Finding:** Low-quality sources lead to generic insights, regardless of persona.

**Evidence:**
- Reports with high-quality sources were much better
- Low-quality sources (blogs, SEO sites) produced generic insights
- Source filtering would have helped significantly

**Solution:**
- Filter sources before analysis
- Prioritize high-quality sources
- Limit low-quality sources to 20% of context

### Insight 3: Competitive Context is Essential
**Finding:** Strategic insights without competitive context are generic.

**Evidence:**
- Insights like "intense competition from Amazon" are obvious
- Comparative insights ("Target's SG&A 2.3% higher than Walmart") are valuable
- Benchmarking enables strategic depth

**Solution:**
- Always gather competitive benchmarks
- Add benchmarks to insights
- Use benchmarks in impact scoring

### Insight 4: Framework Differences Matter
**Finding:** Frameworks are different enough that one-size-fits-all doesn't work.

**Evidence:**
- SWOT needs internal/external validation
- Porter's needs industry structure analysis
- PESTEL needs macro research (no company focus)
- Ansoff needs market/product analysis

**Solution:**
- Split into specialized agents
- Framework-specific components
- Framework-aware planning

### Insight 5: Typed State Prevents Bugs
**Finding:** Dict-based state leads to key collisions and unclear structure.

**Evidence:**
- Hard to know what keys exist
- Easy to overwrite keys
- No validation
- Hard to debug

**Solution:**
- Pydantic models for all state
- ContextStore for safe access
- Type checking

### Insight 6: Observability is Critical
**Finding:** Debugging multi-node workflows without tracing is very hard.

**Evidence:**
- Hard to know which component failed
- Can't track LLM costs
- No performance metrics
- Hard to demo

**Solution:**
- Tracing from day one
- Component execution tracking
- Cost monitoring
- Performance metrics

# Architecture Decisions

## Overview

This document captures key architectural decisions made during development, with rationale and alternatives considered.

---

## Decision 1: Component Library Architecture

**Decision:** Build a modular component library where nodes are reusable, testable units.

**Rationale:**
- Enables code reuse across multiple agents
- Makes testing easier (test components independently)
- Allows experimentation (mix and match components)
- Follows industry best practices (LangChain, Anthropic patterns)

**Alternatives Considered:**
- Monolithic agent with all logic inline
- Framework-specific agents with no shared code
- Single agent with conditional logic

**Status:** âœ… Implemented in design, ready for implementation

---

## Decision 2: Split Into Specialized Agents

**Decision:** Split into 3 specialized agents:
1. Competitive Analysis Agent (SWOT, Porter's)
2. Macro Analysis Agent (PESTEL)
3. Growth Strategy Agent (Ansoff)

**Rationale:**
- Frameworks have different needs (research, analysis, benchmarking)
- Specialized agents can be optimized for their frameworks
- Clearer separation of concerns
- Better code organization

**Alternatives Considered:**
- Single agent for all frameworks (too much conditional logic)
- One agent per framework (too much duplication)
- Two agents (less optimal than three)

**Status:** âœ… Recommended approach

---

## Decision 3: Pydantic Data Models

**Decision:** Use Pydantic models for all state objects instead of dicts.

**Rationale:**
- Type safety and validation
- Better IDE support
- Self-documenting
- Easier debugging
- Automatic serialization

**Alternatives Considered:**
- Dict-based state (too error-prone)
- TypedDict (less validation)
- Custom classes (more boilerplate)

**Status:** âœ… Implemented in foundation

---

## Decision 4: ContextStore for State Management

**Decision:** Use ContextStore class instead of raw dict for state.

**Rationale:**
- Prevents key collisions
- Enforces read/write permissions
- Supports namespacing
- Better debugging (access logging)
- Version tracking

**Alternatives Considered:**
- Raw dict (too brittle)
- TypedDict (no runtime safety)
- State machine library (overkill)

**Status:** âœ… Implemented in foundation

---

## Decision 5: Framework Planner

**Decision:** Add planning layer that determines which components to use for each framework.

**Rationale:**
- Different frameworks need different components
- Avoids wasted component calls
- Enables dynamic workflow construction
- Framework-specific optimizations

**Alternatives Considered:**
- Hardcode component lists in agents
- Use same components for all frameworks
- Conditional logic in components

**Status:** âœ… Implemented in foundation

---

## Decision 6: Source Filtering Before Analysis

**Decision:** Filter and prioritize sources in research node, before analysis.

**Rationale:**
- LLM sees better sources first
- Low-quality sources don't dominate context
- Reduces need for post-hoc warnings

- Filter after analysis (too late)
