# 1. Minor Bug: Double “One Ask”

You have:

```
**One ask:** **One ask:** Review triggered items below...
```

This happens because:

```
_so_what_line()
_next_steps()
report builder
```

are each trying to inject the same phrase.

### Fix

Only let the **report builder** add the label.

Change this:

```python
steps.append("**One ask:** Review triggered items below...")
```

to:

```python
steps.append("Review triggered items below...")
```

Then the report builder will produce:

```
**One ask:** Review triggered items below...
```

Much cleaner.

# 2. Risk Severity Counts Are Incorrect

You show:

```
Open risk events: 4
```

But severity counts show:

```
medium: 2
low: 2
high: 1
critical: 2
```

That totals **7 events**.

That’s confusing to a reader.

### Why this happens

Your risk rollup counts **all events**, while the open count only counts:

```
open
investigating
blocked
```

So resolved risks still appear in severity counts.

### Recommended fix

Add **two sections**:

```
Open risk events
Risk event history
```

Example:

```
Open risk events: 4
By severity (open only): {'critical': 2, 'high': 1, 'medium': 1}

Historical events: 7
```

This prevents executives from thinking risk counts are inconsistent.

# 3. Portfolio Verdict Logic

Your verdict says:

```
Verdict vs target: Attention
(target: healthy)
```

But portfolio status is:

```
healthy_with_watchlist
```

This is actually **not a failure state**.

Right now your logic is:

```
if status == target → On target
else → Attention
```

That’s too coarse.

### Better logic

```
healthy → On target
healthy_with_watchlist → Watch
at_risk → Attention
critical → Escalate
```

Then your report would say:

```
Verdict vs target: Watch
```

Much more realistic.
