````python
"""
Executive report generation: one-pager, exposure by tier/cause, top interventions, next steps (RGOv2 + EXECUTIVE_GRADE).
"""

from typing import Any, Dict, List


def _verdict_from_rei(total_rei: float) -> str:
    """Simple verdict: HIGH / MEDIUM / LOW from risk-weighted exposure (thresholds are illustrative)."""
    if total_rei >= 50000:
        return "HIGH"
    if total_rei >= 15000:
        return "MEDIUM"
    return "LOW"


def _so_what_line(rollup: Dict[str, Any], top_n: int) -> str:
    """One sentence: business impact and who must act."""
    total_rei = rollup.get("total_rei", 0)
    by_cause = rollup.get("exposure_by_root_cause") or {}
    causes = [k for k, v in by_cause.items() if v > 0]
    count = rollup.get("customer_count_with_gaps", 0)
    return (
        f"Risk-weighted revenue exposure ${total_rei:,.0f} across {count} customer(s). "
        f"Top {top_n} interventions prioritized. "
        f"Root causes: {', '.join(causes) or 'N/A'}. "
        "Revenue and Customer Success must act on top priorities."
    )


def _one_ask(top_priority_gaps: List[Dict[str, Any]]) -> str:
    """Single executive action."""
    if not top_priority_gaps:
        return "Review portfolio; no gaps above threshold this run."
    return (
        f"Assign top {min(5, len(top_priority_gaps))} priority customers to owners; "
        "require intervention status within 5 business days."
    )


def _concentration_line(top_priority_gaps: List[Dict[str, Any]], total_rei: float) -> str:
    """Top N account for X% of exposure."""
    if not top_priority_gaps or total_rei <= 0:
        return "N/A"
    top_rei = sum(g.get("rei", 0) for g in top_priority_gaps)
    pct = (top_rei / total_rei) * 100
    return f"Top {len(top_priority_gaps)} customers account for {pct:.0f}% of risk-weighted exposure."


def generate_rgo_v2_report(state: Dict[str, Any]) -> str:
    """
    Build full markdown report from state: goal, rollup, top_priority_gaps, config top_n.
    """
    rollup = state.get("portfolio_rollup") or {}
    top_priority_gaps = state.get("top_priority_gaps") or []
    top_n = 20
    if "top_n_priority" in state:
        top_n = state["top_n_priority"]

    total_gap = rollup.get("total_gap", 0)
    total_rei = rollup.get("total_rei", 0)
    count = rollup.get("customer_count_with_gaps", 0)
    verdict = _verdict_from_rei(total_rei)
    so_what = _so_what_line(rollup, top_n)
    one_ask = _one_ask(top_priority_gaps)
    concentration = _concentration_line(top_priority_gaps, total_rei)

    lines = [
        "# Revenue Gap Orchestrator v2 â€” Executive Report",
        "",
        "## Revenue Exposure Overview",
        "",
        f"- **Verdict:** {verdict}",
        f"- **Total revenue gap:** ${total_gap:,.2f}",
        f"- **Risk-weighted exposure (REI):** ${total_rei:,.2f}",
        f"- **Customers with gaps/at risk:** {count}",
        f"- **Concentration:** {concentration}",
        "",
        f"**So what:** {so_what}",
        "",
        f"**One ask:** {one_ask}",
        "",
        "---",
        "",
        "## Exposure by Tier",
        "",
    ]

    by_tier = rollup.get("exposure_by_tier") or {}
    if by_tier:
        lines.append("| Tier | Risk-weighted exposure (REI) |")
        lines.append("|------|-----------------------------|")
        for tier, val in sorted(by_tier.items(), key=lambda x: -x[1]):
            lines.append(f"| {tier} | ${val:,.2f} |")
    else:
        lines.append("No exposure by tier.")
    lines.append("")

    lines.append("## Exposure by Root Cause")
    lines.append("")
    by_cause = rollup.get("exposure_by_root_cause") or {}
    if by_cause:
        lines.append("| Root cause | Risk-weighted exposure (REI) |")
        lines.append("|------------|-----------------------------|")
        for cause, val in sorted(by_cause.items(), key=lambda x: -x[1]):
            lines.append(f"| {cause} | ${val:,.2f} |")
    else:
        lines.append("No exposure by cause.")
    lines.append("")

    lines.append("## Top Priority Interventions")
    lines.append("")
    lines.append("| # | Customer ID | Gap ($) | REI | Tier | Root cause |")
    lines.append("|---|------------|--------|-----|------|------------|")
    for i, g in enumerate(top_priority_gaps[:top_n], 1):
        cid = g.get("customer_id", "")
        gap = g.get("gap_amount", 0)
        rei = g.get("rei", 0)
        tier = g.get("tier", "")
        rc = g.get("root_cause", "")
        lines.append(f"| {i} | {cid} | {gap:,.2f} | {rei:,.2f} | {tier} | {rc} |")
    lines.append("")

    lines.append("## Next Steps")
    lines.append("")
    lines.append("1. " + one_ask)
    lines.append("2. Review exposure by root cause; assign operational vs demand vs retention actions.")
    lines.append("3. Track recovery on top 20; re-run next cycle for trend.")
    lines.append("")

    return "\n".join(lines)
````
