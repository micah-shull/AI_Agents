````python
from __future__ import annotations

from pathlib import Path
from unittest.mock import patch

import pytest

from config import SalesEnablementOrchestratorV3Config
from agents.SEv3.orchestrator.orchestrator import create_sev3_orchestrator


REPO_ROOT = Path(__file__).resolve().parent
DATA_DIR = REPO_ROOT / "agents" / "data"

# Wired in `create_sev3_orchestrator` (plus __start__ / __end__ on the compiled graph).
EXPECTED_WORKFLOW_NODE_IDS = (
    "goal",
    "planning",
    "data_loading",
    "temporal_metrics",
    "enablement_snapshot",
    "report_generation",
)


def test_sev3_compiled_graph_structure_matches_plan():
    """Graph wiring stays aligned with the intended node chain (no silent drift)."""
    graph = create_sev3_orchestrator(project_root=str(REPO_ROOT))
    node_ids = set(graph.get_graph().nodes.keys())
    for name in EXPECTED_WORKFLOW_NODE_IDS:
        assert name in node_ids, f"missing graph node: {name}"


@patch("agents.SEv3.orchestrator.nodes.load_sev3_inputs", side_effect=RuntimeError("partial failure"))
def test_sev3_graph_surfaces_data_loading_failure(_mock_loader, tmp_path: Path):
    """Integration: loader failure must appear on final state (no silent success)."""
    config = SalesEnablementOrchestratorV3Config(
        data_dir="agents/data",
        reports_dir=str(tmp_path),
    )
    graph = create_sev3_orchestrator(config=config, project_root=str(REPO_ROOT))
    result = graph.invoke({"errors": [], "view_mode": "portfolio"})

    errs = result.get("errors") or []
    assert errs
    assert any("data_loading_node failed" in e for e in errs)
    assert "partial failure" in errs[-1]


@pytest.mark.skipif(not DATA_DIR.exists(), reason="requires agents/data fixtures")
def test_sev3_graph_integration_generates_report(tmp_path: Path):
    config = SalesEnablementOrchestratorV3Config(
        data_dir="agents/data",
        reports_dir=str(tmp_path),
    )
    graph = create_sev3_orchestrator(config=config, project_root=str(REPO_ROOT))
    result = graph.invoke({"errors": [], "view_mode": "portfolio"})

    assert result.get("errors") == []
    assert result.get("goal", {}).get("view_mode") == "portfolio"
    assert len(result.get("plan") or []) == 4
    assert result.get("data_bundle")
    assert result.get("data_counts", {}).get("deals_count", 0) > 0
    assert result.get("reference_datetime_utc")
    assert isinstance(result.get("validation_warnings"), list)
    assert result.get("temporal_metrics", {}).get("pipeline_metrics", {}).get("current_period")
    assert "# Sales Enablement Report" in (result.get("enablement_report_md") or "")

    report_path = Path(result.get("report_file_path") or "")
    assert report_path.exists()
    assert report_path.parent == tmp_path

    period = result["temporal_metrics"]["pipeline_metrics"]["current_period"]
    assert report_path.name == f"sales_enablement_orchestrator_v3_{period}.md"

    content = report_path.read_text(encoding="utf-8")
    assert "# Sales Enablement Orchestrator v3 (SEv3)" in content
    assert "## Executive Summary" in content
    assert "**Immediate priorities:**" in content
    assert "### Top Revenue Risks" in content
    assert "## Run Context" in content
    assert "**Dataset scale:**" in content
    assert "## Trigger Evolution (new signals)" in content
    assert "## Enablement Actions (Current Snapshot)" in content
    assert "### Operational Status Snapshot" in content

    assert isinstance(result.get("processing_time"), (int, float))

    # Narrative ↔ metrics: coverage ratio in markdown matches pipeline_metrics (no print drift).
    cov = result["temporal_metrics"]["pipeline_metrics"].get("coverage_ratio_current")
    if isinstance(cov, (int, float)):
        assert f"{float(cov):.2f}" in content


@pytest.mark.skipif(not DATA_DIR.exists(), reason="requires agents/data fixtures")
def test_sev3_graph_preserves_seed_errors_through_successful_run(tmp_path: Path):
    """Full LangGraph: pre-existing `errors` must not be dropped when the run succeeds."""
    config = SalesEnablementOrchestratorV3Config(
        data_dir="agents/data",
        reports_dir=str(tmp_path),
    )
    graph = create_sev3_orchestrator(config=config, project_root=str(REPO_ROOT))
    seed = "integration: prior warning (seed)"
    result = graph.invoke({"errors": [seed], "view_mode": "portfolio"})

    assert seed in (result.get("errors") or [])
    report_path = Path(result.get("report_file_path") or "")
    assert report_path.exists()
    assert report_path.parent == tmp_path
    assert "# Sales Enablement Orchestrator v3 (SEv3)" in report_path.read_text(encoding="utf-8")
````
