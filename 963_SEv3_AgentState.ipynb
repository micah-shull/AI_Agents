````python
class SalesEnablementOrchestratorV3State(TypedDict, total=False):
    """State for Sales Enablement Orchestrator v3 (temporal revenue intelligence)."""

    # Common orchestration fields
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]
    errors: List[str]
    processing_time: Optional[float]

    # Agent inputs (from project-level orchestrator `options`)
    view_mode: Optional[str]
    project_root: Optional[str]

    # Data loading outputs
    reference_datetime_utc: Optional[str]
    data_counts: Dict[str, Any]
    validation_warnings: List[str]
    data_bundle: Dict[str, Any]

    # Temporal metrics outputs
    temporal_metrics: Dict[str, Any]

    # Enablement outputs
    enablement_report_md: str

    # Final output
    report_file_path: Optional[str]


@dataclass
class SalesEnablementOrchestratorV3Config:
    """Configuration for Sales Enablement Orchestrator v3."""

    # Project-relative paths
    data_dir: str = "agents/data"
    reports_dir: str = "output/SEv3_reports"

    # Enablement knobs
    top_priority_leads_n: int = 10
    enablement_follow_up_overdue_days: int = 3

    # Report knobs
    top_deals_limit: int = 8


````
