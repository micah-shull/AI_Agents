````python
"""LangGraph workflow for Third-Party Risk Orchestrator v2."""

from langgraph.graph import END, StateGraph

from config import ThirdPartyRiskOrchestratorV2Config, ThirdPartyRiskOrchestratorV2State
from agents.third_party_risk_orchestrator_v2.orchestrator.nodes import (
    goal_node,
    planning_node,
    make_data_loading_node,
    risk_scoring_node,
    make_trigger_evaluation_node,
    make_rollup_node,
    make_report_node,
)


def create_third_party_risk_orchestrator_v2(
    config: ThirdPartyRiskOrchestratorV2Config | None = None,
):
    """Build and compile the TPRO v2 graph. Uses config for data/report paths and thresholds."""
    if config is None:
        config = ThirdPartyRiskOrchestratorV2Config()

    data_loading_node = make_data_loading_node(config)
    trigger_evaluation_node = make_trigger_evaluation_node(config)
    rollup_node = make_rollup_node(config)
    report_node = make_report_node(config)

    workflow = StateGraph(ThirdPartyRiskOrchestratorV2State)

    workflow.add_node("goal", goal_node)
    workflow.add_node("planning", planning_node)
    workflow.add_node("data_loading", data_loading_node)
    workflow.add_node("risk_scoring", risk_scoring_node)
    workflow.add_node("trigger_evaluation", trigger_evaluation_node)
    workflow.add_node("rollup", rollup_node)
    workflow.add_node("report_generation", report_node)

    workflow.set_entry_point("goal")
    workflow.add_edge("goal", "planning")
    workflow.add_edge("planning", "data_loading")
    workflow.add_edge("data_loading", "risk_scoring")
    workflow.add_edge("risk_scoring", "trigger_evaluation")
    workflow.add_edge("trigger_evaluation", "rollup")
    workflow.add_edge("rollup", "report_generation")
    workflow.add_edge("report_generation", END)

    return workflow.compile()
````
