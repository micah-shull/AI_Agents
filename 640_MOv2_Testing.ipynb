````python
#!/usr/bin/env python3
"""Run Marketing Orchestrator V2. Usage: python run_marketing_orchestrator_v2.py [--campaign-id CAMP_001]"""
import argparse
import sys
from pathlib import Path

root = Path(__file__).resolve().parent
if str(root) not in sys.path:
    sys.path.insert(0, str(root))

from config import MarketingOrchestratorConfig
from agents.marketing_orchestrator_v2.orchestrator import create_marketing_orchestrator_v2


def main():
    parser = argparse.ArgumentParser(description="Marketing Orchestrator V2")
    parser.add_argument("--campaign-id", type=str, default=None, help="Analyze only this campaign")
    args = parser.parse_args()

    config = MarketingOrchestratorConfig()
    orchestrator = create_marketing_orchestrator_v2(config)
    initial_state = {"errors": []}
    if args.campaign_id:
        initial_state["campaign_id"] = args.campaign_id

    result = orchestrator.invoke(initial_state)

    if result.get("errors"):
        for e in result["errors"]:
            print(e, file=sys.stderr)
        sys.exit(1)
    print("Done.")
    if result.get("report_file_path"):
        print("Report:", result["report_file_path"])
    if result.get("campaign_report"):
        print("\n--- Report preview ---\n")
        print(result["campaign_report"][:2000])
        if len(result["campaign_report"]) > 2000:
            print("\n... (truncated)")
    return 0


if __name__ == "__main__":
    sys.exit(main())
````
