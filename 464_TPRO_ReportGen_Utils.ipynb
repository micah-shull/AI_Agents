````python
"""Report generation utilities for Third-Party Risk Orchestrator

This module contains utilities to generate comprehensive risk assessment reports.
Integrates with toolshed reporting utilities.

All utilities are pure functions, independently testable.

Following MVP-first approach: Rule-based report generation, toolshed integration.
"""

from typing import List, Dict, Any, Optional
from datetime import datetime
from toolshed.reporting.file_handling import save_report


def generate_executive_summary(
    orchestrator_metrics: Dict[str, Any],
    risk_assessments: List[Dict[str, Any]],
    kpi_metrics: Dict[str, Any]
) -> str:
    """
    Generate executive summary section of the report.

    Args:
        orchestrator_metrics: Orchestrator-level metrics
        risk_assessments: List of risk assessments
        kpi_metrics: KPI metrics dictionary

    Returns:
        Executive summary markdown section
    """
    vendors_evaluated = orchestrator_metrics.get("vendors_evaluated", 0)
    assessments_completed = orchestrator_metrics.get("assessments_completed", 0)
    high_risk_count = orchestrator_metrics.get("high_risk_vendors", 0)
    medium_risk_count = orchestrator_metrics.get("medium_risk_vendors", 0)
    low_risk_count = orchestrator_metrics.get("low_risk_vendors", 0)
    human_escalations = orchestrator_metrics.get("human_escalations", 0)

    business_kpis = kpi_metrics.get("business", {})
    total_cost = business_kpis.get("total_run_cost_usd", 0.0)
    net_value = business_kpis.get("net_value_usd", 0.0)
    roi_percentage = business_kpis.get("roi_percentage", 0.0)
    manual_hours_saved = business_kpis.get("estimated_manual_hours_saved", 0.0)

    summary = f"""## Executive Summary

**Run Date:** {orchestrator_metrics.get('run_date', 'N/A')}
**Run ID:** {orchestrator_metrics.get('run_id', 'N/A')}

### Risk Assessment Overview

- **Vendors Evaluated:** {vendors_evaluated}
- **Assessments Completed:** {assessments_completed}
- **High-Risk Vendors:** {high_risk_count} (requiring immediate attention)
- **Medium-Risk Vendors:** {medium_risk_count}
- **Low-Risk Vendors:** {low_risk_count}
- **Human Escalations:** {human_escalations}

### Business Impact

- **Total Cost:** ${total_cost:,.2f}
- **Net Value:** ${net_value:,.2f}
- **ROI:** {roi_percentage:.1f}%
- **Manual Hours Saved:** {manual_hours_saved:.1f} hours

### Key Findings

"""

    # Add key findings based on risk levels
    if high_risk_count > 0:
        summary += f"- âš ï¸ **{high_risk_count} high-risk vendor(s)** require immediate remediation\n"

    if human_escalations > 0:
        summary += f"- ðŸ” **{human_escalations} vendor(s)** escalated for human review\n"

    effectiveness_kpis = kpi_metrics.get("effectiveness", {})
    time_to_identify = effectiveness_kpis.get("time_to_identify_risk_hours", 0.0)
    if time_to_identify > 0:
        summary += f"- â±ï¸ **Risk signals identified** within {time_to_identify:.1f} hours\n"

    if roi_percentage > 0:
        summary += f"- ðŸ’° **Positive ROI** of {roi_percentage:.1f}% demonstrates value\n"

    return summary


def generate_vendor_risk_overview(
    risk_assessments: List[Dict[str, Any]],
    vendor_lookup: Dict[str, Dict[str, Any]],
    approval_history: List[Dict[str, Any]],
    mitigation_actions: List[Dict[str, Any]]
) -> str:
    """
    Generate vendor risk overview section.

    Args:
        risk_assessments: List of risk assessments
        vendor_lookup: Fast vendor lookup
        approval_history: Approval decisions
        mitigation_actions: Mitigation actions

    Returns:
        Vendor risk overview markdown section
    """
    # Sort by risk score (highest first)
    sorted_assessments = sorted(
        risk_assessments,
        key=lambda x: x.get("overall_risk_score", 0.0),
        reverse=True
    )

    section = "## Vendor Risk Overview\n\n"
    section += "### High-Risk Vendors (Requiring Immediate Attention)\n\n"

    high_risk_vendors = [a for a in sorted_assessments if a.get("risk_level") == "high"]
    if high_risk_vendors:
        for assessment in high_risk_vendors:
            vendor_id = assessment.get("vendor_id")
            vendor_data = vendor_lookup.get(vendor_id, {})
            vendor_name = vendor_data.get("vendor_name", vendor_id)

            section += f"#### {vendor_name} ({vendor_id})\n\n"
            section += f"- **Risk Score:** {assessment.get('overall_risk_score', 0.0):.1f}/100\n"
            section += f"- **Risk Level:** {assessment.get('risk_level', 'unknown').upper()}\n"
            section += f"- **Criticality:** {vendor_data.get('criticality', 'unknown').upper()}\n"

            primary_domains = assessment.get("primary_risk_domains", [])
            if primary_domains:
                section += f"- **Primary Risk Domains:** {', '.join(primary_domains)}\n"

            key_drivers = assessment.get("key_drivers", [])
            if key_drivers:
                section += f"- **Key Risk Drivers:**\n"
                for driver in key_drivers[:5]:  # Top 5
                    section += f"  - {driver}\n"

            # Check for approval
            approval = next(
                (a for a in approval_history if a.get("vendor_id") == vendor_id),
                None
            )
            if approval:
                section += f"- **Review Status:** {approval.get('decision', 'unknown').replace('_', ' ').title()}\n"
                if approval.get("conditions"):
                    section += f"- **Remediation Conditions:**\n"
                    for condition in approval.get("conditions", [])[:3]:  # Top 3
                        section += f"  - {condition}\n"

            # Check for mitigation actions
            vendor_actions = [a for a in mitigation_actions if a.get("vendor_id") == vendor_id]
            if vendor_actions:
                section += f"- **Mitigation Actions:** {len(vendor_actions)} action(s) created\n"

            section += f"- **Recommended Action:** {assessment.get('recommended_action', 'N/A')}\n\n"
    else:
        section += "*No high-risk vendors identified.*\n\n"

    section += "### Medium-Risk Vendors\n\n"
    medium_risk_vendors = [a for a in sorted_assessments if a.get("risk_level") == "medium"]
    if medium_risk_vendors:
        section += "| Vendor | Risk Score | Primary Domains | Status |\n"
        section += "|--------|------------|-----------------|--------|\n"
        for assessment in medium_risk_vendors[:10]:  # Top 10
            vendor_id = assessment.get("vendor_id")
            vendor_data = vendor_lookup.get(vendor_id, {})
            vendor_name = vendor_data.get("vendor_name", vendor_id)
            primary_domains = ", ".join(assessment.get("primary_risk_domains", [])[:2])
            section += f"| {vendor_name} | {assessment.get('overall_risk_score', 0.0):.1f} | {primary_domains} | {assessment.get('risk_level', 'unknown')} |\n"
        section += "\n"
    else:
        section += "*No medium-risk vendors identified.*\n\n"

    section += "### Low-Risk Vendors\n\n"
    low_risk_vendors = [a for a in sorted_assessments if a.get("risk_level") == "low"]
    section += f"**Total:** {len(low_risk_vendors)} vendor(s) with low risk scores. Continue standard monitoring.\n\n"

    return section


def generate_kpi_section(
    kpi_metrics: Dict[str, Any],
    orchestrator_metrics: Dict[str, Any]
) -> str:
    """
    Generate KPI metrics section.

    Args:
        kpi_metrics: KPI metrics dictionary
        orchestrator_metrics: Orchestrator metrics

    Returns:
        KPI section markdown
    """
    operational = kpi_metrics.get("operational", {})
    effectiveness = kpi_metrics.get("effectiveness", {})
    business = kpi_metrics.get("business", {})

    section = "## Performance Metrics\n\n"

    section += "### Operational KPIs (Agent Health)\n\n"
    section += "| Metric | Value |\n"
    section += "|--------|-------|\n"
    section += f"| Assessment Completion Rate | {operational.get('completion_rate', 0.0):.1%} |\n"
    section += f"| Avg Assessment Latency | {operational.get('avg_assessment_latency_minutes', 0.0):.1f} minutes |\n"
    section += f"| Human Escalations | {operational.get('human_escalations', 0)} |\n"
    section += f"| Escalation Rate | {operational.get('human_escalation_rate', 0.0):.1%} |\n"
    section += f"| Policy Validation Failures | {operational.get('policy_validation_failures', 0)} |\n"
    section += f"| Audit Log Completeness | {operational.get('audit_log_completeness', 0.0):.1%} |\n\n"

    section += "### Effectiveness KPIs (Risk Management Impact)\n\n"
    section += "| Metric | Value |\n"
    section += "|--------|-------|\n"
    section += f"| Time to Identify Risk | {effectiveness.get('time_to_identify_risk_hours', 0.0):.1f} hours |\n"
    section += f"| Manual Hours Saved | {effectiveness.get('manual_hours_saved', 0.0):.1f} hours |\n"
    section += f"| Manual Review Reduction | {effectiveness.get('manual_review_reduction_percent', 0.0):.1f}% |\n"
    section += f"| Risk Score Consistency | {effectiveness.get('risk_score_consistency', 0.0):.3f} |\n"
    section += f"| Categorization Accuracy | {effectiveness.get('categorization_accuracy', 0.0):.1%} |\n"
    section += f"| Mitigation Speed | {effectiveness.get('mitigation_speed_hours', 0.0):.1f} hours |\n\n"

    section += "### Business KPIs (ROI & Value)\n\n"
    section += "| Metric | Value |\n"
    section += "|--------|-------|\n"
    section += f"| Cost per Assessment | ${business.get('cost_per_assessment_usd', 0.0):.2f} |\n"
    section += f"| Baseline Cost per Assessment | ${business.get('baseline_cost_per_assessment_usd', 0.0):.2f} |\n"
    section += f"| Total Run Cost | ${business.get('total_run_cost_usd', 0.0):,.2f} |\n"
    section += f"| Cost Savings | ${business.get('cost_savings_usd', 0.0):,.2f} |\n"
    section += f"| Estimated Cost Avoidance | ${business.get('estimated_cost_avoidance_usd', 0.0):,.0f} |\n"
    section += f"| Net Value | ${business.get('net_value_usd', 0.0):,.2f} |\n"
    section += f"| ROI | {business.get('roi_percentage', 0.0):.1f}% |\n"
    section += f"| ROI Status | {business.get('roi_status', 'unknown').upper()} |\n\n"

    # Cost breakdown
    section += "#### Cost Breakdown\n\n"
    section += f"- LLM Usage: ${business.get('llm_cost_usd', 0.0):.2f}\n"
    section += f"- API Calls: ${business.get('api_cost_usd', 0.0):.2f}\n"
    section += f"- Human Review: ${business.get('human_review_cost_usd', 0.0):.2f}\n"
    section += f"- Infrastructure: ${business.get('infrastructure_cost_usd', 0.0):.2f}\n\n"

    return section


def generate_mitigation_actions_section(
    mitigation_actions: List[Dict[str, Any]],
    vendor_lookup: Dict[str, Dict[str, Any]]
) -> str:
    """
    Generate mitigation actions section.

    Args:
        mitigation_actions: List of mitigation actions
        vendor_lookup: Fast vendor lookup

    Returns:
        Mitigation actions section markdown
    """
    if not mitigation_actions:
        return "## Mitigation Actions\n\n*No mitigation actions created.*\n\n"

    section = "## Mitigation Actions\n\n"
    section += f"**Total Actions:** {len(mitigation_actions)}\n\n"

    # Group by vendor
    actions_by_vendor = {}
    for action in mitigation_actions:
        vendor_id = action.get("vendor_id")
        if vendor_id not in actions_by_vendor:
            actions_by_vendor[vendor_id] = []
        actions_by_vendor[vendor_id].append(action)

    for vendor_id, actions in actions_by_vendor.items():
        vendor_data = vendor_lookup.get(vendor_id, {})
        vendor_name = vendor_data.get("vendor_name", vendor_id)

        section += f"### {vendor_name} ({vendor_id})\n\n"

        for action in actions:
            section += f"- **{action.get('action_type', 'Action').replace('_', ' ').title()}**\n"
            section += f"  - Status: {action.get('status', 'unknown').replace('_', ' ').title()}\n"
            section += f"  - Target Date: {action.get('target_completion_date', 'N/A')}\n"
            section += f"  - Assigned To: {action.get('assigned_to', 'N/A')}\n"
            if action.get("progress_notes"):
                section += f"  - Notes: {action.get('progress_notes')}\n"
            section += "\n"

    return section


def generate_risk_assessment_report(
    state: Dict[str, Any]
) -> str:
    """
    Generate comprehensive risk assessment report.

    Args:
        state: Complete orchestrator state

    Returns:
        Complete markdown report
    """
    risk_assessments = state.get("risk_assessments", [])
    vendor_lookup = state.get("vendor_lookup", {})
    approval_history = state.get("approval_history", [])
    mitigation_actions = state.get("mitigation_actions", [])
    kpi_metrics = state.get("kpi_metrics", {})
    orchestrator_metrics = state.get("orchestrator_metrics", {})
    run_date = state.get("run_date", datetime.now().strftime("%Y-%m-%d"))

    # Generate report sections
    executive_summary = generate_executive_summary(
        orchestrator_metrics,
        risk_assessments,
        kpi_metrics
    )

    vendor_overview = generate_vendor_risk_overview(
        risk_assessments,
        vendor_lookup,
        approval_history,
        mitigation_actions
    )

    kpi_section = generate_kpi_section(
        kpi_metrics,
        orchestrator_metrics
    )

    mitigation_section = generate_mitigation_actions_section(
        mitigation_actions,
        vendor_lookup
    )

    # Combine into full report
    report = f"""# Third-Party Risk Assessment Report

**Generated:** {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
**Assessment Date:** {run_date}

---

{executive_summary}

---

{vendor_overview}

---

{kpi_section}

---

{mitigation_section}

---

## Appendix

### Report Metadata

- **Run ID:** {orchestrator_metrics.get('run_id', 'N/A')}
- **Vendors Evaluated:** {orchestrator_metrics.get('vendors_evaluated', 0)}
- **Assessments Completed:** {orchestrator_metrics.get('assessments_completed', 0)}
- **External Signals Processed:** {orchestrator_metrics.get('external_signals_processed', 0)}
- **Mitigation Actions Created:** {orchestrator_metrics.get('mitigation_actions_created', 0)}

---

*This report was generated by the Third-Party Risk Orchestrator Agent.*
"""

    return report


def save_risk_assessment_report(
    report_content: str,
    run_id: str,
    reports_dir: str = "output/third_party_risk_orchestrator"
) -> str:
    """
    Save risk assessment report to file.

    Uses toolshed save_report utility.

    Args:
        report_content: Report markdown content
        run_id: Run identifier
        reports_dir: Directory to save reports

    Returns:
        Path to saved report file
    """
    return save_report(
        report_content=report_content,
        report_id=run_id,
        reports_dir=reports_dir,
        prefix="risk_assessment"
    )
````
