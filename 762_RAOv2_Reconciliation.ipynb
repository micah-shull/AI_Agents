````python
"""
Reconcile invoices vs contracts, usage vs contracts, and discount vs approvals.
Returns lists of findings for the detection layer to turn into issues with revenue impact.
"""
from typing import Any, Dict, List, Optional, Tuple


def _expected_invoice_total(quantity: float, unit_price: float, discount_percent: float, tax_rate: float) -> float:
    """Compute expected total: quantity * unit_price * (1 - discount/100) * (1 + tax/100)."""
    subtotal = quantity * unit_price * (1.0 - discount_percent / 100.0)
    return round(subtotal * (1.0 + tax_rate / 100.0), 2)


def reconcile_invoices_contracts(
    invoices: List[Dict[str, Any]],
    contract_by_id: Dict[str, Dict[str, Any]],
    approvals_by_customer: Dict[str, List[Dict[str, Any]]],
) -> List[Dict[str, Any]]:
    """
    For each invoice: check contract exists, unit_price, discount vs contract/approval, total calculation, billing_cycle.
    Returns list of findings: { invoice_id, customer_id, finding_type, detail, expected_value, actual_value, revenue_impact }.
    """
    findings: List[Dict[str, Any]] = []
    for inv in invoices:
        invoice_id = inv.get("invoice_id", "")
        customer_id = inv.get("customer_id", "")
        contract_id = inv.get("contract_id", "")
        quantity = float(inv.get("quantity") or 0)
        unit_price = float(inv.get("unit_price") or 0)
        discount_percent = float(inv.get("discount_percent") or 0)
        tax_rate = float(inv.get("tax_rate") or 0)
        total_amount = float(inv.get("total_amount") or 0)
        billing_cycle = (inv.get("billing_cycle") or "").strip().lower()
        product_name = inv.get("product_name", "")

        contract = contract_by_id.get(contract_id) if contract_id else None

        # Missing contract
        if not contract and contract_id:
            findings.append({
                "finding_type": "missing_contract",
                "invoice_id": invoice_id,
                "customer_id": customer_id,
                "contract_id": contract_id,
                "detail": f"Invoice references contract {contract_id} which does not exist.",
                "revenue_impact": 0.0,
                "product_name": product_name,
            })
            continue

        if not contract:
            continue

        max_discount = float(contract.get("max_discount_percent") or 0)
        contract_unit_price = float(contract.get("unit_price") or 0)
        contract_cycle = (contract.get("billing_cycle") or "").strip().lower()

        # Unit price mismatch
        if unit_price != contract_unit_price:
            expected_total = _expected_invoice_total(quantity, contract_unit_price, discount_percent, tax_rate)
            revenue_impact = abs(expected_total - total_amount)
            findings.append({
                "finding_type": "pricing_violation",
                "invoice_id": invoice_id,
                "customer_id": customer_id,
                "contract_id": contract_id,
                "detail": f"Unit price ${unit_price} does not match contract ${contract_unit_price}.",
                "expected_value": contract_unit_price,
                "actual_value": unit_price,
                "revenue_impact": revenue_impact,
                "product_name": product_name,
            })

        # Discount over max without approval
        if discount_percent > max_discount:
            approvals = approvals_by_customer.get(customer_id, [])
            max_approved = max((a.get("approved_discount_percent") or 0) for a in approvals) if approvals else 0
            if discount_percent > max_approved:
                expected_at_contract_price = quantity * contract_unit_price * (1.0 - max_discount / 100.0) * (1.0 + tax_rate / 100.0)
                actual_total = total_amount
                revenue_impact = max(0.0, expected_at_contract_price - actual_total)
                findings.append({
                    "finding_type": "unauthorized_discount",
                    "invoice_id": invoice_id,
                    "customer_id": customer_id,
                    "contract_id": contract_id,
                    "detail": f"Discount {discount_percent}% exceeds contract max {max_discount}% and approved {max_approved}%.",
                    "revenue_impact": round(revenue_impact, 2),
                    "product_name": product_name,
                })

        # Calculation error (wrong total)
        expected_total = _expected_invoice_total(quantity, unit_price, discount_percent, tax_rate)
        if abs(expected_total - total_amount) > 0.02:  # allow small rounding
            findings.append({
                "finding_type": "calculation_error",
                "invoice_id": invoice_id,
                "customer_id": customer_id,
                "contract_id": contract_id,
                "detail": f"Total ${total_amount} does not match computed ${expected_total}.",
                "expected_value": expected_total,
                "actual_value": total_amount,
                "revenue_impact": abs(expected_total - total_amount),
                "product_name": product_name,
            })

        # Billing cycle mismatch
        if billing_cycle and contract_cycle and billing_cycle != contract_cycle:
            findings.append({
                "finding_type": "billing_cycle_mismatch",
                "invoice_id": invoice_id,
                "customer_id": customer_id,
                "contract_id": contract_id,
                "detail": f"Invoice billing_cycle '{billing_cycle}' does not match contract '{contract_cycle}'.",
                "revenue_impact": 0.0,
                "product_name": product_name,
            })

    return findings


def reconcile_usage(
    usage_records: List[Dict[str, Any]],
    contracts_by_customer: Dict[str, List[Dict[str, Any]]],
) -> List[Dict[str, Any]]:
    """
    Find overage usage (usage_quantity > contracted_quantity) as unbilled/upsell opportunities.
    Revenue impact is estimated from contract unit_price * overage (simplified; no tier logic here).
    """
    findings: List[Dict[str, Any]] = []
    for rec in usage_records:
        if not rec.get("overage_flag"):
            continue
        customer_id = rec.get("customer_id", "")
        product_name = rec.get("product_name", "")
        usage_qty = int(rec.get("usage_quantity") or 0)
        contracted_qty = int(rec.get("contracted_quantity") or 0)
        overage = usage_qty - contracted_qty
        if overage <= 0:
            continue
        contracts = contracts_by_customer.get(customer_id, [])
        # Use first matching contract by product name for unit price
        unit_price = 0.0
        for c in contracts:
            if (c.get("product_name") or "").strip() == (product_name or "").strip():
                unit_price = float(c.get("unit_price") or 0)
                break
        # Rough monthly revenue for overage (assume same price per unit)
        revenue_impact = overage * unit_price if unit_price else 0.0
        findings.append({
            "finding_type": "unbilled_overage",
            "usage_id": rec.get("usage_id", ""),
            "customer_id": customer_id,
            "product_name": product_name,
            "usage_quantity": usage_qty,
            "contracted_quantity": contracted_qty,
            "overage": overage,
            "detail": f"Usage {usage_qty} exceeds contract {contracted_qty} by {overage} units (unbilled).",
            "revenue_impact": round(revenue_impact, 2),
            "usage_period": rec.get("usage_period", ""),
        })
    return findings


def reconcile_discounts(
    invoices: List[Dict[str, Any]],
    contract_by_id: Dict[str, Dict[str, Any]],
    approvals_by_customer: Dict[str, List[Dict[str, Any]]],
) -> List[Dict[str, Any]]:
    """
    Re-check discount vs approvals only (may duplicate reconciliation; kept for clarity).
    Returns findings where discount exceeds approval.
    """
    return []  # Already covered in reconcile_invoices_contracts
````
