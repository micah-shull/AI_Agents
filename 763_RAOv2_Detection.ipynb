````python
"""
Build unified issues list from reconciliation findings and recovery log (for context).
Each issue: issue_type, customer_id, revenue_impact, severity, recommended_owner, source_id, detail, product_name.
"""
from typing import Any, Dict, List


def _severity(revenue_impact: float) -> str:
    if revenue_impact >= 20000:
        return "high"
    if revenue_impact >= 5000:
        return "medium"
    return "low"


def _owner(issue_type: str) -> str:
    if issue_type in ("calculation_error", "pricing_violation", "underbilling", "unbilled_overage"):
        return "Finance"
    if issue_type in ("unauthorized_discount",):
        return "Sales Ops"
    if issue_type in ("upsell_opportunity",):
        return "Account Manager"
    return "Finance"


def build_issues_list(
    invoice_findings: List[Dict[str, Any]],
    usage_findings: List[Dict[str, Any]],
) -> List[Dict[str, Any]]:
    """
    Convert reconciliation findings into a unified issues list with severity and owner.
    """
    issues: List[Dict[str, Any]] = []
    for f in invoice_findings:
        issue_type = f.get("finding_type", "unknown")
        revenue_impact = float(f.get("revenue_impact") or 0)
        issues.append({
            "issue_type": issue_type,
            "customer_id": f.get("customer_id", ""),
            "revenue_impact": revenue_impact,
            "severity": _severity(revenue_impact),
            "recommended_owner": _owner(issue_type),
            "source_id": f.get("invoice_id", ""),
            "detail": f.get("detail", ""),
            "product_name": f.get("product_name", ""),
            "contract_id": f.get("contract_id", ""),
        })
    for f in usage_findings:
        revenue_impact = float(f.get("revenue_impact") or 0)
        issues.append({
            "issue_type": "unbilled_overage",
            "customer_id": f.get("customer_id", ""),
            "revenue_impact": revenue_impact,
            "severity": _severity(revenue_impact),
            "recommended_owner": "Account Manager",
            "source_id": f.get("usage_id", ""),
            "detail": f.get("detail", ""),
            "product_name": f.get("product_name", ""),
            "usage_period": f.get("usage_period", ""),
        })
    return issues
````
