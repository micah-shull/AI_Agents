````python
"""
Compliance Sentinel v2 orchestrator nodes.

Nodes are thin: orchestrate utilities; use closure factories for config-dependent nodes.
"""

from pathlib import Path
from typing import Any, Dict, List

from toolshed.alerting import create_alert, compute_threshold_triggers

from agents.compliance_sentinel_v2.orchestrator.utilities.data_loading import (
    load_all_compliance_sentinel_data,
)
from agents.compliance_sentinel_v2.orchestrator.utilities.rollups import (
    build_portfolio_rollup,
)
from agents.compliance_sentinel_v2.orchestrator.utilities.report_helpers import (
    build_compliance_report,
)


def goal_node(state: Dict[str, Any]) -> Dict[str, Any]:
    """Define the goal for the compliance run."""
    goal = {
        "objective": "Enterprise AI governance and compliance command center",
        "focus_areas": [
            "compliance_findings",
            "financial_exposure",
            "escalations",
            "remediation",
            "human_review",
            "portfolio_risk",
        ],
    }
    return {"goal": goal, "errors": state.get("errors", [])}


def planning_node(state: Dict[str, Any]) -> Dict[str, Any]:
    """Create execution plan."""
    plan = [
        {"step": 1, "name": "data_loading", "description": "Load AI events, systems, findings, exposure, escalations, remediation, reviews, snapshots", "dependencies": []},
        {"step": 2, "name": "portfolio", "description": "Build portfolio rollup and executive triggers", "dependencies": ["data_loading"]},
        {"step": 3, "name": "escalation", "description": "Create alerts when triggers fire", "dependencies": ["portfolio"]},
        {"step": 4, "name": "report", "description": "Generate board-ready compliance report", "dependencies": ["portfolio"]},
    ]
    return {"plan": plan, "errors": state.get("errors", [])}


def make_data_loading_node(config: Any, project_root: str):
    """Closure: load all v2 data and build lookups."""
    def node(state: Dict[str, Any]) -> Dict[str, Any]:
        errors = state.get("errors", [])
        data_dir = state.get("data_dir") or config.data_dir
        try:
            data = load_all_compliance_sentinel_data(
                data_dir,
                project_root,
                ai_output_events_file=config.ai_output_events_file,
                ai_system_registry_file=config.ai_system_registry_file,
                compliance_findings_file=config.compliance_findings_file,
                remediation_actions_file=config.remediation_actions_file,
                escalation_events_file=config.escalation_events_file,
                financial_exposure_models_file=config.financial_exposure_models_file,
                human_review_actions_file=config.human_review_actions_file,
                portfolio_snapshots_file=config.portfolio_snapshots_file,
            )
            systems = data.get("ai_system_registry") or []
            findings = data.get("compliance_findings") or []
            exposure_models = data.get("financial_exposure_models") or []
            systems_lookup = {s["system_id"]: s for s in systems if s.get("system_id")}
            findings_lookup = {f["finding_id"]: f for f in findings if f.get("finding_id")}
            exposure_lookup = {e["finding_id"]: e for e in exposure_models if e.get("finding_id")}
            return {
                "ai_output_events": data.get("ai_output_events"),
                "ai_system_registry": systems,
                "compliance_findings": findings,
                "remediation_actions": data.get("remediation_actions"),
                "escalation_events": data.get("escalation_events"),
                "financial_exposure_models": exposure_models,
                "human_review_actions": data.get("human_review_actions"),
                "portfolio_snapshots": data.get("portfolio_snapshots"),
                "systems_lookup": systems_lookup,
                "findings_lookup": findings_lookup,
                "exposure_lookup": exposure_lookup,
                "data_snapshot_loaded_at": data.get("data_snapshot_loaded_at"),
                "validation_warnings": state.get("validation_warnings") or [],
                "errors": errors,
            }
        except Exception as e:
            return {"errors": errors + [f"data_loading: {str(e)}"]}
    return node


def make_portfolio_node(config: Any):
    """Closure: build rollup and compute executive triggers from config thresholds."""
    def node(state: Dict[str, Any]) -> Dict[str, Any]:
        errors = state.get("errors", [])
        findings = state.get("compliance_findings") or []
        escalations = state.get("escalation_events") or []
        exposure_models = state.get("financial_exposure_models") or []
        snapshots = state.get("portfolio_snapshots") or []
        systems_lookup = state.get("systems_lookup") or {}
        try:
            rollup = build_portfolio_rollup(
                findings,
                escalations,
                exposure_models,
                snapshots,
                systems_lookup,
            )
            threshold_config = {
                "open_cases": {"key": "open_cases", "threshold": config.open_cases_critical, "label": "Open compliance cases"},
                "total_exposure": {
                    "key": "total_estimated_exposure_usd",
                    "threshold": config.total_exposure_usd_critical,
                    "label": "Total estimated exposure",
                    "format_value": lambda v: f"${float(v):,.0f}",
                    "format_threshold": lambda t: f"${float(t):,.0f}",
                },
                "portfolio_risk_score": {"key": "portfolio_risk_score", "threshold": config.portfolio_risk_score_critical, "label": "Portfolio risk score"},
                "critical_findings": {"key": "critical_cases", "threshold": config.critical_findings_critical, "label": "Critical findings"},
                "executive_escalations": {"key": "executive_escalations", "threshold": config.executive_escalations_critical, "label": "Executive escalations"},
            }
            triggers = compute_threshold_triggers(rollup, threshold_config)
            return {
                "portfolio_rollup": rollup,
                "executive_triggers": triggers,
                "errors": errors,
            }
        except Exception as e:
            return {"errors": errors + [f"portfolio: {str(e)}"]}
    return node


def make_escalation_node(config: Any):
    """Closure: create alerts when executive triggers fired."""
    def node(state: Dict[str, Any]) -> Dict[str, Any]:
        errors = state.get("errors", [])
        triggers = state.get("executive_triggers") or []
        alerts = []
        for t in triggers:
            alert = create_alert(
                alert_type="compliance_trigger",
                severity="critical",
                title="Compliance Sentinel: threshold exceeded",
                description=t.get("message", ""),
                metadata={"trigger_type": t.get("trigger_type"), "value": t.get("value"), "threshold": t.get("threshold")},
                alert_id_prefix="cs2",
            )
            alerts.append(alert)
        return {"alerts": alerts, "errors": errors}
    return node


def make_report_node(config: Any, project_root: str):
    """Closure: generate and save compliance report."""
    def node(state: Dict[str, Any]) -> Dict[str, Any]:
        import os
        errors = state.get("errors", [])
        goal = state.get("goal") or {}
        rollup = state.get("portfolio_rollup") or {}
        triggers = state.get("executive_triggers") or []
        findings = state.get("compliance_findings") or []
        open_findings = [f for f in findings if (f.get("status") or "").lower() == "open"]
        escalations = state.get("escalation_events") or []
        remediation = state.get("remediation_actions") or []
        systems_lookup = state.get("systems_lookup") or {}
        data_loaded_at = state.get("data_snapshot_loaded_at")
        validation_warnings = state.get("validation_warnings") or []
        try:
            report = build_compliance_report(
                goal,
                rollup,
                triggers,
                open_findings,
                escalations,
                remediation,
                systems_lookup,
                data_snapshot_loaded_at=data_loaded_at,
                validation_warnings=validation_warnings or None,
                exposure_critical_usd=config.total_exposure_usd_critical,
                exposure_elevated_usd=1_000_000,
            )
            reports_dir = state.get("reports_dir") or config.reports_dir
            base = Path(project_root) / reports_dir
            base.mkdir(parents=True, exist_ok=True)
            from datetime import datetime, timezone
            ts = datetime.now(timezone.utc).strftime("%Y%m%d_%H%M%S")
            path = base / f"compliance_sentinel_v2_report_{ts}.md"
            path.write_text(report, encoding="utf-8")
            return {
                "compliance_report": report,
                "report_file_path": str(path),
                "errors": errors,
            }
        except Exception as e:
            return {"errors": errors + [f"report: {str(e)}"]}
    return node
````
