````python
def report_generation_node(
    state: EvalAsServiceOrchestratorState,
    config: EvalAsServiceOrchestratorConfig
) -> Dict[str, Any]:
    """Generate comprehensive evaluation report with CEO-friendly business metrics."""
    print("  [Report Generation Node] Starting...")
    from toolshed.reporting import generate_mission_report, save_report

    summary = state.get('evaluation_summary', {})
    agent_summaries = state.get('agent_performance_summary', [])
    scores = state.get('evaluation_scores', [])
    evaluations = state.get('executed_evaluations', [])

    # Build report sections
    report_sections = []

    # Executive Summary - Business Value First!
    report_sections.append("## Executive Summary\n\n")
    report_sections.append("### üí∞ Business Impact\n\n")
    report_sections.append(f"- **Total Revenue Impact:** ${summary.get('total_revenue_impact', 0.0):,.2f}\n")
    report_sections.append(f"- **Total Cost:** ${summary.get('total_cost', 0.0):,.2f}\n")
    report_sections.append(f"- **Net ROI:** ${summary.get('total_net_roi', 0.0):,.2f} ({summary.get('overall_roi_percent', 0.0):.1f}% return)\n")
    report_sections.append(f"- **Cost per Evaluation:** ${summary.get('cost_per_evaluation', 0.0):.2f}\n")
    report_sections.append(f"- **Agents with Positive ROI:** {summary.get('agents_with_positive_roi', 0)}/{summary.get('agents_evaluated', 0)}\n")
    report_sections.append(f"- **Agents Needing Optimization:** {summary.get('agents_needing_optimization', 0)}\n\n")

    report_sections.append("### üìä Performance Metrics\n\n")
    report_sections.append(f"- **Total Scenarios Evaluated:** {summary.get('total_scenarios', 0)}\n")
    report_sections.append(f"- **Total Evaluations:** {summary.get('total_evaluations', 0)}\n")
    report_sections.append(f"- **Overall Pass Rate:** {summary.get('overall_pass_rate', 0.0):.1%}\n")
    report_sections.append(f"- **Average Score:** {summary.get('average_score', 0.0):.2f}\n")
    report_sections.append(f"- **Healthy Agents:** {summary.get('healthy_agents', 0)}\n")
    report_sections.append(f"- **Degraded Agents:** {summary.get('degraded_agents', 0)}\n")
    report_sections.append(f"- **Critical Agents:** {summary.get('critical_agents', 0)}\n\n")

    # Key Takeaways
    report_sections.append("### üéØ Key Takeaways\n\n")
    if summary.get('overall_roi_percent', 0) > 100:
        report_sections.append(f"‚úÖ **Excellent ROI:** {summary.get('overall_roi_percent', 0.0):.1f}% return on investment\n")
    elif summary.get('overall_roi_percent', 0) > 0:
        report_sections.append(f"‚úÖ **Positive ROI:** {summary.get('overall_roi_percent', 0.0):.1f}% return on investment\n")
    else:
        report_sections.append(f"‚ö†Ô∏è **Negative ROI:** {summary.get('overall_roi_percent', 0.0):.1f}% - optimization needed\n")

    if summary.get('overall_pass_rate', 0) >= 0.90:
        report_sections.append(f"‚úÖ **High Quality:** {summary.get('overall_pass_rate', 0.0):.1%} pass rate indicates reliable agents\n")
    elif summary.get('overall_pass_rate', 0) >= 0.70:
        report_sections.append(f"‚ö†Ô∏è **Moderate Quality:** {summary.get('overall_pass_rate', 0.0):.1%} pass rate - some improvements needed\n")
    else:
        report_sections.append(f"‚ùå **Low Quality:** {summary.get('overall_pass_rate', 0.0):.1%} pass rate - significant improvements required\n")

    if summary.get('agents_needing_optimization', 0) > 0:
        report_sections.append(f"‚ö†Ô∏è **Optimization Opportunity:** {summary.get('agents_needing_optimization', 0)} agent(s) have ROI ratio < 2.0x\n")
    report_sections.append("\n")

    # Recommended Actions
    report_sections.append("### üéØ Recommended Actions\n\n")
    recommendations = []

    # ROI-based recommendations
    if summary.get('overall_roi_percent', 0) < 0:
        recommendations.append("**URGENT:** Overall ROI is negative - review agent costs and revenue impact immediately")
    elif summary.get('overall_roi_percent', 0) < 50:
        recommendations.append("**PRIORITY:** ROI is below 50% - optimize costs or increase revenue impact")

    # Agent-specific recommendations
    for agent_summary in agent_summaries:
        agent_id = agent_summary.get('agent_id', 'unknown')
        roi_percent = agent_summary.get('roi_percent', 0)
        roi_ratio = agent_summary.get('roi_ratio', 0)

        if roi_percent < 0:
            recommendations.append(f"**{agent_id}:** Negative ROI ({roi_percent:.1f}%) - consider retiring or major optimization")
        elif roi_ratio < 2.0 and roi_ratio != float('inf'):
            recommendations.append(f"**{agent_id}:** Low ROI ratio ({roi_ratio:.2f}x) - optimize costs or improve value")
        elif roi_percent > 200:
            recommendations.append(f"**{agent_id}:** Exceptional ROI ({roi_percent:.1f}%) - consider scaling to more use cases")

    # Quality-based recommendations
    if summary.get('overall_pass_rate', 0) < 0.70:
        recommendations.append("**QUALITY:** Pass rate below 70% - improve agent reliability before scaling")
    elif summary.get('critical_agents', 0) > 0:
        recommendations.append(f"**QUALITY:** {summary.get('critical_agents', 0)} agent(s) in critical status - immediate attention required")

    if not recommendations:
        recommendations.append("‚úÖ **All systems performing well** - continue monitoring and consider scaling successful patterns")

    for i, rec in enumerate(recommendations, 1):
        report_sections.append(f"{i}. {rec}\n")
    report_sections.append("\n")

    # ROI & Business Value Section
    report_sections.append("## üí∞ ROI & Business Value Analysis\n\n")
    report_sections.append("### Overall Business Impact\n\n")
    report_sections.append(f"- **Total Revenue Impact:** ${summary.get('total_revenue_impact', 0.0):,.2f}\n")
    report_sections.append(f"- **Total Cost:** ${summary.get('total_cost', 0.0):,.2f}\n")
    report_sections.append(f"- **Net ROI:** ${summary.get('total_net_roi', 0.0):,.2f}\n")
    report_sections.append(f"- **ROI Percentage:** {summary.get('overall_roi_percent', 0.0):.1f}%\n")
    report_sections.append(f"- **Cost per Evaluation:** ${summary.get('cost_per_evaluation', 0.0):.2f}\n\n")

    report_sections.append("### Agent-Level ROI\n\n")
    report_sections.append("| Agent | Cost | Revenue Impact | Net ROI | ROI % | ROI Ratio | Status |\n")
    report_sections.append("|-------|------|----------------|---------|-------|-----------|--------|\n")
    for agent_summary in agent_summaries:
        agent_id = agent_summary.get('agent_id', 'unknown')
        cost = agent_summary.get('total_cost', 0.0)
        revenue = agent_summary.get('revenue_impact', 0.0)
        net_roi = agent_summary.get('net_roi', 0.0)
        roi_percent = agent_summary.get('roi_percent', 0.0)
        roi_ratio = agent_summary.get('roi_ratio', 0.0)
        roi_category = agent_summary.get('roi_category', 'unknown')

        # Status indicator
        if roi_percent > 100:
            status = "‚úÖ Excellent"
        elif roi_percent > 0:
            status = "‚úÖ Positive"
        else:
            status = "‚ö†Ô∏è Negative"

        roi_ratio_str = f"{roi_ratio:.2f}x" if roi_ratio != float('inf') else "‚àû"
        roi_percent_str = f"{roi_percent:.1f}%" if roi_percent != float('inf') else "‚àû"

        report_sections.append(f"| {agent_id} | ${cost:.2f} | ${revenue:.2f} | ${net_roi:.2f} | {roi_percent_str} | {roi_ratio_str} | {status} |\n")
    report_sections.append("\n")

    # Agent Performance Details
    report_sections.append("## üìä Agent Performance Details\n\n")
    for agent_summary in agent_summaries:
        agent_id = agent_summary.get('agent_id', 'unknown')
        report_sections.append(f"### {agent_id}\n\n")

        # Performance metrics
        report_sections.append("**Performance:**\n")
        report_sections.append(f"- **Status:** {agent_summary.get('health_status', 'unknown')}\n")
        report_sections.append(f"- **Total Evaluations:** {agent_summary.get('total_evaluations', 0)}\n")
        report_sections.append(f"- **Passed:** {agent_summary.get('passed_count', 0)}\n")
        report_sections.append(f"- **Failed:** {agent_summary.get('failed_count', 0)}\n")
        report_sections.append(f"- **Average Score:** {agent_summary.get('average_score', 0.0):.2f}\n")
        report_sections.append(f"- **Average Response Time:** {agent_summary.get('average_response_time', 0.0):.2f}s\n")

        # Business metrics
        if 'total_cost' in agent_summary:
            report_sections.append("\n**Business Value:**\n")
            report_sections.append(f"- **Total Cost:** ${agent_summary.get('total_cost', 0.0):.2f}\n")
            report_sections.append(f"- **Revenue Impact:** ${agent_summary.get('revenue_impact', 0.0):.2f}\n")
            report_sections.append(f"- **Net ROI:** ${agent_summary.get('net_roi', 0.0):.2f}\n")
            report_sections.append(f"- **ROI Percentage:** {agent_summary.get('roi_percent', 0.0):.1f}%\n")
            report_sections.append(f"- **ROI Category:** {agent_summary.get('roi_category', 'unknown')}\n")
            report_sections.append(f"- **Cost per Evaluation:** ${agent_summary.get('cost_per_evaluation', 0.0):.2f}\n")

            # Statistical significance (if we had historical data)
            # For MVP, we'll add a placeholder
            report_sections.append("\n**Statistical Assessment:**\n")
            report_sections.append("- *Historical comparison data not yet available*\n")
            report_sections.append("- *Trend analysis will be available after multiple evaluation runs*\n")

        report_sections.append("\n")

    # Evaluation Results
    report_sections.append("## Evaluation Results\n\n")
    report_sections.append("| Scenario | Agent | Score | Passed | Issues |\n")
    report_sections.append("|----------|-------|-------|--------|--------|\n")

    for score in scores[:20]:  # Limit to first 20 for readability
        scenario_id = score.get('scenario_id', 'N/A')
        agent_id = score.get('target_agent_id', 'N/A')
        overall_score = score.get('overall_score', 0.0)
        passed = "‚úì" if score.get('passed', False) else "‚úó"
        issues = ", ".join(score.get('issues', []))[:50]  # Truncate long issues
        report_sections.append(f"| {scenario_id} | {agent_id} | {overall_score:.2f} | {passed} | {issues} |\n")

    if len(scores) > 20:
        report_sections.append(f"\n*Showing first 20 of {len(scores)} evaluations*\n\n")

    # Combine report
    report = "# Evaluation-as-a-Service Report\n\n"
    report += f"**Generated:** {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n\n"
    report += "".join(report_sections)

    # Save report if enabled
    report_file_path = None
    if config.enable_reporting:
        try:
            report_file_path = save_report(
                report_content=report,
                report_id="evaluation_report",
                reports_dir=config.reports_dir,
                prefix="eval_report"
            )
        except Exception as e:
            errors = state.get('errors', [])
            if not isinstance(errors, list):
                errors = []
            errors.append(f"Report saving error: {str(e)}")
            state['errors'] = errors

    print(f"  [Report Generation Node] Report generated ({len(report)} chars)")
    if report_file_path:
        print(f"  [Report Generation Node] Saved to: {report_file_path}")

    return {
        "evaluation_report": report,
        "report_file_path": report_file_path
    }
````
