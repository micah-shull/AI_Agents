# Test Pattern Detection with Intentional Failures

```python
"""
Test Pattern Detection with Intentional Failures
Tests orchestrator insights when agents fail on specific scenarios
"""

import sys
import logging
from pathlib import Path

# Add project root to path
project_root = Path(__file__).parent.parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from config import EaaSState
from nodes import (
    goal_node,
    planning_node,
    data_ingestion_node,
    scenario_generation_node,
    evaluation_execution_node,
    scoring_node,
    report_node
)

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')
logger = logging.getLogger(__name__)


def test_pattern_detection():
    """Test orchestrator pattern detection with intentional failures"""

    # Initialize state with failure test data
    state: EaaSState = {
        "target_agents": [
            {
                "id": "agent_001",
                "name": "Sentiment Classifier",
                "type": "classification"
            },
            {
                "id": "agent_002",
                "name": "Safety Checker",
                "type": "safety"
            }
        ],
        "evaluation_config": {
            "criteria": ["accuracy", "safety", "latency"],
            "thresholds": {
                "accuracy": 0.8,
                "safety": 0.95,
                "latency_ms": 2000
            }
        },
        "test_data_paths": [
            "data/classification_cases_with_failures.json",  # Has intentional failures
            "data/safety_cases_with_failures.json"          # Has intentional failures
        ],
        "errors": []
    }

    print("\n" + "="*60)
    print("ğŸ§ª Pattern Detection Test - Intentional Failures")
    print("="*60 + "\n")
    print("This test uses scenarios designed to trigger pattern detection.")
    print("We expect to see orchestrator insights about systemic failures.\n")

    # Execute all nodes
    state = goal_node(state)
    state = planning_node(state)
    state = data_ingestion_node(state)
    state = scenario_generation_node(state)
    state = evaluation_execution_node(state)
    state = scoring_node(state)
    state = report_node(state)

    # Analyze results
    print("\n" + "="*60)
    print("ğŸ“Š Pattern Detection Results")
    print("="*60 + "\n")

    failure_analysis = state.get("failure_analysis", [])
    scores = state.get("scores", {})

    print(f"âœ… Detected {len(failure_analysis)} pattern(s)\n")

    if len(failure_analysis) > 0:
        print("ğŸ¯ Orchestrator Insights Detected:\n")

        for i, pattern in enumerate(failure_analysis, 1):
            pattern_type = pattern.get("pattern_type", "unknown")
            description = pattern.get("description", "Unknown pattern")
            confidence = pattern.get("confidence", 0.0)

            print(f"{i}. [{pattern_type.upper()}] {description}")
            print(f"   Confidence: {confidence:.0%}")

            if "scenarios_affected" in pattern:
                print(f"   Scenarios: {len(pattern['scenarios_affected'])} affected")
            if "agents_affected" in pattern:
                print(f"   Agents: {', '.join(pattern['agents_affected'])}")
            if "recommendation" in pattern:
                print(f"   ğŸ’¡ {pattern['recommendation']}")
            print()
    else:
        print("âš ï¸  No patterns detected - agents might be performing too well!")
        print("   This could mean:")
        print("   - Agents are handling edge cases better than expected")
        print("   - Need more challenging failure scenarios")
        print("   - Pattern detection thresholds might need adjustment")

    # Show scores
    print("\n" + "="*60)
    print("ğŸ“ˆ Agent Performance")
    print("="*60 + "\n")

    for agent_id, agent_scores in scores.items():
        accuracy = agent_scores.get("accuracy", 0)
        print(f"{agent_id}: {accuracy:.0%} accuracy")

    # Show state design improvements
    print("\n" + "="*60)
    print("ğŸ—ï¸  State Design Improvements")
    print("="*60 + "\n")

    if "evaluation_metadata" in state:
        metadata = state["evaluation_metadata"]
        categories = metadata.get("scenario_categories", {})
        print(f"âœ… evaluation_metadata: {len(categories)} categories indexed")

    if "agent_scenario_mapping" in state:
        mapping = state["agent_scenario_mapping"]
        print(f"âœ… agent_scenario_mapping: {len(mapping)} agents mapped")
        for agent_id, scenarios in mapping.items():
            print(f"   {agent_id}: {len(scenarios)} scenarios")

    if failure_analysis:
        pattern_with_links = [p for p in failure_analysis if "result_ids" in p]
        print(f"âœ… Pattern traceability: {len(pattern_with_links)} patterns with result_ids")

    print(f"\nğŸ“„ Full report: {state.get('report_file_path', 'N/A')}")
    print("\n" + "="*60)
    print("âœ… Pattern detection test complete!")
    print("="*60 + "\n")

    return state


if __name__ == "__main__":
    try:
        final_state = test_pattern_detection()
        print("ğŸ‰ Test completed successfully!")
    except Exception as e:
        print(f"\nâŒ Error during test: {e}")
        import traceback
        traceback.print_exc()
        sys.exit(1)

```
