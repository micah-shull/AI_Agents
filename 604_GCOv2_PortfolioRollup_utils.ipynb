````python
"""
Portfolio-level governance rollup and executive triggers.

Builds portfolio_rollup from cases, enforcement events, and portfolio summary;
computes executive_triggers when thresholds are exceeded.
"""

from typing import Any, Dict, List


def build_portfolio_rollup(
    governance_cases: List[Dict[str, Any]],
    governance_portfolio_summary: List[Dict[str, Any]],
    risk_scores: Dict[str, Any],
) -> Dict[str, Any]:
    """
    Build a single portfolio_rollup dict for the current view.

    Uses latest portfolio summary run if available; merges with current risk_scores and cases.
    """
    portfolio_summary_list = governance_portfolio_summary or []
    runs = sorted(
        [p for p in portfolio_summary_list if isinstance(p, dict)],
        key=lambda x: (x.get("generated_at") or x.get("run_id") or ""),
        reverse=True,
    )
    latest_run = runs[0] if runs else {}

    open_cases = [c for c in (governance_cases or []) if (c.get("status") or "").lower() == "open"]
    high_risk_agents = list((risk_scores.get("agent_scores") or {}).keys())
    # Filter to agents with risk_score >= 60
    agent_scores = risk_scores.get("agent_scores") or {}
    high_risk_agents = [a for a, s in agent_scores.items() if (s.get("risk_score") or 0) >= 60]

    regulatory_exposure = sum(
        float(c.get("financial_exposure_usd") or 0) for c in (governance_cases or [])
    )
    top_root_causes = latest_run.get("top_root_causes") or []

    return {
        "high_risk_agents": high_risk_agents,
        "open_cases": len(open_cases),
        "open_case_ids": [c.get("case_id") for c in open_cases if c.get("case_id")],
        "blocked_actions": latest_run.get("blocked_actions", 0),
        "human_review_rate": latest_run.get("human_review_rate"),
        "regulatory_exposure_usd": round(regulatory_exposure, 2),
        "overall_risk_score": risk_scores.get("overall_risk_score"),
        "top_root_causes": top_root_causes,
        "executive_summary_snippet": latest_run.get("executive_summary"),
    }


def compute_executive_triggers(
    portfolio_rollup: Dict[str, Any],
    risk_scores: Dict[str, Any],
    executive_risk_threshold: float,
    executive_open_cases_threshold: int,
    executive_regulatory_exposure_threshold_usd: float,
) -> List[Dict[str, Any]]:
    """
    Return list of conditions that warrant executive attention.

    Each trigger: { trigger_type, message, value, threshold }
    """
    triggers: List[Dict[str, Any]] = []
    overall = float(risk_scores.get("overall_risk_score") or 0)
    if overall >= executive_risk_threshold:
        triggers.append({
            "trigger_type": "risk_threshold",
            "message": f"Overall risk score {overall} exceeds executive threshold {executive_risk_threshold}",
            "value": overall,
            "threshold": executive_risk_threshold,
        })
    open_cases = int(portfolio_rollup.get("open_cases") or 0)
    if open_cases >= executive_open_cases_threshold:
        triggers.append({
            "trigger_type": "open_cases",
            "message": f"Open governance cases ({open_cases}) at or above threshold ({executive_open_cases_threshold})",
            "value": open_cases,
            "threshold": executive_open_cases_threshold,
        })
    exposure = float(portfolio_rollup.get("regulatory_exposure_usd") or 0)
    if exposure >= executive_regulatory_exposure_threshold_usd:
        triggers.append({
            "trigger_type": "regulatory_exposure",
            "message": f"Regulatory/financial exposure ${exposure:,.0f} exceeds ${executive_regulatory_exposure_threshold_usd:,.0f}",
            "value": exposure,
            "threshold": executive_regulatory_exposure_threshold_usd,
        })
    return triggers


def build_root_cause_summary(
    governance_cases: List[Dict[str, Any]],
    portfolio_rollup: Dict[str, Any],
) -> List[Dict[str, Any]]:
    """
    Top root causes with agent/case linkage for audit report.

    Each item: { root_cause, agent_name, case_id, severity, financial_exposure_usd }
    """
    top_root_causes = portfolio_rollup.get("top_root_causes") or []
    cases_by_id = {c.get("case_id"): c for c in (governance_cases or []) if c.get("case_id")}
    result: List[Dict[str, Any]] = []
    seen: set = set()
    for cause in top_root_causes[:10]:
        if isinstance(cause, str) and cause not in seen:
            seen.add(cause)
            # Try to link to a case (e.g. by trigger or agent)
            case_match = None
            for c in (governance_cases or []):
                if cause in (c.get("triggered_by") or "") or cause in str(c.get("remediation_steps") or []):
                    case_match = c
                    break
            result.append({
                "root_cause": cause,
                "agent_name": case_match.get("agent_name") if case_match else None,
                "case_id": case_match.get("case_id") if case_match else None,
                "severity": case_match.get("risk_score") if case_match else None,
                "financial_exposure_usd": case_match.get("financial_exposure_usd") if case_match else None,
            })
    return result
````
