````python
"""Deterministic workforce metrics, trends, and executive signals for WDO v3."""

from __future__ import annotations

from collections import Counter
from typing import Any, Dict, List, Optional, Tuple


def _parse_date(s: str) -> Tuple[int, int, int]:
    y, m, d = s.split("-")
    return int(y), int(m), int(d)


def _headcount_by_department(departments: List[Dict[str, Any]]) -> Dict[str, int]:
    out: Dict[str, int] = {}
    for row in departments:
        did = row.get("department_id")
        if did is not None:
            out[str(did)] = int(row.get("headcount") or 0)
    return out


def _employee_dept_lookup(employees: List[Dict[str, Any]]) -> Dict[str, str]:
    return {str(e["employee_id"]): str(e["department_id"]) for e in employees if e.get("employee_id") and e.get("department_id")}


def _snapshots_by_department(snapshots: List[Dict[str, Any]]) -> Dict[str, List[Dict[str, Any]]]:
    by_dept: Dict[str, List[Dict[str, Any]]] = {}
    for row in snapshots:
        did = row.get("department_id")
        if not did:
            continue
        by_dept.setdefault(str(did), []).append(row)
    for did, rows in by_dept.items():
        rows.sort(key=lambda r: _parse_date(str(r.get("date", "1970-01-01"))))
    return by_dept


def _weighted_avg(values_weights: List[Tuple[float, int]]) -> Optional[float]:
    num = sum(v * w for v, w in values_weights if w > 0)
    den = sum(w for _, w in values_weights if w > 0)
    if den <= 0:
        return None
    return num / den


def _snapshot_metrics_row(row: Dict[str, Any]) -> Dict[str, float]:
    return {
        "readiness_score": float(row.get("readiness_score") or 0),
        "skill_gap_index": float(row.get("skill_gap_index") or 0),
        "avg_automation_exposure": float(row.get("avg_automation_exposure") or 0),
        "reskilling_progress_pct": float(row.get("reskilling_progress_pct") or 0),
        "attrition_risk_index": float(row.get("attrition_risk_index") or 0),
    }


def compute_org_period_rollups(
    snapshots: List[Dict[str, Any]],
    headcount_by_dept: Dict[str, int],
) -> Tuple[Optional[Dict[str, Any]], Optional[Dict[str, Any]], List[str]]:
    """
    Headcount-weighted organization metrics for the latest and previous snapshot dates
    (using each department's last and second-to-last snapshot).
    """
    warnings: List[str] = []
    by_dept = _snapshots_by_department(snapshots)
    if not by_dept:
        return None, None, ["No workforce snapshots with department_id."]

    keys = ["readiness_score", "skill_gap_index", "avg_automation_exposure", "reskilling_progress_pct", "attrition_risk_index"]
    cur_weights: Dict[str, List[Tuple[float, int]]] = {k: [] for k in keys}
    prev_weights: Dict[str, List[Tuple[float, int]]] = {k: [] for k in keys}
    latest_dates: List[str] = []
    prev_dates: List[str] = []

    for did, rows in by_dept.items():
        hc = headcount_by_dept.get(did, 0)
        if hc <= 0:
            warnings.append(f"No headcount for department {did}; excluded from org rollup.")
            continue
        if len(rows) < 2:
            warnings.append(f"Department {did} has fewer than 2 snapshots; excluded from trend rollup.")
            continue
        prev_row, cur_row = rows[-2], rows[-1]
        latest_dates.append(str(cur_row.get("date", "")))
        prev_dates.append(str(prev_row.get("date", "")))
        pm, cm = _snapshot_metrics_row(prev_row), _snapshot_metrics_row(cur_row)
        for k in keys:
            prev_weights[k].append((pm[k], hc))
            cur_weights[k].append((cm[k], hc))

    current: Dict[str, Any] = {}
    previous: Dict[str, Any] = {}
    for k in keys:
        ca = _weighted_avg(cur_weights[k])
        pa = _weighted_avg(prev_weights[k])
        if ca is None or pa is None:
            return None, None, warnings + ["Insufficient data for headcount-weighted org rollup."]
        current[k] = ca
        previous[k] = pa

    current["_period_label"] = f"{min(latest_dates)} to {max(latest_dates)} (latest per dept)"
    latest_rows = [rows[-1] for rows in by_dept.values() if len(rows) >= 2]
    prev_rows = [rows[-2] for rows in by_dept.values() if len(rows) >= 2]
    if latest_rows:
        canon_latest = max(latest_rows, key=lambda r: _parse_date(str(r.get("date", "1970-01-01"))))
        current["_run_id"] = canon_latest.get("run_id")
    if prev_rows:
        canon_prev = max(prev_rows, key=lambda r: _parse_date(str(r.get("date", "1970-01-01"))))
        previous["_run_id"] = canon_prev.get("run_id")
    return current, previous, warnings


def compute_deltas(current: Dict[str, Any], previous: Dict[str, Any]) -> Dict[str, float]:
    keys = ["readiness_score", "skill_gap_index", "avg_automation_exposure", "reskilling_progress_pct", "attrition_risk_index"]
    return {f"{k}_delta": float(current[k]) - float(previous[k]) for k in keys}


def classify_workforce_trajectory(readiness_delta: float, skill_gap_delta: float) -> str:
    if readiness_delta > 0 and skill_gap_delta < 0:
        return "improving"
    if readiness_delta < 0 and skill_gap_delta > 0:
        return "at_risk"
    return "stable"


def classify_transformation_health(automation_delta: float, reskilling_velocity: float) -> str:
    if automation_delta <= 0:
        return "healthy_adaptation"
    if reskilling_velocity >= automation_delta:
        return "healthy_adaptation"
    return "transformation_risk"


def classify_workforce_stability(attrition_delta: float) -> str:
    if attrition_delta < 0:
        return "stable_workforce"
    if attrition_delta > 0:
        return "retention_risk"
    return "stable_workforce"


def adaptation_gap_index(automation_delta: float, reskilling_velocity: float) -> float:
    """Net pressure: positive means automation pressure outpacing reskilling (risk)."""
    return float(automation_delta) - float(reskilling_velocity)


def training_roi_stats(training_rows: List[Dict[str, Any]]) -> Dict[str, Any]:
    """
    Raw ROI (doc): uplift_pct / cost_usd.
    Risk-adjusted: (uplift_pct * completion_probability) / cost_usd.
    Also uplift per $1k spend for readability.
    """
    raw_points: List[float] = []
    adj_points: List[float] = []
    per_1k: List[float] = []
    for row in training_rows:
        uplift = float(row.get("expected_productivity_uplift_pct") or 0)
        cost = float(row.get("cost_usd") or 0)
        prob = float(row.get("completion_probability") or 1.0)
        if cost <= 0:
            continue
        raw_points.append(uplift / cost)
        adj_points.append((uplift * prob) / cost)
        per_1k.append((uplift * prob) / (cost / 1000.0))

    def _mean(xs: List[float]) -> Optional[float]:
        return sum(xs) / len(xs) if xs else None

    return {
        "roi_uplift_pct_per_usd_mean": _mean(raw_points),
        "roi_risk_adjusted_uplift_pct_per_usd_mean": _mean(adj_points),
        "risk_adjusted_uplift_pct_per_1000_usd_mean": _mean(per_1k),
        "training_row_count": len(training_rows),
    }


def retraining_coverage_weighted_avg(
    controls: List[Dict[str, Any]],
    headcount_by_dept: Dict[str, int],
) -> Tuple[Optional[float], Optional[float]]:
    """Returns (coverage_weighted_mean, displacement_weighted_mean) or (None, None)."""
    cov_w: List[Tuple[float, int]] = []
    disp_w: List[Tuple[float, int]] = []
    for row in controls:
        did = str(row.get("department_id", ""))
        hc = headcount_by_dept.get(did, 0)
        if hc <= 0:
            continue
        cov = float(row.get("retraining_coverage_pct") or 0)
        disp = float(row.get("displacement_risk_pct") or 0)
        cov_w.append((cov, hc))
        disp_w.append((disp, hc))
    return _weighted_avg(cov_w), _weighted_avg(disp_w)


def skill_gap_counts_by_skill(skills_gaps: List[Dict[str, Any]]) -> Dict[str, int]:
    c = Counter(str(g.get("skill_id", "unknown")) for g in skills_gaps)
    return dict(c.most_common())


def department_analysis_table(
    snapshots: List[Dict[str, Any]],
    headcount_by_dept: Dict[str, int],
    department_filter: Optional[str],
) -> List[Dict[str, Any]]:
    by_dept = _snapshots_by_department(snapshots)
    out: List[Dict[str, Any]] = []
    for did in sorted(by_dept.keys()):
        if department_filter and did != department_filter:
            continue
        rows = by_dept[did]
        if len(rows) < 2:
            out.append(
                {
                    "department_id": did,
                    "headcount": headcount_by_dept.get(did, 0),
                    "trend_available": False,
                }
            )
            continue
        prev_row, cur_row = rows[-2], rows[-1]
        pm, cm = _snapshot_metrics_row(prev_row), _snapshot_metrics_row(cur_row)
        deltas = {k: cm[k] - pm[k] for k in pm}
        out.append(
            {
                "department_id": did,
                "headcount": headcount_by_dept.get(did, 0),
                "trend_available": True,
                "latest_run_id": cur_row.get("run_id"),
                "latest_date": cur_row.get("date"),
                "readiness_score": cm["readiness_score"],
                "readiness_delta": deltas["readiness_score"],
                "skill_gap_index": cm["skill_gap_index"],
                "skill_gap_delta": deltas["skill_gap_index"],
                "avg_automation_exposure": cm["avg_automation_exposure"],
                "automation_delta": deltas["avg_automation_exposure"],
                "reskilling_progress_pct": cm["reskilling_progress_pct"],
                "reskilling_velocity": deltas["reskilling_progress_pct"],
                "attrition_risk_index": cm["attrition_risk_index"],
                "attrition_delta": deltas["attrition_risk_index"],
                "trajectory": classify_workforce_trajectory(
                    deltas["readiness_score"], deltas["skill_gap_index"]
                ),
                "transformation_health": classify_transformation_health(
                    deltas["avg_automation_exposure"], deltas["reskilling_progress_pct"]
                ),
                "stability": classify_workforce_stability(deltas["attrition_risk_index"]),
            }
        )
    return out


def lookup_drivers(
    drivers_rows: List[Dict[str, Any]],
    run_id: str,
    department_id: str,
) -> Dict[str, List[str]]:
    for row in drivers_rows:
        if str(row.get("run_id")) == str(run_id) and str(row.get("department_id")) == str(department_id):
            return {
                "top_positive_drivers": list(row.get("top_positive_drivers") or []),
                "top_negative_drivers": list(row.get("top_negative_drivers") or []),
            }
    return {"top_positive_drivers": [], "top_negative_drivers": []}


def build_recommendations(
    trajectory: str,
    transformation_health: str,
    stability: str,
    adaptation_gap: float,
    skill_counts: Dict[str, int],
    dept_analyses: List[Dict[str, Any]],
    coverage: Optional[float],
) -> List[str]:
    recs: List[str] = []
    if trajectory == "at_risk":
        recs.append(
            "Readiness is slipping while skill gaps widen: prioritize role-targeted upskilling "
            "and reduce automation shock in the worst-hit teams."
        )
    elif trajectory == "improving":
        recs.append(
            "Trajectory is positive (readiness up, gaps narrowing): codify what high-performing "
            "departments did and replicate budgets and cadence."
        )
    else:
        recs.append(
            "Mixed signals org-wide: set explicit targets for readiness and gap closure by quarter, "
            "and review departments with divergent trends separately."
        )

    if transformation_health == "transformation_risk":
        recs.append(
            f"Automation pressure is outpacing reskilling velocity (net adaptation gap {adaptation_gap:+.3f}): "
            "accelerate funded learning paths in high-exposure departments."
        )
    if stability == "retention_risk":
        recs.append(
            "Attrition risk is rising: pair reskilling with workload and change-management interventions "
            "in departments where risk is concentrated."
        )

    if coverage is not None and coverage < 0.55:
        recs.append(
            f"Retraining coverage (~{coverage:.0%}) is below a proactive threshold: expand eligibility "
            "for at-risk cohorts and tie coverage to workforce planning gates."
        )

    worst = [d for d in dept_analyses if d.get("trend_available") and d.get("trajectory") == "at_risk"]
    if worst:
        ids = ", ".join(d["department_id"] for d in worst[:3])
        recs.append(f"Executive focus: departments {ids} show an at-risk trajectory; assign accountable owners and milestones.")

    if skill_counts:
        top_skill = next(iter(skill_counts))
        recs.append(
            f"Largest gap concentration is in “{top_skill}” skills; align learning catalog and hiring profiles to close it."
        )

    return recs


def compute_workforce_analysis(
    data_bundle: Dict[str, Any],
    department_filter: Optional[str] = None,
) -> Tuple[Dict[str, Any], List[Dict[str, Any]], List[str], List[str]]:
    """
    Full org (or single-dept) analysis: current, deltas, signals, ROI, recommendations.

    Returns (metrics, department_analyses, validation_warnings, recommendations).
    """
    snapshots = data_bundle.get("workforce_snapshots") or []
    departments = data_bundle.get("departments") or []
    headcount = _headcount_by_department(departments)

    warnings: List[str] = []
    dept_analyses = department_analysis_table(snapshots, headcount, department_filter)

    if department_filter:
        by_dept = _snapshots_by_department(snapshots)
        rows = by_dept.get(department_filter, [])
        if len(rows) < 2:
            current, previous = None, None
            warnings.append(
                f"Department {department_filter} needs at least two snapshots to compute trajectory."
            )
        else:
            pm, cm = _snapshot_metrics_row(rows[-2]), _snapshot_metrics_row(rows[-1])
            current = {**cm, "_period_label": str(rows[-1].get("date")), "_run_id": rows[-1].get("run_id")}
            previous = {**pm, "_run_id": rows[-2].get("run_id")}
    else:
        current, previous, w_org = compute_org_period_rollups(snapshots, headcount)
        warnings.extend(w_org)

    metrics: Dict[str, Any] = {
        "scope": "department" if department_filter else "organization",
        "department_id": department_filter,
    }

    if current is None or previous is None:
        metrics["status"] = "insufficient_data"
        metrics["validation_warnings"] = warnings
        return metrics, dept_analyses, warnings, []

    deltas = compute_deltas(current, previous)
    readiness_delta = deltas["readiness_score_delta"]
    skill_gap_delta = deltas["skill_gap_index_delta"]
    automation_delta = deltas["avg_automation_exposure_delta"]
    reskilling_velocity = deltas["reskilling_progress_pct_delta"]
    attrition_delta = deltas["attrition_risk_index_delta"]

    trajectory = classify_workforce_trajectory(readiness_delta, skill_gap_delta)
    transformation_health = classify_transformation_health(automation_delta, reskilling_velocity)
    stability = classify_workforce_stability(attrition_delta)
    gap_idx = adaptation_gap_index(automation_delta, reskilling_velocity)

    roi = training_roi_stats(data_bundle.get("training_investments") or [])
    coverage, displacement = retraining_coverage_weighted_avg(
        data_bundle.get("workforce_risk_controls") or [], headcount
    )

    skill_counts = skill_gap_counts_by_skill(data_bundle.get("skills_gaps") or [])

    latest_run = str(current.get("_run_id") or "")
    drivers_payload: Dict[str, Any] = {}
    if department_filter:
        drivers_payload = lookup_drivers(
            data_bundle.get("workforce_snapshot_drivers") or [], latest_run, department_filter
        )
    else:
        # Portfolio: attach drivers per department for latest run
        drivers_by_dept: Dict[str, Any] = {}
        for drow in dept_analyses:
            if not drow.get("latest_run_id"):
                continue
            rid = str(drow["latest_run_id"])
            did = str(drow["department_id"])
            drivers_by_dept[did] = lookup_drivers(
                data_bundle.get("workforce_snapshot_drivers") or [], rid, did
            )
        drivers_payload = {"by_department": drivers_by_dept}

    recs = build_recommendations(
        trajectory,
        transformation_health,
        stability,
        gap_idx,
        skill_counts,
        dept_analyses,
        coverage,
    )

    metrics.update(
        {
            "status": "ok",
            "current_period": {k: v for k, v in current.items() if not str(k).startswith("_")},
            "previous_period": {k: v for k, v in previous.items() if not str(k).startswith("_")},
            "period_labels": {
                "latest_run_id": current.get("_run_id"),
                "previous_run_id": previous.get("_run_id"),
                "note": current.get("_period_label"),
            },
            "deltas": deltas,
            "trajectory_status": trajectory,
            "transformation_health": transformation_health,
            "workforce_stability_signal": stability,
            "adaptation_gap_index": gap_idx,
            "training_roi": roi,
            "retraining_coverage_weighted_mean": coverage,
            "displacement_risk_weighted_mean": displacement,
            "skill_gap_counts_by_skill": skill_counts,
            "automation_signal_count": len(data_bundle.get("automation_signals") or []),
            "drivers": drivers_payload,
            "validation_warnings": warnings,
        }
    )

    return metrics, dept_analyses, warnings, recs
````
