## ðŸŽ¯ Decision Rules & Dependencies

### **Opportunity Detection Rules:**

1. **Routine Gap Detection:**
   - Essential categories: `cleanser`, `moisturizer`, `spf`
   - If missing â†’ routine_gap opportunity
   - Priority: High (essential for basic routine)

2. **Routine Step Detection:**
   - Customer should have products in steps 1-5
   - Missing steps â†’ cross-sell opportunity
   - Adjacent steps preferred (step 1 â†’ step 2, step 2 â†’ step 3, etc.)

3. **Ingredient Matching:**
   - Match customer `skin_concerns` to product `target_concerns`
   - Exact match = high fit
   - Partial match = medium fit
   - No match = low fit (but still possible if routine gap)

4. **Upgrade Detection:**
   - Check if customer has Tier 1 products
   - If product has `upgrades_from` matching customer's Tier 1 products â†’ upgrade opportunity
   - Only for customers with `tier_preference` >= 2

5. **Replenishment Detection:**
   - Calculate days since purchase
   - If days >= `replenishment_cycle_days` â†’ replenishment opportunity
   - If days >= `replenishment_cycle_days - 7` â†’ approaching replenishment

### **Rule Priority (Order Matters):**
1. Replenishment (highest priority - customer needs to repurchase)
2. Routine gaps (essential categories missing)
3. Routine step gaps (missing steps in routine)
4. Ingredient matches (concern-based recommendations)
5. Upgrades (Tier 1 â†’ Tier 2)
6. Cross-sells (adjacent routine steps)

# Node functions for Tier 2 Cross-Sell & Upsell Orchestrator

````python
"""Node functions for Tier 2 Cross-Sell & Upsell Orchestrator"""

from typing import Dict, Any
from config import Tier2CrossSellUpsellState


def goal_node(state: Tier2CrossSellUpsellState) -> Dict[str, Any]:
    """
    Goal Node: Define the goal for cross-sell/upsell analysis.

    This is a simple rule-based goal definition that sets the framework
    for analyzing a customer's routine and identifying opportunities.

    Args:
        state: Current state with customer_id

    Returns:
        Updated state with goal field
    """
    customer_id = state.get("customer_id")

    if not customer_id:
        return {
            "errors": state.get("errors", []) + ["goal_node: customer_id is required"]
        }

    goal = {
        "objective": "Identify cross-sell and upsell opportunities for Tier 2 customer",
        "customer_id": customer_id,
        "analysis_type": "tier2_crosssell_upsell",
        "focus_areas": [
            "routine_completeness",  # Essential categories and routine steps
            "ingredient_matching",   # Match customer concerns to product ingredients
            "upgrade_opportunities",  # Tier 1 â†’ Tier 2 upgrades
            "replenishment_needs",   # Products due for repurchase
            "routine_step_adjacency" # Adjacent routine steps for cross-sell
        ],
        "success_criteria": {
            "identify_at_least_3_opportunities": True,
            "prioritize_high_value_opportunities": True,
            "consider_customer_concerns": True,
            "consider_routine_gaps": True
        }
    }

    return {
        "goal": goal,
        "errors": state.get("errors", [])
    }


def planning_node(state: Tier2CrossSellUpsellState) -> Dict[str, Any]:
    """
    Planning Node: Create execution plan based on goal.

    This creates a step-by-step plan for analyzing the customer and
    identifying opportunities. Rule-based, no LLM needed.

    Args:
        state: Current state with goal

    Returns:
        Updated state with plan field
    """
    goal = state.get("goal")

    if not goal:
        return {
            "errors": state.get("errors", []) + ["planning_node: goal is required"]
        }

    customer_id = goal.get("customer_id")

    plan = [
        {
            "step": 1,
            "name": "data_loading",
            "description": "Load customer data and Tier 2 product catalog",
            "dependencies": [],
            "outputs": ["customer_data", "product_catalog", "product_lookup"]
        },
        {
            "step": 2,
            "name": "routine_analysis",
            "description": "Analyze customer's current routine: products, categories, routine steps, gaps",
            "dependencies": ["data_loading"],
            "outputs": ["customer_products", "customer_categories", "customer_routine_steps",
                       "routine_gaps", "routine_step_gaps", "replenishment_needs"]
        },
        {
            "step": 3,
            "name": "opportunity_detection",
            "description": "Detect opportunities: cross-sell, ingredient matches, upgrades, replenishment",
            "dependencies": ["routine_analysis"],
            "outputs": ["cross_sell_opportunities", "ingredient_opportunities",
                       "upgrade_opportunities", "replenishment_opportunities", "bundle_opportunities"]
        },
        {
            "step": 4,
            "name": "scoring",
            "description": "Score opportunities multi-dimensionally: business value, customer fit, routine completeness, etc.",
            "dependencies": ["opportunity_detection"],
            "outputs": ["scored_opportunities"]
        },
        {
            "step": 5,
            "name": "ranking",
            "description": "Rank opportunities by final score and select top N",
            "dependencies": ["scoring"],
            "outputs": ["ranked_opportunities", "top_opportunities", "opportunity_summary"]
        },
        {
            "step": 6,
            "name": "report_generation",
            "description": "Generate markdown report with recommendations",
            "dependencies": ["ranking"],
            "outputs": ["recommendations_report", "report_file_path"]
        }
    ]

    return {
        "plan": plan,
        "errors": state.get("errors", [])
    }

````

# Tests for Tier 2 Cross-Sell & Upsell Orchestrator nodes

````python
"""Tests for Tier 2 Cross-Sell & Upsell Orchestrator nodes"""

from agents.tier2_crosssell_upsell.nodes import goal_node, planning_node
from config import Tier2CrossSellUpsellState


def test_goal_node():
    """Test goal_node sets goal correctly"""
    print("\n=== Testing goal_node ===")

    # Initial state
    state: Tier2CrossSellUpsellState = {
        "customer_id": "C001",
        "errors": []
    }

    # Test goal_node
    result = goal_node(state)

    # Verify goal is set
    assert "goal" in result, "goal_node should return goal field"
    assert result["goal"]["customer_id"] == "C001", "goal should include customer_id"
    assert result["goal"]["analysis_type"] == "tier2_crosssell_upsell", "goal should have correct analysis_type"
    assert "focus_areas" in result["goal"], "goal should have focus_areas"
    assert len(result["goal"]["focus_areas"]) > 0, "goal should have at least one focus area"

    print("âœ… goal_node test passed!")
    print(f"   Goal: {result['goal']['objective']}")
    print(f"   Focus areas: {result['goal']['focus_areas']}")

    return result


def test_goal_node_missing_customer_id():
    """Test goal_node handles missing customer_id"""
    print("\n=== Testing goal_node with missing customer_id ===")

    state: Tier2CrossSellUpsellState = {
        "errors": []
    }

    result = goal_node(state)

    # Should have error
    assert "errors" in result, "goal_node should return errors field"
    assert len(result["errors"]) > 0, "goal_node should add error for missing customer_id"
    assert "customer_id is required" in result["errors"][0], "Error message should mention customer_id"

    print("âœ… goal_node error handling test passed!")
    print(f"   Error: {result['errors'][0]}")

    return result


def test_planning_node():
    """Test planning_node creates plan correctly"""
    print("\n=== Testing planning_node ===")

    # State with goal (from goal_node)
    state: Tier2CrossSellUpsellState = {
        "customer_id": "C001",
        "goal": {
            "objective": "Identify cross-sell and upsell opportunities for Tier 2 customer",
            "customer_id": "C001",
            "analysis_type": "tier2_crosssell_upsell",
            "focus_areas": ["routine_completeness", "ingredient_matching"],
            "success_criteria": {}
        },
        "errors": []
    }

    # Test planning_node
    result = planning_node(state)

    # Verify plan is set
    assert "plan" in result, "planning_node should return plan field"
    assert isinstance(result["plan"], list), "plan should be a list"
    assert len(result["plan"]) == 6, "plan should have 6 steps"

    # Verify plan structure
    for step in result["plan"]:
        assert "step" in step, "Each plan step should have 'step' field"
        assert "name" in step, "Each plan step should have 'name' field"
        assert "description" in step, "Each plan step should have 'description' field"
        assert "dependencies" in step, "Each plan step should have 'dependencies' field"
        assert "outputs" in step, "Each plan step should have 'outputs' field"

    # Verify step names
    step_names = [step["name"] for step in result["plan"]]
    expected_steps = ["data_loading", "routine_analysis", "opportunity_detection",
                     "scoring", "ranking", "report_generation"]
    assert step_names == expected_steps, f"Plan steps should be {expected_steps}"

    print("âœ… planning_node test passed!")
    print(f"   Plan has {len(result['plan'])} steps:")
    for step in result["plan"]:
        print(f"      Step {step['step']}: {step['name']}")

    return result


def test_planning_node_missing_goal():
    """Test planning_node handles missing goal"""
    print("\n=== Testing planning_node with missing goal ===")

    state: Tier2CrossSellUpsellState = {
        "customer_id": "C001",
        "errors": []
    }

    result = planning_node(state)

    # Should have error
    assert "errors" in result, "planning_node should return errors field"
    assert len(result["errors"]) > 0, "planning_node should add error for missing goal"
    assert "goal is required" in result["errors"][0], "Error message should mention goal"

    print("âœ… planning_node error handling test passed!")
    print(f"   Error: {result['errors'][0]}")

    return result


def test_goal_then_planning():
    """Test goal_node followed by planning_node (integration test)"""
    print("\n=== Testing goal_node â†’ planning_node integration ===")

    # Start with just customer_id
    state: Tier2CrossSellUpsellState = {
        "customer_id": "C001",
        "errors": []
    }

    # Step 1: goal_node
    state = {**state, **goal_node(state)}
    assert "goal" in state, "State should have goal after goal_node"

    # Step 2: planning_node
    state = {**state, **planning_node(state)}
    assert "plan" in state, "State should have plan after planning_node"

    print("âœ… Integration test passed!")
    print(f"   Customer ID: {state['customer_id']}")
    print(f"   Goal: {state['goal']['objective']}")
    print(f"   Plan steps: {len(state['plan'])}")

    return state


if __name__ == "__main__":
    """Run all tests"""
    print("=" * 60)
    print("Testing Tier 2 Cross-Sell & Upsell Orchestrator Nodes")
    print("=" * 60)

    # Test goal_node
    test_goal_node()
    test_goal_node_missing_customer_id()

    # Test planning_node
    test_planning_node()
    test_planning_node_missing_goal()

    # Integration test
    test_goal_then_planning()

    print("\n" + "=" * 60)
    print("âœ… All tests passed!")
    print("=" * 60)

````
