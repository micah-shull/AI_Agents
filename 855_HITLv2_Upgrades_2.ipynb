# Policy Compliance Layer — HITL v2 Governance Upgrade

## Overview

The **Human-in-the-Loop Collaboration Orchestrator (HITL v2)** already routes tasks according to governance policy rules based on:

* task **risk level**
* model **confidence**
* defined **routing policies**

However, routing alone does not guarantee **policy compliance**.

To strengthen the system’s governance capabilities, this upgrade introduces a **Policy Compliance Detection Layer** that verifies whether decisions **actually followed the policy rules that triggered them**.

This enhancement transforms the agent from a policy-aware workflow into a **policy-verifying governance system**.

---

# Motivation

In enterprise environments, it is not enough for systems to *intend* to follow rules.

Organizations must be able to demonstrate that:

* policies were **applied correctly**
* routing decisions were **enforced**
* governance rules were **not bypassed**

Without verification, failures can occur silently due to:

* manual overrides
* integration errors
* incorrect workflow execution
* system bugs
* human reviewers bypassing escalation policies

The Policy Compliance Layer addresses this gap by **verifying policy adherence at runtime**.

---

# Architectural Upgrade

## Existing HITL Pipeline

```
Goal
 ↓
Planning
 ↓
Data Loading
 ↓
Routing Policy Engine
 ↓
Human Simulation
 ↓
Audit Logging
 ↓
Feedback Capture
 ↓
Executive Report
```

This pipeline determines how tasks **should** be handled.

But it does not confirm whether the final action **matched the expected policy outcome**.

---

## Upgraded Governance Pipeline

The Policy Compliance Node is inserted after human review and before audit logging.

```
Goal
 ↓
Planning
 ↓
Data Loading
 ↓
Routing Policy Engine
 ↓
Human Simulation
 ↓
Policy Compliance Check
 ↓
Audit Logging
 ↓
Feedback Capture
 ↓
Executive Report
```

This node compares the **expected policy action** with the **actual outcome**.

---

# Compliance Verification Logic

The compliance node performs a simple but powerful check.

### Step 1 — Determine Expected Action

The routing engine already produces structured routing output:

```
expected_action
rule_id
assigned_human_role
```

Example:

```
Task T002
risk_level: high
confidence_score: 0.92

Triggered rule: RULE_003
Expected action: escalate
```

---

### Step 2 — Capture Actual Outcome

After routing and human interaction, the system records the final decision:

```
auto_approve
human_review
escalate
```

Example:

```
actual_action = auto_approve
```

---

### Step 3 — Compare Expected vs Actual

The compliance node verifies policy adherence:

```
if expected_action != actual_action:
    policy_violation = True
```

Example violation:

```
Expected action: escalate
Actual action: auto_approve
```

---

### Step 4 — Record Violation Event

Violations are recorded in structured logs.

Example violation record:

```
{
  "task_id": "T002",
  "rule_id": "RULE_003",
  "expected_action": "escalate",
  "actual_action": "auto_approve",
  "risk_level": "high",
  "confidence_score": 0.92,
  "violation_type": "policy_breach"
}
```

Violation records can be stored in:

```
policy_violations.json
```

or appended to the existing audit logs.

---

# Executive Reporting

The compliance layer adds a new **Policy Compliance section** to the executive report.

Example output:

```
Policy Compliance

Tasks evaluated: 5
Policy violations detected: 1
```

If violations occur, the report surfaces detailed information:

```
Violation details

Task: T002
Expected action: escalate
Actual action: auto_approve
Triggered rule: RULE_003
Risk level: high
Confidence: 0.92
```

This allows executives to quickly identify governance failures.
