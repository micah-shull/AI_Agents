````python
# ============================================================================
# Revenue Assurance Orchestrator v2
# ============================================================================
# Reconciles contracts, invoices, usage, discount approvals, recovery log.
# Surfaces underbilling, unbilled usage, unauthorized discounts, revenue at risk.
# CFO-grade report: Revenue Recovered, Revenue at Risk, Upsell Pipeline, Recovery Rate, ROI.
# Data: agents/data/ (sample_contracts.json, sample_invoices.csv, usage_data.json,
#       discount_approvals.json, recovery_log.json).

class RAAv2OrchestratorState(TypedDict, total=False):
    """State for Revenue Assurance Orchestrator v2."""

    # Input
    data_dir: Optional[str]
    reports_dir: Optional[str]
    project_root: Optional[str]

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Loaded data (single loader returns one normalized dict)
    contracts: List[Dict[str, Any]]
    invoices: List[Dict[str, Any]]
    usage_records: List[Dict[str, Any]]
    discount_approvals: List[Dict[str, Any]]
    recovery_log: List[Dict[str, Any]]

    # Lookups (built in data_loading node)
    contracts_by_customer: Dict[str, List[Dict[str, Any]]]
    approvals_by_customer: Dict[str, List[Dict[str, Any]]]
    contract_by_id: Dict[str, Dict[str, Any]]

    # Reconciliation & detection
    issues: List[Dict[str, Any]]  # Each: issue_type, customer_id, revenue_impact, severity, owner, etc.
    issues_by_type: Dict[str, List[Dict[str, Any]]]

    # Rollup (for report and triggers)
    revenue_rollup: Dict[str, Any]  # revenue_recovered, revenue_at_risk, upsell_pipeline_value, recovery_rate, total_identified
    executive_triggers: List[Dict[str, Any]]

    # Prioritized output
    top_issues: List[Dict[str, Any]]
    next_steps: List[str]

    # Metadata
    data_snapshot_loaded_at: Optional[str]
    validation_warnings: List[str]

    # Output
    revenue_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class RAAv2OrchestratorConfig:
    """Configuration for Revenue Assurance Orchestrator v2."""
    reports_dir: str = "agents/output/raa_v2_reports"
    data_dir: str = "agents/data"
    contracts_file: str = "sample_contracts.json"
    invoices_file: str = "sample_invoices.csv"
    usage_file: str = "usage_data.json"
    discount_approvals_file: str = "discount_approvals.json"
    recovery_log_file: str = "recovery_log.json"

    # Thresholds for executive triggers
    revenue_at_risk_critical: float = 50000.0
    revenue_at_risk_elevated: float = 20000.0
    open_issues_critical: int = 10
    open_issues_elevated: int = 5

    # Prioritization
    top_issues_count: int = 10
````
