# Compliance Sentinel v2 ‚Äî Enterprise AI Governance Orchestrator

## Purpose

A real-time **AI governance and risk command center** for enterprises deploying generative AI and autonomous agents at scale.

Compliance Sentinel continuously monitors AI outputs, evaluates them against regulatory and internal policies, quantifies financial exposure, escalates high-risk incidents, and produces board-ready executive reports.

---

## What It Does

* **Enterprise surveillance** ‚Äî Observes AI outputs across business units using `ai_output_events_v2.json` and `ai_system_registry_v2.json`.
* **Policy & regulation engine** ‚Äî Evaluates findings in `compliance_findings_v2.json` (GDPR, HIPAA, EEOC, Fair Lending, SOX).
* **Risk scoring** ‚Äî Computes severity, risk_score, and repeat-offender patterns from compliance findings.
* **Financial exposure modeling** ‚Äî Uses `financial_exposure_models_v2.json` to estimate fines, remediation cost, legal exposure, and total impact.
* **Executive escalations** ‚Äî Tracks `escalation_events_v2.json`; configurable thresholds trigger alerts and executive routing.
* **Human review & remediation** ‚Äî Coordinates oversight and corrective action via `human_review_actions_v2.json` and `remediation_actions_v2.json`.
* **Portfolio snapshots** ‚Äî Leverages `portfolio_snapshots_v2.json` for trend analysis and portfolio-level risk scoring.
* **Board-ready reporting** ‚Äî Generates a one-view verdict, ‚Äúso-what‚Äù line, prioritized next steps, R/Y/G segmentation tables, escalation summaries, and remediation status.

---

## Data

All inputs live under `agents/data_v2/` (configurable via `--data-dir`):

| File                                | Content                                                         |
| ----------------------------------- | --------------------------------------------------------------- |
| `ai_output_events_v2.json`          | AI output events (sensitive data, region, risk signal)          |
| `ai_system_registry_v2.json`        | Registered AI systems (business unit, owner, regulated domains) |
| `compliance_findings_v2.json`       | Findings (regulation, severity, risk_score, status)             |
| `remediation_actions_v2.json`       | Remediation actions (finding, status, exposure reduction)       |
| `escalation_events_v2.json`         | Escalation events (trigger, exposure, notified roles)           |
| `financial_exposure_models_v2.json` | Per-finding exposure (fine, legal, remediation, total)          |
| `human_review_actions_v2.json`      | Human review decisions and SLAs                                 |
| `portfolio_snapshots_v2.json`       | Daily portfolio metrics and trends                              |

---

## Running

```bash
# From project root, with venv activated
python run_compliance_sentinel_v2.py
python run_compliance_sentinel_v2.py --data-dir agents/data_v2 --reports-dir agents/output/compliance_sentinel_v2_reports

# Via project-level orchestrator
python run_orchestrator.py --agent compliance_sentinel_v2
python run_orchestrator.py --agent compliance_sentinel_v2 --data-dir agents/data_v2
```

Reports are written to `agents/output/compliance_sentinel_v2_reports/` (or `--reports-dir`) as:

```
compliance_sentinel_v2_report_YYYYMMDD_HHMMSS.md
```

---

## Configuration (`config.py`)

`ComplianceSentinelV2Config` controls:

* data paths and file names
* executive trigger thresholds:

  * `open_cases_critical`
  * `total_exposure_usd_critical`
  * `portfolio_risk_score_critical`
  * `critical_findings_critical`
  * `executive_escalations_critical`

When any threshold is exceeded, the workflow executes the **escalation** node before report generation, and the report highlights executive attention at the top.

* `orchestrator/nodes.py` ‚Äî goal, planning, data loading (closure), portfolio (closure), escalation (closure), reporting (closure).
