````python
"""Decision Analysis Utilities

Analyze orchestrator decisions: patterns, triggers, confidence scores.
"""

from typing import Dict, Any, List
from collections import Counter


def analyze_campaign_decisions(
    campaign_id: str,
    decisions: List[Dict[str, Any]]
) -> Dict[str, Any]:
    """
    Analyze decisions for a single campaign.

    Args:
        campaign_id: Campaign ID
        decisions: List of all orchestrator decisions

    Returns:
        Decision analysis dictionary for the campaign
    """
    campaign_decisions = [
        d for d in decisions
        if d.get("campaign_id") == campaign_id
    ]

    if not campaign_decisions:
        return {
            "campaign_id": campaign_id,
            "total_decisions": 0,
            "automated_decisions": 0,
            "human_overrides": 0,
            "average_confidence": 0.0,
            "common_triggers": [],
            "decision_patterns": []
        }

    # Count automated vs human overrides
    automated = len([d for d in campaign_decisions if not d.get("human_override", False)])
    human_overrides = len([d for d in campaign_decisions if d.get("human_override", False)])

    # Calculate average confidence
    confidence_scores = [d.get("confidence_score", 0.0) for d in campaign_decisions]
    average_confidence = sum(confidence_scores) / len(confidence_scores) if confidence_scores else 0.0

    # Extract common triggers
    triggers = [d.get("trigger", "") for d in campaign_decisions if d.get("trigger")]
    trigger_counts = Counter(triggers)
    common_triggers = [trigger for trigger, count in trigger_counts.most_common(3)]

    # Extract decision patterns from actions
    actions = [d.get("action_taken", "") for d in campaign_decisions if d.get("action_taken")]
    # Simple pattern detection: look for keywords
    patterns = []
    if any("budget" in action.lower() or "allocation" in action.lower() for action in actions):
        patterns.append("budget_reallocation")
    if any("scale" in action.lower() or "prioritize" in action.lower() for action in actions):
        patterns.append("experiment_scaling")
    if any("pause" in action.lower() or "stop" in action.lower() for action in actions):
        patterns.append("campaign_pausing")
    if any("review" in action.lower() or "flag" in action.lower() for action in actions):
        patterns.append("human_escalation")

    return {
        "campaign_id": campaign_id,
        "total_decisions": len(campaign_decisions),
        "automated_decisions": automated,
        "human_overrides": human_overrides,
        "average_confidence": round(average_confidence, 2),
        "common_triggers": common_triggers,
        "decision_patterns": patterns
    }


def analyze_all_decisions(
    campaigns: List[Dict[str, Any]],
    decisions: List[Dict[str, Any]]
) -> List[Dict[str, Any]]:
    """
    Analyze decisions for all campaigns.

    Args:
        campaigns: List of campaign dictionaries
        decisions: List of all orchestrator decisions

    Returns:
        List of decision analysis dictionaries
    """
    insights = []
    for campaign in campaigns:
        campaign_id = campaign.get("campaign_id")
        if campaign_id:
            insight = analyze_campaign_decisions(campaign_id, decisions)
            insights.append(insight)

    return insights
````
