```python
"""
Analysis utilities for Workforce Development Orchestrator v2.

Rule-based: automation risk, skill gap enrichment, learning path recommendations,
role evolution/transition recommendations, workforce summary.
"""

from typing import Any, Dict, List, Optional

# Default thresholds (match config)
DEFAULT_HIGH = 0.7
DEFAULT_MEDIUM = 0.4


def compute_automation_risk_analysis(
    tasks_by_role: Dict[str, List[Dict[str, Any]]],
    tasks_lookup: Dict[str, Dict[str, Any]],
    employees_by_role: Dict[str, List[Dict[str, Any]]],
    roles_lookup: Dict[str, Dict[str, Any]],
    employee_id: Optional[str] = None,
    high_threshold: float = DEFAULT_HIGH,
    medium_threshold: float = DEFAULT_MEDIUM,
) -> List[Dict[str, Any]]:
    """
    Compute automation risk per role (and optionally per employee).

    Returns list of analysis items with role_id, high/medium/low_risk_tasks,
    overall_risk_score, risk_level, affected_employees, recommendations.
    """
    result: List[Dict[str, Any]] = []
    for rid, role_tasks in tasks_by_role.items():
        if employee_id:
            emps = [e for e in employees_by_role.get(rid, []) if e["employee_id"] == employee_id]
            if not emps:
                continue
        else:
            emps = employees_by_role.get(rid, [])

        high_risk = [t["task_id"] for t in role_tasks if t.get("automation_risk_score", 0) >= high_threshold]
        medium_risk = [
            t["task_id"]
            for t in role_tasks
            if medium_threshold <= t.get("automation_risk_score", 0) < high_threshold
        ]
        low_risk = [t["task_id"] for t in role_tasks if t.get("automation_risk_score", 0) < medium_threshold]

        scores = [t.get("automation_risk_score", 0) for t in role_tasks]
        overall = sum(scores) / len(scores) if scores else 0.0
        if overall >= high_threshold:
            risk_level = "high"
        elif overall >= medium_threshold:
            risk_level = "medium"
        else:
            risk_level = "low"

        recs: List[str] = []
        if high_risk:
            recs.append(f"Prioritize reskilling for {len(high_risk)} high-automation-risk task(s).")
        if overall >= medium_threshold:
            recs.append("Review role evolution and learning paths for this role.")

        result.append({
            "role_id": rid,
            "role_name": roles_lookup.get(rid, {}).get("role_name", rid),
            "high_risk_tasks": high_risk,
            "medium_risk_tasks": medium_risk,
            "low_risk_tasks": low_risk,
            "overall_risk_score": round(overall, 2),
            "risk_level": risk_level,
            "affected_employees": [e["employee_id"] for e in emps],
            "recommendations": recs,
        })
    return result


def enrich_skill_gap_analysis(
    skill_gaps: List[Dict[str, Any]],
    employees_lookup: Dict[str, Dict[str, Any]],
    roles_lookup: Dict[str, Dict[str, Any]],
    skills_lookup: Dict[str, Dict[str, Any]],
    learning_paths: List[Dict[str, Any]],
    automation_risk_analysis: List[Dict[str, Any]],
) -> List[Dict[str, Any]]:
    """
    Enrich skill gaps with names, automation context, urgency, recommended path.
    """
    role_risk: Dict[str, str] = {}
    for a in automation_risk_analysis:
        role_risk[a["role_id"]] = a.get("risk_level", "low")

    skill_to_path: Dict[str, Dict[str, Any]] = {}
    for lp in learning_paths:
        skill_to_path[lp["target_skill"]] = lp

    priority_score = {"high": 0.9, "medium": 0.6, "low": 0.3}
    enriched: List[Dict[str, Any]] = []
    for g in skill_gaps:
        emp = employees_lookup.get(g["employee_id"], {})
        role = roles_lookup.get(emp.get("role_id", ""), {})
        skill = skills_lookup.get(g["skill_id"], {})
        risk_level = role_risk.get(emp.get("role_id", ""), "low")
        urgency = priority_score.get(g.get("priority", "medium"), 0.5)
        if g.get("role_requirement"):
            urgency = min(1.0, urgency + 0.1)
        if risk_level == "high":
            urgency = min(1.0, urgency + 0.15)
        rec_path = skill_to_path.get(g["skill_id"], {})
        enriched.append({
            **g,
            "employee_name": emp.get("name", ""),
            "role_id": emp.get("role_id", ""),
            "role_name": role.get("role_name", ""),
            "skill_name": skill.get("skill_name", ""),
            "automation_risk_context": risk_level,
            "urgency_score": round(urgency, 2),
            "recommended_learning_path": rec_path.get("learning_path_id"),
        })
    return enriched


def compute_learning_path_recommendations(
    skill_gap_analysis: List[Dict[str, Any]],
    learning_paths: List[Dict[str, Any]],
    employees_lookup: Dict[str, Dict[str, Any]],
    match_threshold: float = 0.6,
    top_n: int = 10,
) -> List[Dict[str, Any]]:
    """
    For each gap with a recommended_learning_path, produce a recommendation record.
    """
    path_lookup = {lp["learning_path_id"]: lp for lp in learning_paths}
    recs: List[Dict[str, Any]] = []
    for g in skill_gap_analysis:
        lp_id = g.get("recommended_learning_path")
        if not lp_id:
            continue
        lp = path_lookup.get(lp_id, {})
        match_score = g.get("urgency_score", 0.5)
        if match_score < match_threshold:
            continue
        recs.append({
            "employee_id": g["employee_id"],
            "employee_name": g.get("employee_name", ""),
            "learning_path_id": lp_id,
            "learning_path_name": lp.get("path_name", ""),
            "target_skill": g["skill_id"],
            "match_score": round(match_score, 2),
            "priority": g.get("priority", "medium"),
            "estimated_completion_weeks": lp.get("estimated_completion_weeks"),
            "prerequisites_met": True,  # Simplified: assume met
            "rationale": f"Employee missing future/role skill: {g.get('skill_name', g['skill_id'])}",
        })
    order = {"high": 0, "medium": 1, "low": 2}
    recs.sort(key=lambda x: (order.get(x["priority"], 2), -x["match_score"]))
    return recs[:top_n]


def compute_role_evolution_recommendations(
    role_evolution: List[Dict[str, Any]],
    roles_lookup: Dict[str, Dict[str, Any]],
    employees_by_role: Dict[str, List[Dict[str, Any]]],
    top_n: int = 5,
) -> List[Dict[str, Any]]:
    """
    Turn role_evolution data into recommendations with affected_employees and readiness.
    """
    recs: List[Dict[str, Any]] = []
    for ev in role_evolution:
        rid = ev.get("role_id")
        emps = employees_by_role.get(rid, [])
        # Simple readiness: share of employees with 2+ current skills (placeholder)
        readiness = 0.75 if emps else 0.0
        recs.append({
            "role_id": rid,
            "role_name": ev.get("current_role_name", roles_lookup.get(rid, {}).get("role_name", "")),
            "evolution_id": ev.get("evolution_id"),
            "evolution_type": ev.get("evolution_type", ""),
            "description": ev.get("description", ""),
            "affected_employees": [e["employee_id"] for e in emps],
            "automated_tasks": ev.get("automated_tasks", []),
            "new_skills_required": ev.get("new_skills_required", []),
            "readiness_score": readiness,
            "implementation_priority": "high" if ev.get("evolution_type") == "transformed" else "medium",
        })
    recs.sort(key=lambda x: (0 if x["implementation_priority"] == "high" else 1, -x["readiness_score"]))
    return recs[:top_n]


def compute_role_transition_recommendations(
    role_transitions: List[Dict[str, Any]],
    roles_lookup: Dict[str, Dict[str, Any]],
    employees_by_role: Dict[str, List[Dict[str, Any]]],
    employees_lookup: Dict[str, Dict[str, Any]],
    top_n: int = 5,
) -> List[Dict[str, Any]]:
    """
    List feasible role transitions with sample employees who could transition.
    """
    recs: List[Dict[str, Any]] = []
    for rt in role_transitions:
        from_rid = rt.get("from_role_id")
        to_rid = rt.get("to_role_id")
        from_emps = employees_by_role.get(from_rid, [])
        recs.append({
            "from_role_id": from_rid,
            "from_role_name": roles_lookup.get(from_rid, {}).get("role_name", from_rid),
            "to_role_id": to_rid,
            "to_role_name": roles_lookup.get(to_rid, {}).get("role_name", to_rid),
            "transition_probability": rt.get("transition_probability"),
            "training_required_paths": rt.get("training_required_paths", []),
            "time_to_transition_weeks": rt.get("time_to_transition_weeks"),
            "internal_demand_score": rt.get("internal_demand_score"),
            "candidate_employees": [e["employee_id"] for e in from_emps[:3]],
        })
    recs.sort(key=lambda x: -(x.get("transition_probability") or 0))
    return recs[:top_n]


def compute_workforce_summary(
    employees: List[Dict[str, Any]],
    roles: List[Dict[str, Any]],
    tasks: List[Dict[str, Any]],
    skill_gap_analysis: List[Dict[str, Any]],
    automation_risk_analysis: List[Dict[str, Any]],
    learning_path_recommendations: List[Dict[str, Any]],
    role_evolution_recommendations: List[Dict[str, Any]],
    high_threshold: float = DEFAULT_HIGH,
) -> Dict[str, Any]:
    """
    Aggregate summary counts and overall workforce readiness score.
    """
    employees_at_risk = set()
    for a in automation_risk_analysis:
        if a.get("risk_level") == "high" or a.get("overall_risk_score", 0) >= high_threshold:
            for eid in a.get("affected_employees", []):
                employees_at_risk.add(eid)

    high_priority_gaps = sum(1 for g in skill_gap_analysis if g.get("priority") == "high")
    readiness = 1.0 - (len(employees_at_risk) / len(employees)) if employees else 0.0
    readiness = readiness * 0.7 + (1.0 - min(1.0, len(skill_gap_analysis) / max(1, len(employees)))) * 0.3
    readiness = max(0.0, min(1.0, readiness))

    return {
        "total_employees": len(employees),
        "total_roles": len(roles),
        "total_tasks": len(tasks),
        "employees_at_risk": len(employees_at_risk),
        "total_skill_gaps": len(skill_gap_analysis),
        "high_priority_gaps": high_priority_gaps,
        "total_learning_recommendations": len(learning_path_recommendations),
        "roles_requiring_evolution": len(role_evolution_recommendations),
        "overall_workforce_readiness_score": round(readiness, 2),
    }


def prioritize_gaps(gap_analysis: List[Dict[str, Any]], top_n: int = 10) -> List[Dict[str, Any]]:
    """Sort gaps by priority and urgency, return top_n."""
    order = {"high": 0, "medium": 1, "low": 2}
    out = sorted(
        gap_analysis,
        key=lambda g: (order.get(g.get("priority", "low"), 2), -g.get("urgency_score", 0)),
    )
    return out[:top_n]


def prioritize_evolutions(
    evolution_recommendations: List[Dict[str, Any]], top_n: int = 5
) -> List[Dict[str, Any]]:
    """Already sorted from compute; return top_n."""
    return evolution_recommendations[:top_n]
```
