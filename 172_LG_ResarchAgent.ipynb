## **The Three Key Changes Explained:**

### **1. Added `goal_interpretation: Dict[str, any]`**
**Why this change?**
- In the original scaffold, the goal was just passed through as-is
- Now we use an LLM to actually **analyze and structure** the goal
- This field stores the LLM's interpretation of what the user wants, including:
  - Key topics extracted
  - Target audience analysis
  - Success criteria breakdown
  - Suggested search approaches

**What it enables:**
- More intelligent search strategy generation
- Better report tailoring to the actual goal
- Smarter validation against original intent

### **2. Changed `search_results: List[Dict]` ‚Üí `Dict[str, List[Dict]]`**
**Why this change?**
- Original: All search results were lumped together in one list
- New: Results are organized **by search query**
- This matches how real web search works - you get results per query

**What it enables:**
- Track which results came from which search queries
- Better relevance analysis per query
- Debugging and optimization of search strategies
- Parallel search execution (each query gets its own result set)

### **3. Added `validation_result: Dict[str, any]`**
**Why this change?**
- Original: Simple boolean check if goal was met
- New: Comprehensive validation with detailed feedback
- Stores quality scores, completeness metrics, and improvement suggestions

**What it enables:**
- Detailed quality assessment of the final report
- Feedback on what worked well and what didn't
- Confidence scoring for the research results
- Actionable recommendations for improvement

## **Why These Changes Matter:**

1. **Better Data Flow**: Each stage now has richer data to work with
2. **Real Functionality**: Instead of mock data, we're storing actual LLM analysis
3. **Debugging**: You can inspect what the LLM actually understood and produced
4. **Quality Control**: We can measure and improve the research quality
5. **Scalability**: The structure supports more complex research workflows

The key insight is that **state management in LangGraph is crucial** - the state schema determines what data flows between nodes, and richer state enables more sophisticated AI workflows.

````python
"""
Research Agent - Goal Interpretation and Search Strategy Testing
Isolated test for the first two nodes of the research workflow
"""

import os
import sys
from typing import Dict, List, Any
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Add tools directory to path
sys.path.append(os.path.join(os.path.dirname(__file__), '..'))

# Import our tools
from tools.llm_research_tools import LLMResearchTool, LLMConfig

def test_goal_interpretation():
    """Test the goal interpretation functionality"""
    print("=" * 60)
    print("üß™ TESTING GOAL INTERPRETATION")
    print("=" * 60)

    # Initialize LLM tool
    llm_tool = LLMResearchTool()

    # Sample research goal
    sample_goal = {
        "objective": "Identify major AI trends emerging in the industry today",
        "scope": "Industry trends, not academic research",
        "output_format": "Industry standard report",
        "target_audience": "Business professionals",
        "depth": "Comprehensive overview with key insights",
        "sources": "Web-based industry sources, recent articles",
        "success_criteria": "Clear identification of top 5-7 trends with supporting evidence"
    }

    print(f"üìã Original Goal:")
    for key, value in sample_goal.items():
        print(f"   {key}: {value}")

    print(f"\nüéØ Interpreting goal with LLM...")

    try:
        # Interpret the goal
        goal_interpretation = llm_tool.interpret_research_goal(sample_goal)

        print(f"\n‚úÖ Goal Interpretation Results:")
        print(f"   Main Objective: {goal_interpretation.get('main_objective', 'N/A')}")
        print(f"   Key Topics: {goal_interpretation.get('key_topics', [])}")
        print(f"   Target Audience: {goal_interpretation.get('target_audience', 'N/A')}")
        print(f"   Required Depth: {goal_interpretation.get('required_depth', 'N/A')}")
        print(f"   Success Criteria: {goal_interpretation.get('success_criteria', 'N/A')}")
        print(f"   Suggested Queries: {goal_interpretation.get('suggested_queries', [])}")

        print(f"\nüìù LLM Analysis Notes:")
        print(f"   {goal_interpretation.get('analysis_notes', 'No analysis notes')}")

        return goal_interpretation

    except Exception as e:
        print(f"‚ùå Goal interpretation failed: {str(e)}")
        return None

def test_search_strategy_generation(goal_interpretation: Dict[str, Any]):
    """Test the search strategy generation functionality"""
    print("\n" + "=" * 60)
    print("üîç TESTING SEARCH STRATEGY GENERATION")
    print("=" * 60)

    if not goal_interpretation:
        print("‚ùå Cannot test search strategy - no goal interpretation available")
        return

    # Initialize LLM tool
    llm_tool = LLMResearchTool()

    print(f"üìä Using Goal Interpretation:")
    print(f"   Main Objective: {goal_interpretation.get('main_objective', 'N/A')}")
    print(f"   Key Topics: {goal_interpretation.get('key_topics', [])}")
    print(f"   Target Audience: {goal_interpretation.get('target_audience', 'N/A')}")

    print(f"\nüîç Generating search strategy with LLM...")

    try:
        # Generate search strategy
        search_queries = llm_tool.generate_search_strategy(goal_interpretation)

        print(f"\n‚úÖ Search Strategy Results:")
        print(f"   Number of queries: {len(search_queries)}")
        print(f"   Generated queries:")
        for i, query in enumerate(search_queries, 1):
            print(f"     {i}. {query}")

        return search_queries

    except Exception as e:
        print(f"‚ùå Search strategy generation failed: {str(e)}")
        return None

def test_different_goals():
    """Test with different types of research goals"""
    print("\n" + "=" * 60)
    print("üîÑ TESTING DIFFERENT RESEARCH GOALS")
    print("=" * 60)

    # Different sample goals
    test_goals = [
        {
            "name": "AI Trends Research",
            "goal": {
                "objective": "Identify major AI trends emerging in the industry today",
                "scope": "Industry trends, not academic research",
                "output_format": "Industry standard report",
                "target_audience": "Business professionals",
                "depth": "Comprehensive overview with key insights",
                "sources": "Web-based industry sources, recent articles",
                "success_criteria": "Clear identification of top 5-7 trends with supporting evidence"
            }
        },
        {
            "name": "Market Analysis",
            "goal": {
                "objective": "Analyze the competitive landscape for electric vehicles",
                "scope": "Global market analysis",
                "output_format": "Executive summary",
                "target_audience": "C-suite executives",
                "depth": "High-level strategic overview",
                "sources": "Industry reports, financial data, news",
                "success_criteria": "Clear competitive positioning and market opportunities"
            }
        },
        {
            "name": "Technology Research",
            "goal": {
                "objective": "Research emerging cybersecurity threats and defenses",
                "scope": "Technical and business implications",
                "output_format": "Technical brief",
                "target_audience": "IT security professionals",
                "depth": "Detailed technical analysis",
                "sources": "Security research papers, threat intelligence",
                "success_criteria": "Actionable security recommendations"
            }
        }
    ]

    llm_tool = LLMResearchTool()

    for test_case in test_goals:
        print(f"\nüìã Testing: {test_case['name']}")
        print(f"   Objective: {test_case['goal']['objective']}")

        try:
            # Interpret goal
            interpretation = llm_tool.interpret_research_goal(test_case['goal'])

            # Generate search strategy
            queries = llm_tool.generate_search_strategy(interpretation)

            print(f"   ‚úÖ Generated {len(queries)} search queries:")
            for i, query in enumerate(queries[:3], 1):  # Show first 3
                print(f"     {i}. {query}")
            if len(queries) > 3:
                print(f"     ... and {len(queries) - 3} more")

        except Exception as e:
            print(f"   ‚ùå Failed: {str(e)}")

def main():
    """Main testing function"""
    print("üöÄ Research Agent - Goal Interpretation & Search Strategy Testing")
    print("Focus: Testing LLM-powered goal analysis and search query generation")

    # Test 1: Basic goal interpretation
    goal_interpretation = test_goal_interpretation()

    # Test 2: Search strategy generation
    search_queries = test_search_strategy_generation(goal_interpretation)

    # Test 3: Different goal types
    test_different_goals()

    print("\n" + "=" * 60)
    print("üéì Testing Complete!")
    print("Key Concepts Demonstrated:")
    print("‚Ä¢ LLM-powered goal interpretation")
    print("‚Ä¢ Structured data extraction from goals")
    print("‚Ä¢ Intelligent search query generation")
    print("‚Ä¢ Error handling and fallback mechanisms")
    print("=" * 60)

if __name__ == "__main__":
    main()
````

```python
def search_duckduckgo(query):
    url = f"https://duckduckgo.com/html/?q={query}"
    response = requests.get(url, headers={'User-Agent': '...'})
    soup = BeautifulSoup(response.text, 'html.parser')
    
    results = []
    for result in soup.find_all('div', class_='result'):
        title = result.find('a', class_='result__a').text
        url = result.find('a', class_='result__a')['href']
        snippet = result.find('div', class_='result__snippet').text
        results.append({"title": title, "url": url, "snippet": snippet})
    
    return results
```
