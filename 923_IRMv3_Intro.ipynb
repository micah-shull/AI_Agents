## Executive Operating Intelligence for AI Systems

# Scoring Model Design

## Purpose

Executives should not need to read raw logs.

They should see a few signals that answer the most important questions:

* **Are our AI systems stable?**

* **Where are we exposed to risk?**

# The Four Core Scores

The orchestrator calculates four major scores.

Each one answers a different leadership question.

| Score                           | What it answers                              |
| ------------------------------- | -------------------------------------------- |
| Integration Stability Score     | Are the systems connecting reliably?         |
| Dependency Exposure Score       | Are we vulnerable to shared system failures? |
| Governance Responsiveness Score | Are problems getting fixed?                  |
| Automation Value Score          | Are we getting the value we expected?        |

These four scores combine into a final signal:

**Portfolio Health Status**

```
HEALTHY
AT RISK
CRITICAL
```

## Business Question

**Are the systems connecting reliably?**

Most AI failures happen at the integration layer:

* APIs slow down
* schemas change

From:

---

## Core Factors

### Uptime reliability

```
uptime >= 99.5% → excellent
uptime 98–99.5 → moderate risk
uptime <98 → unstable
```

---

### API latency

```
p95 latency < 300ms → healthy
300–800ms → degraded
>800ms → unstable
```

---

### Error rate

```
error rate <1% → healthy
1–5% → concerning
>5% → unstable
```

---

### Schema mismatch events

Frequent schema changes cause workflow failures.

Penalty applied when mismatches occur.

| ------ | ------------- |

# Score Calculation Formulas

Each score is normalized to a **0–100 scale**.

Higher scores always mean **healthier systems**.

All formulas use **weighted components** so the logic is easy to tune later in configuration.

---

# 1 Integration Stability Score

## What It Measures

**Are the systems connecting reliably?**

This score reflects the health of external services, APIs, and system integrations.

---

## Inputs

From:

* `integration_health_logs.json`
* `system_integrations.json`
* `dependency_incidents.json`

---

# Component Metrics

### Uptime Score

```
uptime_score = uptime_percentage
```

Example:

```
99.7% uptime → score = 99.7
```

---

### Error Rate Score

Error rate is inverted because lower errors are better.

```
error_rate_score = 100 - (error_rate * 100)
```

Example:

```
2% error rate → 100 - 2 = 98
```

---

### Latency Score

We normalize latency relative to a **target latency of 300ms**.

```
latency_score = max(0, 100 - (latency_ms / 300) * 20)
```

Examples:

```
150ms → score ≈ 90
300ms → score ≈ 80
800ms → score ≈ 47
```

---

### Schema Stability Score

Frequent schema mismatches reduce reliability.

```
schema_score = max(0, 100 - (schema_mismatch_events * 10))
```

Example:

```
3 mismatches → score = 70
```

---

# Final Formula

```
Integration Stability Score =
(0.40 * uptime_score) +
(0.30 * error_rate_score) +
(0.20 * latency_score) +
(0.10 * schema_score)
```

---

# 2 Dependency Exposure Score

## What It Measures

**How fragile the system architecture is.**

If one integration fails, how much breaks?

---

## Inputs

From:

* `dependency_graph.json`
* `dependency_incidents.json`
* `agents.json`
* `workflows.json`

---

# Component Metrics

### Dependency Concentration Score

If many systems rely on one dependency, exposure increases.

```
concentration_score =
100 - (dependent_workflows * 5)
```

Example:

```
5 workflows rely on SendGrid → score = 75
```

---

### Critical Workflow Impact Score

Mission-critical workflows increase risk.

```
criticality_score =
100 - (critical_workflows * 10)
```

Example:

```
3 critical workflows → score = 70
```

---

### Incident History Score

Frequent outages indicate fragility.

```
incident_score =
100 - (incident_count * 8)
```

Example:

```
4 incidents → score = 68
```

---

### Failover Protection Score

If fallback systems exist, exposure drops.

```
failover_score =
100 if failover_available else 60
```

---

# Final Formula

```
Dependency Exposure Score =
(0.35 * concentration_score) +
(0.30 * criticality_score) +
(0.25 * incident_score) +
(0.10 * failover_score)
```

---

# 3 Governance Responsiveness Score

## What It Measures

**Whether the organization fixes known problems.**

This score reflects **leadership accountability**.

---

## Inputs

From:

* `risk_signals.json`
* `remediation_actions.json`
* `governance_reviews.json`

---

# Component Metrics

### Remediation Completion Score

```
completion_rate = resolved_actions / total_actions
completion_score = completion_rate * 100
```

---

### Overdue Remediation Penalty

```
overdue_penalty = overdue_actions * 5
```

---

### Average Resolution Time Score

```
resolution_score =
max(0, 100 - (avg_days_to_resolve * 2))
```

Example:

```
20 days → score = 60
```

---

### Repeat Incident Penalty

Recurring issues indicate poor governance.

```
repeat_penalty = repeated_incidents * 6
```

---

# Final Formula

```
Governance Responsiveness Score =
(0.50 * completion_score) +
(0.30 * resolution_score) -
(overdue_penalty + repeat_penalty)
```

Then clamp:

```
score = max(0, min(100, score))
```

---

# 4 Automation Value Score

## What It Measures

**Are AI systems delivering the value they promised?**

---

## Inputs

From:

* `expected_vs_actual_value.json`
* `kpis_cost.json`
* `task_execution_logs.json`

---

# Component Metrics

### ROI Realization Score

```
roi_ratio = actual_roi / expected_roi
roi_score = min(100, roi_ratio * 100)
```

Example:

```
ROI achieved 80% → score = 80
```

---

### Cost Overrun Score

```
cost_score =
max(0, 100 - (cost_overrun_percentage))
```

Example:

```
15% overrun → score = 85
```

---

### Manual Labor Score

```
labor_score =
max(0, 100 - (manual_minutes / expected_minutes) * 100)
```

---

### Throughput Performance Score

```
throughput_ratio =
actual_throughput / expected_throughput

throughput_score =
min(100, throughput_ratio * 100)
```

---

# Final Formula

```
Automation Value Score =
(0.40 * roi_score) +
(0.25 * cost_score) +
(0.20 * throughput_score) +
(0.15 * labor_score)
```

---

# Portfolio Health Calculation

The orchestrator combines the four scores into one signal.

---

## Weighted Portfolio Score

```
portfolio_score =
(0.30 * integration_stability) +
(0.25 * dependency_exposure) +
(0.20 * governance_responsiveness) +
(0.25 * automation_value)
```

---

## Status Logic

```
if any score < 60:
    status = CRITICAL

elif portfolio_score < 75:
    status = AT_RISK

else:
    status = HEALTHY
```

---
