## 📊 **RAG Confidence Scoring Strategy**

I'm thinking we implement **multi-layered confidence scoring**:

```python
class RAGConfidence:
    retrieval_confidence: float    # Vector similarity score
    knowledge_coverage: float      # How much of query is covered
    response_confidence: float    # LLM's confidence in response
    overall_confidence: float     # Combined weighted score
```

### **Confidence Thresholds**
- **>0.8**: High confidence → Direct response
- **0.6-0.8**: Medium confidence → Response with disclaimers  
- **<0.6**: Low confidence → Escalate to human

````python
"""
Customer Support Resolution Agent - LangGraph MVP Implementation
Goal-driven support agent with RAG-enhanced responses and escalation logic
"""

import logging
from typing import TypedDict, List, Optional, Dict, Any
from langgraph import StateGraph, END
from langgraph.graph import StateGraph

# Configure logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# State Definition
class SupportAgentState(TypedDict):
    """State object for the Customer Support Resolution Agent"""
    # Customer Context
    customer_id: str
    ticket_id: str

    # Goal and Query
    goal: Dict[str, Any]
    customer_query: str

    # Knowledge and Context
    retrieved_knowledge: List[Dict[str, Any]]
    agent_response: str

    # Resolution Process
    confidence_score: float
    escalation_reason: Optional[str]

    # Output
    final_response: str
    resolution_status: str
    audit_log: List[Dict[str, Any]]

# Node Functions (Placeholder implementations)
def set_support_goal_and_criteria(state: SupportAgentState) -> SupportAgentState:
    """Set the resolution goal and success criteria based on customer query"""
    logger.info("🎯 Setting support goal and criteria...")

    # Extract query intent and set goal
    query = state["customer_query"]

    # Simple goal setting logic (will be enhanced with LLM)
    if "bill" in query.lower() or "charge" in query.lower() or "payment" in query.lower():
        goal = {
            "objective": "Resolve billing inquiry",
            "success_criteria": "Customer understands their charges and next steps",
            "priority": "medium",
            "escalation_threshold": 0.6,
            "category": "billing"
        }
    elif "login" in query.lower() or "password" in query.lower() or "app" in query.lower():
        goal = {
            "objective": "Resolve technical issue",
            "success_criteria": "Customer can access their account or resolve technical problem",
            "priority": "high",
            "escalation_threshold": 0.6,
            "category": "technical"
        }
    else:
        goal = {
            "objective": "Provide general support",
            "success_criteria": "Customer receives helpful response",
            "priority": "low",
            "escalation_threshold": 0.6,
            "category": "general"
        }

    state["goal"] = goal
    logger.info(f"✅ Goal set: {goal['objective']}")
    return state

def retrieve_knowledge_from_rag(state: SupportAgentState) -> SupportAgentState:
    """Retrieve relevant knowledge from RAG system"""
    logger.info("🔍 Retrieving knowledge from RAG system...")

    # Placeholder for RAG retrieval
    # This will be implemented with Chroma + Sentence Transformers
    query = state["customer_query"]
    category = state["goal"]["category"]

    # Mock knowledge retrieval (will be replaced with real RAG)
    mock_knowledge = [
        {
            "id": "kb_001",
            "content": "Sample knowledge base entry",
            "confidence": 0.85,
            "category": category
        }
    ]

    state["retrieved_knowledge"] = mock_knowledge
    logger.info(f"✅ Retrieved {len(mock_knowledge)} knowledge items")
    return state

def generate_support_response(state: SupportAgentState) -> SupportAgentState:
    """Generate support response using retrieved knowledge"""
    logger.info("💬 Generating support response...")

    # Placeholder for LLM response generation
    query = state["customer_query"]
    knowledge = state["retrieved_knowledge"]

    # Mock response generation (will be replaced with LLM)
    if knowledge and knowledge[0]["confidence"] > 0.8:
        response = f"Based on your query '{query}', here's what I found: {knowledge[0]['content']}"
        confidence = 0.85
    else:
        response = f"I understand you're asking about '{query}'. Let me help you with that."
        confidence = 0.65

    state["agent_response"] = response
    state["confidence_score"] = confidence
    logger.info(f"✅ Response generated with confidence: {confidence}")
    return state

def assess_response_confidence(state: SupportAgentState) -> SupportAgentState:
    """Assess overall confidence and determine next steps"""
    logger.info("📊 Assessing response confidence...")

    confidence = state["confidence_score"]
    goal = state["goal"]
    escalation_threshold = goal["escalation_threshold"]

    if confidence >= 0.8:
        resolution_status = "high_confidence"
        escalation_reason = None
    elif confidence >= escalation_threshold:
        resolution_status = "medium_confidence"
        escalation_reason = None
    else:
        resolution_status = "low_confidence"
        escalation_reason = f"Confidence {confidence} below threshold {escalation_threshold}"

    state["resolution_status"] = resolution_status
    state["escalation_reason"] = escalation_reason
    logger.info(f"✅ Confidence assessment: {resolution_status}")
    return state

def generate_final_response(state: SupportAgentState) -> SupportAgentState:
    """Generate final customer-facing response"""
    logger.info("📝 Generating final customer response...")

    response = state["agent_response"]
    confidence = state["confidence_score"]
    escalation_reason = state["escalation_reason"]

    if escalation_reason:
        final_response = f"{response}\n\nI'm connecting you with a human agent to ensure you get the best help possible."
    elif confidence < 0.8:
        final_response = f"{response}\n\nIf this doesn't fully answer your question, please let me know and I can connect you with additional support."
    else:
        final_response = response

    state["final_response"] = final_response
    logger.info("✅ Final response generated")
    return state

def handle_escalation(state: SupportAgentState) -> SupportAgentState:
    """Handle escalation to human agents"""
    logger.info("🚨 Handling escalation to human agent...")

    # Create escalation package
    escalation_package = {
        "customer_id": state["customer_id"],
        "ticket_id": state["ticket_id"],
        "query": state["customer_query"],
        "attempted_response": state["agent_response"],
        "confidence_score": state["confidence_score"],
        "escalation_reason": state["escalation_reason"],
        "priority": state["goal"]["priority"],
        "retrieved_knowledge": state["retrieved_knowledge"]
    }

    # Update final response
    state["final_response"] = f"I'm connecting you with a human agent who can provide more detailed assistance. Your ticket #{state['ticket_id']} has been prioritized."
    state["resolution_status"] = "escalated"

    logger.info("✅ Escalation handled successfully")
    return state

def create_audit_trail(state: SupportAgentState) -> SupportAgentState:
    """Create audit trail for the support interaction"""
    logger.info("📋 Creating audit trail...")

    audit_entry = {
        "timestamp": "2024-01-01T00:00:00Z",  # Will use real timestamp
        "customer_id": state["customer_id"],
        "ticket_id": state["ticket_id"],
        "query": state["customer_query"],
        "goal": state["goal"],
        "confidence_score": state["confidence_score"],
        "resolution_status": state["resolution_status"],
        "escalation_reason": state["escalation_reason"],
        "final_response": state["final_response"]
    }

    if "audit_log" not in state:
        state["audit_log"] = []
    state["audit_log"].append(audit_entry)

    logger.info("✅ Audit trail created")
    return state

# Routing Functions
def route_based_on_confidence(state: SupportAgentState) -> str:
    """Route based on confidence assessment"""
    confidence = state["confidence_score"]
    escalation_threshold = state["goal"]["escalation_threshold"]

    if confidence < escalation_threshold:
        return "escalate"
    else:
        return "generate_response"

def route_based_on_escalation(state: SupportAgentState) -> str:
    """Route based on escalation decision"""
    escalation_reason = state["escalation_reason"]

    if escalation_reason:
        return "escalate"
    else:
        return "audit"

# Main Agent Creation Function
def create_customer_support_agent():
    """Create the Customer Support Resolution Agent workflow - MVP"""
    logger.info("🏗️ Building Customer Support Resolution Agent Workflow (MVP)...")

    # Create the workflow
    workflow = StateGraph(SupportAgentState)

    # Add nodes (core support flow)
    workflow.add_node("set_support_goal", set_support_goal_and_criteria)
    workflow.add_node("retrieve_knowledge", retrieve_knowledge_from_rag)
    workflow.add_node("generate_response", generate_support_response)
    workflow.add_node("assess_confidence", assess_response_confidence)
    workflow.add_node("create_final_response", generate_final_response)
    workflow.add_node("handle_escalation", handle_escalation)
    workflow.add_node("audit_trail", create_audit_trail)

    # Add edges (linear flow with conditional routing)
    workflow.add_edge("set_support_goal", "retrieve_knowledge")
    workflow.add_edge("retrieve_knowledge", "generate_response")
    workflow.add_edge("generate_response", "assess_confidence")

    # Conditional routing based on confidence
    workflow.add_conditional_edges(
        "assess_confidence",
        route_based_on_confidence,
        {
            "generate_response": "create_final_response",
            "escalate": "handle_escalation"
        }
    )

    # All paths lead to audit logging
    workflow.add_edge("create_final_response", "audit_trail")
    workflow.add_edge("handle_escalation", "audit_trail")
    workflow.add_edge("audit_trail", END)

    # Set entry point
    workflow.set_entry_point("set_support_goal")

    # Compile the workflow
    app = workflow.compile()

    logger.info("✅ Customer Support Resolution Agent workflow compiled successfully")
    return app

# Example usage
if __name__ == "__main__":
    # Create the agent
    support_agent = create_customer_support_agent()

    # Example state
    initial_state = {
        "customer_id": "cust_123",
        "ticket_id": "ticket_456",
        "customer_query": "Why is my bill higher this month?",
        "retrieved_knowledge": [],
        "agent_response": "",
        "confidence_score": 0.0,
        "escalation_reason": None,
        "final_response": "",
        "resolution_status": "",
        "audit_log": []
    }

    # Run the agent
    result = support_agent.invoke(initial_state)
    print(f"Final response: {result['final_response']}")
    print(f"Resolution status: {result['resolution_status']}")
````

````python
"""
RAG (Retrieval Augmented Generation) Module for Customer Support Agent
Uses Chroma + Sentence Transformers for knowledge retrieval
"""

import json
import logging
from typing import List, Dict, Any, Optional
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer
import numpy as np

logger = logging.getLogger(__name__)

class SupportRAGSystem:
    """RAG system for customer support knowledge retrieval"""

    def __init__(self,
                 collection_name: str = "support_knowledge",
                 embedding_model: str = "all-MiniLM-L6-v2",
                 persist_directory: str = "./chroma_db"):
        """
        Initialize the RAG system

        Args:
            collection_name: Name of the Chroma collection
            embedding_model: Sentence transformer model name
            persist_directory: Directory to persist Chroma database
        """
        self.collection_name = collection_name
        self.persist_directory = persist_directory

        # Initialize embedding model
        logger.info(f"Loading embedding model: {embedding_model}")
        self.embedding_model = SentenceTransformer(embedding_model)

        # Initialize Chroma client
        logger.info("Initializing Chroma client...")
        self.client = chromadb.PersistentClient(
            path=persist_directory,
            settings=Settings(
                anonymized_telemetry=False,
                allow_reset=True
            )
        )

        # Get or create collection
        try:
            self.collection = self.client.get_collection(name=collection_name)
            logger.info(f"Loaded existing collection: {collection_name}")
        except Exception:
            self.collection = self.client.create_collection(
                name=collection_name,
                metadata={"description": "Customer support knowledge base"}
            )
            logger.info(f"Created new collection: {collection_name}")

    def load_knowledge_base(self, kb_files: List[str]) -> None:
        """
        Load knowledge base from JSON files

        Args:
            kb_files: List of JSON file paths containing knowledge base entries
        """
        logger.info(f"Loading knowledge base from {len(kb_files)} files...")

        all_documents = []
        all_metadatas = []
        all_ids = []

        for kb_file in kb_files:
            try:
                with open(kb_file, 'r', encoding='utf-8') as f:
                    kb_data = json.load(f)

                logger.info(f"Processing {kb_file}: {len(kb_data)} entries")

                for entry in kb_data:
                    # Extract content for embedding
                    content = f"{entry.get('title', '')} {entry.get('content', '')}"

                    # Create metadata
                    metadata = {
                        "category": entry.get("category", "general"),
                        "confidence_level": entry.get("confidence_level", "medium"),
                        "title": entry.get("title", ""),
                        "tags": ",".join(entry.get("tags", []))
                    }

                    all_documents.append(content)
                    all_metadatas.append(metadata)
                    all_ids.append(entry.get("id", f"kb_{len(all_ids)}"))

            except Exception as e:
                logger.error(f"Error loading {kb_file}: {e}")
                continue

        if all_documents:
            # Add documents to collection
            self.collection.add(
                documents=all_documents,
                metadatas=all_metadatas,
                ids=all_ids
            )
            logger.info(f"✅ Loaded {len(all_documents)} knowledge base entries")
        else:
            logger.warning("No documents loaded from knowledge base files")

    def retrieve_knowledge(self,
                          query: str,
                          category: Optional[str] = None,
                          n_results: int = 5,
                          min_score: float = 0.3) -> List[Dict[str, Any]]:
        """
        Retrieve relevant knowledge for a query

        Args:
            query: Customer query
            category: Optional category filter (billing, technical, etc.)
            n_results: Number of results to return
            min_score: Minimum similarity score threshold

        Returns:
            List of relevant knowledge entries with scores
        """
        logger.info(f"Retrieving knowledge for query: '{query[:50]}...'")

        # Prepare query filters
        where_clause = {}
        if category:
            where_clause["category"] = category

        try:
            # Query the collection
            results = self.collection.query(
                query_texts=[query],
                n_results=n_results,
                where=where_clause if where_clause else None
            )

            # Process results
            retrieved_knowledge = []

            if results['documents'] and results['documents'][0]:
                for i, (doc, metadata, distance) in enumerate(zip(
                    results['documents'][0],
                    results['metadatas'][0],
                    results['distances'][0]
                )):
                    # Convert distance to similarity score (Chroma uses cosine distance)
                    similarity_score = 1 - distance

                    if similarity_score >= min_score:
                        knowledge_entry = {
                            "id": results['ids'][0][i],
                            "content": doc,
                            "metadata": metadata,
                            "similarity_score": similarity_score,
                            "category": metadata.get("category", "general"),
                            "confidence_level": metadata.get("confidence_level", "medium")
                        }
                        retrieved_knowledge.append(knowledge_entry)

            logger.info(f"✅ Retrieved {len(retrieved_knowledge)} relevant knowledge entries")
            return retrieved_knowledge

        except Exception as e:
            logger.error(f"Error retrieving knowledge: {e}")
            return []

    def calculate_retrieval_confidence(self,
                                     retrieved_knowledge: List[Dict[str, Any]],
                                     query: str) -> float:
        """
        Calculate overall confidence score for retrieved knowledge

        Args:
            retrieved_knowledge: List of retrieved knowledge entries
            query: Original customer query

        Returns:
            Overall confidence score (0.0 to 1.0)
        """
        if not retrieved_knowledge:
            return 0.0

        # Calculate weighted confidence based on similarity scores
        total_weight = 0.0
        weighted_score = 0.0

        for entry in retrieved_knowledge:
            similarity_score = entry.get("similarity_score", 0.0)
            confidence_level = entry.get("confidence_level", "medium")

            # Weight by confidence level
            confidence_weight = {
                "high": 1.0,
                "medium": 0.8,
                "low": 0.6
            }.get(confidence_level, 0.8)

            weight = similarity_score * confidence_weight
            weighted_score += similarity_score * weight
            total_weight += weight

        if total_weight > 0:
            overall_confidence = weighted_score / total_weight
        else:
            overall_confidence = 0.0

        logger.info(f"Retrieval confidence: {overall_confidence:.3f}")
        return overall_confidence

    def get_collection_stats(self) -> Dict[str, Any]:
        """Get statistics about the knowledge base collection"""
        try:
            count = self.collection.count()

            # Get sample of categories
            sample_results = self.collection.get(limit=100)
            categories = {}
            if sample_results['metadatas']:
                for metadata in sample_results['metadatas']:
                    cat = metadata.get('category', 'unknown')
                    categories[cat] = categories.get(cat, 0) + 1

            return {
                "total_entries": count,
                "categories": categories,
                "collection_name": self.collection_name
            }
        except Exception as e:
            logger.error(f"Error getting collection stats: {e}")
            return {"error": str(e)}

def create_support_rag_system() -> SupportRAGSystem:
    """Create and initialize the support RAG system"""
    logger.info("🏗️ Creating Support RAG System...")

    rag_system = SupportRAGSystem()

    # Load knowledge base files
    kb_files = [
        "validaion_test_docs/billing_kb.json",
        "validaion_test_docs/technical_kb.json"
    ]

    rag_system.load_knowledge_base(kb_files)

    # Print stats
    stats = rag_system.get_collection_stats()
    logger.info(f"📊 Knowledge base stats: {stats}")

    return rag_system

# Global RAG system instance
rag_system = None

def get_rag_system() -> SupportRAGSystem:
    """Get or create the global RAG system instance"""
    global rag_system
    if rag_system is None:
        rag_system = create_support_rag_system()
    return rag_system
````
