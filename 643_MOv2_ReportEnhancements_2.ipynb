````python
def build_headline_verdict(
    open_risks: List[Dict[str, Any]],
    segment_rollups: List[Dict[str, Any]],
    budget_actions: List[Dict[str, Any]],
) -> str:
    """One-line verdict at top of report: ATTENTION vs ON TRACK with counts."""
    n_risks = len(open_risks)
    stop_rollups = [r for r in segment_rollups if (r.get("recommendation") or "").lower() == "stop"]
    n_stops = len(stop_rollups)
    pending_budget = [b for b in budget_actions if (b.get("status") or "").lower() == "proposed"]
    n_pending = len(pending_budget)
    if n_risks > 0 or n_stops > 0 or n_pending > 0:
        parts = []
        if n_risks:
            parts.append(f"{n_risks} open risk{'s' if n_risks != 1 else ''}")
        if n_stops:
            parts.append(f"{n_stops} segment stop{'s' if n_stops != 1 else ''}")
        if n_pending:
            parts.append(f"{n_pending} reallocation{'s' if n_pending != 1 else ''} pending")
        return f"**Marketing health: ATTENTION** â€” {'; '.join(parts)}."
    return "**Marketing health: ON TRACK** â€” No open risks, no pending reallocations."


def build_next_steps(
    open_risks: List[Dict[str, Any]],
    budget_actions: List[Dict[str, Any]],
    segment_rollups: List[Dict[str, Any]],
    campaigns_lookup: Dict[str, Dict[str, Any]],
    segments_lookup: Dict[str, Dict[str, Any]],
    max_bullets: int = 5,
) -> str:
    """2â€“5 concrete next steps from risks, pending budget, and stop recommendations."""
    bullets: List[str] = []
    for r in open_risks:
        rid = r.get("risk_id", "?")
        sev = r.get("severity", "")
        action = r.get("recommended_action", "")
        bullets.append(f"- Review and act on **{rid}** ({sev} â€” {action}).")
    pending = [b for b in budget_actions if (b.get("status") or "").lower() == "proposed"]
    for b in pending:
        bid = b.get("budget_action_id", "?")
        camp_id = b.get("campaign_id", "")
        camp_name = (campaigns_lookup.get(camp_id) or {}).get("name", camp_id)
        reason = b.get("reason_code", "")
        bullets.append(f"- Decide on **{bid}** ({camp_name}: {reason}).")
    stop_recs = [r for r in segment_rollups if (r.get("recommendation") or "").lower() == "stop"]
    for r in stop_recs:
        seg_id = r.get("segment_id", "")
        camp_id = r.get("campaign_id", "")
        seg_name = (segments_lookup.get(seg_id) or {}).get("name", seg_id)
        camp_name = (campaigns_lookup.get(camp_id) or {}).get("name", camp_id)
        bullets.append(f"- Review stop recommendation for **{seg_name}** ({camp_name}).")
    chosen = bullets[:max_bullets]
    if not chosen:
        return "## Next steps\n\nNo immediate actions required.\n\n"
    lines = ["## Next steps", ""] + chosen + [""]
    return "\n".join(lines)
````
