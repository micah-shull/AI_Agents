````python
"""
Cost Calculation Utilities

Calculate costs for evaluation runs including API costs, compute costs, and storage costs.
"""

from typing import Dict, Any, List


def calculate_evaluation_cost(
    execution_time_seconds: float,
    api_calls_count: int = 1,
    cost_per_api_call_usd: float = 0.001,
    cost_per_second_compute_usd: float = 0.0001,
    storage_mb: float = 0.0,
    cost_per_mb_storage_usd: float = 0.01
) -> Dict[str, float]:
    """
    Calculate cost for a single evaluation.

    Args:
        execution_time_seconds: Time taken to execute evaluation
        api_calls_count: Number of API calls made
        cost_per_api_call_usd: Cost per API call
        cost_per_second_compute_usd: Cost per second of compute time
        storage_mb: Storage used in MB
        cost_per_mb_storage_usd: Cost per MB of storage

    Returns:
        Dictionary with cost breakdown
    """
    api_cost = api_calls_count * cost_per_api_call_usd
    compute_cost = execution_time_seconds * cost_per_second_compute_usd
    storage_cost = storage_mb * cost_per_mb_storage_usd
    total_cost = api_cost + compute_cost + storage_cost

    return {
        "api_cost_usd": round(api_cost, 4),
        "compute_cost_usd": round(compute_cost, 4),
        "storage_cost_usd": round(storage_cost, 4),
        "total_cost_usd": round(total_cost, 4)
    }


def calculate_total_costs(
    executed_evaluations: List[Dict[str, Any]],
    agent_performance_summary: Dict[str, Any],
    config: Any
) -> Dict[str, Any]:
    """
    Calculate total costs and cost breakdown for all evaluations.

    Args:
        executed_evaluations: List of executed evaluations
        agent_performance_summary: Performance summary per agent
        config: Orchestrator config with cost settings

    Returns:
        Dictionary with comprehensive cost analysis
    """
    if not config.enable_cost_tracking:
        return {
            "total_cost_usd": 0.0,
            "cost_per_evaluation": 0.0,
            "cost_per_scenario": 0.0,
            "cost_breakdown": {},
            "cost_by_agent": {},
            "cost_by_scenario": {}
        }

    total_cost = 0.0
    total_api_cost = 0.0
    total_compute_cost = 0.0
    total_storage_cost = 0.0

    cost_by_agent = {}
    cost_by_scenario = {}

    # Calculate costs for each evaluation
    for evaluation in executed_evaluations:
        execution_time = evaluation.get("execution_time_seconds", 0.0)
        scenario_id = evaluation.get("scenario_id", "unknown")
        agent_id = evaluation.get("target_agent_id", "unknown")

        # Estimate API calls (1 per evaluation for MVP)
        api_calls = 1

        eval_cost = calculate_evaluation_cost(
            execution_time_seconds=execution_time,
            api_calls_count=api_calls,
            cost_per_api_call_usd=config.cost_per_api_call_usd,
            cost_per_second_compute_usd=config.cost_per_second_compute_usd,
            cost_per_mb_storage_usd=config.cost_per_mb_storage_usd
        )

        total_cost += eval_cost["total_cost_usd"]
        total_api_cost += eval_cost["api_cost_usd"]
        total_compute_cost += eval_cost["compute_cost_usd"]
        total_storage_cost += eval_cost["storage_cost_usd"]

        # Track by agent
        if agent_id not in cost_by_agent:
            cost_by_agent[agent_id] = 0.0
        cost_by_agent[agent_id] += eval_cost["total_cost_usd"]

        # Track by scenario
        if scenario_id not in cost_by_scenario:
            cost_by_scenario[scenario_id] = 0.0
        cost_by_scenario[scenario_id] += eval_cost["total_cost_usd"]

    # Calculate averages
    total_evaluations = len(executed_evaluations)
    cost_per_evaluation = total_cost / total_evaluations if total_evaluations > 0 else 0.0

    # Get unique scenarios
    unique_scenarios = len(set(e.get("scenario_id") for e in executed_evaluations))
    cost_per_scenario = total_cost / unique_scenarios if unique_scenarios > 0 else 0.0

    return {
        "total_cost_usd": round(total_cost, 2),
        "cost_per_evaluation": round(cost_per_evaluation, 4),
        "cost_per_scenario": round(cost_per_scenario, 2),
        "cost_breakdown": {
            "api_costs_usd": round(total_api_cost, 2),
            "compute_costs_usd": round(total_compute_cost, 2),
            "storage_costs_usd": round(total_storage_cost, 2)
        },
        "cost_by_agent": {k: round(v, 2) for k, v in cost_by_agent.items()},
        "cost_by_scenario": {k: round(v, 2) for k, v in cost_by_scenario.items()}
    }
````
