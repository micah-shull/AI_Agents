# Agent State & Configuration

### Product–Customer Fit Discovery Orchestrator v2 (PCFDv2)

This section defines the **operating structure of the orchestrator**.

In agent systems, the state and configuration act as the **control system** for everything the agent does.

Rather than letting AI freely interpret data, this design establishes a **clear contract** for:

* what data the agent can access
* how discoveries are structured
* what thresholds determine decisions
* how results are reported

This approach ensures the system is **predictable, auditable, and trustworthy**.

---

# 🧠 The Role of Agent State

The `PCFDv2State` object acts as the **central memory of the orchestrator**.

Every step of the agent pipeline reads from and writes to this structure.

Think of it as the **shared intelligence layer** between nodes in the orchestration workflow.

### In practical terms it tracks:

• input configuration
• loaded datasets
• intermediate discovery signals
• final opportunity insights
• reporting outputs
• errors and runtime metadata

This allows each node in the pipeline to operate **independently but coherently**.

---

# 🔄 How Data Flows Through the State

The state structure reflects the **logical progression of the discovery engine**.

```text
Goal
  ↓
Plan
  ↓
Load Data
  ↓
Build Lookups
  ↓
Generate Intelligence Layers
  ↓
Detect Opportunities
  ↓
Produce Executive Report
```

Each stage writes its outputs back into the state so downstream nodes can reuse them.

This pattern improves:

* modularity
* transparency
* debugging
* reproducibility

# ⚙️ Configuration — Where Business Logic Lives

The `PCFDv2Config` class defines the **rules that govern the discovery engine**.

Instead of hard-coding logic into the agent, thresholds are defined here.

Examples:

```python

```

These thresholds determine:

* when a feature demand surge is significant
* when a bundle opportunity is actionable

````python
# ============================================================================
# Product–Customer Fit Discovery Orchestrator v2 (PCFDv2)
# ============================================================================
# Discovery engine: customer/product/feature/trend intelligence + strategic
# signals. Rules-first, thresholds in config. Data under agents/data/ (baseline,
# enrichment, trends, signals).

class PCFDv2State(TypedDict, total=False):
    """State for Product–Customer Fit Discovery Orchestrator v2"""

    # Input
    data_dir: Optional[str]
    reports_dir: Optional[str]

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Loaded data (baseline, enrichment, trends, signals)
    customers: List[Dict[str, Any]]
    product_catalog: List[Dict[str, Any]]
    transactions: List[Dict[str, Any]]
    customer_metrics: List[Dict[str, Any]]
    product_metrics: List[Dict[str, Any]]
    feature_usage: List[Dict[str, Any]]
    customer_journeys: List[Dict[str, Any]]
    market_signals: List[Dict[str, Any]]
    product_adoption_history: List[Dict[str, Any]]
    segment_growth_history: List[Dict[str, Any]]
    feature_demand_history: List[Dict[str, Any]]
    bundle_opportunity_signals: List[Dict[str, Any]]

    # Lookups (id -> record)
    customers_lookup: Dict[str, Dict[str, Any]]
    product_lookup: Dict[str, Dict[str, Any]]

    # Record counts (traceability)
    loader_counts: Dict[str, int]
    data_snapshot_loaded_at: Optional[str]
    validation_warnings: List[str]

    # Discovery rollups (rules-based)
    customer_intel: Dict[str, Any]       # high_value_segments, concentration, retention_risk
    product_intel: Dict[str, Any]        # high_margin_products, bundles, adoption
    feature_intel: Dict[str, Any]        # feature_demand_surges, gaps
    trend_intel: Dict[str, Any]          # emerging/declining products, segment growth
    strategic_signals: Dict[str, Any]    # cross_sell_opportunities, bundle opportunities

    # Output
    pcfd_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class PCFDv2Config:
    """Configuration for Product–Customer Fit Discovery Orchestrator v2"""
    reports_dir: str = "agents/output/PCFDv2_reports"
    data_dir: str = "agents/data"
    project_root: Optional[str] = None  # Set by run script or registry for path resolution

    # Subfolders under data_dir (per agents/data/README.md)
    baseline_dir: str = "baseline"
    enrichment_dir: str = "enrichment"
    trends_dir: str = "trends"
    signals_dir: str = "signals"

    # Thresholds (rules-first; configurable)
    adoption_growth_min_pct: float = 10.0       # Min % growth for "emerging" product
    decline_growth_max_pct: float = -5.0        # Max % growth for "declining" product
    high_margin_min_pct: float = 25.0          # Min profit_margin % for high-margin
    opportunity_score_min: float = 0.5         # Min opportunity_score for bundle signal
    segment_growth_min_pct: float = 15.0        # Min % for fast-expanding segment
    feature_demand_min_requests: int = 10      # Min requests for feature demand surge
    top_n_bundles: int = 10
    top_n_segments: int = 5
    top_n_products: int = 10
````
