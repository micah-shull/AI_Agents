````python
"""
Portfolio-level rollup for Mission Control v2.

Uses the latest mission_portfolio_snapshot (or latest by snapshot_date) to build
a single portfolio_rollup for reporting and executive triggers.
"""

from typing import Any, Dict, List


def build_portfolio_rollup(
    mission_portfolio_snapshot: List[Dict[str, Any]],
) -> Dict[str, Any]:
    """
    Build a single portfolio rollup from snapshot list (use latest by snapshot_date).

    Args:
        mission_portfolio_snapshot: List of snapshot dicts (snapshot_date, active_runs,
            completed_runs, missions_at_risk, avg_kpi_improvement_pct, top_bottleneck_*, etc.).

    Returns:
        Dict with: snapshot_date, active_runs, completed_runs, missions_at_risk,
        avg_kpi_improvement_pct, top_bottleneck_task_id, top_bottleneck_task_name,
        escalations_last_24h, human_interventions_last_24h. If empty list, returns
        defaults (zeros and empty strings).
    """
    if not mission_portfolio_snapshot:
        return {
            "snapshot_date": None,
            "active_runs": 0,
            "completed_runs": 0,
            "missions_at_risk": 0,
            "avg_kpi_improvement_pct": 0.0,
            "top_bottleneck_task_id": "",
            "top_bottleneck_task_name": "",
            "escalations_last_24h": 0,
            "human_interventions_last_24h": 0,
        }

    latest = max(
        mission_portfolio_snapshot,
        key=lambda s: s.get("snapshot_date") or "",
    )
    return {
        "snapshot_date": latest.get("snapshot_date"),
        "active_runs": latest.get("active_runs", 0),
        "completed_runs": latest.get("completed_runs", 0),
        "missions_at_risk": latest.get("missions_at_risk", 0),
        "avg_kpi_improvement_pct": latest.get("avg_kpi_improvement_pct", 0),
        "top_bottleneck_task_id": latest.get("top_bottleneck_task_id", ""),
        "top_bottleneck_task_name": latest.get("top_bottleneck_task_name", ""),
        "escalations_last_24h": latest.get("escalations_last_24h", 0),
        "human_interventions_last_24h": latest.get("human_interventions_last_24h", 0),
    }
````
