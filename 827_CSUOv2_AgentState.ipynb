````python
# ============================================================================
# Cross-Sell & Upsell Orchestrator v2
# ============================================================================
# v2 adds: offers + offer_responses, trend/performance rollup, traceability.

class CrossSellUpsellState(TypedDict, total=False):
    """State for Cross-Sell & Upsell Orchestrator Agent v2"""

    # Input
    customer_id: str

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Data Ingestion
    customer_data: Dict[str, Any]
    product_catalog: List[Dict[str, Any]]
    product_lookup: Dict[str, Dict[str, Any]]
    offers: List[Dict[str, Any]]
    offer_responses: List[Dict[str, Any]]
    response_lookup: Dict[str, Dict[str, Any]]  # offer_id -> response dict
    # Loader record counts (traceability)
    customers_count: int
    products_count: int
    offers_count: int
    responses_count: int
    data_snapshot_loaded_at: Optional[str]

    # Routine Analysis
    customer_products: List[str]
    customer_categories: List[str]
    routine_gaps: List[str]
    replenishment_needs: List[Dict[str, Any]]

    # Opportunity Detection
    cross_sell_opportunities: List[Dict[str, Any]]
    upsell_opportunities: List[Dict[str, Any]]

    # Scoring & Ranking
    scored_opportunities: List[Dict[str, Any]]
    ranked_opportunities: List[Dict[str, Any]]
    top_opportunities: List[Dict[str, Any]]
    opportunity_summary: Dict[str, Any]

    # v2: Offer performance / trend
    offer_performance_rollup: Dict[str, Any]
    # e.g. by_offer_type: { cross_sell: { sent, accepted, revenue }, ... }, by_channel: { email: {...} }

    # Output
    recommendations_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class CrossSellUpsellConfig:
    """Configuration for Cross-Sell & Upsell Orchestrator Agent v2"""
    llm_model: str = os.getenv("LLM_MODEL", "gpt-4o-mini")
    temperature: float = 0.3
    reports_dir: str = "agents/output/csuo_v2_reports"
    data_dir: str = "agents/data"

    top_n_opportunities: int = 5
    high_value_score_threshold: float = 15.0
    replenishment_warning_days: int = 7

    # Data file names (under data_dir)
    customers_file: str = "customers.json"
    product_catalog_file: str = "product_catalog.json"
    offers_file: str = "offers.json"
    offer_responses_file: str = "offer_responses.json"
````
