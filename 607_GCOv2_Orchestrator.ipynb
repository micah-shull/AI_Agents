````python
"""
GCO v2 orchestrator: LangGraph linear workflow (goal -> plan -> load -> policy_eval -> risk -> trend -> portfolio -> prioritize -> report -> END).
"""

from typing import Any, Dict, Optional

from config import (
    GovernanceComplianceOrchestratorConfigV2,
    GovernanceComplianceOrchestratorStateV2,
)

from agents.governance_compliance_orchestrator_v2.orchestrator import nodes

try:
    from langgraph.graph import END, StateGraph
    _LANGGRAPH_AVAILABLE = True
except ImportError:
    _LANGGRAPH_AVAILABLE = False


def build_gco_v2_graph(
    config: GovernanceComplianceOrchestratorConfigV2,
    project_root: str,
):
    """
    Build the GCO v2 LangGraph workflow with linear edges.

    Nodes that need config or project_root are wrapped so the graph only passes state.
    Requires langgraph to be installed (see requirements/base.txt).
    """
    if not _LANGGRAPH_AVAILABLE:
        raise RuntimeError("LangGraph is required. Install with: pip install -r requirements/base.txt")

    cfg = config
    proot = project_root

    def load_node(state: GovernanceComplianceOrchestratorStateV2) -> Dict[str, Any]:
        return nodes.data_loading_node(state, cfg, proot)

    def risk_node(state: GovernanceComplianceOrchestratorStateV2) -> Dict[str, Any]:
        return nodes.risk_node(state, cfg)

    def portfolio_node(state: GovernanceComplianceOrchestratorStateV2) -> Dict[str, Any]:
        return nodes.portfolio_node(state, cfg)

    def prioritize_node(state: GovernanceComplianceOrchestratorStateV2) -> Dict[str, Any]:
        return nodes.prioritize_node(state, cfg)

    def report_node(state: GovernanceComplianceOrchestratorStateV2) -> Dict[str, Any]:
        return nodes.report_node(state, cfg, proot)

    workflow = StateGraph(GovernanceComplianceOrchestratorStateV2)

    workflow.add_node("goal", nodes.goal_node)
    workflow.add_node("plan", nodes.plan_node)
    workflow.add_node("load_data", load_node)
    workflow.add_node("policy_eval", nodes.policy_eval_node)
    workflow.add_node("risk", risk_node)
    workflow.add_node("trend", nodes.trend_node)
    workflow.add_node("portfolio", portfolio_node)
    workflow.add_node("prioritize", prioritize_node)
    workflow.add_node("report", report_node)

    workflow.set_entry_point("goal")
    workflow.add_edge("goal", "plan")
    workflow.add_edge("plan", "load_data")
    workflow.add_edge("load_data", "policy_eval")
    workflow.add_edge("policy_eval", "risk")
    workflow.add_edge("risk", "trend")
    workflow.add_edge("trend", "portfolio")
    workflow.add_edge("portfolio", "prioritize")
    workflow.add_edge("prioritize", "report")
    workflow.add_edge("report", END)

    return workflow.compile()


def _run_linear_fallback(
    project_root: str,
    config: GovernanceComplianceOrchestratorConfigV2,
) -> Dict[str, Any]:
    """Run without LangGraph (same order as graph). Use when LangGraph is not installed."""
    state: GovernanceComplianceOrchestratorStateV2 = {"errors": [], "processing_time": 0.0}
    state = {**state, **nodes.goal_node(state)}
    state = {**state, **nodes.plan_node(state)}
    state = {**state, **nodes.data_loading_node(state, config, project_root)}
    if state.get("errors"):
        return state
    state = {**state, **nodes.policy_eval_node(state)}
    state = {**state, **nodes.risk_node(state, config)}
    state = {**state, **nodes.trend_node(state)}
    state = {**state, **nodes.portfolio_node(state, config)}
    state = {**state, **nodes.prioritize_node(state, config)}
    state = {**state, **nodes.report_node(state, config, project_root)}
    return state


def run_governance_compliance_v2(
    project_root: str,
    config: Optional[GovernanceComplianceOrchestratorConfigV2] = None,
) -> Dict[str, Any]:
    """
    Run the Governance & Compliance Orchestrator v2 via LangGraph.

    Steps: goal -> plan -> load_data -> policy_eval -> risk -> trend -> portfolio -> prioritize -> report -> END.
    Returns final state dict (including audit_report, report_file_path, errors, processing_time).
    Falls back to linear execution if LangGraph is not installed.
    """
    cfg = config or GovernanceComplianceOrchestratorConfigV2()

    if _LANGGRAPH_AVAILABLE:
        app = build_gco_v2_graph(cfg, project_root)
        initial_state: GovernanceComplianceOrchestratorStateV2 = {"errors": [], "processing_time": 0.0}
        return app.invoke(initial_state)

    return _run_linear_fallback(project_root, cfg)
````
