# Executive Reporting Layer

This module generates the **final executive intelligence artifact** produced by the orchestrator.

All previous stages in the system perform data ingestion and signal generation:

* telemetry loading
* portfolio rollups
* risk aggregation
* trigger detection

This layer transforms those signals into a **decision-ready report**.

The result is a structured document designed to answer the questions leaders actually ask:

* Is the experimentation program healthy?
* Where is risk emerging?
* Are we learning fast enough?
* What action is required?

The report is deliberately structured as a **board-style one-pager followed by supporting analysis**.

---

# 1. The “So What” Signal

```python
_so_what_line()
```

This function produces the **single most important sentence in the report**.

Executives often skim documents looking for the answer to one question:

> “What is happening, and do I need to act?”

This function delivers that answer.

Example output:

```
Portfolio status: healthy. 1 experiment at risk; 2 open risk events. No critical triggers; continue monitoring.
```

Or, if escalation is required:

```
Portfolio status: at_risk. 3 experiments at risk; 5 open risk events. Executive attention required.
```

This design forces the report to **lead with the conclusion**, not bury it.

Many analytics systems do the opposite — they lead with charts and force readers to interpret results themselves.

---

# 2. Action-Oriented Next Steps

```python
_next_steps()
```

This function converts signals into **clear actions**.

Two scenarios exist.

### When triggers fire

The system produces a **single executive ask**:

```
One ask: Review triggered items and assign owners.
```

This is followed by specific issues.

Example:

```
Experiments at risk (3) above critical threshold (2)
Learning velocity below target
```

This approach aligns perfectly with your **executive reporting philosophy**:

> One ask → clear accountability → measurable follow-up.

---

### When triggers do not fire

The system shifts into monitoring mode.

Example:

```
Continue routine portfolio reviews; monitor learning velocity and risk events.
```

This prevents unnecessary alarm while still keeping leadership informed.

---

# 3. Trend Indicators

```python
_trend_str()
```

This helper function provides **directional context** for portfolio signals.

Example output:

```
Experiments at risk: 3 (↑1 from prior)
Learning velocity: 4 (↓2 from prior)
```

Directional signals are extremely important.

Executives care less about static values and more about **movement**:

* Is risk increasing?
* Are experiments improving?
* Is learning accelerating?

This small helper dramatically improves report readability.

build_experiment_report()

The structure follows a **deliberate information hierarchy**.

# 6. Portfolio Health Section

The next section expands into the **operational state of the experimentation program**.

It answers questions such as:

* How many experiments are running?
* How many have completed?
* How many are planned?
* Are we scaling results or stopping experiments?

This provides a quick sense of **experimentation discipline**.

Healthy programs typically show a mix of:

```
scale decisions
iterate decisions

```

If “stop” decisions are missing, the system may be detecting **confirmation bias or weak governance**.

Most experimentation tools ignore governance entirely — this system makes it a first-class signal.

````python
"""Executive report builder for Experimentation Portfolio Orchestrator v3."""

from typing import Any, Dict, List, Optional


def _so_what_line(
    portfolio_rollup: Dict[str, Any],
    risk_rollup: Dict[str, Any],
    triggers: List[Dict[str, Any]],
) -> str:
    """One sentence: portfolio status, risk, and what needs attention."""
    status = portfolio_rollup.get("portfolio_status", "unknown")
    at_risk = portfolio_rollup.get("experiments_at_risk", 0)
    open_risk = risk_rollup.get("open_count", 0)
    if triggers:
        return (
            f"Portfolio status: {status}. {at_risk} experiment(s) at risk; {open_risk} open risk event(s). "
            "Executive attention required — see Next steps."
        )
    return (
        f"Portfolio status: {status}. {at_risk} experiment(s) at risk; {open_risk} open risk event(s). "
        "No critical triggers; continue monitoring."
    )


def _next_steps(
    triggers: List[Dict[str, Any]],
    portfolio_rollup: Dict[str, Any],
) -> List[str]:
    """Concrete next steps; one ask at top when triggers fire."""
    steps: List[str] = []
    if triggers:
        steps.append("**One ask:** Review triggered items below and assign owners for risk events and at-risk experiments.")
        for t in triggers:
            steps.append(f"- {t.get('message', '')}")
    else:
        steps.append("Continue routine portfolio reviews; monitor learning velocity and risk events.")
        if portfolio_rollup.get("top_risk_type") and portfolio_rollup.get("top_risk_type") != "none":
            steps.append(f"Watch top risk type: {portfolio_rollup.get('top_risk_type')}.")
    return steps


def _trend_str(delta: Optional[int], label: str = "") -> str:
    if delta is None:
        return ""
    if delta > 0:
        return f" (↑{delta} from prior)"
    if delta < 0:
        return f" (↓{abs(delta)} from prior)"
    return " (no change)"


def build_experiment_report(
    portfolio_rollup: Dict[str, Any],
    risk_rollup: Dict[str, Any],
    executive_triggers: List[Dict[str, Any]],
    *,
    runs_count: int = 0,
    snapshots_count: int = 0,
    risk_events_count: int = 0,
    task_logs_count: int = 0,
    data_snapshot_loaded_at: Optional[str] = None,
    validation_warnings: Optional[List[str]] = None,
    target_portfolio_status: str = "healthy",
) -> str:
    """
    Build the full markdown report: one-view, targets, portfolio health,
    risk summary, recent risk events, next steps, traceability.
    """
    so_what = _so_what_line(portfolio_rollup, risk_rollup, executive_triggers)
    next_steps_list = _next_steps(executive_triggers, portfolio_rollup)

    status = portfolio_rollup.get("portfolio_status", "unknown")
    on_target = "On target" if status == target_portfolio_status else "Attention"
    delta_at_risk = portfolio_rollup.get("delta_experiments_at_risk")
    delta_vel = portfolio_rollup.get("delta_learning_velocity")

    # Board one-pager table
    lines: List[str] = [
        "# Experimentation Portfolio — Executive Report",
        "",
        "## One view",
        "",
        f"**Portfolio status:** {status}  \n**Verdict vs target:** {on_target} (target: {target_portfolio_status})",
        "",
        so_what,
        "",
        "**One ask:** " + (next_steps_list[0] if next_steps_list else "Continue monitoring."),
        "",
        "| Metric | Value |",
        "|--------|-------|",
        f"| Portfolio status | {status} |",
        f"| Experiments at risk | {portfolio_rollup.get('experiments_at_risk', 0)}{_trend_str(delta_at_risk)} |",
        f"| High-risk experiments | {portfolio_rollup.get('high_risk_experiments', 0)} |",
        f"| Open risk events | {risk_rollup.get('open_count', 0)} |",
        f"| Learning velocity (last 30d) | {portfolio_rollup.get('learning_velocity_last_30d', 0)}{_trend_str(delta_vel)} |",
        f"| Estimated value in flight (USD) | {portfolio_rollup.get('estimated_annual_value_in_flight_usd', 0):,} |",
        "",
        "---",
        "",
        "## Portfolio health (latest snapshot)",
        "",
        "| Metric | Value | Trend |",
        "|--------|-------|-------|",
        f"| Active experiments | {portfolio_rollup.get('active_experiments', 0)} | |",
        f"| Completed experiments | {portfolio_rollup.get('completed_experiments', 0)} | |",
        f"| Planned experiments | {portfolio_rollup.get('planned_experiments', 0)} | |",
        f"| Experiments at risk | {portfolio_rollup.get('experiments_at_risk', 0)} |{_trend_str(delta_at_risk)} |",
        f"| High-risk experiments | {portfolio_rollup.get('high_risk_experiments', 0)} | |",
        f"| Scale decisions (YTD) | {portfolio_rollup.get('scale_decisions_ytd', 0)} | |",
        f"| Iterate decisions (YTD) | {portfolio_rollup.get('iterate_decisions_ytd', 0)} | |",
        f"| Stop decisions (YTD) | {portfolio_rollup.get('stop_decisions_ytd', 0)} | |",
        f"| Avg primary metric lift % | {portfolio_rollup.get('avg_primary_metric_lift_pct', 0):.1f}% | |",
        f"| Human reviews (last 30d) | {portfolio_rollup.get('human_reviews_last_30d', 0)} | |",
        "",
        "## Risk & governance",
        "",
        f"**Open risk events:** {risk_rollup.get('open_count', 0)}  \n"
        f"**By severity:** {risk_rollup.get('by_severity', {})}  \n"
        f"**By risk type:** {risk_rollup.get('by_risk_type', {})}  \n"
        f"**Top risk type (snapshot):** {portfolio_rollup.get('top_risk_type', 'none')}",
        "",
    ]

    # Recent risk events (top 10 by severity order)
    severity_order = {"critical": 0, "high": 1, "medium": 2, "low": 3}
    events = risk_rollup.get("events", [])
    sorted_events = sorted(
        events,
        key=lambda e: (severity_order.get((e.get("severity") or "").lower(), 99), e.get("detected_at") or ""),
        reverse=False,
    )
    if sorted_events:
        lines.append("### Recent risk events")
        lines.append("")
        lines.append("| Event | Experiment | Type | Severity | Status |")
        lines.append("|-------|------------|------|----------|--------|")
        for e in sorted_events[:10]:
            lines.append(
                f"| {e.get('event_id', '')} | {e.get('experiment_id', '')} | "
                f"{e.get('risk_type', '')} | {e.get('severity', '')} | {e.get('status', '')} |"
            )
        lines.append("")

    # Executive triggers
    if executive_triggers:
        lines.append("## Executive triggers (attention required)")
        lines.append("")
        for t in executive_triggers:
            lines.append(f"- **{t.get('trigger_type', '')}:** {t.get('message', '')}")
        lines.append("")

    # Next steps
    lines.append("## Next steps")
    lines.append("")
    for s in next_steps_list:
        lines.append(s)
    lines.append("")

    # Traceability
    lines.append("---")
    lines.append("")
    lines.append("## Traceability")
    lines.append("")
    lines.append(
        f"Data: {runs_count} runs, {snapshots_count} snapshots, "
        f"{risk_events_count} risk events, {task_logs_count} task execution logs."
    )
    if data_snapshot_loaded_at:
        lines.append(f"  \nSnapshot loaded at: {data_snapshot_loaded_at} (UTC).")
    if validation_warnings:
        lines.append("  \n**Data quality warnings:**")
        for w in validation_warnings:
            lines.append(f"- {w}")
    lines.append("")
    return "\n".join(lines)
````
