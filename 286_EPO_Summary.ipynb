The **Experimentation Portfolio Orchestrator** is an AI system that manages the full lifecycle of experiments across an organization. It transforms experimentation from a scattered, manual effort into a **disciplined, automated, organization-wide capability**.

### Core Capabilities

1. **Portfolio Analysis** - Analyzes all experiments in your portfolio, identifying what's working, what needs attention, and what's missing
2. **Automated Analysis Calculation** - Calculates experiment metrics (lift, confidence, direction) from raw data when analysis is missing
3. **Decision Generation** - Generates recommendations (scale, iterate, retire, do_not_start) based on configurable business rules
4. **Portfolio Insights** - Identifies patterns, risks, and opportunities across experiments
5. **Comprehensive Reporting** - Generates detailed markdown reports with experiment details, analysis, decisions, and insights

### Workflow

```
Goal â†’ Planning â†’ Data Loading â†’ Portfolio Analysis â†’ Decision Generation â†’ Report Generation
```

### **1. MVP-First Approach** â­ CRITICAL LEARNING

**What It Is:**
- Built **rule-based MVP first** (no LLM dependencies)

**Why This Matters:**
- âœ… **Fast to build** - Deterministic logic, no API dependencies

- âœ… Clear separation makes debugging easier

- `nodes.py` - Thin orchestration nodes

### **5. Utilities Before Nodes Pattern** â­ CRITICAL LEARNING

**What It Is:**
- Build utilities first, test them independently
- Then build nodes that orchestrate utilities
- Nodes become simple (just orchestration)

**Why This Matters:**
- âœ… Catch data structure issues early
- âœ… Verify business logic before orchestration
- âœ… Nodes become simpler to test
- âœ… Easier debugging

**Example:**

```python
# Step 1: Build and test utility
def test_calculate_experiment_analysis():
    analysis = calculate_experiment_analysis(exp_id, definition, metrics)
    assert analysis["relative_lift_percent"] == 44.4

# Step 2: THEN build node (which uses utility)
def portfolio_analysis_node(state):
    analysis = calculate_experiment_analysis(...)  # Uses tested utility
    return {"calculated_analyses": [analysis]}
```

**Key Files:**
- `utilities/` - All utilities tested independently
- `nodes.py` - Nodes orchestrate tested utilities

**Learning Point:** Test utilities independently. Nodes become trivial.

### **6. Progressive State Enrichment** â­ IMPORTANT LEARNING

**What It Is:**
- State starts simple and gets richer as it flows through nodes
- Each node adds new dimensions to state
- Enables pattern detection and insights

**Why This Matters:**
- âœ… Clear data flow
- âœ… Each node has clear responsibility
- âœ… State becomes richer, enabling insights
- âœ… Easy to reason about what each node produces

**State Flow:**

```python
# Node 1: Goal
state = {"experiment_id": None, "goal": {...}}

# Node 2: Planning
state = {..., "plan": [...]}

# Node 3: Data Loading
state = {..., "portfolio": [...], "portfolio_lookup": {...}}

# Node 4: Portfolio Analysis
state = {..., "analyzed_experiments": [...], "portfolio_summary": {...}}

# Node 5: Decision Generation
state = {..., "generated_decisions": [...]}

# Node 6: Report Generation
state = {..., "portfolio_report": "...", "report_file_path": "..."}
```

### **7. Error Handling & Validation** â­ IMPORTANT LEARNING

**What It Is:**
- Each node validates required inputs
- Errors collected in state
- Graceful degradation (continue processing other experiments if one fails)

**Why This Matters:**
- âœ… Robust to bad data
- âœ… Clear error messages
- âœ… Partial success possible
- âœ… Easier debugging

**Example:**

```python
def portfolio_analysis_node(state):
    errors = state.get("errors", [])
    portfolio = state.get("portfolio", [])
    
    if not portfolio:
        return {"errors": errors + ["portfolio_analysis_node: portfolio data required"]}
    
    try:
        # Process portfolio
        return {"analyzed_experiments": [...], "errors": errors}
    except Exception as e:
        return {"errors": errors + [f"portfolio_analysis_node: {str(e)}"]}
```

**Key Pattern:** Always validate inputs, collect errors, handle exceptions gracefully.

**Learning Point:** Error handling makes agents production-ready.

### **8. Configurable Business Rules** â­ IMPORTANT LEARNING

**What It Is:**
- Decision thresholds in config (not hardcoded)
- Easy to adjust for different business contexts
- Rule-based logic (deterministic, testable)

**Why This Matters:**
- âœ… Adaptable to different companies/contexts
- âœ… Easy to tune (change thresholds, not code)
- âœ… Deterministic (same input = same output)
- âœ… Testable (can test edge cases)

**Example:**

```python
@dataclass
class ExperimentationPortfolioOrchestratorConfig:
    scale_threshold_lift: float = 20.0  # Configurable!
    iterate_threshold_lift: float = 5.0
    retire_threshold_lift: float = -10.0
```

**Key File:**
- `config.py` - `ExperimentationPortfolioOrchestratorConfig` with all thresholds

**Learning Point:** Make business rules configurable. Hardcoded rules are technical debt.
