````python
# ============================================================================
# B2B Sales Orchestrator v2
# ============================================================================
# Deterministic pipeline intelligence: weighted forecast, risk velocity,
# rep performance trajectory, executive triggers. Data in agents/data/.

class B2BSalesV2State(TypedDict, total=False):
    """State for B2B Sales Orchestrator v2 (pipeline + risk velocity + rep performance)"""

    # Input
    data_dir: Optional[str]
    reports_dir: Optional[str]

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Raw data (from agents/data/)
    deals_snapshot: List[Dict[str, Any]]   # Current snapshot: deal_stage, probability, risk_score_*, trigger_flag_*, etc.
    deals_history: List[Dict[str, Any]]   # Time series: deal_id, period, stage, probability, engagement_score, risk_score
    sales_reps: List[Dict[str, Any]]      # rep_id, quota_3m, closed_won_3m_*, pipeline_value, risk_*

    # Lookups
    deals_by_id: Dict[str, Dict[str, Any]]   # deal_id -> deal
    reps_by_id: Dict[str, Dict[str, Any]]   # rep_id -> rep

    # Pipeline intelligence (deterministic)
    weighted_forecast: float
    stalled_deals: List[Dict[str, Any]]      # Deals stuck (e.g. days_in_stage > threshold)
    stage_bottlenecks: List[Dict[str, Any]] # Stage -> count, total_value, avg_days
    deal_momentum: List[Dict[str, Any]]     # Per-deal or per-stage momentum signals

    # Risk velocity
    deals_risk_rising: List[Dict[str, Any]]  # risk_velocity_flag == "increasing"
    trigger_summary: Dict[str, Any]         # new / persistent / resolved counts
    executive_triggers: List[Dict[str, Any]]

    # Rep performance
    rep_performance: List[Dict[str, Any]]    # quota_attainment, pipeline_trend, coaching_candidate, benchmark

    # Rollups (for report and triggers)
    pipeline_rollup: Dict[str, Any]         # total_forecast, deal_count, by_stage, etc.
    risk_rollup: Dict[str, Any]             # risk_velocity counts, trigger counts
    data_snapshot_loaded_at: Optional[str]
    validation_warnings: List[str]

    # Output
    b2b_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class B2BSalesV2Config:
    """Configuration for B2B Sales Orchestrator v2"""
    reports_dir: str = "output/b2b_sales_v2_reports"
    data_dir: str = "agents/data"
    deals_snapshot_file: str = "deals_snapshot.json"
    deals_history_file: str = "deals_history.json"
    sales_reps_file: str = "sales_reps.json"

    # Pipeline
    stalled_days_threshold: int = 45          # Days in stage to flag as stalled
    stage_order: List[str] = field(default_factory=lambda: [
        "Qualification", "Discovery", "Proposal", "Negotiation", "Closed Won", "Closed Lost"
    ])

    # Risk / triggers (executive triggers when exceeded)
    risk_velocity_critical_count: int = 3     # Trigger if N+ deals with rising risk
    trigger_persistent_count: int = 2         # Trigger if N+ deals with persistent trigger_flag
    weighted_forecast_target: Optional[float] = None  # Optional target for "on track" comparison
````
