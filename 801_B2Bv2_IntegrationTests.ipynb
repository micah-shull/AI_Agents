````python
"""
Integration tests for B2B Sales v2: full graph invoke.
Skip when agents/data is missing so CI without data still passes.
Run from project root: python -m pytest test_b2b_sales_v2_integration.py -v --tb=short
"""
import sys
from pathlib import Path

root = Path(__file__).resolve().parent
if str(root) not in sys.path:
    sys.path.insert(0, str(root))

import pytest

from config import B2BSalesV2Config
from agents.b2b_sales_v2.orchestrator.orchestrator import create_b2b_sales_v2_orchestrator


DATA_DIR = root / "agents" / "data"
HAS_DATA = (
    DATA_DIR.is_dir()
    and (DATA_DIR / "deals_snapshot.json").exists()
    and (DATA_DIR / "deals_history.json").exists()
    and (DATA_DIR / "sales_reps.json").exists()
)


@pytest.mark.skipif(not HAS_DATA, reason="agents/data not present or missing required JSON files")
def test_b2b_sales_v2_full_run_no_errors():
    """Full graph invoke: no errors, report path set, file exists."""
    config = B2BSalesV2Config()
    config.data_dir = str(DATA_DIR)
    config.reports_dir = str(root / "output" / "b2b_sales_v2_reports")
    orchestrator = create_b2b_sales_v2_orchestrator(config)
    initial_state = {
        "data_dir": config.data_dir,
        "reports_dir": config.reports_dir,
        "errors": [],
    }
    result = orchestrator.invoke(initial_state)
    assert result.get("errors") == [], f"Unexpected errors: {result.get('errors')}"
    assert result.get("report_file_path"), "report_file_path should be set"
    report_path = Path(result["report_file_path"])
    assert report_path.exists(), f"Report file should exist: {report_path}"


@pytest.mark.skipif(not HAS_DATA, reason="agents/data not present or missing required JSON files")
def test_b2b_sales_v2_report_has_key_sections():
    """Report content contains required sections and domain phrases."""
    config = B2BSalesV2Config()
    config.data_dir = str(DATA_DIR)
    config.reports_dir = str(root / "output" / "b2b_sales_v2_reports")
    orchestrator = create_b2b_sales_v2_orchestrator(config)
    initial_state = {
        "data_dir": config.data_dir,
        "reports_dir": config.reports_dir,
        "errors": [],
    }
    result = orchestrator.invoke(initial_state)
    assert result.get("errors") == []
    content = result.get("b2b_report") or ""
    assert "Board One-Pager" in content
    assert "One-View" in content
    assert "Pipeline Summary" in content
    assert "Risk Velocity" in content
    assert "Rep Performance" in content
    assert "Next Steps" in content
    assert "Verdict" in content or "ON TRACK" in content or "ATTENTION" in content
    assert "Weighted forecast" in content or "forecast" in content.lower()
````
