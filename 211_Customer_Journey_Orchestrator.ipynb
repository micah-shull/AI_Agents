# data_loader

````python
"""Data loading utilities for Customer Journey Orchestrator"""

import json
from pathlib import Path
from typing import Dict, List, Any, Optional
from datetime import datetime


def load_json_file(file_path: str) -> Any:
    """
    Load a JSON file and return its contents.

    Args:
        file_path: Path to the JSON file

    Returns:
        Parsed JSON data

    Raises:
        FileNotFoundError: If file doesn't exist
        json.JSONDecodeError: If file is not valid JSON
    """
    path = Path(file_path)
    if not path.exists():
        raise FileNotFoundError(f"File not found: {file_path}")

    with open(path, 'r', encoding='utf-8') as f:
        return json.load(f)


def load_customers_data(file_path: str) -> Dict[str, Dict[str, Any]]:
    """
    Load customers data and return as a dictionary indexed by customer_id.

    Args:
        file_path: Path to customers.json

    Returns:
        Dictionary mapping customer_id to customer data
    """
    customers_list = load_json_file(file_path)
    return {customer["customer_id"]: customer for customer in customers_list}


def load_orders_data(file_path: str) -> Dict[str, Dict[str, Any]]:
    """
    Load orders data and return as a dictionary indexed by order_id.

    Args:
        file_path: Path to orders.json

    Returns:
        Dictionary mapping order_id to order data
    """
    orders_list = load_json_file(file_path)
    return {order["order_id"]: order for order in orders_list}


def load_logistics_data(file_path: str) -> Dict[str, Dict[str, Any]]:
    """
    Load logistics data from the nested carrier structure.

    Args:
        file_path: Path to logistics_api.json

    Returns:
        Dictionary mapping order_id to logistics data
    """
    logistics_by_carrier = load_json_file(file_path)
    logistics_by_order = {}

    # Flatten the nested structure: {carrier: {order_id: data}} -> {order_id: data}
    for carrier, orders in logistics_by_carrier.items():
        for order_id, logistics_info in orders.items():
            logistics_by_order[order_id] = logistics_info

    return logistics_by_order


def load_marketing_signals_data(file_path: str) -> Dict[str, Dict[str, Any]]:
    """
    Load marketing signals data and return as a dictionary indexed by customer_id.

    Args:
        file_path: Path to marketing_signals.json

    Returns:
        Dictionary mapping customer_id to marketing signals data
    """
    signals_list = load_json_file(file_path)
    return {signal["customer_id"]: signal for signal in signals_list}


def extract_logistics_by_order_id(
    logistics_data: Dict[str, Dict[str, Any]],
    order_id: str
) -> Optional[Dict[str, Any]]:
    """
    Extract logistics data for a specific order_id.

    Args:
        logistics_data: Dictionary of logistics data (from load_logistics_data)
        order_id: Order ID to look up

    Returns:
        Logistics data for the order, or None if not found
    """
    return logistics_data.get(order_id)


def merge_customer_and_marketing_data(
    customer_data: Dict[str, Any],
    marketing_data: Optional[Dict[str, Any]]
) -> Dict[str, Any]:
    """
    Merge customer data with marketing signals data.

    This creates a unified customer object with all relevant information
    needed for decision-making.

    Args:
        customer_data: Customer data from customers.json
        marketing_data: Marketing signals data (can be None if not found)

    Returns:
        Merged customer data dictionary
    """
    merged = customer_data.copy()

    if marketing_data:
        # Add marketing fields to customer data
        merged["email_engagement"] = marketing_data.get("email_engagement")
        merged["last_email_open_date"] = marketing_data.get("last_email_open_date")
        merged["nps_history"] = marketing_data.get("nps_history", [])
        merged["csat_last_interaction"] = marketing_data.get("csat_last_interaction")
        merged["campaign_responses"] = marketing_data.get("campaign_responses", [])
    else:
        # Set defaults if marketing data is missing
        merged["email_engagement"] = None
        merged["last_email_open_date"] = None
        merged["nps_history"] = []
        merged["csat_last_interaction"] = None
        merged["campaign_responses"] = []

    return merged

````

# data aggregation

````python
"""Node 1: Data Aggregation - Load and merge data from multiple sources"""

from typing import Dict, Any
from datetime import datetime
from config import CustomerJourneyOrchestratorState, CustomerJourneyOrchestratorConfig
from utils.data_loader import (
    load_customers_data,
    load_orders_data,
    load_logistics_data,
    load_marketing_signals_data,
    extract_logistics_by_order_id,
    merge_customer_and_marketing_data
)


def data_aggregation_node(
    state: CustomerJourneyOrchestratorState,
    config: CustomerJourneyOrchestratorConfig
) -> Dict[str, Any]:
    """
    Node 1: Data Aggregation

    This node:
    1. Loads data from multiple sources (customers, orders, logistics, marketing)
    2. Merges customer data with marketing signals
    3. Extracts relevant data for the specific customer and order
    4. Creates a unified data structure for downstream nodes

    Args:
        state: Current orchestrator state
        config: Orchestrator configuration

    Returns:
        Updated state with aggregated data
    """
    errors = state.get("errors", [])
    customer_id = state.get("customer_id")
    order_id = state.get("order_id")

    # Validate required inputs
    if not customer_id:
        errors.append("customer_id is required for data aggregation")
        return {"errors": errors}

    if not order_id:
        errors.append("order_id is required for data aggregation")
        return {"errors": errors}

    try:
        # Load all data sources
        customers_data = load_customers_data(config.customers_file)
        orders_data = load_orders_data(config.orders_file)
        logistics_data = load_logistics_data(config.logistics_file)
        marketing_signals_data = load_marketing_signals_data(config.marketing_signals_file)

        # Extract data for this specific customer and order
        customer_raw = customers_data.get(customer_id)
        if not customer_raw:
            errors.append(f"Customer {customer_id} not found in customers data")
            return {"errors": errors}

        order_raw = orders_data.get(order_id)
        if not order_raw:
            errors.append(f"Order {order_id} not found in orders data")
            return {"errors": errors}

        # Verify order belongs to customer
        if order_raw.get("customer_id") != customer_id:
            errors.append(
                f"Order {order_id} does not belong to customer {customer_id}. "
                f"Order belongs to customer {order_raw.get('customer_id')}"
            )
            return {"errors": errors}

        # Get marketing signals for this customer
        marketing_raw = marketing_signals_data.get(customer_id)

        # Merge customer and marketing data
        customer_merged = merge_customer_and_marketing_data(customer_raw, marketing_raw)

        # Extract logistics data for this order
        logistics_raw = extract_logistics_by_order_id(logistics_data, order_id)
        if not logistics_raw:
            errors.append(f"Logistics data not found for order {order_id}")
            # Don't fail completely - we can still proceed without logistics data
            logistics_raw = {
                "order_id": order_id,
                "status": "unknown",
                "carrier": "unknown",
                "estimated_delivery": None,
                "is_delayed": None
            }

        # Prepare updated state
        updates: Dict[str, Any] = {
            "customer_data": customer_merged,
            "order_data": order_raw,
            "logistics_data": logistics_raw,
            "data_loaded_at": datetime.now().isoformat(),
            "errors": errors,  # Always include errors field (empty list if no errors)
        }

        return updates

    except FileNotFoundError as e:
        errors.append(f"Data file not found: {str(e)}")
        return {"errors": errors}
    except Exception as e:
        errors.append(f"Error during data aggregation: {str(e)}")
        return {"errors": errors}

````

# Architecture Pattern: Node vs Utility Function

**Key Question:** What's the difference between `data_aggregation_node.py` and `merge_customer_and_marketing_data()`?

---

## ğŸ¯ The Short Answer

- **`merge_customer_and_marketing_data()`** = A **utility function** (single-purpose helper)
- **`data_aggregation_node()`** = A **node function** (orchestrator that coordinates multiple steps)

**Think of it like this:**
- Utility = A tool in your toolbox (does one thing well)
- Node = A complete workflow step (uses multiple tools to accomplish a goal)

---

### 1. **Separation of Concerns**

### 2. **Reusability**

**Utilities** can be used by multiple nodes:
```python

```

**Nodes** are specific to the workflow:
```python

```

### 3. **Testability**

**Utilities** = Unit tests (fast, isolated)

# State Management vs Database: Understanding the Difference

**Key Question:** Are we creating a database? Can we view it?

---

## ğŸ¯ Short Answer

**No, we're NOT creating a database.** We're building an **in-memory state structure** that gets passed between nodes.

**Yes, you CAN view it!** Use the inspection tools we created.

---

## ğŸ“Š What We're Actually Doing

### In-Memory State (What We Have)

```
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚   Python Dictionary (in memory)     â”‚
â”‚   {                                  â”‚
â”‚     "customer_id": "C001",           â”‚
â”‚     "customer_data": {...},          â”‚
â”‚     "order_data": {...},             â”‚
â”‚     "logistics_data": {...}          â”‚
â”‚   }                                  â”‚
â”‚   â† Lives in RAM during execution    â”‚
â”‚   â† Gets passed between nodes        â”‚
â”‚   â† Disappears when program ends     â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜

- âœ… Fast (in-memory)
- âœ… Temporary (only during execution)

```
