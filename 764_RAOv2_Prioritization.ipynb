````python
"""
Prioritize issues by revenue impact and severity; build rollup and executive triggers.
"""
from typing import Any, Dict, List


def prioritize_issues(issues: List[Dict[str, Any]], top_n: int = 10) -> List[Dict[str, Any]]:
    """Sort by revenue_impact descending, then severity (high > medium > low). Return top N."""
    severity_order = {"high": 3, "medium": 2, "low": 1}

    def key_fn(i: Dict[str, Any]) -> tuple:
        rev = float(i.get("revenue_impact") or 0)
        sev = severity_order.get(i.get("severity", "low"), 0)
        return (-rev, -sev)

    sorted_issues = sorted(issues, key=key_fn)
    return sorted_issues[:top_n]


def build_rollup(
    issues: List[Dict[str, Any]],
    recovery_log: List[Dict[str, Any]],
) -> Dict[str, Any]:
    """
    Build revenue_rollup: revenue_recovered, revenue_at_risk, upsell_pipeline_value, recovery_rate, total_identified, open_issues_count.
    """
    total_recovered = sum(float(r.get("recovered_amount") or 0) for r in recovery_log)
    total_identified = sum(float(r.get("identified_amount") or 0) for r in recovery_log)
    recovery_rate = (total_recovered / total_identified * 100.0) if total_identified else 0.0

    revenue_at_risk = sum(float(i.get("revenue_impact") or 0) for i in issues)
    upsell_types = ("unbilled_overage", "upsell_opportunity")
    upsell_pipeline_value = sum(
        float(i.get("revenue_impact") or 0)
        for i in issues
        if i.get("issue_type") in upsell_types
    )

    return {
        "revenue_recovered": round(total_recovered, 2),
        "revenue_at_risk": round(revenue_at_risk, 2),
        "upsell_pipeline_value": round(upsell_pipeline_value, 2),
        "recovery_rate": round(recovery_rate, 1),
        "total_identified": round(total_identified, 2),
        "open_issues_count": len(issues),
        "recovery_log_count": len(recovery_log),
    }


def compute_executive_triggers(
    rollup: Dict[str, Any],
    revenue_at_risk_critical: float,
    revenue_at_risk_elevated: float,
    open_issues_critical: int,
    open_issues_elevated: int,
) -> List[Dict[str, Any]]:
    """Return list of { trigger_type, message, value, threshold } when thresholds exceeded."""
    triggers: List[Dict[str, Any]] = []
    rar = rollup.get("revenue_at_risk", 0) or 0
    open_count = rollup.get("open_issues_count", 0) or 0

    if rar >= revenue_at_risk_critical:
        triggers.append({
            "trigger_type": "revenue_at_risk",
            "message": f"Revenue at risk ${rar:,.0f} exceeds critical threshold ${revenue_at_risk_critical:,.0f}.",
            "value": rar,
            "threshold": revenue_at_risk_critical,
        })
    elif rar >= revenue_at_risk_elevated:
        triggers.append({
            "trigger_type": "revenue_at_risk",
            "message": f"Revenue at risk ${rar:,.0f} exceeds elevated threshold ${revenue_at_risk_elevated:,.0f}.",
            "value": rar,
            "threshold": revenue_at_risk_elevated,
        })

    if open_count >= open_issues_critical:
        triggers.append({
            "trigger_type": "open_issues",
            "message": f"Open issues count {open_count} exceeds critical threshold {open_issues_critical}.",
            "value": open_count,
            "threshold": open_issues_critical,
        })
    elif open_count >= open_issues_elevated:
        triggers.append({
            "trigger_type": "open_issues",
            "message": f"Open issues count {open_count} exceeds elevated threshold {open_issues_elevated}.",
            "value": open_count,
            "threshold": open_issues_elevated,
        })

    return triggers
````
