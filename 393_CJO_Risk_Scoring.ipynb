# Risk Scoring Utilities for Customer Journey Orchestrator

````python
"""
Risk Scoring Utilities for Customer Journey Orchestrator

Utilities for calculating comprehensive risk scores for customers based on
signals, journey state, and customer value.
"""

from typing import Dict, Any, List, Optional


def calculate_risk_score(
    customer_id: str,
    aggregated_signals: Optional[Dict[str, Any]],
    journey_evaluation: Optional[Dict[str, Any]],
    customer_data: Dict[str, Any],
    risk_weights: Dict[str, float],
    risk_tier_thresholds: Dict[str, float]
) -> Dict[str, Any]:
    """
    Calculate comprehensive risk score for a customer.

    Args:
        customer_id: Customer identifier
        aggregated_signals: Aggregated signals for this customer
        journey_evaluation: Journey evaluation for this customer
        customer_data: Customer data dictionary
        risk_weights: Weights for different risk factors
        risk_tier_thresholds: Thresholds for risk tiers

    Returns:
        Risk score dictionary
    """
    # Extract risk components
    signal_risk = aggregated_signals.get("aggregated_risk_score", 0.0) if aggregated_signals else 0.0
    max_signal_strength = aggregated_signals.get("max_signal_strength", 0.0) if aggregated_signals else 0.0
    signal_count = aggregated_signals.get("total_signals", 0) if aggregated_signals else 0

    # Time in state risk (normalized to 0-1)
    days_in_state = journey_evaluation.get("days_in_state", 0) if journey_evaluation else 0
    typical_duration = journey_evaluation.get("typical_duration", 30) if journey_evaluation else 30
    time_risk = min(1.0, days_in_state / (typical_duration * 2)) if typical_duration > 0 else 0.0

    # Customer value risk (higher value = higher risk if at risk)
    account_value = customer_data.get("account_value", 0)
    # Normalize account value to 0-1 (assuming max value around 200k)
    value_risk = min(1.0, account_value / 200000.0)

    # Signal count risk (more signals = higher risk)
    signal_count_risk = min(1.0, signal_count / 5.0)  # Normalize to 5 signals = max risk

    # Calculate weighted overall risk score
    overall_risk_score = (
        signal_risk * risk_weights.get("signal_strength", 0.35) +
        time_risk * risk_weights.get("time_in_state", 0.25) +
        value_risk * risk_weights.get("customer_value", 0.20) +
        signal_count_risk * risk_weights.get("signal_count", 0.20)
    )

    # Clamp to 0-1 range
    overall_risk_score = max(0.0, min(1.0, overall_risk_score))

    # Determine risk tier
    if overall_risk_score >= risk_tier_thresholds.get("high", 0.70):
        risk_tier = "high"
    elif overall_risk_score >= risk_tier_thresholds.get("medium", 0.40):
        risk_tier = "medium"
    else:
        risk_tier = "low"

    # Determine urgency
    if overall_risk_score >= 0.80:
        urgency = "high"
    elif overall_risk_score >= 0.50:
        urgency = "medium"
    else:
        urgency = "low"

    # Collect risk factors
    risk_factors = []
    if signal_risk > 0.5:
        risk_factors.append("high_signal_strength")
    if signal_count >= 2:
        risk_factors.append("multiple_negative_signals")
    if time_risk > 0.5:
        risk_factors.append("long_time_in_state")
    if journey_evaluation and journey_evaluation.get("friction_detected", False):
        risk_factors.append("journey_friction_detected")

    # Calculate churn risk (subset of overall risk, focused on signals and friction)
    churn_risk_score = (
        signal_risk * 0.50 +
        (1.0 if journey_evaluation and journey_evaluation.get("friction_detected", False) else 0.0) * 0.30 +
        time_risk * 0.20
    )
    churn_risk_score = max(0.0, min(1.0, churn_risk_score))

    return {
        "customer_id": customer_id,
        "overall_risk_score": round(overall_risk_score, 3),
        "churn_risk_score": round(churn_risk_score, 3),
        "friction_score": round(time_risk, 3),
        "risk_factors": risk_factors,
        "risk_tier": risk_tier,
        "urgency": urgency
    }


def calculate_all_risk_scores(
    customers_lookup: Dict[str, Dict[str, Any]],
    aggregated_signals: List[Dict[str, Any]],
    journey_evaluations: List[Dict[str, Any]],
    risk_weights: Dict[str, float],
    risk_tier_thresholds: Dict[str, float]
) -> List[Dict[str, Any]]:
    """
    Calculate risk scores for all customers.

    Args:
        customers_lookup: Lookup dictionary for customer data
        aggregated_signals: List of aggregated signal dictionaries
        journey_evaluations: List of journey evaluation dictionaries
        risk_weights: Weights for different risk factors
        risk_tier_thresholds: Thresholds for risk tiers

    Returns:
        List of risk score dictionaries
    """
    # Build lookups for fast access
    signals_lookup = {agg["customer_id"]: agg for agg in aggregated_signals}
    evaluations_lookup = {eval["customer_id"]: eval for eval in journey_evaluations}

    risk_scores = []

    for customer_id, customer_data in customers_lookup.items():
        aggregated_signal = signals_lookup.get(customer_id)
        journey_evaluation = evaluations_lookup.get(customer_id)

        risk_score = calculate_risk_score(
            customer_id,
            aggregated_signal,
            journey_evaluation,
            customer_data,
            risk_weights,
            risk_tier_thresholds
        )
        risk_scores.append(risk_score)

    return risk_scores

````

# Risk Scoring Node

````python
def risk_scoring_node(
    state: CustomerJourneyOrchestratorState,
    config: CustomerJourneyOrchestratorConfig
) -> Dict[str, Any]:
    """
    Risk Scoring Node: Calculate comprehensive risk scores.

    Calculates risk scores for each customer based on aggregated signals,
    journey evaluations, and customer value.
    """
    errors = state.get("errors", [])
    customers_lookup = state.get("customers_lookup", {})
    aggregated_signals = state.get("aggregated_signals", [])
    journey_evaluations = state.get("journey_evaluations", [])

    if not customers_lookup:
        return {
            "errors": errors + ["risk_scoring_node: customers_lookup is required"]
        }

    try:
        risk_scores = calculate_all_risk_scores(
            customers_lookup,
            aggregated_signals,
            journey_evaluations,
            config.risk_scoring_weights,
            config.risk_tier_thresholds
        )

        return {
            "risk_scores": risk_scores,
            "errors": errors
        }
    except Exception as e:
        return {
            "errors": errors + [f"risk_scoring_node: {str(e)}"]
        }
````
