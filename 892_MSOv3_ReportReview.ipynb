# 1️⃣ The First Problem: "Unknown Mission"

```
Mission: Unknown Mission
```

This is the **first thing an executive sees**, so it should never happen.

The issue is simply that `mission_name` is not being loaded into the state.

Right now the code falls back to:

```
state.get("mission_name") or state.get("mission_id")
```

But neither exists.

### Easy Fix

Add this to the telemetry loader or config:

```
state["mission_name"] = mission_kpis.get("mission_name")
```

or load it from a mission catalog.

For example:

```
Mission: Customer Support Resolution
```

This will immediately improve the report.

---

# 2️⃣ Efficiency Score Not Computed

The report says:

```
Performance efficiency cannot be scored yet
(insufficient KPI configuration).
```

But we can see KPI data exists:

```
Latest KPI snapshot:
{'resolution_time_hours': 3.2, 'csat': 4.6}
```

So the problem is **KPI structure**.

Your system expects:

```
baseline
target
actual
higher_is_better
```

But your telemetry currently only contains:

```
actual KPI values
```

Example:

```
resolution_time_hours = 3.2
csat = 4.6
```

Without baseline or target, the system cannot compute efficiency.

### This is actually good

Your system correctly **refused to guess**.

That is exactly the type of **deterministic discipline** that builds trust.

### Suggested improvement

Add KPI config like:

```
mission_kpis.json

{
  "support_resolution": {
      "resolution_time_hours": {
          "baseline": 4.5,
          "target": 2.5,
          "higher_is_better": false
      }
  }
}
```

Once that exists, efficiency scoring will work automatically.

# 5️⃣ Trend Section Reveals a Bug

```
Recent values:
[3.1, 2.6, 2.1, None, None, None, None, None]
```

Those `None` values should **not appear in the trend series**.

The dataset likely contains runs where the KPI key is missing.

Your trend function currently does:

```
series = [run.get("kpis", {}).get(kpi_key)]
```

But then prints the raw list.

### Better behavior

Filter missing values before display.

Example output:

```
Recent values: [3.1, 2.6, 2.1]
```

This will make the report look much cleaner.
