```python
"""
Board-ready PDF export for Workforce Development Orchestrator v2.

Converts markdown report to PDF using markdown -> HTML -> xhtml2pdf.
"""

from pathlib import Path
from typing import List, Optional


def save_report_pdf(
    markdown_content: str,
    markdown_file_path: str,
    errors: Optional[List[str]] = None,
) -> Optional[str]:
    """
    Convert markdown report to PDF and save next to the .md file.

    Args:
        markdown_content: Full markdown report text.
        markdown_file_path: Path to the .md file (PDF will use same base name with .pdf).
        errors: Optional list to append error message if conversion fails.

    Returns:
        Path to the saved PDF file, or None if conversion failed.
    """
    try:
        import markdown
        from xhtml2pdf import pisa
    except ImportError as e:
        if errors is not None:
            errors.append(f"pdf_export: missing dependency ({e}). Install: pip install markdown xhtml2pdf")
        return None

    pdf_path = str(Path(markdown_file_path).with_suffix(".pdf"))
    html = markdown.markdown(
        markdown_content,
        extensions=["tables", "nl2br"],
    )
    style = """
    <style>
        body { font-family: Helvetica, Arial, sans-serif; font-size: 11pt; line-height: 1.4; margin: 1in; }
        table { border-collapse: collapse; width: 100%; margin: 0.5em 0; }
        th, td { border: 1px solid #ddd; padding: 6px 8px; text-align: left; }
        th { background: #f5f5f5; font-weight: bold; }
        h1 { font-size: 18pt; margin-top: 0.5em; }
        h2 { font-size: 14pt; margin-top: 0.8em; }
        h3 { font-size: 12pt; margin-top: 0.6em; }
        ul, ol { margin: 0.3em 0; }
        p { margin: 0.4em 0; }
    </style>
    """
    full_html = f"<!DOCTYPE html><html><head><meta charset='utf-8'/>{style}</head><body>{html}</body></html>"

    try:
        with open(pdf_path, "wb") as pdf_file:
            pisa_status = pisa.CreatePDF(
                full_html.encode("utf-8"),
                dest=pdf_file,
                encoding="utf-8",
            )
        if pisa_status.err:
            if errors is not None:
                errors.append(f"pdf_export: xhtml2pdf returned errors writing {pdf_path}")
            return None
        return pdf_path
    except Exception as e:
        if errors is not None:
            errors.append(f"pdf_export: {e!s}")
        return None
```
