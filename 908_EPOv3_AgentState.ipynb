# Experimentation Portfolio Orchestrator v3

## Code Review & Architecture Explanation

This code defines the **core operating structure** for the Experimentation Portfolio Orchestrator v3.

It establishes two essential pieces of infrastructure:

1. **The State Model** — the structured memory the agent uses while running
2. **The Configuration Model** — the explicit rules and thresholds that govern behavior

Together, these create a **transparent and auditable decision system** for managing experimentation portfolios.

Rather than hiding logic inside prompts or opaque AI outputs, the orchestrator keeps **business signals explicit, measurable, and configurable**.

This is the foundation that allows the agent to produce **executive-ready intelligence rather than ad-hoc analysis**.

# 1. The State Object — The Agent’s Working Memory

```python
class EPOv3OrchestratorState(TypedDict, total=False)
```

The state object defines **everything the orchestrator knows during execution**.

You can think of it as the **shared memory of the entire workflow**.

Each node in the orchestrator:

* reads values from the state
* computes new insights
* writes results back to the state

This pattern ensures **clean data flow between nodes** without hidden dependencies.

````python
# ============================================================================
# Experimentation Portfolio Orchestrator v3 (EPOv3)
# ============================================================================
# Portfolio-level experimentation intelligence: effectiveness, design quality,
# risk exposure, learning velocity. Data in agents/data/.

class EPOv3OrchestratorState(TypedDict, total=False):
    """State for Experimentation Portfolio Orchestrator v3."""

    # Input
    data_dir: Optional[str]
    reports_dir: Optional[str]
    view_mode: Optional[str]  # "portfolio" | "experiment" (future: single experiment focus)

    # Goal & Planning
    goal: Dict[str, Any]
    plan: List[Dict[str, Any]]

    # Raw data (from agents/data/)
    experiment_runs: List[Dict[str, Any]]
    experiment_portfolio_snapshots: List[Dict[str, Any]]
    experiment_risk_events: List[Dict[str, Any]]
    experiment_task_execution_logs: List[Dict[str, Any]]

    # Loader contract: record counts for traceability
    runs_count: int
    snapshots_count: int
    risk_events_count: int
    task_logs_count: int

    # Rollups
    portfolio_rollup: Dict[str, Any]   # latest snapshot + deltas, experiments_at_risk, learning_velocity, etc.
    risk_rollup: Dict[str, Any]       # by severity, by risk_type, open_count
    executive_triggers: List[Dict[str, Any]]

    data_snapshot_loaded_at: Optional[str]
    validation_warnings: List[str]

    # Output
    experiment_report: str
    report_file_path: Optional[str]

    # Metadata
    errors: List[str]
    processing_time: Optional[float]


@dataclass
class EPOv3OrchestratorConfig:
    """Configuration for Experimentation Portfolio Orchestrator v3."""
    reports_dir: str = "agents/output/epo_v3_reports"

    data_dir: str = "agents/data"
    experiment_runs_file: str = "experiment_runs.json"
    experiment_portfolio_snapshots_file: str = "experiment_portfolio_snapshots.json"
    experiment_risk_events_file: str = "experiment_risk_events.json"
    experiment_task_execution_logs_file: str = "experiment_task_execution_logs.json"

    # Portfolio / risk thresholds (executive triggers)
    experiments_at_risk_critical: int = 2
    high_risk_experiments_critical: int = 1
    open_risk_events_critical: int = 5
    learning_velocity_min_target: int = 2  # min learning_velocity_last_30d we want

    # Report targets (for "on target" comparison)
    target_portfolio_status: str = "healthy"  # healthy | healthy_with_watchlist | at_risk
````
