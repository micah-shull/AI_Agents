## ðŸ”§ ROI Leaderboard Logic

Add helper:

```python
def _roi_leaderboard(
    playbook_performance_log,
    playbooks_lookup,
):
    rows = []

    for p in playbook_performance_log:
        pb = playbooks_lookup.get(p["playbook_id"], {})
        rows.append({
            "playbook": pb.get("name", p["playbook_id"]),
            "segment": p.get("segment"),
            "runs": p.get("runs"),
            "success_rate": p.get("success_rate"),
            "avg_cost": p.get("avg_cost"),
            "avg_value": p.get("avg_value"),
            "roi_ratio": p.get("roi_ratio"),
        })

    return sorted(rows, key=lambda r: r["roi_ratio"], reverse=True)[:10]
```
