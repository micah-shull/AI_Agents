````python
"""
Executive report builder for PDO V2.

Structure: one-view headline, targets, segment table (risk tier / bottlenecks),
so-what line, next steps, optional executive callout when triggers fire.
"""
from typing import Any, Dict, List


def _build_one_view(rollup: Dict[str, Any], triggers: List[Dict[str, Any]]) -> str:
    if triggers:
        return "Document portfolio: **Attention required** â€” thresholds exceeded; see next steps."
    return "Document portfolio: **Within guardrails** â€” key metrics within configured thresholds."


def _build_so_what(rollup: Dict[str, Any], triggers: List[Dict[str, Any]]) -> str:
    r = rollup
    parts = [
        f"{r.get('total_documents', 0)} document(s);",
        f"{r.get('executive_visibility_count', 0)} requiring executive visibility;",
        f"{r.get('high_risk_count', 0)} high-risk;",
        f"{r.get('stage_anomaly_count', 0)} stage anomaly(ies);",
        f"Reviewer cost ${r.get('total_reviewer_cost_usd', 0):,.0f}.",
    ]
    if triggers:
        parts.append(" Triggers exceeded; escalation alerts created.")
    return " ".join(parts)


def _build_next_steps(triggers: List[Dict[str, Any]], rollup: Dict[str, Any]) -> List[str]:
    steps = []
    if triggers:
        for t in triggers:
            steps.append(t.get("message", "Review triggered metric."))
        steps.append("Review escalation alerts and assign owners for high-risk documents.")
    else:
        steps.append("Continue monitoring portfolio rollup; no immediate action required.")
    if rollup.get("bottleneck_stages"):
        top = rollup["bottleneck_stages"][:2]
        for b in top:
            name = b.get("stage_name", "?")
            fr = b.get("failure_rate")
            steps.append(f"Review stage '{name}' (failure rate: {fr}).")
    return steps[:5]


def _build_segment_table(rollup: Dict[str, Any]) -> str:
    risk = rollup.get("risk_tier_summary") or {}
    by_tier = risk.get("by_tier") or {}
    lines = [
        "| Risk tier | Document count |",
        "|-----------|----------------|",
    ]
    for tier in sorted(by_tier.keys()):
        lines.append(f"| {tier} | {by_tier[tier]} |")
    return "\n".join(lines)


def _build_bottleneck_table(bottleneck_stages: List[Dict[str, Any]]) -> str:
    if not bottleneck_stages:
        return "_No bottleneck stages identified._"
    lines = [
        "| Stage | Documents | Failures | Failure rate | Avg duration (min) |",
        "|-------|-----------|----------|--------------|--------------------|",
    ]
    for b in bottleneck_stages[:8]:
        name = b.get("stage_name", "?")
        doc_count = b.get("document_count", 0)
        fail = b.get("failure_count", 0)
        fr = b.get("failure_rate", 0)
        avg = b.get("avg_duration_minutes")
        avg_str = f"{avg:.1f}" if avg is not None else "â€”"
        lines.append(f"| {name} | {doc_count} | {fail} | {fr:.0%} | {avg_str} |")
    return "\n".join(lines)


def _build_reviewer_economics_section(reviewer_economics: Dict[str, Any]) -> str:
    total = reviewer_economics.get("total_reviewer_cost_usd") or 0
    by_role = reviewer_economics.get("by_role") or {}
    lines = [f"**Total reviewer cost:** ${total:,.2f}", ""]
    for role, data in sorted(by_role.items()):
        cost = data.get("cost_usd", 0)
        count = data.get("review_count", 0)
        lines.append(f"- **{role}**: ${cost:,.2f} ({count} review(s))")
    return "\n".join(lines)


def build_executive_report(
    portfolio_rollup: Dict[str, Any],
    executive_triggers: List[Dict[str, Any]],
) -> str:
    """Build full markdown executive report."""
    sections: List[str] = []

    # Callout when triggers fire (AGENT_LESSONS_LEARNED: unmissable at top)
    if executive_triggers:
        sections.append("> **Executive attention required.** Triggers exceeded; see next steps below.")
        sections.append("")

    # One-view and so-what
    sections.append("## One-view")
    sections.append(_build_one_view(portfolio_rollup, executive_triggers))
    sections.append("")
    sections.append("### So what")
    sections.append(_build_so_what(portfolio_rollup, executive_triggers))
    sections.append("")

    # Next steps
    sections.append("## Next steps")
    for i, step in enumerate(_build_next_steps(executive_triggers, portfolio_rollup), 1):
        sections.append(f"{i}. {step}")
    sections.append("")

    # Risk tier segment table
    sections.append("## Portfolio by risk tier")
    sections.append(_build_segment_table(portfolio_rollup))
    sections.append("")

    # Bottlenecks
    sections.append("## Stage bottlenecks")
    sections.append(_build_bottleneck_table(portfolio_rollup.get("bottleneck_stages") or []))
    sections.append("")

    # Reviewer economics
    sections.append("## Reviewer economics")
    sections.append(_build_reviewer_economics_section(portfolio_rollup.get("reviewer_economics") or {}))
    sections.append("")

    # Stage anomalies (if any)
    anomalies = portfolio_rollup.get("stage_anomalies") or []
    if anomalies:
        sections.append("## Stage anomalies (duration vs baseline)")
        sections.append("| Document | Stage | Actual (min) | Expected (min) | Variance % |")
        sections.append("|----------|-------|--------------|----------------|------------|")
        for a in anomalies[:10]:
            sections.append(
                f"| {a.get('document_id', '?')} | {a.get('stage_name', '?')} | "
                f"{a.get('actual', '?')} | {a.get('expected', '?')} | {a.get('variance_pct', '?')}% |"
            )
        sections.append("")

    return "\n".join(sections)
````
