```python
# ---------------------------------------------------------------------------
# CEO / Business Manager enhancements
# ---------------------------------------------------------------------------


def compute_department_summary(
    departments: List[Dict[str, Any]],
    workforce_risk_controls: List[Dict[str, Any]],
    employees: List[Dict[str, Any]],
    automation_risk_analysis: List[Dict[str, Any]],
    skill_gap_analysis: List[Dict[str, Any]],
    roles_lookup: Dict[str, Dict[str, Any]],
    departments_lookup: Dict[str, Dict[str, Any]],
) -> List[Dict[str, Any]]:
    """
    Build per-department summary: headcount, criticality, displacement risk,
    retraining coverage, fairness/regulatory flags, gap count, risk heatmap.
    """
    risk_by_dept: Dict[str, Dict[str, Any]] = {}
    for a in automation_risk_analysis:
        role = roles_lookup.get(a["role_id"], {})
        did = role.get("department_id")
        if not did:
            continue
        if did not in risk_by_dept:
            risk_by_dept[did] = {"high_risk_roles": 0, "medium_risk_roles": 0, "employees_in_risk_roles": 0}
        if a.get("risk_level") == "high":
            risk_by_dept[did]["high_risk_roles"] += 1
        elif a.get("risk_level") == "medium":
            risk_by_dept[did]["medium_risk_roles"] += 1
        risk_by_dept[did]["employees_in_risk_roles"] = risk_by_dept[did].get("employees_in_risk_roles", 0) + len(a.get("affected_employees", []))

    gaps_by_dept: Dict[str, int] = {}
    for g in skill_gap_analysis:
        emp = next((e for e in employees if e.get("employee_id") == g.get("employee_id")), {})
        did = emp.get("department_id")
        if did:
            gaps_by_dept[did] = gaps_by_dept.get(did, 0) + 1

    controls_lookup: Dict[str, Dict[str, Any]] = {c["department_id"]: c for c in workforce_risk_controls}

    result: List[Dict[str, Any]] = []
    for d in departments:
        did = d["department_id"]
        ctrl = controls_lookup.get(did, {})
        risk = risk_by_dept.get(did, {})
        gap_count = gaps_by_dept.get(did, 0)
        headcount = d.get("headcount", 0) or sum(1 for e in employees if e.get("department_id") == did)
        disp_pct = ctrl.get("displacement_risk_pct")
        retrain_pct = ctrl.get("retraining_coverage_pct")
        fairness = ctrl.get("fairness_flag", False)
        regulatory = ctrl.get("regulatory_exposure", "low")
        human_override = ctrl.get("human_override_required", False)
        high_risk_roles = risk.get("high_risk_roles", 0)
        medium_risk_roles = risk.get("medium_risk_roles", 0)
        if high_risk_roles > 0 or (disp_pct is not None and disp_pct >= 0.2):
            heat = "high"
        elif medium_risk_roles > 0 or gap_count > 0 or (disp_pct is not None and disp_pct >= 0.1):
            heat = "medium"
        else:
            heat = "low"
        result.append({
            "department_id": did,
            "name": d.get("name", did),
            "headcount": headcount,
            "criticality": d.get("criticality", "medium"),
            "displacement_risk_pct": disp_pct,
            "retraining_coverage_pct": retrain_pct,
            "fairness_flag": fairness,
            "human_override_required": human_override,
            "regulatory_exposure": regulatory,
            "skill_gap_count": gap_count,
            "high_risk_roles": high_risk_roles,
            "medium_risk_roles": medium_risk_roles,
            "risk_heatmap_level": heat,
        })
    return result


def compute_trend_summary(
    workforce_snapshots: List[Dict[str, Any]],
    snapshots_by_department: Dict[str, List[Dict[str, Any]]],
    departments_lookup: Dict[str, Dict[str, Any]],
) -> Dict[str, Any]:
    """
    Compare latest vs previous snapshot by department: readiness delta,
    reskilling progress delta. Overall verdict: improving / stable / declining.
    """
    if not workforce_snapshots:
        return {"has_trends": False, "department_trends": [], "verdict": "No historical snapshots yet."}
    runs = sorted(set(s["run_id"] for s in workforce_snapshots), reverse=True)
    if len(runs) < 2:
        return {"has_trends": False, "department_trends": [], "verdict": "Only one run; add more snapshots for trend."}
    latest_run, prev_run = runs[0], runs[1]
    department_trends: List[Dict[str, Any]] = []
    for did, snap_list in snapshots_by_department.items():
        latest = next((s for s in snap_list if s.get("run_id") == latest_run), None)
        prev = next((s for s in snap_list if s.get("run_id") == prev_run), None)
        if not latest or not prev:
            continue
        readiness_delta = (latest.get("readiness_score") or 0) - (prev.get("readiness_score") or 0)
        reskilling_delta = (latest.get("reskilling_progress_pct") or 0) - (prev.get("reskilling_progress_pct") or 0)
        dept_name = departments_lookup.get(did, {}).get("name", did)
        department_trends.append({
            "department_id": did,
            "department_name": dept_name,
            "latest_readiness": latest.get("readiness_score"),
            "previous_readiness": prev.get("readiness_score"),
            "readiness_delta": readiness_delta,
            "reskilling_progress_latest": latest.get("reskilling_progress_pct"),
            "reskilling_progress_previous": prev.get("reskilling_progress_pct"),
            "reskilling_delta": reskilling_delta,
        })
    if not department_trends:
        return {"has_trends": False, "department_trends": [], "verdict": "No department trends in snapshots."}
    deltas = [t["readiness_delta"] for t in department_trends]
    avg_delta = sum(deltas) / len(deltas) if deltas else 0
    if avg_delta > 2:
        verdict = "Workforce readiness is improving across tracked departments."
    elif avg_delta < -2:
        verdict = "Workforce readiness has declined; accelerate reskilling and review at-risk roles."
    else:
        verdict = "Workforce readiness is stable; maintain reskilling investment to improve."
    return {
        "has_trends": True,
        "latest_run": latest_run,
        "previous_run": prev_run,
        "department_trends": department_trends,
        "verdict": verdict,
    }


def compute_training_investment_summary(
    training_investments: List[Dict[str, Any]],
    learning_paths_lookup: Optional[Dict[str, Dict[str, Any]]] = None,
) -> Dict[str, Any]:
    """
    Aggregate current L&D: total spend, by status, expected uplift, completion rate.
    """
    if not training_investments:
        return {
            "total_invested_usd": 0,
            "employees_enrolled": 0,
            "completed": 0,
            "in_progress": 0,
            "planned": 0,
            "completion_rate_pct": 0.0,
            "avg_expected_uplift_pct": 0.0,
            "has_data": False,
        }
    total = sum(t.get("cost_usd", 0) for t in training_investments)
    by_status: Dict[str, int] = {}
    uplifts: List[float] = []
    probs: List[float] = []
    for t in training_investments:
        st = t.get("status", "planned")
        by_status[st] = by_status.get(st, 0) + 1
        if t.get("expected_productivity_uplift_pct") is not None:
            uplifts.append(t["expected_productivity_uplift_pct"])
        if t.get("completion_probability") is not None:
            probs.append(t["completion_probability"])
    completed = by_status.get("completed", 0)
    in_progress = by_status.get("in_progress", 0)
    planned = by_status.get("planned", 0)
    n = len(training_investments)
    completion_rate = (completed / n * 100) if n else 0
    avg_uplift = sum(uplifts) / len(uplifts) if uplifts else 0
    return {
        "has_data": True,
        "total_invested_usd": total,
        "employees_enrolled": n,
        "completed": completed,
        "in_progress": in_progress,
        "planned": planned,
        "completion_rate_pct": round(completion_rate, 1),
        "avg_expected_uplift_pct": round(avg_uplift, 1),
        "avg_completion_probability": round(sum(probs) / len(probs), 2) if probs else None,
    }


def compute_business_impact_summary(
    automation_risk_analysis: List[Dict[str, Any]],
    skill_gap_analysis: List[Dict[str, Any]],
    department_summary: List[Dict[str, Any]],
    workforce_summary: Dict[str, Any],
    roles_lookup: Dict[str, Dict[str, Any]],
    departments_lookup: Dict[str, Dict[str, Any]],
    top_risks_n: int = 3,
    top_recommendations_n: int = 3,
) -> Dict[str, Any]:
    """
    Top risks and top recommendations in business language; one call to action.
    """
    criticality_order = {"high": 0, "medium": 1, "low": 2}
    top_risks: List[str] = []
    for d in department_summary:
        if d.get("fairness_flag"):
            top_risks.append(f"{d.get('name', d['department_id'])} has a fairness flag â€” recommend HR/Legal review.")
        if d.get("regulatory_exposure") and d.get("regulatory_exposure") != "low":
            top_risks.append(f"{d.get('name', d['department_id'])} has {d.get('regulatory_exposure')} regulatory exposure.")
        if d.get("risk_heatmap_level") == "high" and d.get("criticality") == "high":
            top_risks.append(f"High-criticality {d.get('name', d['department_id'])} has high automation/ displacement risk â€” prioritize reskilling.")
    for a in automation_risk_analysis:
        if a.get("risk_level") != "high":
            continue
        role_name = a.get("role_name", a.get("role_id"))
        n_emps = len(a.get("affected_employees", []))
        top_risks.append(f"{n_emps} employee(s) in high-risk role '{role_name}' â€” fund learning paths and role evolution.")
    top_risks = list(dict.fromkeys(top_risks))[:top_risks_n]

    high_priority_gaps = [g for g in skill_gap_analysis if g.get("priority") == "high"]
    top_recommendations: List[str] = []
    if high_priority_gaps:
        top_recommendations.append(f"Fund reskilling for {len(high_priority_gaps)} high-priority skill gaps (AI tools, automation workflows).")
    high_risk_depts = [d for d in department_summary if d.get("risk_heatmap_level") == "high"]
    if high_risk_depts:
        names = [d.get("name", d["department_id"]) for d in high_risk_depts]
        top_recommendations.append(f"Prioritize L&D budget for high-risk departments: {', '.join(names)}.")
    if any(d.get("fairness_flag") for d in department_summary):
        top_recommendations.append("Review displacement and fairness in HR with Legal; ensure human override where required.")
    top_recommendations = top_recommendations[:top_recommendations_n]

    call_to_action = "Approve L&D budget for high-priority gaps and review department risk heatmap; schedule fairness/regulatory review for flagged departments."
    if not top_risks and not high_risk_depts:
        call_to_action = "Maintain current reskilling investment; monitor readiness and add snapshots to track trend."
    return {
        "top_risks": top_risks,
        "top_recommendations": top_recommendations,
        "call_to_action": call_to_action,
    }
```
