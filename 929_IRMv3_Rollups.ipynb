# IRMv3 Rollups & Executive Triggers

## Architecture Review & Operational Explanation

The rollup module is responsible for converting raw score outputs into **clear operational signals for leadership**.

After the scoring system evaluates the AI ecosystem, the rollups layer answers the most important question:

**Which issues require attention right now?**

Instead of forcing leaders to interpret multiple metrics, the orchestrator produces two structured outputs:

1. **Score Rollup** – a consolidated snapshot of system health
2. **Executive Triggers** – explicit alerts requiring leadership attention

This step transforms technical telemetry into **decision-ready operational intelligence**.

---

# Score Rollup

The `build_score_rollup()` function consolidates all score outputs into a single structured object.

```python
build_score_rollup()
```

This produces a unified structure containing:

| Metric                    | Meaning                                 |
| ------------------------- | --------------------------------------- |
| Integration Stability     | Reliability of system integrations      |
| Dependency Exposure       | Architectural fragility                 |
| Governance Responsiveness | Organizational ability to fix issues    |
| Automation Value          | Whether AI systems deliver expected ROI |
| Portfolio Score           | Weighted overall health                 |
| Portfolio Status          | HEALTHY / AT_RISK / CRITICAL            |

### Why this design matters

Operational dashboards often scatter metrics across multiple systems.

This function produces **one authoritative snapshot** of the system’s condition.

That snapshot becomes the core input for:

* executive reports
* dashboards
* trend analysis
* audit records

This simplifies the interface between **technical monitoring and executive decision-making**.

---

# Executive Triggers

The `build_executive_triggers()` function is where the orchestrator becomes **action-oriented**.

Scores alone are not enough.

Leaders need to know:

**What requires attention now?**

This function converts score thresholds into **explicit escalation signals**.

Each trigger contains:

```python
{
    "trigger_type": "...",
    "message": "...",
    "value": ...,
    "threshold": ...
}
```

This structure ensures every alert clearly communicates:

* what the issue is
* the metric value
* the threshold that was crossed

There is no ambiguity.

---

# Portfolio-Level Triggers

The system first evaluates **overall system health**.

Possible triggers:

```text
portfolio_critical
portfolio_at_risk
```

These signals indicate that the AI ecosystem as a whole may be unstable.

For leadership, this answers a fundamental question:

**Is our AI operation safe to scale right now?**

---

# Operational Risk Triggers

The system then checks each core dimension independently.

Possible alerts include:

| Trigger                   | Meaning                         |
| ------------------------- | ------------------------------- |
| integration_stability     | Systems connecting unreliably   |
| dependency_exposure       | Architecture fragile            |
| governance_responsiveness | Issues not being fixed          |
| automation_value          | Automation not delivering value |

These signals allow leadership to understand **where the operational issue originates**.

For example:

Integration instability suggests **technical infrastructure problems**, while governance failures indicate **organizational process issues**.

---

# Governance Escalation Signals

Two triggers focus specifically on **operational discipline**.

| Trigger              | Meaning                     |
| -------------------- | --------------------------- |
| open_risks           | Too many unresolved issues  |
| overdue_remediations | Fixes not happening on time |

These triggers measure something most AI systems ignore:

**whether the organization is maintaining operational control of its AI environment**.

Without these signals, problems may accumulate silently.

# Overall Assessment

This rollups module is **exactly what executive-facing AI systems need**.

It converts raw telemetry and scores into:

* clear operational signals
* structured escalation triggers
* actionable insights for leadership

Most AI agents stop at **analysis**.

IRMv3 continues one step further:

It creates **governance signals**.

That distinction is important because as organizations deploy more AI systems, the biggest challenge will not be building automation.

It will be **maintaining control over it**.

The rollups layer is what ensures that control remains visible.


````python
"""Rollups and executive triggers for IRMv3."""

from typing import Any, Dict, List


def build_score_rollup(
    integration_stability: float,
    dependency_exposure: float,
    governance_responsiveness: float,
    automation_value: float,
    portfolio_score: float,
    portfolio_status: str,
) -> Dict[str, Any]:
    """Single dict with all scores and status for report and state."""
    return {
        "integration_stability_score": integration_stability,
        "dependency_exposure_score": dependency_exposure,
        "governance_responsiveness_score": governance_responsiveness,
        "automation_value_score": automation_value,
        "portfolio_score": portfolio_score,
        "portfolio_status": portfolio_status,
    }


def build_executive_triggers(
    integration_stability: float,
    dependency_exposure: float,
    governance_responsiveness: float,
    automation_value: float,
    portfolio_status: str,
    open_risks_count: int,
    overdue_remediations_count: int,
    integration_critical: float = 60.0,
    dependency_critical: float = 60.0,
    governance_critical: float = 60.0,
    automation_critical: float = 60.0,
    open_risks_critical: int = 8,
    overdue_remediations_critical: int = 3,
) -> List[Dict[str, Any]]:
    """Build list of executive triggers from scores and counts."""
    triggers: List[Dict[str, Any]] = []

    if portfolio_status == "CRITICAL":
        triggers.append({
            "trigger_type": "portfolio_critical",
            "message": "Portfolio status is CRITICAL; at least one core score is below threshold.",
            "value": portfolio_status,
            "threshold": "any score < 60",
        })
    elif portfolio_status == "AT_RISK":
        triggers.append({
            "trigger_type": "portfolio_at_risk",
            "message": "Portfolio is AT RISK; address low scores or portfolio score below target.",
            "value": portfolio_status,
            "threshold": "portfolio < 75 or two scores < 70",
        })

    if integration_stability < integration_critical:
        triggers.append({
            "trigger_type": "integration_stability",
            "message": f"Integration stability score ({integration_stability}) is below critical threshold.",
            "value": integration_stability,
            "threshold": integration_critical,
        })
    if dependency_exposure < dependency_critical:
        triggers.append({
            "trigger_type": "dependency_exposure",
            "message": f"Dependency exposure score ({dependency_exposure}) indicates fragile architecture.",
            "value": dependency_exposure,
            "threshold": dependency_critical,
        })
    if governance_responsiveness < governance_critical:
        triggers.append({
            "trigger_type": "governance_responsiveness",
            "message": f"Governance responsiveness ({governance_responsiveness}) is below threshold.",
            "value": governance_responsiveness,
            "threshold": governance_critical,
        })
    if automation_value < automation_critical:
        triggers.append({
            "trigger_type": "automation_value",
            "message": f"Automation value score ({automation_value}) indicates value leakage.",
            "value": automation_value,
            "threshold": automation_critical,
        })
    if open_risks_count >= open_risks_critical:
        triggers.append({
            "trigger_type": "open_risks",
            "message": f"Open risks ({open_risks_count}) exceed critical threshold.",
            "value": open_risks_count,
            "threshold": open_risks_critical,
        })
    if overdue_remediations_count >= overdue_remediations_critical:
        triggers.append({
            "trigger_type": "overdue_remediations",
            "message": f"Overdue remediations ({overdue_remediations_count}) require attention.",
            "value": overdue_remediations_count,
            "threshold": overdue_remediations_critical,
        })

    return triggers
````
