````python
from __future__ import annotations

from datetime import datetime, timezone
from pathlib import Path
from typing import Dict, Any

from config import CBOv2Config, CBOv2State


def _ensure_reports_dir(config: CBOv2Config) -> Path:
    root = Path(__file__).resolve().parents[4]
    reports_dir = root / config.reports_dir
    reports_dir.mkdir(parents=True, exist_ok=True)
    return reports_dir


def build_markdown_report(state: CBOv2State, config: CBOv2Config) -> str:
    """Build an executive-facing markdown report from state."""
    portfolio = state.get("portfolio_rollup") or {}
    triggers = state.get("executive_triggers") or []
    top_risk = state.get("top_risk_customers") or []
    top_growth = state.get("top_growth_customers") or []

    lines = []
    lines.append("# Cross-Business Customer Intelligence Orchestrator (CBOv2)")
    lines.append("")
    lines.append("## Executive Summary")
    lines.append("")
    lines.append(
        "- **Purpose**: Detect deterministic, explainable cross-business risk and expansion signals "
        "across Finance, Retail, and Healthcare before they appear in financial statements."
    )
    lines.append("")

    if portfolio:
        lines.append("## Portfolio Rollup (90d view)")
        lines.append("")
        lines.append(f"- **Total customers analyzed**: {portfolio.get('customers_total', 'N/A')}")
        lines.append(
            f"- **Customers with coordinated deterioration (≥2 domains)**: "
            f"{portfolio.get('customers_with_negative_alignment', 0)} "
            f"({portfolio.get('pct_negative_alignment', 0):.1%})"
        )
        lines.append(
            f"- **Customers with coordinated expansion (≥2 domains)**: "
            f"{portfolio.get('customers_with_positive_alignment', 0)} "
            f"({portfolio.get('pct_positive_alignment', 0):.1%})"
        )
        for key in ("customers_count", "finance_signals_count", "retail_signals_count", "healthcare_signals_count"):
            if key in portfolio:
                lines.append(f"- **{key}**: {portfolio[key]}")
        lines.append("")

    if triggers:
        lines.append("## Executive Triggers Fired")
        lines.append("")
        for trig in triggers:
            lines.append(f"- **{trig.get('name')}**: {trig.get('verdict')} — {trig.get('reason')}")
        lines.append("")

    if top_risk:
        lines.append("## Top Accelerating Risk Trajectories")
        lines.append("")
        for rec in top_risk:
            cid = rec.get("customer_id")
            domains_negative = rec.get("domains_negative", 0)
            lines.append(f"- **Customer {cid}** — negative in {domains_negative} domains")
        lines.append("")

    if top_growth:
        lines.append("## Top Accelerating Growth Trajectories")
        lines.append("")
        for rec in top_growth:
            cid = rec.get("customer_id")
            domains_positive = rec.get("domains_positive", 0)
            lines.append(f"- **Customer {cid}** — positive in {domains_positive} domains")
        lines.append("")

    if config.include_thresholds_in_report:
        lines.append("## Thresholds & Parameters (for auditability)")
        lines.append("")
        lines.append(f"- **Deterioration slope threshold**: {config.deterioration_slope_threshold}")
        lines.append(f"- **Improvement slope threshold**: {config.improvement_slope_threshold}")
        lines.append(f"- **Min negative domains for risk alignment**: {config.min_negative_domains_for_risk}")
        lines.append(f"- **Min positive domains for expansion alignment**: {config.min_positive_domains_for_growth}")
        lines.append("")

    return "\n".join(lines)


def save_report(markdown: str, config: CBOv2Config) -> str:
    """Save report to disk and return path as string."""
    reports_dir = _ensure_reports_dir(config)
    ts = datetime.now(timezone.utc).strftime("%Y%m%dT%H%M%SZ")
    path = reports_dir / f"CBOv2_report_{ts}.md"
    path.write_text(markdown, encoding="utf-8")
    return str(path)

````
