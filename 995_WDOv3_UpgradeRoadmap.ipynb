## Cross-doc synthesis

The four files repeat a few themes under different names. Treat these as **one workstream each**:

| Theme | Appears in | Single outcome |

## Worth pursuing (recommended)

**Phase 1 — High ROI, fits current architecture**

1. **Config thresholds** — Replace magic numbers in `metrics.py` with `WDOv3OrchestratorConfig` (readiness/skill gap/reskilling/automation bands you already documented in ideas_2).
2. **Reporting upgrades (ideas_3)** — Split verdict vs watch, **“What this means”** line, **Key signals** (primary risk / strongest performer from dept analyses), **adaptation gap → plain English**, lead with **per-$1k ROI** not raw ratio, **priority tags** on recommendations, **Executive ask** (one sentence), optional **time sensitivity** heuristic.
3. **Structured recommendations** — `List[Dict]` with `priority`, `message` (or `category`) so the report can render `[High Priority]` and you can pick a single “primary” ask deterministically.
4. **Loader: required files + critical failure** — Config-driven `required_files`; if `workforce_snapshots` or `departments` is empty → **error + halt** (not silent report).
5. **Light schema validation** — Required fields on snapshot rows; accumulate warnings.
6. **Data freshness** — Use max snapshot `date` vs today; warn if older than N days (configurable).
7. **`data_quality_score`** — Simple function of warning count + optional schema hits (cap at e.g. 0.3–1.0).
8. **Graph: halt + conditional edges** — After `data_loading`, if critical failure → skip analysis → minimal or “INCOMPLETE” report (ideas_4).
9. **Department ID validation** — If `department_id` set and unknown → structured error or warning before analysis.
10. **Mixed / conflicting signals** — Extend trajectory beyond improving/stable/at_risk where readiness ↑ and gaps ↑ (ideas_2 metrics #3).

**Phase 2 — Strong additions, slightly more design**

11. **Confidence scoring** — Rules: delta magnitudes, signal alignment, `data_quality_score`, mixed signals → `high | medium | low`; show in verdict (**ideas_1B + ideas_3**).
12. **Driver validation** — For each driver label, compare to sign of relevant delta; append `driver_validation_warnings` or inline flags (**ideas_1C**).
13. **Scenario outlook** — Deterministic: for each scenario, `projected_gap = adaptation_gap * (automation_multiplier - 1)` style + narrative line; no full Monte Carlo (**ideas_1D + ideas_3 #7**).
14. **State clarity (ideas_2)** — Either document that `workforce_metrics` holds these, or add explicit `trajectory_summary` / `metric_deltas` **views** populated in analysis (avoid duplicating everything twice).

**Phase 3 — Nice-to-have / observability**

15. **Node-level `node_status` + execution metadata** — Useful for dashboards; add when you have a consumer.
16. **Structured error objects** — `{node, type, message}` — when you add logging/aggregation.
17. **Retry hooks** — Only if I/O becomes flaky or you add network loaders.
18. **Summary node + UUID `run_id`** — Audit trail; defer until multi-run analytics matters.
19. **NODE_METADATA registry** — Dev tooling; defer.

---

## Defer or keep minimal

- **Extreme value capping** — Quick guard in metrics if you see bad data; not urgent for demo data.
- **Logging hook** — One `logger.warning` in loader when you add Phase 1; full pipelines later.
- **“Execution summary node”** — Overlaps with enriching final state in `report` node unless you need a separate artifact.
