# Product–Customer Fit Discovery Orchestrator v2

## Data Generation Plan

---

# Purpose

The **Product–Customer Fit Discovery Orchestrator v2** analyzes relationships between customers, products, features, and usage behavior to uncover **hidden revenue opportunities (“ghost demand”)**.

Unlike traditional analytics systems that only analyze current activity, this agent incorporates **historical trend signals** to identify:

* emerging demand
* declining products
* high-growth customer segments
* feature demand acceleration
* product portfolio gaps

The dataset used in this project is **synthetic but realistic**, designed to simulate the types of signals product strategy teams analyze when making product roadmap decisions.

The dataset intentionally embeds **hidden discovery patterns** so the orchestrator can demonstrate real strategic insights.

---

# Dataset Architecture

The v2 dataset consists of **three layers of intelligence**:

1️⃣ **Customer Layer**
Who the customers are.

2️⃣ **Product & Usage Layer**
How customers interact with products.

3️⃣ **Strategic Trend Layer**
How demand evolves over time.

---

# Recommended Format Strategy

## Use **CSV** for tabular datasets

CSV is ideal when:

* rows represent records
* columns are consistent
* data is relational
* tables may be joined

This fits most of your datasets.

Use **CSV for:**

```
customers.csv
customer_metrics.csv
product_catalog.csv
product_metrics.csv
transactions.csv
feature_usage.csv
product_adoption_history.csv
segment_growth_history.csv
feature_demand_history.csv
```

These behave like **database tables**.

Advantages:

* easy to view
* easy to load with pandas
* easy to join
* compact

Example load:

```python
df = pd.read_csv("customers.csv")
```

---

# Use **JSON** for nested / event data

JSON shines when data contains:

* variable fields
* nested structures
* sequences of actions
* objects with attributes

Your **customer journey dataset** is perfect for JSON.

Instead of flattening events like:

```
customer_journey_events.csv
```

We can do:

```
customer_journeys.json
```

Example:

```json
{
  "Customer_ID": "C012",
  "journey": [
    {"event": "trial", "product": "P02", "date": "2024-02-01"},
    {"event": "purchase", "product": "P02", "date": "2024-02-10"},
    {"event": "upgrade", "product": "P05", "date": "2024-06-18"}
  ]
}
```

Advantages:

* cleaner representation
* realistic SaaS event logs
* easier to simulate sequences
