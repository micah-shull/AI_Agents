# 1️⃣ Concern Alignment Engine (High ROI, Low Complexity)

### Why This Matters

You already store:

* `skin_concerns` in customers
* `target_concerns` in products

Right now your scoring does not deeply leverage this.

That’s a missed opportunity.

---

## What to Add

### New Scoring Dimension (V2)

Add:

```python
concern_alignment_score
```

### Logic:

```python
matching_concerns = set(customer_skin_concerns)
                     ∩ set(product_target_concerns)

score = len(matching_concerns) * concern_weight
```

Example:

* Customer concern: `["hyperpigmentation"]`
* Product targets: `["hyperpigmentation", "brightness"]`
* Match count = 1 → strong alignment

---

### Suggested Weight

Add as **15–20% of total score**

To avoid bloating:

Reduce:

* Routine Completeness from 20% → 15%
* Keep Business Value dominant

---

### Why This Is Powerful

It shifts the system from:

> “Fill routine gaps”

to

> “Solve customer problems.”

This increases:

* Conversion probability
* Customer satisfaction
* Retention

Without ML.

# 3️⃣ Churn-Aware Mode (Strategic Behavior Switch)

## V2 Addition: Adaptive Scoring Mode

| Normal Mode      | Churn Mode              |
| ---------------- | ----------------------- |
| Emphasize margin | Emphasize fit           |

# 🎯 Short Answer

Adding **historical trend analysis** is:

**Technically easy. Architecturally moderate. Strategically powerful.**

If we do it cleanly, this becomes a *growth intelligence engine* — not just a recommendation engine.

---

# 🔍 What “Historical Data” Should Mean (For MVP v2)

We do NOT need:

* 3 years of transactional logs
* ML time-series forecasting
* Customer cohort modeling
* Real-time behavior streams

For v2, we just need:

> Lightweight historical snapshots that reveal direction.

---

# 🧠 What Trends Actually Matter?

For a Cross-Sell & Upsell Orchestrator, only 4 trend types are strategically valuable:

---

## 1️⃣ Spend Velocity Trend

Are they:

* Spending more?
* Spending less?
* Flat?

You already store:

* `purchase_history`
* `lifetime_value`
* `rfm.monetary_90d`

So you don’t even need new data — you just need derived metrics.

---

### Simple MVP Trend Logic

```python
last_30_days_spend
previous_30_days_spend

spend_trend = (last_30 - previous_30) / previous_30
```

Classify as:

* "accelerating"
* "stable"
* "declining"

---

## 2️⃣ Routine Expansion Trend

Is the customer:

* Adding new categories?
* Stuck with one product?
* Dropping steps?

You already store:

* `categories`
* `purchase_history`

We just need historical category count over time.

---

### MVP Version

Store a simple snapshot per month:

```python
customer_history = [
    {"month": "2025-09", "category_count": 1},
    {"month": "2025-10", "category_count": 3},
    {"month": "2025-11", "category_count": 5}
]
```

Then:

```python
if category_count increasing:
    routine_trend = "expanding"
```

---

## 3️⃣ Replenishment Discipline Trend

Are they:

* Buying before due?
* Buying late?
* Skipping refills?

You already compute:

* `days_since_purchase`
* `replenishment_cycle_days`

You just need to track average delay.

---

Example:

```python
delay = days_since_purchase - replenishment_cycle
```

Trend:

* Improving
* Stable
* Worsening

---

## 4️⃣ Tier Migration Trend

Are they:

* Moving toward higher-tier products?
* Downgrading?
* Stagnant?

You already store:

* `tier_preference`
* Product `tier`

Compute:

```python
avg_product_tier_last_3_purchases
```

Then compare to previous window.
