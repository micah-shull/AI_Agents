# Orchestrator Agents: Strategic Guide

**Purpose:** Understanding what makes orchestrator agents different, valuable, and worth building in the AI era.

---

## üéØ What Are Orchestrator Agents?

**Orchestrator agents** are AI systems that coordinate, connect, and create value across multiple business units, data sources, or systems. Unlike single-purpose agents that solve one problem, orchestrators create **network effects** by connecting isolated capabilities.

### The Key Difference

| Standard Agent | Orchestrator Agent |
|----------------|-------------------|
| Solves one problem in one domain | Connects multiple domains/systems |
| Works with single data source | Integrates data from multiple sources |
| Creates local value | Creates **compounding value** across units |
| Example: "Summarize this article" | Example: "Connect customer data across retail, finance, and healthcare to find cross-sell opportunities" |

---

## üí° Why They're Valuable (The Strategic Context)

Based on the HBR article you shared, orchestrator agents embody the shift from **industrial-era logic** to **intelligence-era logic**:

### 1. **Portfolio Logic: Shared Data = Shared Advantage**

**Old Way (Industrial Era):**
- Business units operate independently
- Each unit has its own customer database
- No cross-unit insights
- "We're big, so we win"

**Orchestrator Way (AI Era):**
- Business units share data through the orchestrator
- Unified customer view across all units
- Cross-unit opportunities emerge automatically
- "Our assets learn from each other"

**Example:** A customer buys from your retail division. The orchestrator agent sees they don't have a credit card from your finance division, identifies them as a good fit, and recommends a cross-sell. This is **impossible** without orchestration.

### 2. **Operating Model: AI as Company-Wide Capability**

**Old Way:**
- AI is a department (IT, Analytics)
- Each division builds its own models
- No reuse, no sharing
- "We have AI in 5 departments"

**Orchestrator Way:**
- AI orchestrator enables data flow company-wide
- Models and insights are reusable assets
- Knowledge transfers automatically
- "AI is our operating system"

**Example:** The orchestrator learns that customers who buy Product A in retail are 3x more likely to need Service B in healthcare. This insight becomes a reusable pattern that all units can leverage.

### 3. **Algorithmic Advantage: Defensible Moat**

**Old Way:**
- Competitive advantage = brand, scale, capital
- Easy to copy
- Diminishing returns

**Orchestrator Way:**
- Competitive advantage = proprietary data network effects
- Hard to copy (requires same data access)
- Compounding returns (more data ‚Üí better insights ‚Üí more value)

**Example:** Your orchestrator agent has learned from 5 years of cross-business data. A competitor can't replicate this without the same data relationships. This is your **defensible moat**.

```python
def match_customer_identities(customer_a, customer_b):
    """Match customers across business units"""
    confidence = 0.0
    
    # Exact match on unique identifier
    if customer_a["email"] == customer_b["email"]:
        confidence += 0.5
    
    # Phone match
    if normalize_phone(customer_a["phone"]) == normalize_phone(customer_b["phone"]):
        confidence += 0.3
    
    # Name similarity
    if name_similarity(customer_a["name"], customer_b["name"]) > 0.8:
        confidence += 0.2
    
    return confidence >= 0.7  # Threshold for match
```

```python
def find_cross_sell_opportunities(customer_graph):

        # Pattern: Customers with Service A often need Service B
```

## üéØ Core Identity Fields (Required for Matching)

These fields are **critical** for matching customers across business units:

| `email` | string | Primary identifier (most reliable) | `john.smith@email.com` |

- **Email**: Most reliable unique identifier (rarely changes)

- **Name**: For fuzzy matching when other fields differ
- **Address**: Additional confidence for matches
- **Customer ID**: Varies by business unit (Retail might use `RTL-123`, Finance might use `FIN-456`)

## üîç What Makes Good Test Data

### 1. **Identity Variations** (Tests matching logic)
- Same customer with slightly different names: `John Smith` vs `J. Smith` vs `John A. Smith`
- Same email, different phone formats: `555-123-4567` vs `(555) 123-4567`
- Same customer, different addresses: `123 Main St` vs `123 Main Street`

### 2. **Cross-Unit Opportunities** (Tests opportunity detection)
- Customer with Retail + Finance, but no Healthcare ‚Üí Healthcare opportunity
- Customer with basic tier in one unit, premium in another ‚Üí Upsell opportunity
- High-value customer in one unit, not in others ‚Üí Cross-sell opportunity

### 3. **Data Quality Variations** (Tests robustness)
- Some records missing phone numbers
- Some records with incomplete addresses
- Some records with typos in names
- Some records with different date formats

### 4. **Realistic Distribution**
- Mix of customer segments (high-value, medium, low)
- Mix of engagement levels (active, inactive)
- Mix of tenure (new, established, long-term)
- Geographic diversity

# Goal Node

````python
"""Goal node - defines the agent's objective and structure"""

import logging
from typing import Dict, Any
from config import CustomerGraphState

logger = logging.getLogger(__name__)


def goal_node(state: CustomerGraphState) -> CustomerGraphState:
    """
    Define the goal for the Customer Graph Orchestrator Agent.

    Reads: data_sources, business_units
    Writes: goal
    """
    logger.info("üéØ Defining agent goal...")

    try:
        data_sources = state.get("data_sources", [])
        business_units = state.get("business_units", [])

        goal = {
            "objective": "Build a unified customer identity graph across multiple business units and identify cross-sell/upsell opportunities",
            "scope": {
                "data_sources": data_sources,
                "business_units": business_units,
                "expected_outputs": [
                    "unified_customer_graph",
                    "cross_sell_opportunities",
                    "upsell_opportunities",
                    "personalization_insights"
                ]
            },
            "success_criteria": {
                "identity_resolution": "Match customers across business units with confidence scores",
                "opportunity_detection": "Identify cross-sell and upsell opportunities",
                "graph_quality": "Create unified customer profiles with merged data"
            }
        }

        state["goal"] = goal
        logger.info(f"‚úÖ Goal defined for {len(business_units)} business units")

    except Exception as e:
        logger.error(f"‚ùå Error in goal_node: {e}")
        state.setdefault("errors", []).append(f"Goal definition failed: {str(e)}")

    return state

````

# Data Ingestion Node

````python
"""Data ingestion node - loads customer data from multiple sources"""

import logging
from pathlib import Path
from typing import Dict, Any, List
from config import CustomerGraphState, CustomerGraphConfig

from customer_graph_agent.utils.file_parser import parse_csv_file

logger = logging.getLogger(__name__)


def data_ingestion_node(state: CustomerGraphState) -> CustomerGraphState:
    """
    Load and parse customer data from all business units.

    Reads: data_sources, business_units
    Writes: ingested_data
    """
    logger.info("üì• Ingesting customer data from all sources...")

    try:
        data_sources = state.get("data_sources", [])
        business_units = state.get("business_units", [])

        if len(data_sources) != len(business_units):
            raise ValueError(
                f"Mismatch: {len(data_sources)} data sources but {len(business_units)} business units"
            )

        ingested_data = []
        project_root = Path(__file__).parent.parent.parent

        for file_path, business_unit in zip(data_sources, business_units):
            # Resolve relative paths from project root
            if not Path(file_path).is_absolute():
                full_path = project_root / file_path
            else:
                full_path = Path(file_path)

            try:
                parsed_data = parse_csv_file(str(full_path), business_unit)
                ingested_data.append(parsed_data)

            except FileNotFoundError as e:
                logger.error(f"‚ùå File not found: {e}")
                state.setdefault("errors", []).append(f"File not found: {file_path}")
            except Exception as e:
                logger.error(f"‚ùå Error parsing {file_path}: {e}")
                state.setdefault("errors", []).append(f"Error parsing {file_path}: {str(e)}")

        state["ingested_data"] = ingested_data

        total_records = sum(item["record_count"] for item in ingested_data)
        logger.info(f"‚úÖ Ingested {total_records} total customer records from {len(ingested_data)} sources")

    except Exception as e:
        logger.error(f"‚ùå Error in data_ingestion_node: {e}")
        state.setdefault("errors", []).append(f"Data ingestion failed: {str(e)}")
        state["ingested_data"] = []

    return state

````

# Graph Builder Node

````python
"""Graph builder node - creates unified customer identity graph"""

import logging
from typing import Dict, Any, List, Set
from config import CustomerGraphState, CustomerGraphConfig

from customer_graph_agent.utils.identity_matcher import find_matching_customers, calculate_match_confidence

logger = logging.getLogger(__name__)


def _create_unified_id(customer_id: str, business_unit: str) -> str:
    """Create a unified customer ID"""
    return f"{business_unit}:{customer_id}"


def _merge_customer_profiles(
    customers: List[Dict[str, Any]],
    business_units: List[str]
) -> Dict[str, Any]:
    """
    Merge multiple customer records into a unified profile.

    Args:
        customers: List of customer records from different business units
        business_units: List of business unit names

    Returns:
        Merged customer profile
    """
    if not customers:
        return {}

    # Start with the first customer as base
    merged = customers[0].copy()

    # Track which business units this customer appears in
    merged["business_units"] = []
    merged["customer_ids_by_unit"] = {}

    # Merge data from all customers
    for customer in customers:
        business_unit = customer.get("_business_unit", "Unknown")
        customer_id = customer.get("customer_id", "")

        merged["business_units"].append(business_unit)
        merged["customer_ids_by_unit"][business_unit] = customer_id

        # Merge fields, preferring non-null values
        for key, value in customer.items():
            if key.startswith("_"):  # Skip internal fields
                continue

            if key not in merged or not merged[key]:
                merged[key] = value
            elif value and isinstance(value, list) and isinstance(merged[key], list):
                # Merge lists (unique items)
                merged[key] = list(set(merged[key] + value))

    # Ensure business_units is unique
    merged["business_units"] = list(set(merged["business_units"]))

    return merged


def graph_builder_node(state: CustomerGraphState) -> CustomerGraphState:
    """
    Build unified customer identity graph by matching customers across business units.

    Reads: ingested_data
    Writes: customer_graph
    """
    logger.info("üîó Building unified customer identity graph...")

    try:
        ingested_data = state.get("ingested_data", [])
        config = CustomerGraphConfig()
        thresholds = config.matching_thresholds

        if not ingested_data:
            logger.warning("‚ö†Ô∏è  No ingested data available")
            state["customer_graph"] = {
                "total_unique_customers": 0,
                "identity_mappings": {},
                "customer_profiles": {},
                "cross_unit_connections": []
            }
            return state

        # Collect all customers with their business unit info
        all_customers = []
        for unit_data in ingested_data:
            business_unit = unit_data["business_unit"]
            for customer in unit_data["raw_data"]:
                customer["_business_unit"] = business_unit
                customer["_unified_id"] = _create_unified_id(
                    customer.get("customer_id", ""), business_unit
                )
                all_customers.append(customer)

        logger.info(f"üìä Processing {len(all_customers)} customer records...")

        # Build identity mappings
        identity_mappings = {}  # unified_id -> [all_related_unified_ids]
        processed = set()
        cross_unit_connections = []
        customer_groups = []  # Groups of customers that are the same person

        for i, customer1 in enumerate(all_customers):
            if customer1["_unified_id"] in processed:
                continue

            # Find all matching customers
            other_customers = [c for j, c in enumerate(all_customers) if j != i]
            matches = find_matching_customers(
                customer1, other_customers, thresholds, min_confidence=thresholds["low_confidence"]
            )

            if matches:
                # Create a group of matching customers
                group = [customer1]
                group_ids = {customer1["_unified_id"]}

                for matched_customer, confidence, evidence in matches:
                    matched_id = matched_customer["_unified_id"]
                    if matched_id not in processed and matched_id not in group_ids:
                        group.append(matched_customer)
                        group_ids.add(matched_id)
                        processed.add(matched_id)

                        # Track cross-unit connections
                        if customer1["_business_unit"] != matched_customer["_business_unit"]:
                            cross_unit_connections.append({
                                "customer_id": customer1.get("customer_id", ""),
                                "business_units": sorted(list(set([
                                    customer1["_business_unit"],
                                    matched_customer["_business_unit"]
                                ]))),
                                "confidence": confidence,
                                "evidence": evidence
                            })

                # Create identity mapping for this group
                primary_id = customer1["_unified_id"]
                related_ids = [c["_unified_id"] for c in group[1:]]
                identity_mappings[primary_id] = [primary_id] + related_ids

                # Store the group for profile merging
                customer_groups.append((primary_id, group))

                processed.add(primary_id)
            else:
                # Single-unit customer (no matches)
                primary_id = customer1["_unified_id"]
                identity_mappings[primary_id] = [primary_id]
                customer_groups.append((primary_id, [customer1]))
                processed.add(primary_id)

        # Build unified customer profiles
        customer_profiles = {}
        for primary_id, group in customer_groups:
            business_units = [c["_business_unit"] for c in group]
            profile = _merge_customer_profiles(group, business_units)
            profile["unified_id"] = primary_id
            customer_profiles[primary_id] = profile

        # Calculate summary metrics
        total_customers = len(all_customers)
        unique_customers = len(customer_profiles)
        multi_unit_customers = len([
            pid for pid, profile in customer_profiles.items()
            if len(profile.get("business_units", [])) > 1
        ])

        # Calculate data quality score (simplified: based on match confidence and completeness)
        avg_confidence = (
            sum(conn["confidence"] for conn in cross_unit_connections) / len(cross_unit_connections)
            if cross_unit_connections else 0.0
        )
        completeness = unique_customers / total_customers if total_customers > 0 else 0.0
        data_quality_score = (avg_confidence * 0.6 + completeness * 0.4)

        customer_graph = {
            "total_unique_customers": unique_customers,
            "total_customer_records": total_customers,
            "identity_mappings": identity_mappings,
            "customer_profiles": customer_profiles,
            "cross_unit_connections": cross_unit_connections,
            "summary": {
                "total_customers": total_customers,
                "unique_customers": unique_customers,
                "multi_unit_customers": multi_unit_customers,
                "cross_unit_connection_count": len(cross_unit_connections),
                "data_quality_score": round(data_quality_score, 3)
            }
        }

        state["customer_graph"] = customer_graph
        state["graph_summary"] = customer_graph["summary"]

        logger.info(f"‚úÖ Built customer graph:")
        logger.info(f"   - {unique_customers} unique customers from {total_customers} records")
        logger.info(f"   - {multi_unit_customers} customers appear in multiple business units")
        logger.info(f"   - {len(cross_unit_connections)} cross-unit connections found")
        logger.info(f"   - Data quality score: {data_quality_score:.3f}")

    except Exception as e:
        logger.error(f"‚ùå Error in graph_builder_node: {e}")
        import traceback
        logger.error(traceback.format_exc())
        state.setdefault("errors", []).append(f"Graph building failed: {str(e)}")
        state["customer_graph"] = {}

    return state

````

# Opportunity Analyzer Node

````python
"""Opportunity analyzer node - identifies cross-sell and upsell opportunities"""

import logging
from typing import Dict, Any, List, Set
from config import CustomerGraphState, CustomerGraphConfig

logger = logging.getLogger(__name__)


def _get_available_services_by_unit() -> Dict[str, List[str]]:
    """Define available services/products by business unit"""
    return {
        "Retail": ["Home", "Beauty", "Fashion", "Electronics", "Sports", "Outdoors"],
        "Finance": ["checking", "savings", "credit_card", "loan"],
        "Healthcare": ["primary_care", "dental", "specialist"]
    }


def _get_service_tiers_by_unit() -> Dict[str, Dict[str, List[str]]]:
    """Define service tiers by business unit"""
    return {
        "Retail": {
            "none": [],
            "bronze": ["basic_discounts"],
            "silver": ["basic_discounts", "free_shipping"],
            "gold": ["basic_discounts", "free_shipping", "exclusive_access", "priority_support"]
        },
        "Finance": {
            "none": [],
            "basic": ["checking"],
            "standard": ["checking", "savings"],
            "premium": ["checking", "savings", "credit_card"]
        },
        "Healthcare": {
            "none": [],
            "basic": ["primary_care"],
            "standard": ["primary_care", "dental"],
            "premium": ["primary_care", "dental", "specialist"]
        }
    }


def _analyze_cross_sell_opportunities(
    customer_graph: Dict[str, Any],
    ingested_data: List[Dict[str, Any]]
) -> List[Dict[str, Any]]:
    """Identify cross-sell opportunities (customers missing services from other units)"""
    opportunities = []
    available_services = _get_available_services_by_unit()

    profiles = customer_graph.get("customer_profiles", {})

    for unified_id, profile in profiles.items():
        business_units = profile.get("business_units", [])

        # Only consider customers in at least one unit
        if not business_units:
            continue

        # Find services they're missing from other units
        for target_unit, services in available_services.items():
            if target_unit in business_units:
                continue  # Already a customer in this unit

            # Determine what services they might need
            recommended_services = []
            confidence = "medium"
            rationale = f"Customer uses {', '.join(business_units)} but not {target_unit}"

            # Simple logic: recommend basic services for new units
            if target_unit == "Finance":
                recommended_services = ["checking", "savings"]
            elif target_unit == "Healthcare":
                recommended_services = ["primary_care"]
            elif target_unit == "Retail":
                # Recommend based on their preferences in other units
                recommended_services = ["Home", "Beauty"]

            # Calculate confidence based on customer value
            loyalty_tier = profile.get("loyalty_tier", "none")
            account_balance = profile.get("account_balance", 0)
            total_spent = profile.get("total_spent", 0)

            if loyalty_tier == "gold" or account_balance > 10000 or total_spent > 5000:
                confidence = "high"
            elif loyalty_tier in ["silver", "bronze"] or account_balance > 5000 or total_spent > 2000:
                confidence = "medium"
            else:
                confidence = "low"

            # Estimate value (simplified)
            estimated_value = None
            if target_unit == "Finance":
                estimated_value = 500.0  # Annual account fees
            elif target_unit == "Healthcare":
                estimated_value = 1200.0  # Annual insurance premium
            elif target_unit == "Retail":
                estimated_value = 1000.0  # Estimated annual spend

            opportunities.append({
                "customer_id": unified_id,
                "current_services": business_units,
                "recommended_services": recommended_services,
                "business_units": [target_unit],
                "confidence": confidence,
                "rationale": rationale,
                "estimated_value": estimated_value
            })

    return opportunities


def _analyze_upsell_opportunities(
    customer_graph: Dict[str, Any]
) -> List[Dict[str, Any]]:
    """Identify upsell opportunities (customers on lower tiers)"""
    opportunities = []
    service_tiers = _get_service_tiers_by_unit()

    profiles = customer_graph.get("customer_profiles", {})

    for unified_id, profile in profiles.items():
        business_units = profile.get("business_units", [])

        for unit in business_units:
            # Check tier in this unit
            if unit == "Retail":
                current_tier = profile.get("loyalty_tier", "none")
                tiers = ["none", "bronze", "silver", "gold"]
                if current_tier in tiers:
                    tier_index = tiers.index(current_tier)
                    if tier_index < len(tiers) - 1:
                        recommended_tier = tiers[tier_index + 1]
                        total_spent = profile.get("total_spent", 0)

                        # Only recommend if they're spending enough
                        if total_spent > 1000 or current_tier == "none":
                            opportunities.append({
                                "customer_id": unified_id,
                                "current_tier": current_tier,
                                "recommended_tier": recommended_tier,
                                "business_unit": unit,
                                "confidence": "high" if total_spent > 2000 else "medium",
                                "rationale": f"Customer spends ${total_spent:.2f}, eligible for {recommended_tier} tier benefits",
                                "estimated_value": 200.0 if recommended_tier == "gold" else 100.0
                            })

            elif unit == "Finance":
                products = profile.get("products_owned", [])
                if isinstance(products, str):
                    # Handle string representation
                    products = []

                if "checking" in products and "savings" not in products:
                    opportunities.append({
                        "customer_id": unified_id,
                        "current_tier": "basic",
                        "recommended_tier": "standard",
                        "business_unit": unit,
                        "confidence": "high",
                        "rationale": "Customer has checking account, could benefit from savings account",
                        "estimated_value": 150.0
                    })
                elif "credit_card" not in products and len(products) >= 2:
                    opportunities.append({
                        "customer_id": unified_id,
                        "current_tier": "standard",
                        "recommended_tier": "premium",
                        "business_unit": unit,
                        "confidence": "medium",
                        "rationale": "Customer has multiple accounts, eligible for credit card",
                        "estimated_value": 300.0
                    })

            elif unit == "Healthcare":
                insurance_plan = profile.get("insurance_plan", "none")
                plans = ["none", "basic", "standard", "premium"]
                if insurance_plan in plans:
                    plan_index = plans.index(insurance_plan)
                    if plan_index < len(plans) - 1:
                        recommended_plan = plans[plan_index + 1]
                        services_used = profile.get("services_used", [])
                        if isinstance(services_used, str):
                            services_used = []

                        # Recommend upgrade if they're using multiple services
                        if len(services_used) >= 2 or insurance_plan == "none":
                            opportunities.append({
                                "customer_id": unified_id,
                                "current_tier": insurance_plan,
                                "recommended_tier": recommended_plan,
                                "business_unit": unit,
                                "confidence": "high" if len(services_used) >= 2 else "medium",
                                "rationale": f"Customer uses {len(services_used)} services, {recommended_plan} plan would provide better coverage",
                                "estimated_value": 500.0 if recommended_plan == "premium" else 250.0
                            })

    return opportunities


def _analyze_personalization_insights(
    customer_graph: Dict[str, Any]
) -> List[Dict[str, Any]]:
    """Generate personalization insights based on customer behavior"""
    insights = []
    profiles = customer_graph.get("customer_profiles", {})

    for unified_id, profile in profiles.items():
        business_units = profile.get("business_units", [])

        # Multi-unit customer insight
        if len(business_units) > 1:
            insights.append({
                "customer_id": unified_id,
                "insight_type": "cross_unit_presence",
                "description": f"Customer uses {len(business_units)} business units: {', '.join(business_units)}",
                "recommended_actions": [
                    "Create unified customer experience across all touchpoints",
                    "Offer bundled services across units",
                    "Use cross-unit data for personalized recommendations"
                ],
                "business_units_affected": business_units
            })

        # High-value customer insight
        total_spent = profile.get("total_spent", 0)
        account_balance = profile.get("account_balance", 0)
        if total_spent > 5000 or account_balance > 20000:
            insights.append({
                "customer_id": unified_id,
                "insight_type": "high_value_customer",
                "description": f"High-value customer (spent ${total_spent:.2f}, balance ${account_balance:.2f})",
                "recommended_actions": [
                    "Assign dedicated account manager",
                    "Offer premium tier benefits",
                    "Provide priority support"
                ],
                "business_units_affected": business_units
            })

        # Engagement pattern insight
        purchase_count = profile.get("purchase_count", 0)
        appointment_count = profile.get("appointment_count", 0)
        if purchase_count > 30 or appointment_count > 5:
            insights.append({
                "customer_id": unified_id,
                "insight_type": "high_engagement",
                "description": f"Highly engaged customer ({purchase_count} purchases, {appointment_count} appointments)",
                "recommended_actions": [
                    "Send loyalty rewards",
                    "Offer exclusive promotions",
                    "Request testimonials or referrals"
                ],
                "business_units_affected": business_units
            })

    return insights


def opportunity_analyzer_node(state: CustomerGraphState) -> CustomerGraphState:
    """
    Analyze customer graph to identify opportunities.

    Reads: customer_graph, ingested_data
    Writes: cross_sell_opportunities, upsell_opportunities, personalization_insights, graph_summary
    """
    logger.info("üîç Analyzing opportunities...")

    try:
        customer_graph = state.get("customer_graph", {})
        ingested_data = state.get("ingested_data", [])

        if not customer_graph:
            logger.warning("‚ö†Ô∏è  No customer graph available")
            state["cross_sell_opportunities"] = []
            state["upsell_opportunities"] = []
            state["personalization_insights"] = []
            return state

        # Find cross-sell opportunities
        cross_sell_opportunities = _analyze_cross_sell_opportunities(customer_graph, ingested_data)
        state["cross_sell_opportunities"] = cross_sell_opportunities

        # Find upsell opportunities
        upsell_opportunities = _analyze_upsell_opportunities(customer_graph)
        state["upsell_opportunities"] = upsell_opportunities

        # Generate personalization insights
        personalization_insights = _analyze_personalization_insights(customer_graph)
        state["personalization_insights"] = personalization_insights

        # Update graph summary with opportunity counts
        graph_summary = state.get("graph_summary", {})
        graph_summary["cross_sell_count"] = len(cross_sell_opportunities)
        graph_summary["upsell_count"] = len(upsell_opportunities)

        # Calculate total opportunity value
        total_value = (
            sum(opp.get("estimated_value", 0) or 0 for opp in cross_sell_opportunities) +
            sum(opp.get("estimated_value", 0) or 0 for opp in upsell_opportunities)
        )
        graph_summary["total_opportunity_value"] = round(total_value, 2)

        state["graph_summary"] = graph_summary

        logger.info(f"‚úÖ Opportunity analysis complete:")
        logger.info(f"   - {len(cross_sell_opportunities)} cross-sell opportunities")
        logger.info(f"   - {len(upsell_opportunities)} upsell opportunities")
        logger.info(f"   - {len(personalization_insights)} personalization insights")
        logger.info(f"   - Total opportunity value: ${total_value:,.2f}")

    except Exception as e:
        logger.error(f"‚ùå Error in opportunity_analyzer_node: {e}")
        import traceback
        logger.error(traceback.format_exc())
        state.setdefault("errors", []).append(f"Opportunity analysis failed: {str(e)}")
        state["cross_sell_opportunities"] = []
        state["upsell_opportunities"] = []
        state["personalization_insights"] = []

    return state

````

# Report Node

````python
"""Report node - generates insights report"""

import logging
from datetime import datetime
from pathlib import Path
from typing import Dict, Any
from jinja2 import Environment, FileSystemLoader
from config import CustomerGraphState, CustomerGraphConfig

logger = logging.getLogger(__name__)


def report_node(state: CustomerGraphState) -> CustomerGraphState:
    """
    Generate insights report from customer graph analysis.

    Reads: customer_graph, cross_sell_opportunities, upsell_opportunities,
           personalization_insights, graph_summary, business_units
    Writes: insights_report, report_file_path
    """
    logger.info("üìÑ Generating insights report...")

    try:
        config = CustomerGraphConfig()
        customer_graph = state.get("customer_graph", {})
        cross_sell_opportunities = state.get("cross_sell_opportunities", [])
        upsell_opportunities = state.get("upsell_opportunities", [])
        personalization_insights = state.get("personalization_insights", [])
        graph_summary = state.get("graph_summary", {})
        business_units = state.get("business_units", [])
        ingested_data = state.get("ingested_data", [])

        # Prepare data for template
        unit_data = {}
        for unit_info in ingested_data:
            unit_data[unit_info["business_unit"]] = {
                "record_count": unit_info["record_count"]
            }

        template_data = {
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "business_units": business_units,
            "unit_data": unit_data,
            "graph_summary": graph_summary,
            "cross_sell_opportunities": cross_sell_opportunities,
            "upsell_opportunities": upsell_opportunities,
            "personalization_insights": personalization_insights
        }

        # Load template
        template_dir = Path(__file__).parent.parent / "templates"
        env = Environment(loader=FileSystemLoader(str(template_dir)))
        template = env.get_template("insights_report.md.j2")

        # Render report
        report_content = template.render(**template_data)
        state["insights_report"] = report_content

        # Save report to file
        output_dir = Path(__file__).parent.parent.parent / config.customer_graph_reports_dir
        output_dir.mkdir(parents=True, exist_ok=True)

        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        report_filename = f"customer_graph_insights_{timestamp}.md"
        report_path = output_dir / report_filename

        with open(report_path, 'w', encoding='utf-8') as f:
            f.write(report_content)

        state["report_file_path"] = str(report_path)

        logger.info(f"‚úÖ Report generated: {report_path}")
        logger.info(f"   - {len(cross_sell_opportunities)} cross-sell opportunities")
        logger.info(f"   - {len(upsell_opportunities)} upsell opportunities")
        logger.info(f"   - {len(personalization_insights)} personalization insights")

    except Exception as e:
        logger.error(f"‚ùå Error in report_node: {e}")
        import traceback
        logger.error(traceback.format_exc())
        state.setdefault("errors", []).append(f"Report generation failed: {str(e)}")
        state["insights_report"] = f"Error generating report: {str(e)}"

    return state

````

````python
"""Cross-Business Customer Graph Orchestrator Agent - LangGraph Workflow"""

import sys
from pathlib import Path

# Add project root to path (for direct execution)
project_root = Path(__file__).parent.parent
sys.path.insert(0, str(project_root))

from langgraph.graph import StateGraph, END
from config import CustomerGraphState
from customer_graph_agent.nodes import (
    goal_node,
    planning_node,
    data_ingestion_node,
    graph_builder_node,
    opportunity_analyzer_node,
    report_node
)


def create_customer_graph_agent():
    """Create and compile the Customer Graph Orchestrator Agent"""

    # Create workflow
    workflow = StateGraph(CustomerGraphState)

    # Add nodes
    workflow.add_node("goal", goal_node)
    workflow.add_node("planning", planning_node)
    workflow.add_node("data_ingestion", data_ingestion_node)
    workflow.add_node("graph_building", graph_builder_node)
    workflow.add_node("opportunity_analysis", opportunity_analyzer_node)
    workflow.add_node("report_generation", report_node)

    # Linear flow (MVP pattern)
    workflow.add_edge("goal", "planning")
    workflow.add_edge("planning", "data_ingestion")
    workflow.add_edge("data_ingestion", "graph_building")
    workflow.add_edge("graph_building", "opportunity_analysis")
    workflow.add_edge("opportunity_analysis", "report_generation")
    workflow.add_edge("report_generation", END)

    # Set entry point
    workflow.set_entry_point("goal")

    # Compile and return
    agent = workflow.compile()
    return agent


# For direct execution
if __name__ == "__main__":
    import logging

    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
    )

    # Example usage
    initial_state: CustomerGraphState = {
        "data_sources": [
            "data/retail_customers.csv",
            "data/finance_customers.csv",
            "data/healthcare_customers.csv"
        ],
        "business_units": ["Retail", "Finance", "Healthcare"],
        "errors": []
    }

    agent = create_customer_graph_agent()
    result = agent.invoke(initial_state)

    print("\n" + "="*60)
    print("‚úÖ Agent execution complete!")
    print(f"Report saved to: {result.get('report_file_path', 'N/A')}")
    print("="*60)

````
