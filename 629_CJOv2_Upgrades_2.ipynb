## Tier 2 implemented

### 1. **Executive blockquote when above target**
- **Placement:** Right after the title and ‚ÄúGenerated‚Äù line, before the first `---`.
- **Condition:** Shown when **either** customers at high risk > `target_customers_at_high_risk_max` **or** revenue at risk > `target_revenue_at_risk_max`.
- **Content:**  
  `> **Executive attention required.** Customers at high risk / revenue at risk exceeds target; see next steps below.`
- With current targets (50 and \$3M) and current numbers (41, \$2.41M), the blockquote does **not** appear. To see it, raise targets in config or use a run where metrics exceed them.

### 2. **Progress over time**
- **Placement:** New section after ‚ÄúPortfolio summary‚Äù, before ‚ÄúSegment view‚Äù.
- **Logic:** Uses `portfolio_snapshots` sorted by date; compares **oldest** vs **latest** snapshot.
  - **Improving:** Customers at high risk or revenue at risk down by ‚â•5% over the period.
  - **Declining:** Either up by ‚â•5%.
  - **Stable:** Otherwise.
- **Content:** One-line **Trend** (Improving / Stable / Declining), a short **detail** line (e.g. ‚ÄúCustomers at high risk down 29%; revenue at risk down 24% over period‚Äù), and a **mini table** with the last 3 snapshots: Snapshot ID, Date, Customers at high risk, Revenue at risk.
- With your four snapshots (58‚Üí52‚Üí47‚Üí41, revenue at risk decreasing), the report will show **Trend: Improving** and the table.

### 3. **Segment R/Y/G level**
- **Config** (`config.py`):
  - `segment_revenue_at_risk_critical` = **200,000** ‚Üí **R** (critical) when segment revenue at risk ‚â• this.
  - `segment_revenue_at_risk_elevated` = **50,000** ‚Üí **Y** (elevated) when ‚â• this and < critical; **G** (monitor) below.
- **Report:** Segment table now has a **Level** column (R / Y / G) per segment.
- **Legend:** Under the table:  
  `Level: **R** = critical, **Y** = elevated, **G** = monitor (by segment revenue at risk vs config thresholds).`
- With current segment revenue (Enterprise \$220k, Mid-Market \$252k, SMB \$54k), Enterprise and Mid-Market will be **R**, SMB **Y**.
