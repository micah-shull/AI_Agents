````python
"""Report generation utilities"""

from typing import Dict, List, Any, Optional
from toolshed.reporting import save_report


def generate_executive_summary(
    ecosystem_summary: Dict[str, Any],
    prioritized_issues: List[Dict[str, Any]]
) -> str:
    """Generate executive summary section"""
    sections = []
    sections.append("## Executive Summary\n\n")

    # Business Impact
    sections.append("### ðŸ’° Business Impact\n\n")
    sections.append(f"- **Total Cost (30d):** ${ecosystem_summary.get('total_cost_30d', 0.0):,.2f}\n")
    sections.append(f"- **Total ROI Estimate (30d):** ${ecosystem_summary.get('total_roi_estimate', 0.0):,.2f}\n")
    net_roi = ecosystem_summary.get('total_roi_estimate', 0.0) - ecosystem_summary.get('total_cost_30d', 0.0)
    sections.append(f"- **Net ROI:** ${net_roi:,.2f}\n")
    sections.append(f"- **ROI Ratio:** {ecosystem_summary.get('total_roi_estimate', 0.0) / ecosystem_summary.get('total_cost_30d', 1.0):.2f}x\n\n")

    # Ecosystem Health
    sections.append("### ðŸ“Š Ecosystem Health\n\n")
    sections.append(f"- **Overall Health Score:** {ecosystem_summary.get('overall_health_score', 0.0):.1f}/100\n")
    sections.append(f"- **Active Agents:** {ecosystem_summary.get('active_agents', 0)}/{ecosystem_summary.get('total_agents', 0)}\n")
    sections.append(f"- **Healthy Systems:** {ecosystem_summary.get('healthy_systems', 0)}/{ecosystem_summary.get('total_systems', 0)}\n")
    sections.append(f"- **Degraded Systems:** {ecosystem_summary.get('degraded_systems', 0)}\n")
    sections.append(f"- **Critical Systems:** {ecosystem_summary.get('critical_systems', 0)}\n\n")

    # Risk Overview
    sections.append("### âš ï¸ Risk Overview\n\n")
    sections.append(f"- **Total Issues:** {ecosystem_summary.get('total_risks', 0)}\n")
    sections.append(f"- **High Priority Issues:** {ecosystem_summary.get('high_priority_risks', 0)}\n\n")

    # Top Priority Actions
    if prioritized_issues:
        sections.append("### ðŸŽ¯ Top Priority Actions\n\n")
        for i, issue in enumerate(prioritized_issues[:5], 1):
            severity_emoji = "ðŸ”´" if issue.get("severity") == "critical" else "ðŸŸ¡" if issue.get("severity") == "high" else "ðŸŸ¢"
            sections.append(f"{i}. {severity_emoji} **{issue.get('type', 'unknown').title()} Issue:** {issue.get('description', 'N/A')}\n")
            sections.append(f"   - Priority Score: {issue.get('priority_score', 0.0):.1f}\n")
            if issue.get("recommended_action"):
                sections.append(f"   - Action: {issue.get('recommended_action')}\n")
        sections.append("\n")

    return "".join(sections)


def generate_integration_health_section(integration_health: List[Dict[str, Any]]) -> str:
    """Generate integration health section"""
    sections = []
    sections.append("## Integration Health Analysis\n\n")

    if not integration_health:
        sections.append("No integration data available.\n\n")
        return "".join(sections)

    sections.append("| System ID | Health Status | Overall Score | Uptime | Latency | Auth | Affected Agents |\n")
    sections.append("|-----------|---------------|---------------|--------|---------|------|-----------------|\n")

    for health in sorted(integration_health, key=lambda x: x.get("overall_score", 0.0)):
        status_emoji = "âœ…" if health.get("health_status") == "healthy" else "âš ï¸" if health.get("health_status") == "degraded" else "ðŸ”´"
        affected = ", ".join(health.get("affected_agents", []))[:30] + "..." if len(health.get("affected_agents", [])) > 1 else (health.get("affected_agents", [""])[0] if health.get("affected_agents") else "None")

        sections.append(
            f"| {health.get('system_id', 'N/A')} | {status_emoji} {health.get('health_status', 'unknown')} | "
            f"{health.get('overall_score', 0.0):.1f} | {health.get('uptime_score', 0.0):.1f} | "
            f"{health.get('latency_score', 0.0):.1f} | {health.get('auth_score', 0.0):.1f} | {affected} |\n"
        )

    sections.append("\n")
    return "".join(sections)


def generate_risk_analysis_section(risk_assessments: List[Dict[str, Any]]) -> str:
    """Generate risk analysis section"""
    sections = []
    sections.append("## Risk Assessment\n\n")

    if not risk_assessments:
        sections.append("No risk data available.\n\n")
        return "".join(sections)

    sections.append("| Agent ID | Risk Level | Risk Score | Integration Risks | Operational Risks | Cost Risks |\n")
    sections.append("|----------|------------|------------|-------------------|-------------------|------------|\n")

    for assessment in sorted(risk_assessments, key=lambda x: x.get("total_risk_score", 0.0), reverse=True):
        risk_level = assessment.get("risk_level", "unknown")
        level_emoji = "ðŸ”´" if risk_level == "critical" else "ðŸŸ¡" if risk_level == "high" else "ðŸŸ¢"

        sections.append(
            f"| {assessment.get('agent_id', 'N/A')} | {level_emoji} {risk_level} | "
            f"{assessment.get('total_risk_score', 0.0):.1f} | "
            f"{len(assessment.get('integration_risks', []))} | "
            f"{len(assessment.get('operational_risks', []))} | "
            f"{len(assessment.get('cost_risks', []))} |\n"
        )

    sections.append("\n")
    return "".join(sections)


def generate_value_leakage_section(value_leakage_analysis: List[Dict[str, Any]]) -> str:
    """Generate value leakage analysis section (v2)"""
    sections = []
    sections.append("## Value Leakage Analysis\n\n")

    if not value_leakage_analysis:
        sections.append("No value leakage data available.\n\n")
        return "".join(sections)

    sections.append("| Agent ID | Leakage Score | Expected ROI | Actual ROI | ROI Gap | Cost Overrun | Trend |\n")
    sections.append("|----------|---------------|--------------|------------|---------|--------------|-------|\n")

    for analysis in sorted(value_leakage_analysis, key=lambda x: x.get("value_leakage_score", 0.0), reverse=True):
        trend_emoji = analysis.get("trend", {}).get("indicator", "â€”") if isinstance(analysis.get("trend"), dict) else analysis.get("trend", "â€”")

        sections.append(
            f"| {analysis.get('agent_id', 'N/A')} | {analysis.get('value_leakage_score', 0.0):.1f} | "
            f"${analysis.get('expected_roi', 0.0):,.2f} | ${analysis.get('actual_roi', 0.0):,.2f} | "
            f"${analysis.get('roi_gap', 0.0):,.2f} ({analysis.get('roi_gap_percent', 0.0):.1f}%) | "
            f"${analysis.get('cost_overrun', 0.0):,.2f} | {trend_emoji} |\n"
        )

    sections.append("\n")
    return "".join(sections)


def generate_prioritized_issues_section(prioritized_issues: List[Dict[str, Any]]) -> str:
    """Generate prioritized issues section"""
    sections = []
    sections.append("## Prioritized Issues\n\n")

    if not prioritized_issues:
        sections.append("No issues identified.\n\n")
        return "".join(sections)

    sections.append("| Priority | Type | Agent/System | Severity | Score | Description |\n")
    sections.append("|----------|------|--------------|----------|-------|-------------|\n")

    for i, issue in enumerate(prioritized_issues[:20], 1):  # Top 20
        severity_emoji = "ðŸ”´" if issue.get("severity") == "critical" else "ðŸŸ¡" if issue.get("severity") == "high" else "ðŸŸ¢"
        entity = issue.get("agent_id") or issue.get("system_id") or issue.get("workflow_id") or "N/A"

        sections.append(
            f"| {i} | {issue.get('type', 'unknown')} | {entity} | {severity_emoji} {issue.get('severity', 'unknown')} | "
            f"{issue.get('priority_score', 0.0):.1f} | {issue.get('description', 'N/A')[:50]}... |\n"
        )

    sections.append("\n")
    return "".join(sections)


def generate_risk_management_report(
    ecosystem_summary: Dict[str, Any],
    integration_health: List[Dict[str, Any]],
    risk_assessments: List[Dict[str, Any]],
    value_leakage_analysis: List[Dict[str, Any]],
    prioritized_issues: List[Dict[str, Any]],
    historical_trends: Optional[List[Dict[str, Any]]] = None
) -> str:
    """Generate complete risk management report"""
    sections = []

    # Title
    sections.append("# Integration & Risk Management Report\n\n")
    sections.append(f"**Generated:** {__import__('datetime').datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n\n")
    sections.append("---\n\n")

    # Executive Summary
    sections.append(generate_executive_summary(ecosystem_summary, prioritized_issues))
    sections.append("\n---\n\n")

    # Integration Health
    sections.append(generate_integration_health_section(integration_health))
    sections.append("\n---\n\n")

    # Risk Assessment
    sections.append(generate_risk_analysis_section(risk_assessments))
    sections.append("\n---\n\n")

    # Value Leakage (v2)
    sections.append(generate_value_leakage_section(value_leakage_analysis))
    sections.append("\n---\n\n")

    # Prioritized Issues
    sections.append(generate_prioritized_issues_section(prioritized_issues))

    return "".join(sections)
````

````python

def report_generation_node(
    state: IntegrationRiskManagementOrchestratorState,
    config: IntegrationRiskManagementOrchestratorConfig
) -> Dict[str, Any]:
    """Report Generation Node: Generate executive risk management report"""
    errors = state.get("errors", [])
    ecosystem_summary = state.get("ecosystem_summary", {})
    integration_health = state.get("integration_health", [])
    risk_assessments = state.get("risk_assessments", [])
    value_leakage_analysis = state.get("value_leakage_analysis", [])
    prioritized_issues = state.get("prioritized_issues", [])
    historical_trends = state.get("historical_trends", [])

    if not ecosystem_summary:
        return {
            "errors": errors + ["report_generation_node: ecosystem_summary required"]
        }

    try:
        # Generate report
        report = generate_risk_management_report(
            ecosystem_summary,
            integration_health,
            risk_assessments,
            value_leakage_analysis,
            prioritized_issues,
            historical_trends
        )

        # Save report
        report_id = f"risk_report_{__import__('datetime').datetime.now().strftime('%Y%m%d_%H%M%S')}"
        report_file_path = save_report(
            report,
            report_id,
            reports_dir=config.reports_dir,
            prefix="integration_risk_report"
        )

        return {
            "risk_management_report": report,
            "report_file_path": report_file_path,
            "errors": errors
        }
    except Exception as e:
        return {
            "errors": errors + [f"report_generation_node: {str(e)}"]
        }
````
