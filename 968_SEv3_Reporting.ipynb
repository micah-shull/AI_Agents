````python
from __future__ import annotations

from datetime import datetime
from typing import Any, Dict, List, Optional


def _fmt_usd(x: Any) -> str:
    try:
        return f"${float(x):,.0f}"
    except Exception:
        return "$0"


def _fmt_pct(x: Any, decimals: int = 0) -> str:
    try:
        return f"{float(x) * 100:.{decimals}f}%"
    except Exception:
        return "0%"


def _trigger_summary_from_deal_metric(deal_metric: Dict[str, Any]) -> List[str]:
    return deal_metric.get("fired_triggers_current") or []


def build_sev3_report_md(
    *,
    data_counts: Dict[str, Any],
    temporal_metrics: Dict[str, Any],
    enablement_report_md: str,
) -> str:
    deal_metrics_by_deal_id: Dict[str, Any] = temporal_metrics.get("deal_metrics_by_deal_id") or {}
    rep_metrics_by_rep_id: Dict[str, Any] = temporal_metrics.get("rep_metrics_by_rep_id") or {}
    pipeline_metrics: Dict[str, Any] = temporal_metrics.get("pipeline_metrics") or {}
    pipeline_triggers: Dict[str, Any] = temporal_metrics.get("pipeline_triggers") or {}
    forecast_stability: Dict[str, Any] = temporal_metrics.get("forecast_stability") or {}

    pipeline_period = pipeline_metrics.get("current_period") or "unknown-period"
    revenue_trajectory = pipeline_metrics.get("revenue_trajectory") or "stable"

    # Top deals: prioritize fired triggers, then risk score.
    deals_sorted = sorted(
        deal_metrics_by_deal_id.values(),
        key=lambda d: (len(d.get("fired_triggers_current") or []), float(d.get("deal_risk_score_current") or 0.0)),
        reverse=True,
    )
    top_deals = deals_sorted[:8]

    # Deal table.
    deal_lines: List[str] = []
    deal_lines.append("| Deal | Stage | Value | Prob | Prob Trend | Risk Score | Risk Velocity | Momentum | Stalled? | Triggers |")
    deal_lines.append("|---|---|---:|---:|---:|---:|---|---|---:|---|")
    for d in top_deals:
        deal_id = d.get("deal_id", "")
        stage = d.get("stage_current", "")
        val = _fmt_usd(d.get("deal_value_usd_current") or 0.0)
        prob = d.get("probability_current", 0.0)
        prob_trend = d.get("probability_trend_current", 0.0)
        risk_score = d.get("deal_risk_score_current", 0.0)
        risk_vel = d.get("risk_velocity_current", "stable")
        momentum = d.get("momentum_current", "stable")
        stalled = "YES" if d.get("stagnation_flag_current") else "no"
        triggers = ", ".join(d.get("fired_triggers_current") or []) or "—"

        deal_lines.append(
            f"| {deal_id} | {stage} | {val} | {prob:.2f} | {prob_trend:.3f} | {risk_score:.3f} | {risk_vel} | {momentum} | {stalled} | {triggers} |"
        )

    # Rep table.
    reps_sorted = sorted(
        rep_metrics_by_rep_id.values(),
        key=lambda r: (len(r.get("fired_triggers_current") or []), float(r.get("rep_score_current") or 0.0)),
        reverse=True,
    )
    rep_lines: List[str] = []
    rep_lines.append("| Rep | Close Rate | Close Trend | Pipeline Value | Productivity | Volatility | Rep Risk | Triggers |")
    rep_lines.append("|---|---:|---:|---:|---:|---:|---|---|")
    for r in reps_sorted[:6]:
        rep_id = r.get("rep_id", "")
        close_rate = float(r.get("close_rate_current") or 0.0)
        close_trend = float(r.get("close_rate_trend_current") or 0.0)
        pipe_val = _fmt_usd(r.get("pipeline_value_current") or 0.0)
        prod = float(r.get("pipeline_productivity_current") or 0.0)
        vol = float(r.get("performance_volatility_std_dev") or 0.0)
        rep_risk = "YES" if r.get("rep_risk_flag_current") else "no"
        triggers = ", ".join(r.get("fired_triggers_current") or []) or "—"
        rep_lines.append(
            f"| {rep_id} | {_fmt_pct(close_rate, 0)} | {close_trend:.3f} | {pipe_val} | {_fmt_usd(prod)} | {vol:.3f} | {rep_risk} | {triggers} |"
        )

    coverage_ratio = pipeline_metrics.get("coverage_ratio_current")
    coverage_ratio_s = f"{coverage_ratio:.2f}" if isinstance(coverage_ratio, (int, float)) else "n/a"

    avg_stability = forecast_stability.get("avg_deal_forecast_stability_std_dev", 0.0)

    new_deal_triggers = 0
    for dm in deal_metrics_by_deal_id.values():
        for _, tmeta in (dm.get("triggers") or {}).items():
            if tmeta.get("new_trigger_flag"):
                new_deal_triggers += 1

    new_pipeline_triggers = 0
    for _, tmeta in (pipeline_triggers or {}).items():
        if tmeta.get("new_trigger_flag"):
            new_pipeline_triggers += 1

    return "\n".join(
        [
            "# Sales Enablement Orchestrator v3 (SEv3)",
            "",
            "## Run Context",
            f"- **Current Period:** {pipeline_period}",
            f"- **Revenue Trajectory:** **{revenue_trajectory}**",
            f"- **Forecast Stability (avg deal std dev of probability):** {avg_stability:.3f}",
            f"- **Dataset scale:** deals={data_counts.get('deals_count', 0)}, reps={data_counts.get('reps_count', 0)}, interactions={data_counts.get('interactions_count', 0)}, deal_history_deals={data_counts.get('deal_history_deals_count', 0)}",
            "",
            "## Deal Layer (Temporal Risk & Momentum)",
            "",
            *deal_lines,
            "",
            "## Rep Layer (Execution Visibility)",
            "",
            *rep_lines,
            "",
            "## Pipeline Layer (Executive Signals)",
            "",
            "| Metric | Value |",
            "|---|---:|",
            f"| Total pipeline value (current) | {_fmt_usd(pipeline_metrics.get('total_pipeline_value_current') or 0)} |",
            f"| Pipeline growth rate (vs prev) | {pipeline_metrics.get('pipeline_growth_rate', 0):.2%} |",
            f"| Weighted pipeline value (current) | {_fmt_usd(pipeline_metrics.get('weighted_pipeline_value_current') or 0)} |",
            f"| Weighted pipeline trend (vs avg prev 3) | {pipeline_metrics.get('weighted_pipeline_trend', 0):,.0f} |",
            f"| Win rate (current) | {_fmt_pct(pipeline_metrics.get('win_rate_current') or 0, 0)} |",
            f"| Win rate trend (vs avg prev 3) | {pipeline_metrics.get('win_rate_trend', 0):.3f} |",
            f"| Coverage ratio (total pipeline / sum quota) | {coverage_ratio_s} (threshold {pipeline_metrics.get('coverage_ratio_threshold', 'n/a')}) |",
            "",
            "## Trigger Evolution (new signals)",
            "",
            f"- **New deal triggers fired this run:** {new_deal_triggers}",
            f"- **New pipeline triggers fired this run:** {new_pipeline_triggers}",
            "",
            "## Enablement Actions (Current Snapshot)",
            "",
            enablement_report_md,
        ]
    )

````
