# 🧾 **One-Line Summary (Portfolio Gold)**

> **This engine transforms structured sales data into deterministic, time-aware intelligence that reveals trajectory, detects emerging risk, and enables confident revenue decisions.**




````python
from __future__ import annotations

import math
from dataclasses import dataclass
from datetime import date, datetime, timedelta, timezone
from statistics import mean, pstdev
from typing import Any, Dict, List, Optional, Tuple


def _avg(values: List[float]) -> float:
    return float(mean(values)) if values else 0.0


def _std_dev(values: List[float]) -> float:
    # Use population stdev for deterministic results with small N.
    return float(pstdev(values)) if values else 0.0


def _parse_period_end(period: str) -> date:
    """Convert YYYY-MM into UTC month-end date."""
    year_s, month_s = period.split("-")
    year, month = int(year_s), int(month_s)
    if month == 12:
        next_month = datetime(year + 1, 1, 1, tzinfo=timezone.utc)
    else:
        next_month = datetime(year, month + 1, 1, tzinfo=timezone.utc)
    end_dt = next_month - timedelta(days=1)
    return end_dt.date()


def _parse_iso_datetime_to_utc_date(dt_raw: str) -> Optional[date]:
    if not dt_raw:
        return None
    try:
        dt = datetime.fromisoformat(dt_raw.replace("Z", "+00:00")).astimezone(timezone.utc)
        return dt.date()
    except Exception:
        return None


def _consecutive_true_ending_at_end(flags: List[bool]) -> int:
    """Count consecutive True values at the end of a boolean list."""
    n = 0
    for f in reversed(flags):
        if f:
            n += 1
        else:
            break
    return n


def _trigger_evolution(flags: List[bool]) -> Dict[str, Any]:
    if not flags:
        return {
            "trigger_status_current": "not_fired",
            "trigger_status_previous": "not_fired",
            "trigger_duration_periods": 0,
            "new_trigger_flag": False,
        }
    current = bool(flags[-1])
    prev = bool(flags[-2]) if len(flags) >= 2 else False
    duration = _consecutive_true_ending_at_end(flags)
    return {
        "trigger_status_current": "fired" if current else "not_fired",
        "trigger_status_previous": "fired" if prev else "not_fired",
        "trigger_duration_periods": duration,
        "new_trigger_flag": current and not prev,
    }


def _risk_velocity_class(risk_delta: float, *, threshold: float = 0.1) -> str:
    if risk_delta > threshold:
        return "increasing"
    if risk_delta < -threshold:
        return "decreasing"
    return "stable"


def compute_deal_temporal_metrics(
    *,
    deals: List[Dict[str, Any]],
    deal_history: List[Dict[str, Any]],
    interactions: List[Dict[str, Any]],
    signals_lookup: Dict[str, Dict[str, Any]],
    stage_actions: List[Dict[str, Any]],
    thresholds: Dict[str, Any],
) -> Tuple[Dict[str, Any], Dict[str, Any]]:
    """
    Compute deal-layer metrics and deal-level trigger evolution.

    Returns:
      (deal_metrics_by_deal_id, portfolio_prob_trend)
    """
    stalled_threshold = float(thresholds.get("stalled_deal_days_in_stage", 21))
    min_interaction_recency_days = int(thresholds.get("min_interaction_recency_days", 7))
    high_value_min_usd = float(thresholds.get("high_value_deal_min_usd", 100000))

    stage_actions_by_stage = {s.get("stage"): s for s in (stage_actions or []) if s.get("stage")}

    # Index deals for created_date / current values.
    deals_by_id = {d.get("deal_id"): d for d in deals if d.get("deal_id")}

    # Index interactions by deal_id and by date for quick period-based slicing.
    interactions_by_deal: Dict[str, List[date]] = {}
    interactions_sorted_by_deal: Dict[str, List[date]] = {}
    for i in interactions or []:
        deal_id = i.get("deal_id")
        if not deal_id:
            continue
        d = _parse_iso_datetime_to_utc_date(i.get("datetime") or "")
        if d is None:
            continue
        interactions_by_deal.setdefault(deal_id, []).append(d)
    for deal_id, dates in interactions_by_deal.items():
        interactions_sorted_by_deal[deal_id] = sorted(dates)

    # Compute risk + probability trend per deal over its deal_history timeline.
    deal_metrics: Dict[str, Any] = {}

    probability_trends_for_portfolio: List[float] = []

    for dh in deal_history or []:
        deal_id = dh.get("deal_id")
        if not deal_id:
            continue
        history = dh.get("monthly_history") or []
        if not history:
            continue

        # Sort by period.
        points = sorted(history, key=lambda x: x.get("period") or "")
        if not points:
            continue

        current_point = points[-1]
        prev_point = points[-2] if len(points) >= 2 else None

        deal = deals_by_id.get(deal_id)
        if not deal:
            # If we don't have current deal metadata, we can't compute days-in-stage or momentum triggers.
            continue

        created_date = None
        try:
            created_date = datetime.fromisoformat(deal.get("created_date")).date()
        except Exception:
            created_date = None
        if created_date is None:
            # Fallback: approximate created_date as first period end.
            created_date = _parse_period_end(points[0].get("period"))

        lead_id = deal.get("lead_id")
        sig = signals_lookup.get(lead_id) or {}
        engagement_score = float(sig.get("engagement_score", 0.5))
        deal_risk_score_input = float(sig.get("deal_risk_score", 0.5))

        # Precompute per-period metric series.
        prob_series: List[float] = []
        velocity_series: List[float] = []
        probability_trend_series: List[float] = []
        risk_score_series: List[float] = []
        risk_delta_series: List[float] = []
        risk_velocity_class_series: List[str] = []
        interaction_gap_series: List[int] = []
        interactions_last_14_series: List[int] = []
        stage_changed_recently_series: List[bool] = []
        deal_value_series: List[float] = []
        days_in_stage_series: List[int] = []
        stagnation_flag_series: List[bool] = []
        momentum_series: List[str] = []

        interactions_dates = interactions_sorted_by_deal.get(deal_id, [])

        last_stage = None
        for idx, p in enumerate(points):
            period = p.get("period")
            period_end = _parse_period_end(period)
            days_in_stage = max(0, (period_end - created_date).days)
            days_in_stage_series.append(days_in_stage)
            deal_value = float(p.get("deal_value_usd") or deal.get("deal_value_usd") or 0.0)
            deal_value_series.append(deal_value)

            stage = p.get("stage") or deal.get("stage")
            exp_days = stage_actions_by_stage.get(stage, {}).get("if_stalled_days")
            if exp_days is None:
                exp_days = stalled_threshold  # fallback: use stagnation threshold
            exp_days_f = float(exp_days) if exp_days else float(stalled_threshold)
            velocity_ratio = (days_in_stage / exp_days_f) if exp_days_f > 0 else 1.0
            velocity_series.append(float(velocity_ratio))

            prob = float(p.get("probability") or deal.get("probability") or 0.0)
            prob_series.append(prob)

            # Probability trend uses avg of previous 3 probabilities within the deal history.
            prev_probs = prob_series[:-1][-3:]
            probability_trend = prob - _avg(prev_probs)
            probability_trend_series.append(float(probability_trend))

            risk_score = (
                (1.0 - engagement_score) * 0.3
                + deal_risk_score_input * 0.3
                + max(0.0, -probability_trend) * 0.2
                + min(1.0, velocity_ratio / 2.0) * 0.2
            )
            risk_score_series.append(float(risk_score))

            if idx == 0:
                risk_delta_series.append(0.0)
                risk_velocity_class_series.append("stable")
            else:
                risk_delta = risk_score_series[idx] - risk_score_series[idx - 1]
                risk_delta_series.append(float(risk_delta))
                risk_velocity_class_series.append(_risk_velocity_class(risk_delta))

            # Interactions features for this period.
            # Count interactions in the last 14 days ending at the period end date.
            from_date = period_end - timedelta(days=14)
            last_14_count = sum(1 for d in interactions_dates if from_date <= d <= period_end)
            interactions_last_14_series.append(int(last_14_count))

            # Interaction gap = days since last interaction on/before period end.
            last_interaction_date = None
            for d in reversed(interactions_dates):
                if d <= period_end:
                    last_interaction_date = d
                    break
            if last_interaction_date is None:
                gap_days = 9999
            else:
                gap_days = (period_end - last_interaction_date).days
            interaction_gap_series.append(int(gap_days))

            # Stage change from previous period.
            stage_changed_recently = False
            if last_stage is not None:
                stage_changed_recently = (stage != last_stage)
            stage_changed_recently_series.append(bool(stage_changed_recently))
            last_stage = stage

            stagnation_flag = days_in_stage > stalled_threshold
            stagnation_flag_series.append(bool(stagnation_flag))

            # Momentum definition from spec.
            if last_14_count >= 2 and stage_changed_recently:
                momentum_series.append("accelerating")
            elif last_14_count == 0:
                momentum_series.append("stalled")
            else:
                momentum_series.append("stable")

        # Collect portfolio probability trend (for revenue trajectory).
        if probability_trend_series:
            probability_trends_for_portfolio.append(probability_trend_series[-1])

        # Triggers across periods (for evolution).
        high_value_deteriorating_flags: List[bool] = []
        stalled_flags: List[bool] = []
        no_interaction_flags: List[bool] = []
        risk_velocity_increasing_flags: List[bool] = []

        for idx in range(len(points)):
            deal_value = deal_value_series[idx]
            probability_trend = probability_trend_series[idx]
            high_value_deteriorating_flags.append(
                deal_value >= high_value_min_usd and probability_trend < 0.0
            )
            stalled_flags.append(days_in_stage_series[idx] > stalled_threshold)
            no_interaction_flags.append(interaction_gap_series[idx] > min_interaction_recency_days)
            risk_velocity_increasing_flags.append(
                risk_velocity_class_series[idx] == "increasing"
            )

        latest_idx = len(points) - 1
        prev_idx = latest_idx - 1

        deal_metrics[deal_id] = {
            "deal_id": deal_id,
            "current_period": points[latest_idx].get("period"),
            "stage_current": points[latest_idx].get("stage"),
            "deal_value_usd_current": deal_value_series[latest_idx],
            "probability_current": prob_series[latest_idx],
            "probability_trend_current": probability_trend_series[latest_idx],
            "velocity_ratio_current": velocity_series[latest_idx],
            "deal_risk_score_current": risk_score_series[latest_idx],
            "risk_velocity_current": risk_velocity_class_series[latest_idx],
            "risk_delta_current": risk_delta_series[latest_idx],
            "forecast_stability_std_dev": _std_dev(prob_series),
            "stagnation_flag_current": stagnation_flag_series[latest_idx],
            "interaction_last_14_days_count": interactions_last_14_series[latest_idx],
            "interaction_gap_days": interaction_gap_series[latest_idx],
            "momentum_current": momentum_series[latest_idx],
            "risk_velocity_class_history": risk_velocity_class_series,
            "probability_history": prob_series,
            "triggers": {
                "deal_high_value_deteriorating": {
                    **_trigger_evolution(high_value_deteriorating_flags),
                    "trigger_name": "high_value + deteriorating probability",
                    "high_value_threshold_usd": high_value_min_usd,
                },
                "deal_stalled_beyond_threshold": {
                    **_trigger_evolution(stalled_flags),
                    "trigger_name": "stalled beyond threshold",
                    "stalled_threshold_days": stalled_threshold,
                },
                "no_interaction_in_x_days": {
                    **_trigger_evolution(no_interaction_flags),
                    "trigger_name": f"no interaction for > {min_interaction_recency_days} days",
                    "min_interaction_recency_days": min_interaction_recency_days,
                },
                "risk_velocity_increasing": {
                    **_trigger_evolution(risk_velocity_increasing_flags),
                    "trigger_name": "risk velocity increasing",
                    "risk_delta_threshold": 0.1,
                },
            },
            # Convenience: fired triggers at current period.
            "fired_triggers_current": [
                t_key
                for t_key, t_val in {
                    "deal_high_value_deteriorating": high_value_deteriorating_flags[latest_idx],
                    "deal_stalled_beyond_threshold": stalled_flags[latest_idx],
                    "no_interaction_in_x_days": no_interaction_flags[latest_idx],
                    "risk_velocity_increasing": risk_velocity_increasing_flags[latest_idx],
                }.items()
                if t_val
            ],
        }

    portfolio_probability_trend = _avg(probability_trends_for_portfolio)
    return deal_metrics, {"avg_deal_probability_trend_current": portfolio_probability_trend}


def compute_rep_temporal_metrics(
    *,
    deals: List[Dict[str, Any]],
    sales_reps: List[Dict[str, Any]],
    rep_performance_history: List[Dict[str, Any]],
    deal_history: List[Dict[str, Any]],
    thresholds: Dict[str, Any],
) -> Dict[str, Any]:
    """
    Rep-layer temporal metrics + rep-level trigger evolution.

    Note: deal_history is used only to infer "high-value deals consistently lost".
    """
    high_value_min_usd = float(thresholds.get("high_value_deal_min_usd", 100000))

    # Current active deals per rep (from deals.json snapshot).
    active_deals_by_rep: Dict[str, int] = {}
    deals_value_by_rep: Dict[str, float] = {}
    for d in deals or []:
        rep_id = d.get("rep_id")
        if not rep_id:
            continue
        if d.get("status") == "active":
            active_deals_by_rep[rep_id] = active_deals_by_rep.get(rep_id, 0) + 1
            deals_value_by_rep[rep_id] = deals_value_by_rep.get(rep_id, 0.0) + float(
                d.get("deal_value_usd", 0.0)
            )

    reps_by_id = {r.get("rep_id"): r for r in sales_reps or [] if r.get("rep_id")}

    # Build high-value lost deal counts per rep per period from deal_history.
    # deal_history monthly_history provides stage/probability/deal_value per period.
    # We count deals with stage == "Closed Lost" and deal_value >= threshold.
    lost_counts_by_rep_period: Dict[str, Dict[str, int]] = {}

    for dh in deal_history or []:
        deal_id = dh.get("deal_id")
        points = dh.get("monthly_history") or []
        if not points:
            continue
        # Need rep_id for this deal_id from deals snapshot.
        rep_id = None
        for d in deals or []:
            if d.get("deal_id") == deal_id:
                rep_id = d.get("rep_id")
                break
        if not rep_id:
            continue

        points_sorted = sorted(points, key=lambda x: x.get("period") or "")
        for p in points_sorted:
            period = p.get("period")
            if not period:
                continue
            stage = p.get("stage")
            deal_value = float(p.get("deal_value_usd") or 0.0)
            is_lost = stage == "Closed Lost"
            is_high = deal_value >= high_value_min_usd
            if is_lost and is_high:
                lost_counts_by_rep_period.setdefault(rep_id, {}).setdefault(period, 0)
                lost_counts_by_rep_period[rep_id][period] += 1

    rep_metrics: Dict[str, Any] = {}
    # For productivity normalization across reps.
    productivity_by_rep: Dict[str, float] = {}

    for rep_entry in rep_performance_history or []:
        rep_id = rep_entry.get("rep_id")
        if not rep_id:
            continue
        hist = rep_entry.get("monthly_history") or []
        if not hist:
            continue
        points = sorted(hist, key=lambda x: x.get("period") or "")
        current = points[-1]
        current_close = float(current.get("close_rate") or 0.0)
        current_pipeline = float(current.get("pipeline_value") or 0.0)

        prev_close_rates = [float(x.get("close_rate") or 0.0) for x in points[-4:-1]]  # up to 3
        prev_avg_close = _avg(prev_close_rates)
        close_trend = current_close - prev_avg_close

        prev_pipeline_values = [float(x.get("pipeline_value") or 0.0) for x in points[-4:-1]]
        prev_avg_pipeline = _avg(prev_pipeline_values)
        pipeline_growth = current_pipeline - prev_avg_pipeline

        active_deals = active_deals_by_rep.get(rep_id) or int(
            reps_by_id.get(rep_id, {}).get("active_deals") or 0
        ) or 1
        productivity = current_pipeline / max(active_deals, 1)
        productivity_by_rep[rep_id] = productivity

        volatility = _std_dev([float(x.get("close_rate") or 0.0) for x in points])

        rep_metrics[rep_id] = {
            "rep_id": rep_id,
            "current_period": current.get("period"),
            "close_rate_current": current_close,
            "close_rate_trend_current": close_trend,
            "pipeline_value_current": current_pipeline,
            "pipeline_growth_current": pipeline_growth,
            "active_deals_current": active_deals,
            "pipeline_productivity_current": productivity,
            "performance_volatility_std_dev": volatility,
            "timeline": {
                "periods": [x.get("period") for x in points],
                "close_rates": [float(x.get("close_rate") or 0.0) for x in points],
                "pipeline_values": [float(x.get("pipeline_value") or 0.0) for x in points],
            },
        }

    # Normalize productivity + compute rep score + triggers with evolution metadata.
    max_prod = max(productivity_by_rep.values()) if productivity_by_rep else 1.0
    max_pos_growth = 0.0
    for rep_id, rm in rep_metrics.items():
        growth = float(rm.get("pipeline_growth_current") or 0.0)
        max_pos_growth = max(max_pos_growth, growth)
    if max_prod <= 0:
        max_prod = 1.0

    # Evaluate trigger evolution per rep.
    for rep_entry in rep_performance_history or []:
        rep_id = rep_entry.get("rep_id")
        if not rep_id or rep_id not in rep_metrics:
            continue
        hist = rep_entry.get("monthly_history") or []
        points = sorted(hist, key=lambda x: x.get("period") or "")
        if len(points) < 2:
            continue

        # Pull boolean series.
        close_rates = [float(x.get("close_rate") or 0.0) for x in points]
        pipeline_values = [float(x.get("pipeline_value") or 0.0) for x in points]
        periods = [x.get("period") for x in points]

        # Close rate declining for 2+ periods (two consecutive declines) ending at i.
        close_declining_flags: List[bool] = []
        pipeline_shrinking_flags: List[bool] = []
        high_value_lost_consistently_flags: List[bool] = []

        # Precompute lost high-value flag per period for this rep.
        lost_high_value_per_period = {
            period: (lost_counts_by_rep_period.get(rep_id, {}).get(period, 0) > 0)
            for period in periods
        }

        for i in range(len(points)):
            if i >= 2:
                close_declining = close_rates[i] < close_rates[i - 1] and close_rates[i - 1] < close_rates[i - 2]
                close_declining_flags.append(bool(close_declining))
                pipeline_shrinking = pipeline_values[i] < pipeline_values[i - 1] and pipeline_values[i - 1] < pipeline_values[i - 2]
                pipeline_shrinking_flags.append(bool(pipeline_shrinking))
            else:
                close_declining_flags.append(False)
                pipeline_shrinking_flags.append(False)

            if i >= 1:
                high_value_lost_consistent = lost_high_value_per_period.get(periods[i], False) and lost_high_value_per_period.get(
                    periods[i - 1], False
                )
                high_value_lost_consistently_flags.append(bool(high_value_lost_consistent))
            else:
                high_value_lost_consistently_flags.append(False)

        latest_idx = len(points) - 1

        rm = rep_metrics[rep_id]
        productivity = float(rm.get("pipeline_productivity_current") or 0.0)
        productivity_normalized = productivity / max_prod

        pipeline_growth = float(rm.get("pipeline_growth_current") or 0.0)
        pipeline_growth_normalized = (pipeline_growth / max_pos_growth) if max_pos_growth > 0 else 0.0
        pipeline_growth_normalized = max(0.0, pipeline_growth_normalized)

        close_rate_current = float(rm.get("close_rate_current") or 0.0)
        rep_score = (
            close_rate_current * 0.4 + productivity_normalized * 0.3 + pipeline_growth_normalized * 0.3
        )

        rm.update(
            {
                "rep_score_current": float(rep_score),
                "rep_risk_flag_current": bool(
                    close_declining_flags[latest_idx]
                    and (pipeline_values[latest_idx] < pipeline_values[latest_idx - 1])
                )
                if latest_idx >= 1
                else False,
                "triggers": {
                    "close_rate_declining_2plus": {
                        **_trigger_evolution(close_declining_flags),
                        "trigger_name": "close rate declining 2+ periods",
                    },
                    "pipeline_shrinking": {
                        **_trigger_evolution(pipeline_shrinking_flags),
                        "trigger_name": "pipeline shrinking",
                    },
                    "high_value_deals_consistently_lost": {
                        **_trigger_evolution(high_value_lost_consistently_flags),
                        "trigger_name": "high-value deals consistently lost",
                        "high_value_threshold_usd": high_value_min_usd,
                    },
                },
                "fired_triggers_current": [
                    t_key
                    for t_key, fired in {
                        "close_rate_declining_2plus": close_declining_flags[latest_idx],
                        "pipeline_shrinking": pipeline_shrinking_flags[latest_idx],
                        "high_value_deals_consistently_lost": high_value_lost_consistently_flags[latest_idx],
                    }.items()
                    if fired
                ],
            }
        )

    return rep_metrics


def compute_pipeline_temporal_metrics(
    *,
    pipeline_snapshots: List[Dict[str, Any]],
    deals: List[Dict[str, Any]],
    sales_reps: List[Dict[str, Any]],
    deal_history: List[Dict[str, Any]],
    thresholds: Dict[str, Any],
    deal_probability_trend_avg: float,
) -> Dict[str, Any]:
    """Compute pipeline-layer metrics + trigger evolution."""
    thresholds = thresholds or {}
    coverage_ratio_threshold = float(thresholds.get("pipeline_coverage_ratio_min", 0.8))

    reps_total_quota = sum(float(r.get("quota_usd") or 0.0) for r in (sales_reps or []))

    points = sorted(pipeline_snapshots or [], key=lambda x: x.get("period") or "")
    if not points:
        return {"pipeline_metrics": {}, "triggers": {}}

    total_series = [float(p.get("total_pipeline_value") or 0.0) for p in points]
    weighted_series = [float(p.get("weighted_pipeline_value") or 0.0) for p in points]
    win_rate_series = [float(p.get("win_rate") or 0.0) for p in points]
    periods = [p.get("period") for p in points]

    current_total = total_series[-1]
    prev_total = total_series[-2] if len(total_series) >= 2 else None
    pipeline_growth_rate = ((current_total - prev_total) / prev_total) if prev_total else 0.0

    # Weighted pipeline trend using up to previous 3 periods.
    prev_weighted = weighted_series[-4:-1]
    weighted_trend = weighted_series[-1] - _avg(prev_weighted)

    prev_win_rate = win_rate_series[-4:-1]
    win_rate_trend = win_rate_series[-1] - _avg(prev_win_rate)

    # Revenue trajectory derived from weighted pipeline trend + avg deal probability trend.
    if weighted_trend > 0.0 and deal_probability_trend_avg > 0.0:
        revenue_trajectory = "increasing"
    elif weighted_trend < 0.0 and deal_probability_trend_avg < 0.0:
        revenue_trajectory = "declining"
    else:
        revenue_trajectory = "stable"

    coverage_ratio_current = (current_total / reps_total_quota) if reps_total_quota else None

    # Trigger booleans across periods for evolution.
    pipeline_shrinking_flags: List[bool] = []
    weighted_pipeline_declining_flags: List[bool] = []
    win_rate_declining_flags: List[bool] = []
    coverage_ratio_low_flags: List[bool] = []

    for i in range(len(points)):
        if i >= 2:
            pipeline_shrinking = total_series[i] < total_series[i - 1] and total_series[i - 1] < total_series[i - 2]
            pipeline_shrinking_flags.append(bool(pipeline_shrinking))
        else:
            pipeline_shrinking_flags.append(False)

        if i >= 1:
            weighted_declining = weighted_series[i] < weighted_series[i - 1]
            win_declining = win_rate_series[i] < win_rate_series[i - 1]
            weighted_pipeline_declining_flags.append(bool(weighted_declining))
            win_rate_declining_flags.append(bool(win_declining))
        else:
            weighted_pipeline_declining_flags.append(False)
            win_rate_declining_flags.append(False)

        if reps_total_quota and coverage_ratio_threshold is not None:
            cov = total_series[i] / reps_total_quota
            coverage_ratio_low_flags.append(bool(cov < coverage_ratio_threshold))
        else:
            coverage_ratio_low_flags.append(False)

    latest_idx = len(points) - 1
    fired = {
        "pipeline_shrinking_2_periods": pipeline_shrinking_flags[latest_idx],
        "weighted_pipeline_declining": weighted_pipeline_declining_flags[latest_idx],
        "win_rate_declining": win_rate_declining_flags[latest_idx],
        "coverage_ratio_below_threshold": coverage_ratio_low_flags[latest_idx],
    }

    return {
        "pipeline_metrics": {
            "current_period": periods[-1],
            "total_pipeline_value_current": current_total,
            "pipeline_growth_rate": float(pipeline_growth_rate),
            "weighted_pipeline_value_current": weighted_series[-1],
            "weighted_pipeline_trend": float(weighted_trend),
            "win_rate_current": win_rate_series[-1],
            "win_rate_trend": float(win_rate_trend),
            "revenue_trajectory": revenue_trajectory,
            "coverage_ratio_current": coverage_ratio_current,
            "coverage_ratio_threshold": coverage_ratio_threshold,
        },
        "triggers": {
            "pipeline_shrinking_2_periods": {
                **_trigger_evolution(pipeline_shrinking_flags),
                "trigger_name": "pipeline shrinking 2 periods",
            },
            "weighted_pipeline_declining": {
                **_trigger_evolution(weighted_pipeline_declining_flags),
                "trigger_name": "weighted pipeline declining",
            },
            "win_rate_declining": {
                **_trigger_evolution(win_rate_declining_flags),
                "trigger_name": "win rate declining",
            },
            "coverage_ratio_below_threshold": {
                **_trigger_evolution(coverage_ratio_low_flags),
                "trigger_name": "coverage ratio below threshold",
                "coverage_ratio_min": coverage_ratio_threshold,
            },
        },
        "fired_triggers_current": [k for k, v in fired.items() if v],
        "trigger_status_current_map": {k: ("fired" if v else "not_fired") for k, v in fired.items()},
    }


def compute_forecast_stability_portfolio(
    *,
    deal_metrics: Dict[str, Any],
) -> Dict[str, Any]:
    """Compute a portfolio-level stability proxy from per-deal probability history."""
    stabilities: List[float] = []
    for dm in deal_metrics.values():
        stabilities.append(float(dm.get("forecast_stability_std_dev") or 0.0))
    return {
        "avg_deal_forecast_stability_std_dev": _avg(stabilities),
        "max_deal_forecast_stability_std_dev": max(stabilities) if stabilities else 0.0,
    }


def compute_temporal_metrics(
    *,
    deals: List[Dict[str, Any]],
    deal_history: List[Dict[str, Any]],
    interactions: List[Dict[str, Any]],
    signals_lookup: Dict[str, Dict[str, Any]],
    stage_actions: List[Dict[str, Any]],
    thresholds: Dict[str, Any],
    pipeline_snapshots: List[Dict[str, Any]],
    sales_reps: List[Dict[str, Any]],
    rep_performance_history: List[Dict[str, Any]],
) -> Dict[str, Any]:
    """
    Compute all v3 temporal intelligence outputs.

    This is deterministic: every value is derived from JSON histories + explicit formulas.
    """
    deal_metrics, portfolio_prob = compute_deal_temporal_metrics(
        deals=deals,
        deal_history=deal_history,
        interactions=interactions,
        signals_lookup=signals_lookup,
        stage_actions=stage_actions,
        thresholds=thresholds,
    )

    rep_metrics = compute_rep_temporal_metrics(
        deals=deals,
        sales_reps=sales_reps,
        rep_performance_history=rep_performance_history,
        deal_history=deal_history,
        thresholds=thresholds,
    )

    pipeline = compute_pipeline_temporal_metrics(
        pipeline_snapshots=pipeline_snapshots,
        deals=deals,
        sales_reps=sales_reps,
        deal_history=deal_history,
        thresholds=thresholds,
        deal_probability_trend_avg=float(portfolio_prob.get("avg_deal_probability_trend_current") or 0.0),
    )

    stability = compute_forecast_stability_portfolio(deal_metrics=deal_metrics)

    return {
        "deal_metrics_by_deal_id": deal_metrics,
        "portfolio_probability_trend_avg_current": float(portfolio_prob.get("avg_deal_probability_trend_current") or 0.0),
        "rep_metrics_by_rep_id": rep_metrics,
        "pipeline_metrics": pipeline.get("pipeline_metrics") or {},
        "pipeline_triggers": pipeline.get("triggers") or {},
        "pipeline_fired_triggers_current": pipeline.get("fired_triggers_current") or [],
        "forecast_stability": stability,
    }

````
