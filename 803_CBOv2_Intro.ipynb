# Cross-Business Trajectory Signal Framework (v2)

## I. Domain-Level Trend Signals

### Finance

* **Liquidity Direction** – 90d / 180d balance slope
* **Credit Deterioration** – score decline or risk band downgrade
* **Payment Stress** – missed payment acceleration
* **Product Expansion** – new financial product adoption

### Retail

* **Spend Velocity** – 30d / 90d spend delta
* **Engagement Frequency** – purchase frequency change
* **Loyalty Movement** – tier upgrade/downgrade
* **Spend Volatility** – instability in purchasing behavior

### Healthcare

* **Utilization Shift** – appointment frequency delta
* **Care Gap Risk** – days since last visit widening
* **Complexity Growth** – expanding health concern profile
* **Visit Volatility** – irregular utilization pattern

---

## II. Cross-Business Alignment Signals

### 1️⃣ Deterioration Alignment

Multiple negative signals across ≥2 business units.

### 2️⃣ Liquidity Stress Pattern

Declining balance + rising missed payments + retail contraction.

### 3️⃣ Expansion Momentum

Rising spend + rising engagement + stable/improving financial health.

### 4️⃣ Silent Disengagement

Engagement falling while financial metrics remain stable.

### 5️⃣ Service Escalation Risk

Rising support friction + satisfaction decline + behavioral instability.

---

## III. Acceleration & Stability Flags

### Acceleration Flag

Rate of decline or growth increasing vs prior window.

### Volatility Flag

Behavior becoming unstable relative to historical baseline.
