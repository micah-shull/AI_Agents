````python
"""
Orchestrator nodes for Workforce Development Orchestrator v2.

Nodes are thin: they orchestrate and update state; utilities do the work.
"""

import os
from pathlib import Path
from typing import Any, Dict

from agents.workforce_development_orchestrator_v2.orchestrator.utilities import data_loading
from agents.workforce_development_orchestrator_v2.orchestrator.utilities import analysis
from agents.workforce_development_orchestrator_v2.orchestrator.utilities import reporting


def goal_node_v2(state: Dict[str, Any]) -> Dict[str, Any]:
    """Set fixed goal for v2 run (MVP: template-based)."""
    return {
        "goal": {
            "objective": "workforce_development_v2",
            "scope": "full",
            "employee_id": state.get("employee_id"),
            "department_id": state.get("department_id"),
        }
    }


def planning_node_v2(state: Dict[str, Any]) -> Dict[str, Any]:
    """Create execution plan from goal (MVP: fixed plan)."""
    goal = state.get("goal")
    if not goal:
        return {"errors": (state.get("errors") or []) + ["planning_node_v2: goal is required"]}
    plan = [
        {"step": 1, "name": "data_loading", "description": "Load workforce data and build lookups", "outputs": ["employees", "roles", "tasks", "skills", "lookups"]},
        {"step": 2, "name": "analysis", "description": "Automation risk, skill gaps, recommendations", "outputs": ["automation_risk_analysis", "skill_gap_analysis", "learning_path_recommendations", "role_evolution_recommendations", "role_transition_recommendations", "workforce_summary"]},
        {"step": 3, "name": "report", "description": "Generate markdown report", "outputs": ["workforce_report", "report_file_path"]},
    ]
    return {"plan": plan}


def data_loading_node_v2(state: Dict[str, Any]) -> Dict[str, Any]:
    """Load all workforce JSON and build lookups. Uses project_root from state (set by runner)."""
    errors = list(state.get("errors") or [])
    project_root = state.get("project_root") or os.getcwd()
    if isinstance(project_root, Path):
        project_root = str(project_root)
    data_dir = "agents/data"
    try:
        loaded = data_loading.load_workforce_data_v2(project_root, data_dir)
        return {**loaded, "errors": errors}
    except Exception as e:
        errors.append(f"data_loading_node_v2: {e!s}")
        return {"errors": errors}


def analysis_node_v2(state: Dict[str, Any]) -> Dict[str, Any]:
    """Run automation risk, skill gap, learning path, role evolution/transition, and summary."""
    errors = list(state.get("errors") or [])
    if "tasks_by_role" not in state:
        errors.append("analysis_node_v2: data not loaded (run data_loading first)")
        return {"errors": errors}
    cfg = state.get("_config") or {}
    high = cfg.get("automation_risk_thresholds", {}).get("high", 0.7)
    medium = cfg.get("automation_risk_thresholds", {}).get("medium", 0.4)
    top_n_gaps = cfg.get("top_n_gaps", 10)
    top_n_learning = cfg.get("top_n_learning_paths", 5)
    top_n_evolutions = cfg.get("top_n_role_evolutions", 5)
    match_threshold = cfg.get("learning_path_match_threshold", 0.6)

    try:
        automation_risk_analysis = analysis.compute_automation_risk_analysis(
            state["tasks_by_role"],
            state["tasks_lookup"],
            state["employees_by_role"],
            state["roles_lookup"],
            employee_id=state.get("employee_id"),
            high_threshold=high,
            medium_threshold=medium,
        )
        skill_gap_analysis = analysis.enrich_skill_gap_analysis(
            state["skill_gaps"],
            state["employees_lookup"],
            state["roles_lookup"],
            state["skills_lookup"],
            state["learning_paths"],
            automation_risk_analysis,
        )
        learning_path_recommendations = analysis.compute_learning_path_recommendations(
            skill_gap_analysis,
            state["learning_paths"],
            state["employees_lookup"],
            match_threshold=match_threshold,
            top_n=top_n_learning,
        )
        role_evolution_recommendations = analysis.compute_role_evolution_recommendations(
            state["role_evolution"],
            state["roles_lookup"],
            state["employees_by_role"],
            top_n=top_n_evolutions,
        )
        role_transition_recommendations = analysis.compute_role_transition_recommendations(
            state["role_transitions"],
            state["roles_lookup"],
            state["employees_by_role"],
            state["employees_lookup"],
            top_n=5,
        )
        workforce_summary = analysis.compute_workforce_summary(
            state["employees"],
            state["roles"],
            state["tasks"],
            skill_gap_analysis,
            automation_risk_analysis,
            learning_path_recommendations,
            role_evolution_recommendations,
            high_threshold=high,
        )
        prioritized_gaps = analysis.prioritize_gaps(skill_gap_analysis, top_n=top_n_gaps)
        prioritized_recommendations = learning_path_recommendations[:top_n_learning]
        prioritized_evolutions = analysis.prioritize_evolutions(role_evolution_recommendations, top_n=top_n_evolutions)

        return {
            "automation_risk_analysis": automation_risk_analysis,
            "skill_gap_analysis": skill_gap_analysis,
            "learning_path_recommendations": learning_path_recommendations,
            "role_evolution_recommendations": role_evolution_recommendations,
            "role_transition_recommendations": role_transition_recommendations,
            "workforce_summary": workforce_summary,
            "prioritized_gaps": prioritized_gaps,
            "prioritized_recommendations": prioritized_recommendations,
            "prioritized_evolutions": prioritized_evolutions,
            "errors": errors,
        }
    except Exception as e:
        errors.append(f"analysis_node_v2: {e!s}")
        return {"errors": errors}


def report_node_v2(state: Dict[str, Any]) -> Dict[str, Any]:
    """Generate markdown report and optionally save to file."""
    errors = list(state.get("errors") or [])
    cfg = state.get("_config") or {}
    reports_dir = cfg.get("reports_dir", "output/workforce_development_v2_reports")

    workforce_summary = state.get("workforce_summary") or {}
    automation_risk_analysis = state.get("automation_risk_analysis") or []
    skill_gap_analysis = state.get("skill_gap_analysis") or []
    learning_path_recommendations = state.get("learning_path_recommendations") or []
    role_evolution_recommendations = state.get("role_evolution_recommendations") or []
    role_transition_recommendations = state.get("role_transition_recommendations") or []
    prioritized_gaps = state.get("prioritized_gaps") or []
    prioritized_recommendations = state.get("prioritized_recommendations") or []
    prioritized_evolutions = state.get("prioritized_evolutions") or []

    report_content = reporting.generate_workforce_report_v2(
        workforce_summary=workforce_summary,
        automation_risk_analysis=automation_risk_analysis,
        skill_gap_analysis=skill_gap_analysis,
        learning_path_recommendations=learning_path_recommendations,
        role_evolution_recommendations=role_evolution_recommendations,
        role_transition_recommendations=role_transition_recommendations,
        prioritized_gaps=prioritized_gaps,
        prioritized_recommendations=prioritized_recommendations,
        prioritized_evolutions=prioritized_evolutions,
    )
    report_file_path = None
    try:
        Path(reports_dir).mkdir(parents=True, exist_ok=True)
        from datetime import datetime
        fname = f"workforce_report_v2_{datetime.now().strftime('%Y%m%d_%H%M')}.md"
        report_file_path = str(Path(reports_dir) / fname)
        Path(report_file_path).write_text(report_content, encoding="utf-8")
    except Exception as e:
        errors.append(f"report_node_v2 save: {e!s}")

    return {
        "workforce_report": report_content,
        "report_file_path": report_file_path,
        "errors": errors,
    }
````
