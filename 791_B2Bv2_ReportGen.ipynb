````python
"""Executive-grade B2B Sales v2 report: one-pager, pipeline, risk velocity, rep performance."""

from typing import Any, Dict, List, Optional


def _verdict_and_risk(executive_triggers: List[Dict[str, Any]]) -> tuple:
    """Verdict: ON TRACK vs ATTENTION REQUIRED. Risk: LOW / MEDIUM / HIGH."""
    if not executive_triggers:
        return "ON TRACK", "LOW"
    critical = any(
        t.get("trigger_type") == "risk_velocity" or t.get("trigger_type") == "trigger_persistent"
        for t in executive_triggers
    )
    return "ATTENTION REQUIRED", "HIGH" if critical else "MEDIUM"


def _so_what_line(
    executive_triggers: List[Dict[str, Any]],
    pipeline_rollup: Dict[str, Any],
    risk_rollup: Dict[str, Any],
) -> str:
    """One sentence: business impact and who must act."""
    if not executive_triggers:
        return "Pipeline and risk within thresholds. No immediate action required."
    parts = []
    rising = risk_rollup.get("deals_risk_rising_count", 0)
    if rising:
        parts.append(f"{rising} deal(s) with rising risk")
    persistent = risk_rollup.get("trigger_persistent_count", 0)
    if persistent:
        parts.append(f"{persistent} persistent trigger(s)")
    stalled = pipeline_rollup.get("stalled_deal_count", 0)
    if stalled:
        parts.append(f"{stalled} stalled deal(s)")
    return " â€” ".join(parts) + ". Review deals and rep performance below."


def _one_ask(executive_triggers: List[Dict[str, Any]], rep_performance: List[Dict[str, Any]]) -> str:
    """Single executive action."""
    if executive_triggers:
        return "Review triggered deals and rep performance; assign ownership and follow-up dates."
    coaching = [r for r in rep_performance if r.get("coaching_candidate")]
    if coaching:
        return "Schedule coaching for reps flagged below; monitor pipeline trend."
    return "No immediate ask. Monitor next run for trend changes."


def build_b2b_report(state: Dict[str, Any]) -> str:
    """Build full markdown report from state (executive one-pager, pipeline, risk, reps, next steps)."""
    pipeline_rollup = state.get("pipeline_rollup") or {}
    risk_rollup = state.get("risk_rollup") or {}
    executive_triggers = state.get("executive_triggers") or []
    rep_performance = state.get("rep_performance") or []
    stalled_deals = state.get("stalled_deals") or []
    deals_risk_rising = state.get("deals_risk_rising") or []
    trigger_summary = state.get("trigger_summary") or {}
    weighted_forecast = state.get("weighted_forecast") or 0
    target_forecast = state.get("_config_weighted_forecast_target")  # optional, passed from node
    data_loaded_at = state.get("data_snapshot_loaded_at") or "â€”"
    validation_warnings = state.get("validation_warnings") or []

    verdict, risk_level = _verdict_and_risk(executive_triggers)
    so_what = _so_what_line(executive_triggers, pipeline_rollup, risk_rollup)
    one_ask = _one_ask(executive_triggers, rep_performance)

    # Trend: simple from triggers
    trend = "Stable" if not executive_triggers else "Attention required"

    lines: List[str] = [
        "# B2B Sales Pipeline â€” Executive Summary",
        "",
        "## Board One-Pager",
        "",
        "| | |",
        "|---|--|",
        f"| **Verdict** | {verdict} |",
        f"| **Risk level** | {risk_level} |",
        f"| **Weighted forecast** | ${weighted_forecast:,.0f} |",
    ]
    if target_forecast is not None:
        on_track = "On target" if weighted_forecast >= target_forecast else "Below target"
        lines.append(f"| **Target** | ${target_forecast:,.0f} â€” {on_track} |")
    lines.extend([
        f"| **Open deals** | {pipeline_rollup.get('open_deal_count', 0)} |",
        f"| **Deals with rising risk** | {risk_rollup.get('deals_risk_rising_count', 0)} |",
        f"| **Trend** | {trend} |",
        f"| **So-what** | {so_what} |",
        f"| **One ask** | {one_ask} |",
        "",
        "**Traceability:** Data snapshot loaded: " + data_loaded_at + ".",
        "",
    ])
    if validation_warnings:
        lines.append("**Data quality:** " + "; ".join(validation_warnings[:3]) + ".")
        lines.append("")

    # Targets vs actuals (if we have targets)
    if executive_triggers:
        lines.extend([
            "## Targets vs Actuals",
            "",
            "| Metric | Actual | Threshold | Status |",
            "|--------|--------|-----------|--------|",
        ])
        for t in executive_triggers:
            val, thresh = t.get("value"), t.get("threshold")
            status = "Above threshold" if (val is not None and thresh is not None and val >= thresh) else "â€”"
            lines.append(f"| {t.get('trigger_type', '')} | {val} | {thresh} | {status} |")
        lines.append("")

    # One-view narrative
    lines.extend([
        "## One-View",
        "",
        f"{verdict}. Weighted forecast ${weighted_forecast:,.0f}. "
        f"Open deals: {pipeline_rollup.get('open_deal_count', 0)}; "
        f"deals with rising risk: {risk_rollup.get('deals_risk_rising_count', 0)}. "
        + so_what,
        "",
    ])

    # Pipeline summary
    lines.extend([
        "## Pipeline Summary",
        "",
        f"- Weighted forecast: **${weighted_forecast:,.0f}**",
        f"- Open deals: **{pipeline_rollup.get('open_deal_count', 0)}**",
        f"- Stalled deals (attention): **{pipeline_rollup.get('stalled_deal_count', 0)}**",
        "",
    ])
    bottlenecks = pipeline_rollup.get("stage_bottlenecks") or []
    if bottlenecks:
        lines.append("| Stage | Deals | Total value | Avg days in stage |")
        lines.append("|-------|-------|-------------|-------------------|")
        for b in bottlenecks:
            lines.append(f"| {b.get('stage', '')} | {b.get('deal_count', 0)} | ${b.get('total_value', 0):,.0f} | {b.get('avg_days_in_stage', 0)} |")
        lines.append("")

    # Risk velocity
    lines.extend([
        "## Risk Velocity",
        "",
        f"- Deals with **rising risk**: {len(deals_risk_rising)}",
        f"- Triggers â€” New: {trigger_summary.get('new_count', 0)}, Persistent: {trigger_summary.get('persistent_count', 0)}, Resolved: {trigger_summary.get('resolved_count', 0)}",
        "",
    ])
    if deals_risk_rising:
        lines.append("| Deal | Company | Stage | Risk (current) | Risk delta |")
        lines.append("|------|---------|-------|----------------|------------|")
        for d in deals_risk_rising[:10]:
            lines.append(
                f"| {d.get('deal_id', '')} | {d.get('company_name', '')} | {d.get('deal_stage', '')} | "
                f"{d.get('risk_score_current', '')} | {d.get('risk_delta', '')} |"
            )
        lines.append("")

    # Stalled deals (short table)
    if stalled_deals:
        lines.append("## Stalled Deals")
        lines.append("")
        lines.append("| Deal | Company | Stage | Days in stage |")
        lines.append("|------|---------|-------|---------------|")
        for d in stalled_deals[:10]:
            lines.append(f"| {d.get('deal_id', '')} | {d.get('company_name', '')} | {d.get('deal_stage', '')} | {d.get('stalled_days', d.get('days_in_stage', ''))} |")
        lines.append("")

    # Accountability: rep performance
    lines.extend([
        "## Rep Performance",
        "",
        "| Rep | Quota attainment | Pipeline trend | Risk velocity | Coaching? | Benchmark? |",
        "|-----|------------------|----------------|----------------|-----------|------------|",
    ])
    for r in rep_performance:
        lines.append(
            f"| {r.get('rep_name', '')} | {r.get('quota_attainment_pct', 0):.1f}% | "
            f"{r.get('pipeline_trend', '')} | {r.get('risk_velocity_flag', '')} | "
            f"{'Yes' if r.get('coaching_candidate') else 'No'} | "
            f"{'Yes' if r.get('benchmark') else 'No'} |"
        )
    lines.append("")

    # Next steps
    lines.extend([
        "## Next Steps",
        "",
        f"1. {one_ask}",
        "",
    ])
    if executive_triggers:
        for t in executive_triggers:
            lines.append(f"- {t.get('message', '')}")
        lines.append("")

    return "\n".join(lines)
````
