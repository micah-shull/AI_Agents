````python
class CBOv2State(TypedDict, total=False):
    """
    State for Cross-Business Customer Intelligence Orchestrator (CBOv2).

    MVP focus:
    - Deterministic, explainable trend and alignment signals across Finance, Retail, Healthcare
    - Portfolio rollups + executive-ready markdown report
    """

    # Input / runtime options
    data_dir: Optional[str]                 # Where raw JSON lives (default from config)
    reports_dir: Optional[str]              # Where to save markdown reports
    run_id: Optional[str]                   # Optional run identifier for traceability

    # Goal & Plan
    goal: Dict[str, Any]                    # Fixed goal definition
    plan: List[Dict[str, Any]]             # Execution plan steps

    # Raw data (loaded from JSON)
    customers: List[Dict[str, Any]]        # Master customer list with cross-business IDs
    finance_signals: List[Dict[str, Any]]  # Per-customer finance metrics history
    retail_signals: List[Dict[str, Any]]   # Per-customer retail metrics history
    healthcare_signals: List[Dict[str, Any]]  # Per-customer healthcare metrics history

    # Derived per-customer trends & patterns
    customer_trends: List[Dict[str, Any]]  # One dict per customer with trend metrics + flags
    cross_business_patterns: List[Dict[str, Any]]  # Coordinated movement patterns (risk/expansion/etc.)

    # Portfolio-level rollups (executive view)
    portfolio_rollup: Dict[str, Any]       # Aggregate metrics (% improving, % deteriorating, etc.)
    top_risk_customers: List[Dict[str, Any]]
    top_growth_customers: List[Dict[str, Any]]

    # Executive triggers (deterministic, explainable)
    executive_triggers: List[Dict[str, Any]]  # Fired triggers with thresholds + rationale

    # Output
    report_markdown: str                   # Final markdown report body
    report_file_path: Optional[str]        # Path to saved markdown report

    # Metadata
    errors: List[str]                      # Any errors encountered
    processing_time: Optional[float]       # Time taken to process (seconds)


@dataclass
class CBOv2Config:
    """
    Configuration for Cross-Business Customer Intelligence Orchestrator (CBOv2).

    Keep all thresholds and file paths here so nodes stay deterministic and easy to test.
    """

    # LLM settings (reserved for future LLM-enhanced summaries; MVP is deterministic)
    llm_model: str = os.getenv("LLM_MODEL", "gpt-4o-mini")
    temperature: float = 0.2

    # Data & reports
    data_dir: str = "agents/data/CBOv2"
    reports_dir: str = "agents/output/CBOv2_reports"

    # Expected data files (under data_dir)
    customers_file: str = "customers.json"
    finance_signals_file: str = "finance_signals.json"
    retail_signals_file: str = "retail_signals.json"
    healthcare_signals_file: str = "healthcare_signals.json"

    # Trend windows (days)
    finance_short_window_days: int = 90
    finance_long_window_days: int = 180
    retail_short_window_days: int = 30
    retail_long_window_days: int = 90
    healthcare_short_window_days: int = 90
    healthcare_long_window_days: int = 180

    # Thresholds for flags (signed deltas; negative = deterioration)
    deterioration_slope_threshold: float = -0.15  # e.g., -15% vs prior window
    improvement_slope_threshold: float = 0.10     # e.g., +10% vs prior window

    # Cross-business pattern thresholds
    min_negative_domains_for_risk: int = 2       # ≥2 domains deteriorating → risk alignment
    min_positive_domains_for_growth: int = 2     # ≥2 domains improving → expansion alignment

    # Portfolio rollup & reporting
    top_n_risk_customers: int = 10
    top_n_growth_customers: int = 10

    # Traceability / audit
    include_data_counts_in_report: bool = True
    include_thresholds_in_report: bool = True

````
